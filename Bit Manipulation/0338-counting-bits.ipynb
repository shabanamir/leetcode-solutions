## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the numbers in the range 0 to n and return the number of set bits (1's) in their binary form. 

One way is to generate the binary form of each number and count the set bits → O(n) time

To do this in constant time, use the formula <br>
set bits in a number x = set bits in x/2 + (1 if x is odd else 0) <br>
*(we add 1 if x is odd because the rightmost bit (LSB) is set in odd numbers. Dividing a number by 2 shifts the bits towards right and in this process, the rightmost bit is lost.)*


## Approach
<!-- Describe your approach to solving the problem. -->
- initialize an array of size n + 1 with zeroes to count set bits for each number in range 0 to n
    - *index = number, value = total set bits*
- start the loop from 1 because number of set bits in 0 is 0
- for each number in the range 1 to n
    - find the number of set bits using the formula and add it to res
        - *set bits in num = (set bits in num / 2) + LSB*
            - *num & 1 will be zero if the number is even because 0 & 1 = 0 and will be 1 if the number is odd because 1 & 1 = 1*
            - *you can also use the mod operator to check for even and odd*
            - *another way to get num/2 is using right shift i.e num >> 1*
- return res


## Complexity
- Time complexity: O(creating result array + array traversal) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result array) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def countBits(self, n: int) -> List[int]:
        res = [0] * (n + 1)
        for i in range(1, n + 1):
            res[i] = res[i // 2] + (i & 1)
        return res
```
