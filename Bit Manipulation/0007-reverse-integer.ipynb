# <center> 7. Reverse Integer </center>


## Problem Description
[Click here](https://leetcode.com/problems/reverse-integer/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To reverse an integer x, each time get a digit from the right side of x, add it to the left side of the result, and remove the digit from x. 

To get the rightmost digit, mod the number by 10 <br>
To remove the rightmost digit, divide the number by 10 <br>
To add a digit to the left side, multiply the number by 10 and then add the digit

For example: 123
- x = 123, res = 0  → x = 12, res = 3
    - 123 % 10 = 3
    - 123 / 10 = 12
- x = 12, res = 3 → x = 1, res = 32
    - 12 % 10 = 2
    - 12 / 10 = 1
    - 3 * 10 + 2 = 30 + 2 = 32
- x = 1, res = 32 → x = 0, res = 321
    - 1 % 10 = 1
    - 1 / 10 = 0
    - 32 * 10 + 1 = 320 + 1 = 321

*Note: check for integer overflow during reversing*


## Approach
<!-- Describe your approach to solving the problem. -->
- set int_min = integer underflow value
- set int_max = integer overflow value
- set sign to -1 if x is negative, else +1
- make x positive. This is to handle the mod of negative values because, in Python, the mod of -ve and +ve values is different
- set res = 0 to track the reverse integer
- loop until x is 0 i.e there are digits left to process
    - multiply res by 10 to get the correct place to put the next digit. get the rightmost digit of x and add it to res
    - return 0 if there's overflow
    - divide x by 10 to remove the rightmost digit from x
- return res


## Complexity
- Time complexity: O(loop) → O(log$_{10}$x) → O(logx)
    - *log$_{10}$x because each time x is divided by 10*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def reverse(self, x: int) -> int:
        int_min = -2 ** 31
        int_max = 2 ** 31 - 1
        sign = -1 if x < 0 else 1
        x *= sign
        res = 0
        while x:
            res = res * 10 + x % 10
            if res > int_max or res * sign < int_min:
                return 0
            x //= 10
        return res * sign
```
