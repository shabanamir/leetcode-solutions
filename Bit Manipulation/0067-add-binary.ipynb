# <center> 67. Add Binary </center>


## Problem Description
[Click here](https://leetcode.com/problems/add-binary/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Follow the addition rules. Start from the right and add two bits at a time. If the sum is two-bit, store the rightmost bit and pass the leftmost bit as carry.

0 + 0 = 0  (carry 0) <br>
0 + 1 = 1  (carry 0) <br>
1 + 1 = 10 (carry 1) <br>
1 + 1 + 1 = 11 (carry 1)


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = an array to store the final sum digits
- set i = rightmost index in a and j = rightmost index in b to track the index in strings a and b
- set carry = 0
- loop until i or j or carry is 0 
    - get the rightmost bits of a and b
    - add the bits and carry
    - get the rightmost bit of sum using % and add it to the res 
       - *(% 2 because it's binary)*
    - get leftmost bit and add it to carry 
    - decrement i and j
- reverse the result array, convert it to string, and return it


## Complexity
- Time complexity: O(loop + reverse + join) → O(max(m, n) + m + n + m + n) → O(m + n)
    - *m = string a size, n = string b size*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(result array) → O(m + n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def addBinary(self, a: str, b: str) -> str:
        res = []
        i, j = len(a) - 1, len(b) - 1
        carry = 0
        while i >= 0 or j >= 0 or carry:
            bit_a = int(a[i]) if i >= 0 else 0
            bit_b = int(b[j]) if j >= 0 else 0
            bit_sum = bit_a + bit_b + carry
            res.append(str(bit_sum % 2))
            carry = bit_sum // 2
            i -= 1
            j -= 1
        return ''.join(res[::-1])
```
