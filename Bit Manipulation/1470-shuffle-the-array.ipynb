## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
input = [x1, x2, ..., y1, y2,...]
output = x and y in alternating fashion i.e [x1, y1, x2, y2...]

- O(n) space approach <br>
Use two pointers, the first starting at the beginning and the other starting at (n+1)th position. Alternate between them and create the new array.

- O(1) space appraoch <br>
The value range is 1 to 1000. In most languages, 32 bits are assigned to a number. 1000 takes 10 bits because 2$^{10}$ = 1024 = 1024. Use bit manipulation.


## Approach
<!-- Describe your approach to solving the problem. -->
**Two-Pointers Approach**
- set res = an array to store the shuffle array
- set left pointer l = first position i.e at index 0, right pointer r = (n + 1)th position i.e at index n
- loop n times
    - add x element at index l to the result
    - add the corresponding y element at index r to the result
    - increment l and r
- return the result array

**Bit Manipulation**
- loop n times. In each iteration i
    - shift the x value at index i 10 bit towards left to make space for the corresponding y value at index i+n.
    - add the y value with x at index i using or operator
- *after the loop, all corresponding x and y will be stored in the left half of the array as pairs. We need to extract x and y from each pair and store them in their positions
last pair positions: y = -1, x = -2
second last pair positions: y = -3, x = -4*
- set j = rightmost position
- traverse the left half in reverse order 
    - extract y using the & operator and store it at position j.
        - *(right 10-bit represents y and 1024 is represented by 10 bits so y & 1024 will give the y value)*
    - extract x by shifting 10 bits to the right and store it at position j - 1.
        - *(left 10-bits represents x, by shifting 10-bits to right, we will remove the y bits and will get x)*
    - decrement j by 2
- return nums


## Complexity
- Time complexity:
    - Two Pointers O(loop) → O(n)
    - Bit Manipulation O(for loop + for loop) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - Two Pointers O(result array) → O(2n) → O(n)
        - *input array size is 2n* 
    - Bit Manipulation O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def shuffle(self, nums: List[int], n: int) -> List[int]:
        
        # two-pointers
        res = []
        l, r = 0, n
        for i in range(n):
            res.append(nums[l])
            res.append(nums[r])
            l += 1
            r += 1
        return res

        # bit manipulation
        for i in range(n):
            nums[i] = nums[i] << 10
            nums[i] = nums[i] | nums[i+n]
        j = len(nums) - 1
        for i in range(n-1, -1, -1):
            nums[j] = nums[i] & 1023 
            nums[j-1] = nums[i] >> 10  
            j -= 2
        return nums
```
