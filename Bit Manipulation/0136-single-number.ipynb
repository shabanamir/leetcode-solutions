## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To check for duplicates in constant space, use xor operator (^) because xor of two same bits results in 0. All the duplicates (same elements) will result in 0 as the same elements have the same bits.<br>
a ^ b ^ a = a ^ a ^ b = 0 ^ b = b

xor properties
- xor of two same bits is 0 and vice versa <br>
a ^ a = 0
- xor of 0 and an element is the element itself <br>
0 ^ a = a
- xor is commutative i.e the order doesn't matter <br>
a ^ b = b ^ a


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the single number
- traverse the array
    - take xor between current element and res (prev answer of xor)
- return res


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def singleNumber(self, nums: List[int]) -> int:
        res = 0
        for n in nums:
            res ^= n
        return res
```
