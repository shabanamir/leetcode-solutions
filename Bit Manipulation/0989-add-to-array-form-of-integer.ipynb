# <center> 989. Add to Array-Form of Integer </center>


## Problem Description
[Click here](https://leetcode.com/problems/add-to-array-form-of-integer/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Follow the addition rules
- start from the right side and add the corresponding two digits
- if sum > 9, store the rightmost digit and pass the leftmost digit as carry


## Approach
<!-- Describe your approach to solving the problem. -->
- we will treat k as carry here
- traverse the digits in the number from right to left
    - add carry (k) and digit
    - store the sum rightmost digit (sum % 10) and pass the carry (sum // 10) 
        - *you can also use divmod()*
    - if carry is 0, break the loop
- if there's a carry after processing all digits, add the digits in k to the left/starting of the num array. For this:
    - either reverse the array, add the digits in k to the end of array, and again reverse the array
        - *time complexity = O(reverse + loop to add k + reverse) → O(n + k + n) → O(n)*
        - *space complexity = O(1)*
    - or use a deque because it takes O(1) time for insertion at both ends
        - *time complexity O(deque creation + k) → O(n + k) → O(n)*
        - *space complexity O(deque) → O(n)*
- return num


## Complexity
- Time complexity: O(traversal + reverse + while loop + reverse) → O(n + n + k + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def addToArrayForm(self, num: List[int], k: int) -> List[int]:
        for i in range(len(num) - 1, -1, -1):
            sum = num[i] + k
            num[i], k = sum % 10, sum // 10
            if not k: 
                break
        if k:
            num.reverse()
            while k:
                digit, k = k % 10, k // 10
                num.append(digit)
            num.reverse()
        return num
```
