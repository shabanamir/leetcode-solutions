## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Brian Kernighan’s Algorithm.

Start the loop, and each time unset the rightmost set bit i.e convert the rightmost 1 to 0 *(unset = 0, set = 1)*. The loop will end when there is no set bit left i.e the num has become 0. The total no. of 1's i.e set bits is the number of times the loop executes.

For example: n = 101100 <br>
iteration 1, n = 101**1**00 → 101**0**00 <br>
iteration 2, n = 10**1**000 → 10**0**000 <br>
iteration 3, n = **1**00000 → **0**00000 <br>
total set bits = total iterations = 3


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to count the set bits
- loop until the number is zero
    - unset the rightmost set bit. For this, subtract 1 from the number and do bitwise & with itself
    - increment res
- return res


## Complexity
- Time complexity: O(max number of set bits) → O(32) → O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def hammingWeight(self, n: int) -> int:
        res = 0
        while n:
            n &= n-1
            res += 1
        return res
```
