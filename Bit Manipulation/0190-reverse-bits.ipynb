## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
It is the same as [Reverse Integer](https://leetcode.com/problems/reverse-integer/description/), but here we have to reverse bits instead of digits. 

To reverse bits in a number x, each time get a bit from the right side of x, add it to the result, and remove the bit from x.

Use AND operator to get a bit from the right side, LEFT SHIFT operator to add a bit, and RIGHT SHIFT operator to remove a bit from the right side.

Steps to reverse 1101 to 1011
- n = 1101, res = 0000
    - get the rightmost bit of n 
    1011 & 1 = 1, so the rightmost bit is 1
        *(you can also do even-odd check using mod operator to get the rightmost bit because rightmost bit is 1 in odd numbers and 0 in even numbers)*
    - shift bits of res to left to make space to add the bit <br>
    0000 << 1 = 0000
    - add the rightmost bit to res <br>
    0000 + 1 = 0001
    - shift bits of n to right to remove the rightmost bit <br>
    1101 >> 1 = 0110
    - n = 0110 and res = 0001
- n = 0110, res = 0001 
    - rightmost bit of n = 0110 & 1 = 0
    - left shift res = 0001 << 1 = 0010
    - adding rightmost bit to res = 0010 + 0 = 0010
    - right shift n = 0110 >> 1 = 0011
    - n = 0011, res = 0010
- n = 0011, res = 0010 
    - rightmost bit of n = 0011 & 1 = 1
    - left shift res = 0010 << 1 = 0100
    - adding rightmost bit to res = 0100 + 1 = 0101
    - right shift n = 0011 >> 1 = 0001
    - n = 0001, res = 0101
- n = 0001, res = 0101 
    - rightmost bit of n = 0001 & 1 = 1
    - left shift res = 0101 << 1 = 1010
    - adding rightmost bit to res = 1010 + 1 = 1011
    - right shift n = 0001 >> 1 = 0000
    - n = 0000, res = 1011


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to store the result i.e reverse binary form of n
- loop 32 times because n is a 32-bit integer
    - left shift res by 1
        - *(this is to shift bits to the left side to make space for adding the new bit)*
    - if n is odd, the rightmost bit is 1
        - add 1 to res
            - *(you can also use mod operator for even-odd check)*
    - else n is even, the rightmost bit is 0
        - don't do anything because the corresponding bit in res is already 0 and 0 + 0 = 0
    - right shift n to remove a bit from the right side
        - *(another way to do right shift is divide the number by 2)*
- return res

## Code

```python
class Solution:

    def reverseBits(self, n: int) -> int:
        res = 0
        for _ in range(32):
            res <<= 1
            if n & 1:
                res += 1
            n >>= 1
        return res
```
