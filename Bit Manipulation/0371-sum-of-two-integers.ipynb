## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To add two numbers, we start from the right side and add digits. If the result is a two-digit number, we keep the rightmost digit and pass the leftmost as carry to the next digit.

To add numbers without the + operator, use bit manipulation. Add bits using XOR, get carry using AND, and pass the carry using LEFT SHIFT operator. <br> 
a + b = a XOR b + ((a AND b) << 1) 

XOR is used to get the rightmost bit of the result because it returns 1 if bits are different and 0 if bits are the same. 

AND is used to get carry because it returns 1 if both bits are 1 else 0. In binary addition, you only get a carry if both bits are 1.

0 + 0 = 0     rightmost bit 0 carry 0 <br> 
1 + 1 = 10    rightmost bit 0 carry 1 <br> 
0 + 1 = 1     rightmost bit 1 carry 0 <br> 
1 + 0 = 1     rightmost bit 1 carry 0 


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if any of the two numbers is 0, return the other number because 
x + 0 = x
- initialize a mask variable
    - *0xffffffff is a 32-bit mask in hexadecimal and is equal to 2147483647 in decimal. You can use the binary representation also*
    - *Python's default-int size is not 32-bit, it is a large number. So to prevent overflow and stop running into an infinite loop, we need to use a 32-bit mask to limit int size to 32-bit (0xffffffff)*
- while there is carry and the result is in the range of 32-bit
    - get carry using AND
    - add a and b using XOR and store the rightmost bit in a
    - pass carry to the next bit using LEFT SHIFT and store it in b
- check for overflow and return a


## Complexity
- Time complexity: (loop) → O(32) → O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def getSum(self, a: int, b: int) -> int:
        if a == 0:
            return b
        if b == 0:
            return a
        mask = 0xffffffff 
        while b & mask:
            carry = a & b 
            a = a ^ b
            b = carry << 1
        return (a & mask) if b > 0 else a
```
