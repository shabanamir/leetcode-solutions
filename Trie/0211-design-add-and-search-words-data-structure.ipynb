## Approach
<!-- Describe your approach to solving the problem. -->
**Class TrieNode**

**init()**
- set children = a hashmap to store the child nodes
    - *key = characters*
    - *value = child nodes*
- set end = false to track if a node is the last node (character) of  a word

**Class wordDictionary**

**init()**
- create a root node

**insert()** <br>
*start from the root node and create a new node for each character in the word if the node is not already there. Mark the last node i.e end of the word by setting the last node end attribute to true*
- set cur = root node to track current node
- for each character in the word
    - if the character is not a child node of the current node, add it as the child node
    - else set the current node to the child node of the character to move to the next node
- at the end of the loop, the current node will point to the last node (character) of the word. So, set the end attribute to true

**search()** <br>
*searching a word in trie is like finding a root-leaf node path i.e DFS. It is mentioned that a word may contain dots '.' where dots can be matched with any letter. Therefore we have to explore all possible paths i.e we have to search all paths from child nodes to leaf nodes.*
- define a helper function for DFS <br>
dfs(current node, current character index in the word)
    - if the current index is the last index
        - if the current node is the last node (character) of the word, return true, else false
    - get the character at the current index in the search word
    - if the character is not a '.'
        - if the character is a child node, search for the character in the child nodes
        - else, return false
    - else the character is a '.', search in all child nodes of the current node to find a match
- start DFS from the root node with index 0


## Complexity
- Time complexity:
    - init() O(1)
    - addWord() O(word) → O(n)
    - search() O(search word) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - init() O(1)
    - addWord() O(word) → O(n) 
        - *because for each character in the word of size n, a new node is added*
    - search() O(trie DFS) → O(n)

## Code

```python
class TrieNode:

    def __init__(self):
        self.children = defaultdict(TrieNode)
        self.last = False


class WordDictionary:

    def __init__(self):
        self.root = TrieNode()

    def addWord(self, word: str) -> None:
        cur = self.root
        for c in word:
            if c not in cur.children:
                cur.children[c] = TrieNode()
            cur = cur.children[c]
        cur.last = True

    def search(self, word: str) -> bool:

        def dfs(node: TrieNode, index: int) -> bool:
            if index == len(word):
                return node.last
            c = word[index]
            if c != '.':
                if c in node.children:
                    return dfs(node.children[c], index + 1)
                return False
            return any(dfs(child, index + 1) for child in node.children.values())

        return dfs(self.root, 0)
        
    
        


# Your WordDictionary object will be instantiated and called as such:
# obj = WordDictionary()
# obj.addWord(word)
# param_2 = obj.search(word)
```
