# <center> 12. Integer to Roman </center>


## Problem Description
[Click here](https://leetcode.com/problems/integer-to-roman/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use a hashmap to map the roman numeral/symbol to their decimal values. Traverse the hashmap in reverse and check if the numeral can be added to the result.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a hashmap and map the roman numerals
- set res = empty string to track the result roman number
- traverse hashmap in reverse to start from largest base value (the thousands place in roman)
    - count the number of times we need the current roman symbol in the result by dividing the num by base value (symbol value)
    - if the count is greater than 0
        - add the symbol count times to the res
        - update num to remove the current symbol value from num
- return res


## Complexity
- Time complexity: O(hashmap traversal) → O(13) → O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(13) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def intToRoman(self, num: int) -> str:
        roman = {'I': 1, 'IV': 4, 'V': 5, 'IX': 9, 'X': 10, 'XL': 40, 'L': 50, 'XC': 90, 'C': 100, 'CD': 400, 'D': 500, 'CM': 900, 'M': 1000}
        res = ''
        for symbol, val in reversed(roman.items()):
            count = num // val
            if count:
                res += symbol * count
                num = num % val
        return res
```
