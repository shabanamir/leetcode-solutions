## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Matrix rotation by 90

Two easy solutions that can be used for clockwise and anti-clockwise rotation

1. transpose and reverse the matrix O(n² + n²) -> O(n²) <br>
clockwise = first take transpose and then reverse the rows
anticlockwise = vice versa

2. rotate the cycles/squares in the matrix O(n²) <br>
An n x n matrix has n/2 square cycles. The first cycle is formed by the first row, last column, last row, and first column. The second cycle is formed by the second row, second-last column, second-last row, and second column, etc. <br>
clockwise = for each cycle swap the elements in a clockwise direction i.e from top to right, right to bottom, bottom to left, and from left to top. <br>
anticlockwise = vice versa

## Approach
<!-- Describe your approach to solving the problem. -->
- set n = total rows
- use two loops
    - *outer loop to traverse the cycle, inner to traverse the elements in the cycle*
    - *rotate a group of 4 elements at a time <br>* 
    - *total cycles = n/2*
    - *total groups of 4 elements in a cycle = n - cycle number - 1*
    - for each cycle
        - for each group
            - swap the elements clockwise (top left with top right, top right with bottom right, bottom right with bottom left, bottom left with top left)
                - *top left      a[i][j]*
                - *top right     a[~j][i]  or a[n-1-j][i]*
                - *bottom left   a[~i][~j] or a[n-1-i][n-1-j]*
                - *bottom right  a[j][~i]  or a[j][n-1-i]*


## Complexity
- Time complexity: O(outer loop * inner loop) → O(n * n) → O(n²)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        n = len(matrix)
        for i in range(n // 2):
            for j in range(i, n - i - 1):
                # ~num = n - 1 - num
                matrix[i][j], matrix[~j][i], matrix[~i][~j], matrix[j][~i] = matrix[~j][i], matrix[~i][~j], matrix[j][~i], matrix[i][j]
```
