# <center> 73. Set Matrix Zeroes </center>


## Problem Description
[Click here](https://leetcode.com/problems/set-matrix-zeroes/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Traverse the matrix. If a value is 0, mark the values at its top and left zero. <br>
*top = same column, first row* <br>
*left = same row, first column*

Traverse the matrix again but in reverse order. If a value on top or left is zero, mark the current value as 0

## Approach
<!-- Describe your approach to solving the problem. -->
- set col0 = 1 to check if any value in the first column (col index 0) is 0
    - *1 means the first col doesn't contain a 0, and 0 means it does*
    - *we need to check the first column separately because we will not check the first column in the nested loop. This is to avoid marking all values as 0*
- set m = total rows, n = total columns
- traverse the matrix
    - for each row i
        - if value in the first col of this row is 0
            - set col0 to 0
        - traverse all columns except the first 
            - if current cell at index (i, j) is 0, mark the values at its left (ith row first col) and top (first row jth col) as 0
- traverse the matrix again in reverse order
    - for each row i
        - for each column j
            - if the value at top or left is 0
                - set the current cell to 0
        - if col0 is 0
            - set the value in the first col of this row to 0


## Complexity
- Time complexity: O(matrix traversal + matrix traversal) → O(rows * cols + rows * cols) → O(m * n + m * n) → O(mn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        col0 = 1
        m = len(matrix)
        n = len(matrix[0])
        for i in range(m):
            if matrix[i][0] == 0:
                col0 = 0
            for j in range(1, n):
                if matrix[i][j] == 0:
                    matrix[i][0] = 0
                    matrix[0][j] = 0
        for i in range(m - 1, -1, -1):
            for j in range(n - 1, 0, -1):
                if matrix[i][0] == 0 or matrix[0][j] == 0:
                    matrix[i][j] = 0
            if col0 == 0:
                matrix[i][0] = 0
```
