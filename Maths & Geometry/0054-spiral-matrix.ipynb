# <center> 54. Spiral Matrix </center>


## Problem Description
[Click here](https://leetcode.com/problems/spiral-matrix/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
An matrix can be divided into squares/cycles. The first square is formed by the first row, last column, last row, and first column. The second square is formed by the second row, second-last column, second-last row, and second column, etc.

Divide the matrix into squares/cycles and get elements in each square in clockwise direction i.e top → right → bottom → left. 


## Approach
<!-- Describe your approach to solving the problem. -->
- initialize four variables to track the indices of the corner points 
    - *top = 0*
    - *left = 0*
    - *right = rightmost index i.e total columns - 1*
    - *bottom = bottom index i.e total rows - 1*
- set res = a list to store the spiral matrix
- run a loop until we get all squares i.e while left <= right and top <= bottom
    - *get the elements in the square in clockwise direction i.e top row → right column → bottom row → left column, and add it to res*
    - get the top row of the square by moving left to right using a loop
    - increment top after the loop to point to the top of the next square
    - get the right col by moving top to bottom using a loop
    - decrement the right after the loop to point to the right column of the next square
    - if the pointers are out of bounds, break the loop
        - *this is helpful in the case of a row or column matrix, or we are left with one row or column in the square*
    - get the bottom row in reverse by moving right to left using a loop
    - decrement bottom
    - get the left col in reverse by moving bottom to top using a loop 
    - increment left
- return res


## Complexity
- Time complexity: O(outer loop) → O(rows * column) → O(m * n)
    - *the inner loops will not run m or n times for every outer iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        left, right = 0, len(matrix[0]) - 1
        top, bottom = 0, len(matrix) - 1
        res = []
        while left <= right and top <= bottom:
            # move left to right and get top row elements
            for c in range(left, right + 1):
                res.append(matrix[top][c])
            top += 1
            # move top to bottom and get right col elements
            for r in range(top, bottom + 1):
                res.append(matrix[r][right])
            right -= 1
            if not (left <= right and top <= bottom):
                break
            # move right to left and get bottom row elements 
            for c in range(right, left - 1, -1):
                res.append(matrix[bottom][c])
            bottom -= 1
            # move bottom to top and get left col elements
            for r in range(bottom, top - 1, -1):
                res.append(matrix[r][left])
            left += 1
        return res
```
