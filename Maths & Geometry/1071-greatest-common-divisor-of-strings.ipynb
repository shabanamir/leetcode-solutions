## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The substring in str2 that divides both strings starts at index 0 and ends at the index equal to the gcd of the two strings. Find gcd using length of the strings.

For example: <br>
str1 = 'ABCABC',  length = 6 <br>
str2 = 'ABC', length = 3 <br>
gcd(6, 3) = 3 <br>
longest substring dividing str1 and str2 = str2[0:3] = 'ABC'


## Approach
<!-- Describe your approach to solving the problem. -->
- check edge cases
    - if both strings are the same, gcd is the length of the string, return any of the two strings
    - if str1 + str2 is not equal to str2 + str1, no substring can divide the strings
        - return an empty string
- For gcd, either use a library function math.gcd() or define a function using gcd logic
- find the gcd of the length of string1 and string2
- return the substring in str2 starting from index 0 and ending at the index equal to the gcd number


## Complexity
- Time complexity: O(base cases check + gcd) → O(min(str1, str2) + log(min(str1, str2)) → (n + logn) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result string + gcd recursion stack) → O(n + logn) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def gcdOfStrings(self, str1: str, str2: str) -> str:
        if str1 == str2:
            return str1
        elif str1 + str2 != str2 + str1:
            return ''
        
        def gcd(a: int, b: int) -> int:
            if not b:
                return a
            return gcd(b, a % b)

        g = gcd(len(str1), len(str2))
        return str2[:g]
```
