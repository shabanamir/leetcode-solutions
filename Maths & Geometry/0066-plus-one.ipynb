# <center> 66. Plus One </center>


## Problem Description
[Click here](https://leetcode.com/problems/plus-one/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use addition rules. In addition, we start from the right side, add digits, and pass the carry to the next digit.

There is a carry if the sum is a two-digit number. In this problem, we add 1, so there will be a carry only if the digit is 9 because 9 + 1 = 10


## Approach
<!-- Describe your approach to solving the problem. -->
- set carry = 1
    - *(initially the carry is 1 because we have to add 1)*
- traverse the digits array from right side
    - if current digit at position i is equal to 9
        - replace it with 0
            - *(9 + 1 = 10, we keep the rightmost digit i.e 0, and pass the carry i.e 1)*
    - else digit is less than 9
        - add 1 to it 
        - make carry 0 and break the loop
- if carry is still 1 after the loop, insert 1 at position 0
- return the digits array


## Complexity
- Time complexity: O(input array traversal + insert) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def plusOne(self, digits: List[int]) -> List[int]:
        carry = 1
        for i in range(len(digits) -1, -1, -1):
            if digits[i] == 9:
                digits[i] = 0
            else:
                digits[i] += 1
                carry = 0
                break
        if carry:
            digits.insert(0, 1)
        return digits
```
