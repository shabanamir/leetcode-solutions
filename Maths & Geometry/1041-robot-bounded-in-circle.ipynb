## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
north = +y, south = -y, east = +x, west = -x

A robot is bound in a circle if after all instructions:
- it returns to the initial position (0, 0) 
- it doesn't return to the initial direction north (0, 1)

Instruction G changes the position (x, y)
- increment y if the direction is north and decrements y if it is south <br>
north (0, 1) and (x, y) → (x, y + 1) <br>
south (0, -1) and (x, y) → (x, y - 1)
- increments x if direction is east and decrements x if it is west <br>
east (1, 0) and (x, y) → (x + 1, y) <br>
west (-1, 0) and (x, y) → (x - 1, y)
- x → x + direction x <br>
y → y + direction y

Instruction L and R change the direction
- L changes it to anti-clockwise i.e 90 degrees to the left <br>
north (0, 1) → west (-1, 0) <br>
south (0, -1) → east (1, 0) <br>
east (1, 0) → north (0, 1)  <br>
west (-1, 0) → south (0, -1) <br>
x → -1 * y <br>
y → x
- R changes it to clockwise i.e 90 degrees to the right <br>
north (0, 1) → east (1, 0) <br>
south (0, -1) → west (-1, 0) <br>
east (1, 0) → south (0, -1) <br>
west (-1, 0) → north (0, 1) <br>
x → y <br>
y → -1 * x


## Approach
<!-- Describe your approach to solving the problem. -->
- set direction variables dir_x = 0, dir_y = 1 because the robot's initial direction is north (0, 1)
- set position variables x = 0, y = 0 because the robot's initial position is (0, 0)
- traverse the instructions list
    - if the instruction is 'G'
        - update position (x, y)
    - if the instruction 'L'
        - update direction (dir_x, dir_y)
    - if the instruction is 'R'
        - update direction (dir_x, dir_y)
- return true if the robot returns to the initial position or any direction other than initial, else false


## Complexity
- Time complexity: O(input array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isRobotBounded(self, instructions: str) -> bool:
        dir_x, dir_y = 0, 1
        x, y = 0, 0
        for i in instructions:
            if i == 'G':
                x, y = x + dir_x, y + dir_y
            elif i == 'L':
                dir_x, dir_y = -1 * dir_y, dir_x
            elif i == 'R':
                dir_x, dir_y = dir_y, -1 * dir_x
        return (x, y) == (0, 0) or (dir_x, dir_y) != (0, 1)
```
