# <center> 13. Roman to Integer </center>


## Problem Description
[Click here](https://leetcode.com/problems/roman-to-integer/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use a hashmap to map the roman numeral/symbol to their decimal values and apply roman numeral rules.

1. if the left numeral is smaller than the right numeral, subtract the value 
2. if the left numeral is greater than the right numeral, add the value


## Approach
<!-- Describe your approach to solving the problem. -->
- create a hashmap to map roman numerals
- set res = 0 to track the running total i.e result integer
- traverse the string
    - if i is inbound and the current (left) symbol value is less than the next (right) symbol value, subtract it from the res
    - else add it to the res
- return res


## Complexity
- Time complexity: O(traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(7) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def romanToInt(self, s: str) -> int:
        roman = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
        res = 0
        for i in range(len(s)):
            if i + 1 < len(s) and roman[s[i]] < roman[s[i + 1]]:
                res -= roman[s[i]]
            else:
                res += roman[s[i]]
        return res
```
