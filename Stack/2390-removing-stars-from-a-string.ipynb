# <center> 2390. Removing Stars From a String </center>


## Problem Description
[Click here](https://leetcode.com/problems/removing-stars-from-a-string/description/)


# Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
For each star, we need to remove the element towards its left i.e the last inserted element. It follows the last in first out (LIFO) structure, use stack.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = an empty list to represent stack
- traverse the string <br>
for each character in string s
    - if the char is a star
        - pop stack to remove the stack top element i.e the left closest non-star character
    - else the char is a non-star char
        - push it to the stack
- return the string form of the stack


## Complexity
- Time complexity: O(string traversal + join()) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def removeStars(self, s: str) -> str:
        res = []
        for c in s:
            if c == '*':
                res.pop()
            else:
                res.append(c)
        return ''.join(res)
```
