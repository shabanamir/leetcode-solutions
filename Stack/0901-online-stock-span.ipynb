## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
current day span = 1 + number of previous days for which price <= current day price

If we get a high price, we will keep going backward in time until we find a price greater than the current price. So, we will stop if we find the previous greater price i.e we have to find the previous greater element. 

To find previous greater or next greater element, use monotonic decreasing stack.

*In a monotonic decreasing stack, the elements are kept in a decreasing order from bottom to top*
- *if the new element is smaller than the previous element i.e. stack top, it is pushed to the stack*
- *if the new element is greater, the stack is popped until the new element becomes smaller i.e. until all elements that are smaller than the new are removed*
- *the new element is the next greater element for each popped element*
- *after popping all smaller elements, the stack top will contain the previous greater element*


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- create an empty stack to store price and span as a pair
    - *index 0 = price*
    - *index 1 = span*

**next()**
- set span = 1 to track the number of consecutive days for which price was <= current day price
    - *(1 because the current day is included)*
- loop until we find the previous greater price i.e until current price becomes smaller i.e while stack top element price <= current price
    - *(for previous greater element, the stack should be in strictly decreasing order i.e equal elements are not allowed that's why the loop condition is <= instead of < to remove equal elements also)*
    - add the stack top element span to the current span
    - pop stack to remove stack top element
- current price is smaller now, push the current price and span as a pair to the stack
- return span


## Complexity
- Time complexity:
    - init() O(1)
    - next() O(loop) → amortized constant time → O(1)
    - *amortized because the operation runs mostly fast but on some occasions or once in a while runs slow*
        - *the loop runs until previous price < current price*
        - *if every next price is smaller than the previous, the loop will not execute and the span will be 1*
        - *if every next price is greater than the previous, the loop will run only 1 time and the span will be 2*
        - *if we get a greater next price after n smaller prices, the loop will execute n times but that will be once in a while and is very rare*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - init() O(1) 
    - next() O(1) because only 1 item is pushed 
    - in the worst case if every new price is smaller than the previous, stack size will be n after n calls to next()
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class StockSpanner:

    def __init__(self):
        self.stack = []

    def next(self, price: int) -> int:
        span = 1
        while self.stack and self.stack[-1][0] <= price:
            span += self.stack[-1][1]
            self.stack.pop()
        self.stack.append((price, span))
        return span
        


# Your StockSpanner object will be instantiated and called as such:
# obj = StockSpanner()
# param_1 = obj.next(price)
```
