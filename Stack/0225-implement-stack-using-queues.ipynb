## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In the stack, insertion and deletion occur at the right end because it removes the last inserted element first (LIFO). <br>
In the queue, insertion takes place on the right side and deletion on the left side because the queue removes the first inserted element first (FIFO). 

To implement stack using queue, we need to move the last inserted element from the right side to the left side of the queue to delete the last inserted element first. 

Stack can be implemented using two queues by making:
1. pop and top operations costly O(n) or
2. push operation costly 

We will use the second approach because it's better to make one operation costly rather than two.


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- create two empty queues, q1 and q2
    - *q1 (queue1) is to represent the stack*
    - *q2 (queue2) is to maintain the order of the elements according to stack in q1*

**push()**
- add the new element in q2 
    - *(q2 is always empty when push() is called)*
- pop all the elements from q1 and add them to q2. In this way, the new (last) inserted element will be on the left side of q2 and q1 will become empty
- now we have to make q2 empty again for the next push operation, so, swap the queues

**pop()**
- remove the last inserted element i.e pop an element from the left of q1 and return it
    - *(no need to check if q1 is empty because the description says that the calls to top() and pop() are valid)*

**top()**
- return the stack top element i.e the leftmost element from q1

**empty()**
- check if stack is empty i.e q1 length is 0 or not and return the result


## Complexity
- Time complexity:
    - init() O(1)
    - push() O(loop) → O(queue1) → O(n)
    - pop() O(1)
    - top() O(1)
    - empty() O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - init() O(1)
    - push() O(queue1 + queue2) → O(n + n) → O(n)
        - *queues will contain n elements after n calls to push*
    - pop() O(1)
    - top() O(1)
    - empty() O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class MyStack:

    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x: int) -> None:
        self.q2.append(x)
        while self.q1:
            self.q2.append(self.q1.popleft())
        self.q1, self.q2 = self.q2, self.q1
        
    def pop(self) -> int:
        return self.q1.popleft()
        
    def top(self) -> int:
        return self.q1[0]

    def empty(self) -> bool:
        return len(self.q1) == 0
        


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```
