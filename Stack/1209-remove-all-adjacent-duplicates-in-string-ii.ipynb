## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
For each character, we need to check if the previous k characters are equal or not.

To find the previous equal or next equal element, use stack.


## Approach
<!-- Describe your approach to solving the problem. -->
- create an empty stack
- traverse the staring <br>
for each character in string s
    - if the previous character (stack top) is not the same as the current char
        - push a pair of current char and its count to stack
    - else the previous char is the same as the current
        - update the count of the char
        - if we get k same characters, i.e stack top character count is equal to k
            - pop stack to remove the stack top
- multiply each character with its count to get the result string and return it
    - create an array
        - *(we have used an array because string concatenation is expensive)*
    - for each character and its count in the stack
        - multiply the char with its count and add it to the array
    - convert the array to a string


## Complexity
- Time complexity: O(string traversal + creating result array + converting array to string) → O(n + n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(stack + result array) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def removeDuplicates(self, s: str, k: int) -> str:
        stack = []
        for c in s:
            if not stack or stack[-1][0] != c:
                stack.append([c, 1])
            else:
                stack[-1][1] += 1
                if stack[-1][1] == k:
                    stack.pop()
        return ''.join(c * count for c, count in stack)
```
