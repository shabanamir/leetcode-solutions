# <center> 84. Largest Rectangle in Histogram </center>


## Problem Description
[Click here](https://leetcode.com/problems/largest-rectangle-in-histogram/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
For each bar, find the max rectangle ending with the bar, and return the max of all rectangles.

To find the max area rectangle ending with a bar, we will extend the bar to the left to maximize the width but we can only include bars with height >= current bar i.e we will keep extending until we find a smaller bar. So, we will stop if we find a smaller height bar i.e we have to find the previous smaller element.

To find the previous smaller or next smaller element, use monotonic increasing stack.

*In a monotonic increasing stack, the elements are kept in an increasing order from bottom to top*
- *if the new element is greater than the previous element i.e stack top, it is pushed to the stack*
- *if the new element is smaller, the stack is popped until the new element becomes greater i.e until all elements that are greater than the new are removed*
- *the new element is the next smaller element for each popped element*
- *after popping all greater elements, the stack top will contain the previous smaller element*


## Approach
<!-- Describe your approach to solving the problem. -->
- create an empty stack
- append 0 at the end of the input array 
    - *this is to empty the stack and calculate the area for every bar in one loop*
    - *0 because the min height is 0 and the stack is popped until the current height becomes greater than the stack top*
    - *you can skip this but you'll have to check the remaining elements in the stack at the end in a separate loop*
- set res = 0 to track the max area
- traverse the input array <br>
for current bar index i, current bar height h in the heights array
    - set j = current index to find how far to the left we can extend the current bar
    - loop until we find the previous smaller element i.e until the current height becomes greater than the stack top i.e while stack top element height >= current height
        - *(to find the previous smaller element, the stack should be in strictly increasing order i.e equal elements are not allowed that's why the loop condition is >= instead of > to remove equal elements also)*
        - pop stack to get the greater bar height and its index
            - *(stack top contains a bar with height >= current bar, we can include this bar)*
        - find the area and update res by taking max of current area and previous max area
            - *area = height * width = height * (current bar index i - previous bar index j)*
    - current bar is now greater, push a pair of the extended index j and current height to the stack
- return max_area i.e res


## Complexity
- Time complexity: O(array traversal * stack top check) → O(for loop * while loop) → O(n * constt) → O(n)
    - *the inner while loop will not run n times for every outer iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def largestRectangleArea(self, heights: List[int]) -> int:
        stack = []
        heights.append(0)
        res = 0
        for i, h in enumerate(heights):
            j = i
            while stack and stack[-1][1] >= h:
                j, prev_h = stack.pop()
                res = max(res, prev_h * (i - j))
            stack.append((j, h))
        return res
```
