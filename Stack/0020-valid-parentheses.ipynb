# <center> 20. Valid Parentheses </center>


## Problem Description
[Click here](https://leetcode.com/problems/valid-parentheses/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Parenthesis is valid if the last opening bracket is closed by the same type of closing bracket i.e Last in first out (LIFO). Use stack.

If you get an opening bracket, push it to stack. If you get a closing bracket and the stack top is the opening bracket of the same type, pop the stack. In the end, if the stack is empty, then the brackets are balanced.


## Approach
<!-- Describe your approach to solving the problem. -->
- set stack = an empty list to represent the stack
- set match = a hashmap with closing brackets as keys and opening brackets as values
    - *(this is to simplify the bracket checking, you can skip hashmap and use if-else inside the loop)* 
- traverse the string
    - if the current character is not in the hashmap, it is an opening bracket
        - push it to stack
    - else it is a closing bracket
        - if the stack is empty or the current character doesn't match the opening bracket at the stack top
            - return false
        - else it matches the opening bracket at stack top, pop the stack
- check if the stack is empty and return the result i.e true if it is empty else false

## Complexity
- Time complexity: O(string traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(stack + hashmap) → O(n + 3) → O(n)
    - *3 because there are three types of brackets stored in the hashmap*


## Code

```python
class Solution:
    
    def isValid(self, s: str) -> bool:
        stack = []
        match = {
                    ')': '(', 
                    '}': '{', 
                    ']' : '['
                }
        for c in s:
            if c not in match:
                stack.append(c)
                continue
            if not stack or stack[-1] != match[c]:
                return False
            stack.pop()
        return not stack
```
