## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In the stack, insertion and deletion occur at the right end because it removes the last inserted element first (LIFO). <br>
In the queue, insertion takes place on the right side and deletion on the left side because the queue removes the first inserted element first (FIFO).

To implement queue using stack, we need to move the first inserted element to the right side of the stack to delete the first inserted element first.

Queue can be implemented using two stacks by making:
1. pop and top operations costly O(n) or
2. push operation costly 

We will use the second approach because it's better to make one operation costly rather than two.


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- initialize two empty stacks, s1 and s2
    - *s1 (stack1) is to represent the queue*
    - *s2 (stack2) is to maintain the order of the elements according to queue in s1*

**push()** <br>
*move all elements from stack1 to stack2 and then back to stack1, so that the first inserted element is always at the front of the queue i.e at the top of stack1*
- loop and pop all elements from stack1 and add them to stack2
    - *(stack1 contains elements according to queue order i.e first inserted element is at the top to delete it first and the last inserted element is at the bottom. To maintain this order and add the new element before the last inserted element i.e at the bottom, we need to empty stack1 first)*
- stack1 is empty, so add the new element in stack1
- the new element is now at the bottom of stack1, so add the previous elements of stack1 back to maintain the queue order i.e pop all the elements from stack2 and add them to stack1

**pop()**
- remove the first inserted element i.e pop an element from the right of stack1 and return it
    - *(no need to check if stack1 is empty because the description says that the calls to peek() and pop() are valid)*

**peek()**
- return the front (peek) element of queue i.e the rightmost element from stack1

**empty()**
- check if the queue is empty i.e stack1 length is 0 or not and return the result


## Complexity
- Time complexity:
    - O(init) O(1)
    - push() O(loop1 + loop2) → O(stack1 + stack2) → O(n + n) → O(n)
    - pop() O(1)
    - peek() O(1)
    - empty() O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - O(init) O(1)
    - push() O(stack1 + stack2) → O(n + n) → O(n)
        - *stacks will contain n elements after n calls to push*
    - pop() O(1)
    - peek() O(1)
    - empty() O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class MyQueue:

    def __init__(self):
        self.s1 = []
        self.s2 = []

    def push(self, x: int) -> None:
        while self.s1:
            self.s2.append(self.s1.pop())
        self.s1.append(x)
        while self.s2:
            self.s1.append(self.s2.pop())

    def pop(self) -> int:
        return self.s1.pop()
        
    def peek(self) -> int:
        return self.s1[-1]

    def empty(self) -> bool:
        return len(self.s1) == 0
        


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```
