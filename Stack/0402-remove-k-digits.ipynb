# <center> 402. Remove K Digits </center>


## Problem Description
[Click here](https://leetcode.com/problems/remove-k-digits/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To get the minimum number, remove greater digits from the left of a digit because the left digits have higher place values (like tens, hundreds etc).

To remove the greater elements from the left, we have to find the next smaller element because we can remove the greater elements from the left if the current element is the next smaller element.

To find the previous smaller or next smaller element, use monotonic increasing stack. 

*In a monotonic increasing stack, the elements are kept in an increasing order from bottom to top*
- *if the new element is greater than the previous element i.e stack top, it is pushed to the stack*
- *if the new element is smaller, the stack is popped until the new element becomes greater i.e until all elements that are greater than the new are removed*
- *the new element is the next smaller element for each popped element*
- *after popping all greater elements, the stack top will contain the previous smaller element*


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if k is equal to the length of the number (number of digits), we need to remove all digits, return 0 
- set res = a list to represent stack
- for each digit in the number
    - loop until we can remove digits i.e while k is not 0 and until current digit becomes greater i.e while previous (stack top) digit is greater than the current digit
        - pop stack to remove the greater digit
        - decrement k
    - add the current digit to the stack if it is a trailing zero or it is not 0
        - *(the stack not empty condition is to prevent the leading zeroes. 0 will be added only if it is a trailing 0 that's when the stack is not empty)*
- if k is still not 0, remove k digits from stack top
    - loop until k is 0 and there are digits in the stack
        - pop stack to remove a digit
        - decrement k
- return the string form of the stack if the stack is not empty (i.e there are still digits left in the num), else return 0


## Complexity
- Time complexity: O(for loop * while loop + while loop + join()) → O(n * constt + k + n - k) → O(n + k + n)  → O(n)
    - *the inner while loop will not run n times for every outer iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def removeKdigits(self, num: str, k: int) -> str:
        if len(num) == k:
            return '0'
        stack = []
        for digit in num:
            while k > 0 and stack and stack[-1] > digit:
                stack.pop()
                k -= 1
            if stack or digit is not '0':
                stack.append(digit)
        while k and stack:
            stack.pop()
            k -= 1
        return ''.join(stack) if stack else '0'
```
