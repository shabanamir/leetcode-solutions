## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
All operations are related to the last score i.e Last in first out (LIFO). Use a stack and perform the operations.


## Approach
<!-- Describe your approach to solving the problem. -->
- set scores = an empty list to represent stack
- traverse the input array <br>
for each character (operation) in the input list
    - if the current character is '+'
        - add the stack top two scores (last and second last element) to get sum of the last two scores and push the sum to the stack
    - if the current character is 'D'
        - multiply the stack top score by 2 to double the last score and push the result to the stack
    - if the current character is 'C'
        - pop the stack top to remove the last score
    - else the current character is int
        - convert it to int and push it to the stack
- return the sum of the scores in the stack


## Complexity
- Time complexity: O(array traversal + stack sum) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def calPoints(self, operations: List[str]) -> int:
        scores = []
        for o in operations:
            if o == '+':
                scores.append(scores[-1] + scores[-2])
            elif o == 'D':
                scores.append(scores[-1] * 2)
            elif o == 'C':
                scores.pop()
            else:
                scores.append(int(o))
        return sum(scores)
```
