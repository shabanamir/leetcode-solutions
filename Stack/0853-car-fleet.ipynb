# <center> 853. Car Fleet </center>


## Problem Description
[Click here](https://leetcode.com/problems/car-fleet/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
**The faster car will slow down to match the slower car's speed**

For every element (fast car), we need to find the previous smaller elements (slow cars).

Approaches:
1. To find previous smaller or next smaller element, use monotonic increasing stack.
2. If the arrival time of a car is earlier i.e <= the current slowest car, it will join the current fleet. Else it will be the new slowest car and lead a new fleet.
arrival time = target - position / speed


## Approach
<!-- Describe your approach to solving the problem. -->
- pair the position and speed, sort the pairs (cars) in descending order to start with the car closest to the target first, and calculate each car's arrival time using the formula
- set res = 0 to count the number of fleets
- set max_time = 0 to track the current max time i.e the current slowest car time
- for each car's arrival time
    - if the current arrival time > max_time, the car is the next slowest and will lead a new fleet
        - increment fleet count (res) 
        - update max_time
    - *else the current arrival time <= max_time, the car will join the current fleet*
- return res


## Complexity
- Time complexity: O(sort + array traversal) → O(nlogn + n) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort + arrival times array) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def carFleet(self, target: int, position: List[int], speed: List[int]) -> int:
        arrival_times = [(target - p) / s for p, s in sorted(zip(position, speed), reverse=True)]
        res = 0
        max_time = 0
        for time in arrival_times:
            if time > max_time:
                res += 1
                max_time = time
        return res
```
