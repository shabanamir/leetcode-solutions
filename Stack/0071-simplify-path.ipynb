# <center> 71. Simplify Path </center>


## Problem Description
[Click here](https://leetcode.com/problems/simplify-path/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To jump back to the parent directory, we need to remove the last directory. We need a strucutre to remove last inserted element in O(1) i.e a LIFO strucutre. Use stack.

- pop if you get ".." to jump back to the parent directory
- push if you get a valid directory name 


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to represent stack
- split the path at each "/" and traverse
    - continue if you get "" or "." to move to the next directory
    - if you get ".."
        - if the stack is not empty
            - pop the stack top to 
    - else the string is a directory name
        - push it to the stack
- join the stack elements using a forward slash, add a forward slash to the start of the joined stack result, and return it


## Complexity
- Time complexity: O(split + traversal + join) → O(n + n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(split + stack) → O(n + n) → O(n)
    - *split() creates a list*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def simplifyPath(self, path: str) -> str:
        res = []
        for str in path.split('/'):
            if str in ('', '.'):
                continue
            if str == '..':
                if res:
                    res.pop()
            else:
                res.append(str)
        return '/' + '/'.join(res)
```
