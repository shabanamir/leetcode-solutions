## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use hashmap to track the most frequent element, and stack to push & pop it.


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- create a hashmap to store the frequency of each element
    - *key = element*
    - *value = frequency*
- create another hashmap to store all elements with the same frequency
    - *key = frequency*
    - *value = list/stack of elements with the same freq*
- set max_freq = 0 to track the current max frequency i.e most frequent element frequency

**push()**
- increment the frequency of the current element in the freq hashmap
- update max_freq by taking the max of current element frequency and previous max frequency
- add the current element to the list/stack of elements that have the same frequency in the hashmap (freq_stack)

**pop()**
- pop an element from the list/stack of the most frequent elements i.e elements with frequency equal to max_freq 
- decrement the frequency of the popped element because we have removed its one occurrence
- update max_freq if we have removed all elements with max_freq
- return the popped element


## Complexity
- Time complexity: 
    - init() O(1)
    - push() O(1)
    - pop() O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - init() O(1)
    - push() O(1)
    - pop() O(1)
    - hashmap and stack size will be n after n calls to push()
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class FreqStack:

    def __init__(self):
        self.freq = defaultdict(int)
        self.freq_stack = defaultdict(list)
        self.max_freq = 0
        
    def push(self, val: int) -> None:
        self.freq[val] += 1
        self.max_freq = max(self.max_freq, self.freq[val])
        self.freq_stack[self.freq[val]].append(val)

    def pop(self) -> int:
        val = self.freq_stack[self.max_freq].pop()
        self.freq[val] -= 1
        if not self.freq_stack[self.max_freq]:
            self.max_freq -= 1
        return val
        


# Your FreqStack object will be instantiated and called as such:
# obj = FreqStack()
# obj.push(val)
# param_2 = obj.pop()
```
