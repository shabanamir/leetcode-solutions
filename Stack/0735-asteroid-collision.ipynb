# <center> 735. Asteroid Collision </center>


## Problem Description
[Click here](https://leetcode.com/problems/asteroid-collision/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
- asteroids moving in the same direction will not collide
- asteroid moving in opposite directions will collide 
    - if both are of same size, both will explode
    - if both are of different sizes, the smaller one will explode

The collision and type of collision depends on the current and previous (last inserted) asteroid i.e we need a LIFO structure. Use a stack.

If asteroid is moving right (+ sign), push it to stack. If it is moving left (-sign), check for collision.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = an empty list to represent stack
- traverse the asteroids <br>
for each asteroid in asteroids
    - if asteroid is moving right, the direction is the same as the previous one, and they will not collide. So, push it to stack 
    - else asteroid is moving left, the direction can be opposite to the previous asteroid and they can collide
        - if previous asteroid (stack top) is a right-moving smaller size asteroid, both will collide and the previous asteroid will explode. So, pop all right-moving smaller-size asteroids from the stack
        - if stack is empty or previous asteroid is a left-moving asteroid, they will not collide because both are moving in the same direction. So, push the current asteroid
        - if previous is a right-moving equal-size asteroid, both will collide and explode. So, pop stack to remove the previous asteroid
        - else skip because the previous is a right-moving larger-size asteroid, both will collide and the current asteroid will explode because its size is smaller
- return res


## Complexity
- Time complexity: O(traversal) → O(for loop * while loop) → O(n * constt) → O(n)
    - *the inner loop will not run n times for every outer iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def asteroidCollision(self, asteroids: List[int]) -> List[int]:
        res = []
        for a in asteroids:
            if a > 0:
                res.append(a)
            else:
                while res and res[-1] > 0 and res[-1] < abs(a):
                    res.pop()
                if not res or res[-1] < 0:
                    res.append(a)
                elif res[-1] == abs(a):
                    res.pop()
        return res
```
