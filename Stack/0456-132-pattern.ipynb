# <center> 456. 132 Pattern </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find a sequence num_i < num_k < num_j.

We need to find num_i, the next greater number *(num_k)* for num_i, and the next greater number *(num_j)* for num_k.

To find previous greater or next greater element, use monotonic decreasing stack.

*In a monotonic decreasing stack, the elements are kept in a decreasing order from bottom to top*
- *if the new element is smaller than the previous element i.e stack top, it is pushed to the stack*
- *if the new element is greater, the stack is popped until the new element becomes smaller i.e until all elements that are smaller than the new are removed*
- *the new element is the next greater element for each popped element*
- *after popping all smaller elements, the stack top will contain the previous greater element*


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there are not at least 3 elements, return false
- create an empty stack
- set num_k = a very small number to track num_k
- traverse the array in reverse to get num_i and num_j
    - if current number < num_k, it is num_i and it must also be less than num_j
        - return true because the condition num_i < num_k < num_j is satisfied
    - else current number > num_k, the current number is num_j
        - *find num_k*
            - loop until the current element becomes smaller i.e while stack top < current element
                - pop stack to get num_k
                    - *(current number is num_j, stack top number is num_k becasue it is less than the current number)*
        - current number is smaller now, push it to stack
- if the loop completes without returning true, there is no 132 pattern, return false


## Complexity
- Time complexity: O(array traversal) → O(for loop * while loop) → O(n * constt) → O(n)
    - *the inner loop will not run n times for every outer iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def find132pattern(self, nums: List[int]) -> bool:
        if len(nums) < 3:
            return False
        stack = []
        num_k = float('-inf')
        for i in range(len(nums) - 1, -1, -1):
            if nums[i] < num_k:
                return True
            while stack and stack[-1] < nums[i]:
                num_k = stack.pop()
            stack.append(nums[i])
        return False
```
