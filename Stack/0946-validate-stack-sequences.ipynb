# <center> 946. Validate Stack Sequences </center>


## Problem Description
[Click here](https://leetcode.com/problems/validate-stack-sequences/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
- Approach 1: O(n) space <br>
Use a stack. Each time push an element from pushed array to the stack. Pop stack if the stack top is equal to the top of the popped array. If the stack is empty after processing all elements in pushed, the pushed and popped arrays are the result of push and pop operations on a stack.

- Approach 2: O(1) space <br>
Simulate the stack operations using the pushed array.


## Approach
<!-- Describe your approach to solving the problem. -->
**Stack Approach**
- set res = an empty list to represent stack
- set i = 0 to track the top element in the popped array
- traverse the pushed array
    - push the element to stack
    - pop stack while stack top is equal to the popped array element
- check if the stack is empty and return the result

**Stack Simulation Approach**
- set i = 0 to track the index in the simulated stack (pushed array)
- set j = 0 to track the index of the top element in the popped array
- traverse the pushed array <br>
for each element x in pushed
    - push x to the simulated stack by setting pushed[i] value to x
    - increment i 
    - pop simulated stack by decrementing i while the simulated stack top (pushed[i-1]) is equal to the popped array element 
- check if simulated stack is empty (i == 0) and return the result


## Complexity
- Time complexity: 
    - Stack Approach O(n)
    - Stack Simulation Approach O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - Stack Approach O(n)
    - Stack Simulation Approach O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def validateStackSequences(self, pushed: List[int], popped: List[int]) -> bool:
        
        # O(n) space
        res = []
        i = 0
        for num in pushed:
            res.append(num)
            while res and res[-1] == popped[i]:
                res.pop()
                i += 1
        return not res

        # O(1) space
        i, j = 0, 0
        for num in pushed:
            pushed[i] = num
            i += 1
            while i > 0 and pushed[i - 1] == popped[j]:
                i -= 1
                j += 1
        return not i
```
