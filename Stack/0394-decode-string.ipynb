# <center> 394. Decode String </center>


## Problem Description
[Click here](https://leetcode.com/problems/decode-string/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To repeat a string k times, we need k and k is on the left side i.e the last inserted element. Use LIFO structure i.e stack to access the last inserted element in O(1)
 
- if you get an opening bracket, push the encoded string and k
- if you get a closing bracket, start decoding. Pop stack to get k and encoded string. Decode string by multiplying it with k.


## Approach
<!-- Describe your approach to solving the problem. -->
- set stack = a list to represent stack
- set res = an empty string to track the current encoded/decoded string
- set k = an empty string to track the current number k
- traverse the input <br>
for each character c in string s
    - if c is a digit, it is part of k
        - add it to k
            - *(instead of a string concatenation, you can also use (k * 10 + int(c)) to get the current number k. Here string concatenation method is used to easily find k and convert it to int at once not multiple times. k <= 300, at most there will be 3 digits i.e 3 concatenations)*
    - if c is an opening bracket
        - push a pair of current number k and current string res 
        - reset res & k to find the next encoded string and k
    - if c is a closing bracket
        - pop stack to get the previous decoded string & previous k and
        - decode res by multiplying it with k and add the result with the previous decoded string
    - else c is a letter, it is part of the current encoded string
        - add it to res
- return res as the final decoded string


## Complexity
- Time complexity: O(traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(stack + result string) → O(n + m) → O(m)
    - *m > n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def decodeString(self, s: str) -> str:
        stack = [] 
        res, k = '', ''
        for c in s:
            if c.isdigit():
                k += c
            elif c == '[':
                stack.append((res, int(k)))
                res, k = '', ''
            elif c == ']':
                prev_res, prev_k = stack.pop()
                res = prev_res + prev_k * res
            else:
                res += c
        return res
```
