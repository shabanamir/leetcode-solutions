## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
For every element (temperature), we need to find the next greater element. 

To find previous greater or next greater element, use monotonic decreasing stack.

*In a monotonic decreasing stack, the elements are kept in a decreasing order from bottom to top*
- *if the new element is smaller than the previous element i.e stack top, it is pushed to the stack*
- *if the new element is greater, the stack is popped until the new element becomes smaller i.e until all elements that are smaller than the new are removed*
- *the new element is the next greater element for each popped element*
- *after popping all smaller elements, the stack top will contain the previous greater element*


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = an array of size n with zeroes to store the number of days for each temperature
- create an empty stack
- traverse the temperatures
    - loop until current temperature becomes smaller i.e while stack top < current temp
        - pop stack to get the index of the previous temperature
            - *(current temp is the next greater temp for the stack top temp)*
        - find the difference between current and previous temperature indices to find the number of days we have to wait and add the difference at the corresponding index in the result array
    - the current temp is smaller now, add its index to the stack
- return res


## Complexity
- Time complexity: O(traversal) → O(for loop * while loop) → O(n * constt) → O(n)
    - *the inner loop will not run n times for every outer iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(stack + result array) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        res = [0] * len(temperatures)
        stack = []
        for cur_t in range(len(temperatures)):
            while stack and temperatures[stack[-1]] < temperatures[cur_t]:
                prev_t = stack.pop()
                res[prev_t] = cur_t - prev_t
            stack.append(cur_t)
        return res
```
