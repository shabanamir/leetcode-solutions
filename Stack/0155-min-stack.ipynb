## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Other operations take O(1) time in stack. Finding the minimum element takes O(n) in stack. To make it O(1), use a variable to track the minimum value.


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- initialize an empty list to represent stack
- initialize a variable min = largest possible value to store the minimum value in the stack

**push()** <br>
*(store 2nd smallest value below 1st smallest value)*
- if the new element <= current minimum
    - the new element becomes the 1st minimum and the current minimum becomes the second minimum. So add the second minimum first and then the first minimum
    - update min
- else only add the new element

**pop()**
- get the stack top element 
- remove the stack top (last element)
- if the removed element is equal to min, it means the minimum has been removed and we need to find the new minimum. As we have already stored the second min below the first, so the new stack top contains the new minimum
    - update min 
    - remove stack top again

**top()**
- return the last/rightmost element in the stack 

**getMin()**
- return min 


## Complexity
- Time complexity: 
    - init() O(1)
    - push() O(1)
    - pop() O(1)
    - top() O(1)
    - getMin() O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - init() O(1)
    - push() O(1)
    - pop() O(1)
    - top() O(1)
    - getMin() O(1)
    - stack size will be n after n push operations
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.min = float('inf')

    def push(self, val: int) -> None:
        if val <= self.min:
            self.stack.append(self.min)
            self.min = val
        self.stack.append(val)
        
    def pop(self) -> None:
        top = self.stack[-1]
        self.stack.pop()
        if top == self.min:
            self.min = self.stack[-1]
            self.stack.pop()

    def top(self) -> int:
        return self.stack[-1]
        
    def getMin(self) -> int:
        return self.min
        

# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```
