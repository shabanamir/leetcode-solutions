# <center> 22. Generate Parentheses </center>


## Problem Description
[Click here](https://leetcode.com/problems/generate-parentheses/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use DFS and generate all valid (balanced) combinations by adding a left (opening) bracket only after a valid right (closing) bracket.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to store the valid combinations
- define a helper function for DFS
    - dfs(left brackets count, right brackets count, current string)
        - if the left and right brackets count is equal to n, the current string/combination is valid because a string of n parenthesis contains n opening brackets and n closing bracket
            - add current string to the result list
            - return
        - else the current combination is invalid
            - if left brackets are less
                - increment left bracket count, add an opening bracket to the current string, and call dfs()
            - if right brackets are less
                - increment right bracket count, add a closing bracket to the current string, and call dfs()
- start DFS with initial count of left and right brackets as 0 and an empty string
- return res


## Complexity
- Time complexity: O(recursion) → O(calls at each level ^ total levels) → O(2$^n$)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(result list + recursion stack) → O(total valid combinations + recursion depth/levels) → O(nth Catalan number + n) → O(C$_n$ + n) → O(C$_n$)
    - *n = 1, valid parenthesis = () = 1 <br>
       n = 2, valid parenthesis = ()(), (()) = 2 <br>
       n = 3, valid parenthesis = ()()(), (())(), ()(()), (()()), ((())) = 5 <br>
       n = 4, valid parenthesis = 14 <br>
       catalan sequence*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def generateParenthesis(self, n: int) -> List[str]:
        res = []
        
        def dfs(left, right, s):
            if left == right == n:
                res.append(s)
                return
            if left < n:
                dfs(left + 1, right, s + '(')
            if right < left:
                dfs(left, right + 1, s + ')')

        dfs(0, 0, '')
        return res
```
