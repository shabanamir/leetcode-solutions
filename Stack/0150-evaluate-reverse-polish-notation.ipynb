# <center> 150. Evaluate Reverse Polish Notation </center>


## Problem Description
[Click here](https://leetcode.com/problems/evaluate-reverse-polish-notation/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In reverse polish/postfix notation, the operator appears after operands. For each operator, we need the last two elements to perform the operation. It follows last in first out (LIFO) structure, use stack.

Push the operands to the stack. If you get an operator, pop two elements from the stack, perform the operation, and push the result to the stack. At last, return the stack top as it contains the final answer.

*(You can use a hashmap for operators instead of multiple if conditions and two variables instead of a stack, but the question is designed for the stack data structure.)*


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there is only one element in the array, return it
- set res = an empty list to represent stack
- traverse the input array
    - if the char is an operator
        - pop two elements from the stack, perform the operation, and push the result to the stack
    - else the char is a number
        - convert it to int and push it to the stack
- return the stack top i.e last element in the res 


## Complexity
- Time complexity: O(array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def evalRPN(self, tokens: List[str]) -> int:
        if len(tokens) == 1:
            return int(tokens[0])
        res = []
        for c in tokens:
            if c == '+':
                a, b = res.pop(), res.pop()
                res.append(a + b)
            elif c == '-':
                a, b = res.pop(), res.pop()
                res.append(b - a)
            elif c == '*':
                a, b = res.pop(), res.pop()
                res.append(a * b)
            elif c == '/':
                a, b = res.pop(), res.pop()
                res.append(int(b / a))
            else: 
                res.append(int(c))
        return res[-1]
```
