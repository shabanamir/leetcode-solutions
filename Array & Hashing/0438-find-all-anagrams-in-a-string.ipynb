# <center> 438. Find All Anagrams in a String </center>

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Two strings are anagrams if they have the same elements and the same frequency.

Use the sliding window approach. Search for a window (substring) of size p in string s. If the window is valid i.e characters and frequency are the same as string p, add the window start to the result.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - if string p size is greater than s, there can’t be an anagram of p in s, return an empty set
- set res = a set to store the start index of the anagrams
- set count = a hashmap to store the character frequencies of string p for tracking the required characters
    - *(either loop to fill the hashmap or use a library function)*
- set left and right pointers to track the left and right sides of the window i.e set l = 0, r = 0
- traverse the string s using the right pointer
    - if the right character count is not 0, we found an anagram character
      - decrement its count because we have found a required character and it is a part of current window
          - *(remember the rule: decrement count of character if it gets inside the window and increment if it gets outside)*
      - if current window size is valid i.e size is equal to anagram size
        - add left pointer to res as the start index of the anagram
      - move right pointer
    - if left and right pointers are in the same position
      - move both pointers
    - else the current window is not valid because the current character count is 0 i.e it is not part of the anagram. We need to slide the window by removing a character from the left
      - add the left character back to the hashmap because it is a required character and we have removed it
          - *(remember the rule)*
      - move left pointer   
- return res


## Complexity
- Time complexity: O(counting characters in string p + traversing string s) → O(m + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(unique characters in string p) → O(lowercase english letters) → O(26) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findAnagrams(self, s: str, p: str) -> List[int]:
        if len(p) > len(s):
            return []
        res = []
        count = Counter(p)
        l, r = 0, 0
        while r < len(s):
            if count[s[r]] > 0:
                count[s[r]] -= 1
                if r - l + 1 == len(p):
                    res.append(l)
                r += 1
            elif l == r:
                l += 1
                r += 1
            else:
                count[s[l]] += 1
                l += 1
        return res 
```
