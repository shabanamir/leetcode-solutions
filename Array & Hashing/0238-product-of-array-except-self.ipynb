# <center> 238. Product of Array Except Self </center>


## Problem Description
[Click here](https://leetcode.com/problems/product-of-array-except-self/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Prefix and suffix product

answer[i] <br>
= product of all the elements except nums[i] <br>
= product of all element before nums[i] i.e upto index i-1 * product of all elements after nums[i] i.e from i+1 to n <br>
= prefix product * suffix product


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = size of the input array
- set res = an array of size n with default value 1 
- set prefix = 1 to track the prefix product
- set suffix = 1 to track the suffix product
- traverse the array from start to end and add the prefix product for each element
  - for i in range 0 to n
    - multiply the current prefix product with value at index i in the result array to get the prefix product for index i
    - update the prefix product by multiplying it with the value at index i in the input list
- traverse the array in reverse to add the suffix product for each element
  - for i in range n to 0
    - multiply the current suffix product with value at index i in the result array to get the suffix product for index i
    - update the suffix product by multiplying it with the value at index i in the input list
- return res


## Complexity
- Time complexity: O(creating result array + loop for prefix product + loop for suffix product) → O(n + n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result array) → O(n) but according to the follow-up the output array doesn't count as extra space, so space complexity is O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def productExceptSelf(self, nums: List[int]) -> List[int]:
        n = len(nums)
        res = [1] * len(nums)
        prefix, suffix = 1, 1
        for i in range(n):
            res[i] *= prefix
            prefix *= nums[i]
        for i in range(n - 1, -1, -1):
            res[i] *= suffix
            suffix *= nums[i]
        return res 
```
