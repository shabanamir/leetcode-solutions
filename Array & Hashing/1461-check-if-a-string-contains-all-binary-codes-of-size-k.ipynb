# <center> 1461. Check If a String Contains All Binary Codes of Size K </center>


## Problem Description
[Click here](https://leetcode.com/problems/check-if-a-string-contains-all-binary-codes-of-size-k/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Check if the number of k-size substrings in the string equals the total combinations (binary codes) of size k.

Total k-size binary codes (combinations) = $2^k$

A substring is formed by removing or adding characters from left or right


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = required number of substrings i.e $2^k$ 
- if the size of the string is less than m, return false
- set codes = a hashset to store the unique k-size substrings (binary codes)
- traverse string, *loop until size of string - k + 1 to keep index inbound*
    - get a k-size substring and add it to the set
- check if the number of substrings in the set is equal to the required substrings i.e m


## Complexity
- Time complexity: O(iterations * substring) → O(n - k * k) → O(n * k)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashset + substring) → O(total k-size substrings + k) -> O(n-k + k) → O(n + k) → O(n)
    - *in each iteration, we add a substring to the set and total iterations are n - k*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def hasAllCodes(self, s: str, k: int) -> bool:
        m = 1 << k       # 1 << k = 2**k
        if len(s) < m:
            return False
        codes = set()
        for i in range(len(s) - k + 1):
            codes.add(s[i: i + k])
        return len(codes) == m  
```
