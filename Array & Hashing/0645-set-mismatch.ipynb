## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the duplicate and missing number in the range 1 to n.

It is combination of:
[442. Find All Duplicates in an Array](https://leetcode.com/problems/find-all-duplicates-in-an-array/description/)
[448. Find All Numbers Disappeared in an Array](https://leetcode.com/problems/find-all-numbers-disappeared-in-an-array/description/)

The numbers in the array are in the range 1 to n and the indices are 0 to n - 1.

Traverse the array and mark the index of the current number i.e make the number at that index negative. If you encounter a marked index i.e a negative number, it is the duplicate number.

Traverse the array again and check the indices. If you encounter a non-marked index i.e a positive number, it is the missing number.


## Approach
<!-- Describe your approach to solving the problem. -->
- set duplicate = an outbound number i.e 0 to track the duplicate number
- traverse the array
    - find the index of the current number
    - if the index is already marked i.e the number at that index is negative, the current number is a duplicate
        - update duplicate, *(use the absolute value because we may have made the number negative for marking)*
    - else mark the index by making the number at the index negative
- traverse the array again
    - if index is not marked i.e its value is positive, it is the missing number
        - return the duplicate number and missing number (i.e current index) as a list


## Complexity
- Time complexity: O(traversal to mark indices + traversal to check indices) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findErrorNums(self, nums: List[int]) -> List[int]:
        duplicate = 0
        for n in nums:
            i = abs(n)
            if nums[i - 1] < 0:
                duplicate = abs(n)
            else:
                nums[i - 1] *= -1
        for i in range(len(nums)):
            if nums[i] > 0:
                return [duplicate, i + 1]
```
