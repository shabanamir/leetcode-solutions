## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The result of the difference of two arrays is the same as the difference of two sets. 

Convert the arrays to sets to remove duplicates and find the difference.

1. Find the difference using the - operator <br>
*(because of the output format, this solution involves too many conversions, the output should be in array form and - operator returns a set)*
2. Find the difference using traversal


## Approach
<!-- Describe your approach to solving the problem. -->
**Approach 1**
- convert the arrays to sets 
- subtract set1 from set2 to get the numbers in the first array that are not present in the second array and vice versa
- convert the difference result to an array and return it
    - *index 0 = set1 difference result*
    - *index 2 = set2 difference result*

**Approach 2**
- convert the arrays to sets
- set res = a 2d array of size 2 to store the difference result
    - *index 0 = array to store first array difference*
    - *index 2 = array to store second array difference*
- traverse the first array set
    - if the current number is not in the second array set
        - add the number to the array at index 0 in res
- traverse the second array set
    - if the current number is not in the first array set
        - add the number to the list at index 1 in res
- return res  


## Complexity
- Time complexity: 
    - Approach 1 O(conversion from array to set + finding difference + conversion from set to array) → O(nums1 to set1 + nums2 to set2 + set1 difference + set2 difference + set1 difference result to array + set2 difference result to array) → O(m + n + m + n + m + n) → O(m + n)
    - Approach 2 O(conversion from array to set + array traversal) → O(nums1 to set1 + nums2 to set2 + nums1 traversal + nums2 traversal) → O(m + n + m + n) → O(m + n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - Approach 1 O(set1 + set2 + set1 difference + set2 difference + result list) → O(m + n + m + n + m + n) → O(m + n)
    - Approach 2 O(result array) → O(m + n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def findDifference(self, nums1: List[int], nums2: List[int]) -> List[List[int]]:
        # appraoch 1
        set1, set2 = set(nums1), set(nums2)
        return [list(set1 - set2), list(set2 - set1)]

        # appraoch 2
        set1, set2 = set(nums1), set(nums2)
        res = [[], []]
        for n in set1:
            if n not in set2:
                res[0].append(n)
        for n in set2:
            if n not in set1:
                res[1].append(n)
        return res
```
