## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use an array or a hashmap to store the parking spaces. 

We can park a car if the available car space is greater than zero.

## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- set spaces = an array of size 3 to store the available car spaces  
    - *3 because total car types are 3*
    - *index 0 = big, index 1 = medium, index 2 = small*

**addCar()**
- if space is available for the specified car type i.e value at index carType - 1 is not 0 *(carType - 1 because car types are in range 1 to 3 and indices are in range 0 to 2)*
    - park the car by decreasing the car space by 1
    - return true because we have parked the car
- else we can't park the car, return false


## Complexity
- Time complexity: 
    - init() O(array creation) → O(3) → O(1)
    - addCar() O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
    - init() O(array) → O(3) → O(1)
    - addCar() O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class ParkingSystem:

    def __init__(self, big: int, medium: int, small: int):
        self.spaces = [big, medium, small]

    def addCar(self, carType: int) -> bool:
        if self.spaces[carType - 1]:
            self.spaces[carType - 1] -= 1
            return True
        return False
        


# Your ParkingSystem object will be instantiated and called as such:
# obj = ParkingSystem(big, medium, small)
# param_1 = obj.addCar(carType)
```
