## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The intersection of two arrays is the same as the intersection of two sets.

1. Find the intersection using & operator <br>
*(this solution involves too many conversions)*
2. Find the intersection using traversal


## Approach
<!-- Describe your approach to solving the problem. -->
#### Approach 1
- Convert the arrays to sets to remove duplicates, find the intersection using & operator, convert the result to a list, and return it

#### Approach 2
- set res = a list to store the intersection result
- convert the array1 to a set to remove duplicates and search in O(1) time
- traverse array2
    - if the current number in array2 is in the array1 set
        - add it to the result list
        - remove the number from the array1 set to avoid duplicates in the result
- return the result list


## Complexity
- Time complexity: O(converting array1 to set + traversing array2) → O(m + n) 
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(array1 set) → O(array1 size) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def intersection(self, nums1: List[int], nums2: List[int]) -> List[int]:
        
        # approach 1
        return list(set(nums1) & set(nums2))
        
        #approach 2
        res = []
        seen = set(nums1)
        for n in nums2:
            if n in seen:
                res.append(n)
                seen.remove(n)
        return res
```
