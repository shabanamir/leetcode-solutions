# <center> 724. Find Pivot Index </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-pivot-index/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
For each element, we need to find the left and right sum.

left sum = sum of elements from index 0 until current index = prefix sum
right sum = sum of elements from last until current index


## Approach
<!-- Describe your approach to solving the problem. -->
- set left = 0 to track the left sum
- set right = total sum of the array to track the right sum
- traverse the array <br>
for each index, number in the array
  - exclude the current element from the right sum to get right sum for the current index
  - if left sum is equal to the right sum
    - return the current index
  - update the left sum
- if the loop completes without returning anything, there's no pivot, return -1


## Complexity
- Time complexity: O(finding sum + array traversal) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def pivotIndex(self, nums: List[int]) -> int:
        left = 0
        right = sum(nums)
        for i, n in enumerate(nums):
            right -= n
            if left == right:
                return i
            left += n
        return -1
```
