# <center> 128. Longest Consecutive Sequence </center>


## Problem Description
[Click here](https://leetcode.com/problems/longest-consecutive-sequence/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
- search for the starting point of a sequence. A number x is the starting of a subsequence if x-1 is not in the input
- for each starting point, find the number of consecutive elements 
- return the max count

Use a hash set to search in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the length of the longest seq
- convert the nums array into a set
- traverse the set
    - if current number - 1 is not present in the set, the current number is the starting point of a seq
        - set size = 1 to track the length of the current seq
        - search for consecutive numbers in the subsequence and update the size
        - update res by taking max of current seq size and previous max size
- return res


## Complexity
- Time complexity: O(converting array to set + set traversal * finding seq numbers) → O(n + for loop * while loop) → O(n + n * constt) → O(n)
  - *the while loop will not run n times for every outer iteration
  for example: [100,4,200,1,3,2] (n = 6), the while loop will only execute in case of 100 and 1 for 2 and 4 times, respectively*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(set) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def longestConsecutive(self, nums: List[int]) -> int:
        res = 0
        nums = set(nums)
        for n in nums:
            if n - 1 not in nums:
                size = 1
                while n + size in nums:
                    size += 1
                res = max(res, size)
        return res
```
