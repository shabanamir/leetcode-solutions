## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to put 0's in the left, 2's in the right and 1's in the middle. This can be done in two ways:

1. Multiple loops
count the occurrences of 0, 1, and 2 (x, y, and z). Replace the first x elements of the array with 0, the next y elements with 1, and the remaining with 2.

2. Single loop
use pointers to track the left and right indices. Move 0's to left and 2's to right, 1's will be auto-moved to the middle.


## Approach
<!-- Describe your approach to solving the problem. -->
- use three pointers i, l, r to keep track of current, left, and right element indices
- while i is inbound
    - if current element is 0
      - move it to the left by swapping current element with the left index element
      - move current and left index pointers i.e increment l and i
    - if current element is 2
      - move it to the right by swapping it with the right index element
      - move right index pointer i.e decrement r
    - else the current element is 1
      - no need to move it because it is already in the middle
      - move current index pointer i.e increment i


## Complexity
- Time complexity: O(loop) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)

## Code

```python
class Solution:

    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        i, l, r = 0, 0, len(nums) - 1
        while i <= r:
            if nums[i] == 0:
                nums[i], nums[l] = nums[l], nums[i]
                i += 1
                l += 1
            elif nums[i] == 2:
                nums[i], nums[r] = nums[r], nums[i]
                r -= 1
            else:
                i += 1
```
