## Problem Description
[Click here](https://leetcode.com/problems/repeated-dna-sequences/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use a hashset and search for repeated sequences (10-size substring). Add the repeated sequences to the result.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a hash set to store the repeated sequences and avoid duplicates
- set seen = a hash set to track the repeated sequences
- traverse the string, *loop size - 9 times to keep index inbound*
    - get current sequence/substring of size 10
    - if the sequence is in the seen set, it is a repeated sequence
        - add it to the result set
    - add the sequence to the seen set
- convert the result set to an array and return it 


## Complexity
- Time complexity: O(string traversal * substring creation + converting result set to array) → O(n-9 * 10 + n/10) → O(n + n) → O(n)
    - *iterations = n - 9*
    - *substring size = 10*
    - *if every sequence is repeated twice, the result hashset will contain n/10 sequences*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(seen hashset + result hashset + substring) → O(n-9 + n/10 + 10) → O(n + n + 10) → O(n)
    - *if every sequence is unique, the seen hashset will contain n-9 sequences because total iterations are n - 9 and in each iteration, a sequence is added*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findRepeatedDnaSequences(self, s: str) -> List[str]:
        res = set()
        seen = set()
        for i in range(len(s) - 9):
            seq = s[i:i+10]
            if seq in seen:
                res.add(seq)
            seen.add(seq)
        return list(res)
```
