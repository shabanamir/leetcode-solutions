# <center> 535. Encode and Decode TinyURL </center>


## Problem Description
[Click here](https://leetcode.com/problems/encode-and-decode-tinyurl/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Generate a random 5-char alphanumeric code to encode the url. To avoid encoding the same url multiple times and to easily decode, use hashmap. 


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- set url_code = a hashmap to map the url (longUrl) to code (shortUrl)
    - *key = long url*
    - *value = short url*
- set code_url = a hashmap to map the code (shortUrl) to url (longUrl)
    - *key = short url*
    - *value = long url*
- set pool = an alphanumeric string containing all small letters, capital letters, and digits to generate the random code

**encode()**
- if the url is not in the hashmap, it is not encoded before
    - generate a random 5-character code
    - create short url *i.e short url = 'http://tiny_url.com' + random code*
    - add the short url as the key and long url as its value to the code_url hashmap and vice versa
    - return the short url
- else return the encoded form of the url using the url_code hashmap i.e longUrl value

**decode()**
- return the decoded form of the url using the code_url hashmap i.e shortUrl value


## Complexity
- Time complexity: 
  - encode() O(random.choices() + join()) → O(5 + 5) → O(1)
  - decode() O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
  - encode() O(1)
  - decode  O(1)
  - *the hashmap will contain n urls after n calls to encode()*


## Code

```python
class Codec:
    
    def __init__(self):
        self.url_code = {}
        self.code_url = {}
        self.pool = string.ascii_letters + string.digits

    def encode(self, longUrl: str) -> str:
        """Encodes a URL to a shortened URL.
        """
        if longUrl not in self.url_code:
            random_str = ''.join(random.choices(self.pool, k=5))
            code = 'http://tiny_url.com' + random_str
            self.code_url[code] = longUrl
            self.url_code[longUrl] = code
            return code
        return self.url_code[longUrl]

    def decode(self, shortUrl: str) -> str:
        """Decodes a shortened URL to its original URL.
        """
        return self.code_url[shortUrl]
        

# Your Codec object will be instantiated and called as such:
# codec = Codec()
# codec.decode(codec.encode(url))
```
