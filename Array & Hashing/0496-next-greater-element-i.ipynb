# <center> 496. Next Greater Element I </center>


## Problem Description
[Click here](https://leetcode.com/problems/next-greater-element-i/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
For each element x in nums1, if x is present in nums2, we have to find the next greater element of x in nums2.

To find previous greater or next greater element, use monotonic decreasing stack.

*In a monotonic decreasing stack, the elements are kept in a decreasing order from bottom to top*
- *if the new element is smaller than the previous element i.e stack top, it is pushed to the stack*
- *if the new element is greater, the stack is popped until the new element becomes smaller i.e until all elements that are smaller than the new are removed*
- *the new element is the next greater element for each popped element*
- *after popping all smaller elements, the stack top will contain the previous greater element*


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = an array of size equal to nums1 initialized with -1 to store the next greater elements for nums1 elements
- set mapping = a hashmap and add the indices of nums1
    - *key = nums1 element*
    - *value = index*
- create a stack 
- traverse nums2 array
  - loop until the current element becomes smaller i.e while stack top < current element 
    - pop the stack to get the stack top element and find its index using the hashmap
    - insert the current element at that index in the res array as the next greater element
  - current element is now smaller, add it to the stack if it is in nums1 i.e hashmap because we only need to find the next greater elements for nums1
- return res


## Complexity
- Time complexity: O(creating result array + creating hashmap + for loop * while loop) → O(nums1 + nums1 + nums2 * constt) → O(m + m + n * constt) → O(m + n)
  - *the inner while loop doesn't run n times for every outer loop iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(result array + hashmap + stack) → O(nums1 + nums1 + nums2) → O(m + m + n) → O(m + n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def nextGreaterElement(self, nums1: List[int], nums2: List[int]) -> List[int]:
        res = [-1] * len(nums1)
        mapping = {n:i for i, n in enumerate(nums1)}
        stack = []
        for n in nums2:
            while stack and stack[-1] < n:
                i = mapping[stack.pop()]
                res[i] = n
            if n in mapping:
                stack.append(n)
        return res
```
