# <center> 554. Brick Wall </center>


## Problem Description
[Click here](https://leetcode.com/problems/brick-wall/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To find a line that will cross the minimum bricks, we need to find where the most brick edges line up i.e which index has the highest number of edges because passing through the edges of a brick doesn't count as crossing a brick.

To find an edge, use the prefix sum <br>
For example, if row is [1, 2, 2, 1], then the inside edges = [0+1, 1+2, 1+2+2] = [1, 3, 5] <br>
*(we didn't consider the last brick (i.e 1, sum is 6) because its edge is the right edge of the wall and we can't consider the wall edge)*


## Approach
<!-- Describe your approach to solving the problem. -->
- set maxi = 0 to track the index with max edges
- set edges = a hashmap to count the number of edges for an index
    - *key = edge index*
    - *value = total number of edges at that index*
- for each row in the wall
    - set pre_sum = 0 to track the prefix sum to find the edge index
    - for each column (brick) in the row, *(skip the last brick because its edge is the wall edge)*
        - update prefix sum by adding the current column (brick)
        - the prefix sum represents the edge index. We got a new edge, so increment the count of edges for the edge index in the hashmap 
        - update maxi 
- return the difference between the length of the wall and the max edges as the minimum number of crossed bricks


## Complexity
- Time complexity: O(rows * columns) → O(n * constt) → O(n)
  - *each row has a different number of columns (bricks), so the nested loop will not run n times for every outer iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashmap) → O(total edge indices) → O(max sum row) → O(m)
  - *because the total edge indices can't be more than the max sum and here each row sum is the same*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def leastBricks(self, wall: List[List[int]]) -> int:
        maxi = 0
        edges = defaultdict(int)
        for row in wall:
            pre_sum = 0
            for col in range(len(row) - 1):
                pre_sum += row[col]
                edges[pre_sum] += 1
                maxi = max(maxi, edges[pre_sum])
        return len(wall) - maxi
```
