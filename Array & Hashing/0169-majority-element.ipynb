## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
1. Sorting - O(nlogn) <br>
The majority element will always be in the middle of the sorted list.

2. Hashmap - O(n) <br>
Count the occurrences of elements using a hashmap and return the element with count = n / 2

3. Moore’s Voting Algorithm - O(n) time O(1) space <br>
According to the algo, if there is a majority element in an array, it will always remain in the lead i.e its votes (count) will never become 0. 
Select a candidate (element) from a set of candidates (array). Traverse the array. If the current element is the same as the chosen one, increase the votes (count), else decrease the votes. If votes become 0, select a new candidate.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = array first element to track the majority element
- set count = 1 to track the count of the current majority element
- traverse the array from the second element
  - if count is 0, update the majority element i.e set res as the current element
  - if the current element is the same as res (majority element), increment the count else decrement the count
- return res


## Complexity
- Time complexity: O(array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def majorityElement(self, nums: List[int]) -> int:
        res = nums[0]
        count = 1
        for i in range(1, len(nums)):
            if count == 0:
                res = nums[i]
            count += 1 if nums[i] == res else -1
        return res
```
