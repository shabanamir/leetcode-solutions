## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Calculate the prefix sum for each index. For the sum range, return the difference between the prefix sum of the right and left indices. 


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- compute the prefix sum of array
    - set pre_sum = a list to store the prefix sum for each index
    - set cur_sum = 0 to track the current sum
    - traverse the input array
        - add the current number to cur_sum to get the prefix sum for the current index
        - add the prefix sum i.e cur_sum to prefix_sum array

**sumRange()**
- if left is 0, return the prefix sum of right
- else return the difference between values at position right and left-1 in the prefix sum array


## Complexity
- Time complexity:
    - init() O(array traversal) → O(n) one-time
    - sumRange(): O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - init() O(prefix sum array) → O(n) one-time
    - sumRange(): O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class NumArray:

    def __init__(self, nums: List[int]):
        self.pre_sum = []
        cur_sum = 0
        for n in nums:
            cur_sum += n
            self.pre_sum.append(cur_sum)
        
    def sumRange(self, left: int, right: int) -> int:
        if not left:
            return self.pre_sum[right]
        return self.pre_sum[right] - self.pre_sum[left - 1]
        


# Your NumArray object will be instantiated and called as such:
# obj = NumArray(nums)
# param_1 = obj.sumRange(left,right)
```
