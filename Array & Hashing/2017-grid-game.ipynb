# <center> 2017. Grid Game </center>

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Robot 1 can take n possible paths. Robot 2 can only move down (bottom row) or right (top row) of the column that robot 1 traversed and can collect the top row points or bottom row points.

Take the max of the top and bottom row sum because robot 2 wants to maximize the points. Take the minimum of the maximum points because robot 1 wants to minimize the points.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a very large number to track the total points
- set bottom = 0 to track the bottom sum 
- set top = 0 to track the top sum
- traverse the values in the first and second rows in parallel, *(the grid has two rows of size n, either loop in range 0 to n or use zip() for parallel traversal)*
    - decrement the top sum
    - update res i.e total points 
        - *(take max of bottom and top row points and then take min of their result and previous total points)*
    - increment the bottom sum
- return res


## Complexity
- Time complexity: O(loop) → O(min(row1 size, row2 size)) → O(min(n, n)) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def gridGame(self, grid: List[List[int]]) -> int:
        res = math.inf
        bottom, top = 0, sum(grid[0])
        for a, b in zip(grid[0], grid[1]):
            top -= a
            res = min(res, max(top, bottom))
            bottom += b
        return res
```
