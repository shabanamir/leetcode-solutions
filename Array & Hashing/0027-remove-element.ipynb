# <center> 27. Remove Element </center>


## Problem Description
[Click here](https://leetcode.com/problems/remove-element/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a number and have to remove all its occurrences from the array in place i.e we can't use another array. 

Move all the occurrences of the number to be removed towards the end of the array because the problem statement says that the elements beyond the new length of the array can be anything.


## Approach
<!-- Describe your approach to solving the problem. -->
- set k = 0 to track the index and count numbers that are not equal to the val (number to be removed)
- traverse the array
    - if the number is not equal to val
        - insert the number at the kth index
        - increment k
- return k


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def removeElement(self, nums: List[int], val: int) -> int:
        k = 0
        for i in range(len(nums)):
            if nums[i] != val:
                nums[k] = nums[i]
                k += 1
        return k
```
