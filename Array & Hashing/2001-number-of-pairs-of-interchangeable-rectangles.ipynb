# <center> 2001. Number of Pairs of Interchangeable Rectangles </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Interchangeable rectangles have the same width-to-height ratio.

Store the rectangle height and width ratio in a hashmap. Traverse the ratios, and for each ratio count, find the number of pairs.

For n rectangle, there are n(n-1)/2 pairs.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to store the total pair count
- set ratio_count = a hashmap to count the same ratios (similar rectangle)
    - *key = ratio i.e width/height*
    - *value = frequency*
- traverse the rectangles array and fill the hashmap
    - for height, width in the rectangles array 
        - get the ratio and increment the ratio count in the hashmap
- traverse the hashmap
    - for each ratio count
        - find the number of pairs using the formula and update res 
- return res


## Complexity
- Time complexity: O(array traversal + hashmap traversal) → O(n + hashmap keys i.e unique ratios) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def interchangeableRectangles(self, rectangles: List[List[int]]) -> int:
        res = 0
        ratio_count = defaultdict(int)
        for w, h in rectangles:
            ratio_count[w/h] += 1
        for c in ratio_count.values():
            res += (c * (c - 1)) // 2
        return res
```
