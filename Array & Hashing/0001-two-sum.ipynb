# <center> 1. Two Sum </center>


## Problem Description
[Click here](https://leetcode.com/problems/two-sum/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
n1 + n2 = target 
n2 = target - n1

For each number n1, check if we have n2 i.e target - n1 using a hashmap.


## Approach
- create a hashmap for lookup and storing indices 
    - *key = number*
    - *value = index*
    - *we need a hashmap instead of a hashset because we need to find the second number n2 and return its index*
- traverse the array <br>
for each index, number (n1) in the nums list
    - find n2 
    - if n2 is in the hashmap
        - return the indices of n1 and n2
    - else add the number and its index to the hashmap


## Complexity
- Time complexity: O(array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        num_map = {}
        for i, n in enumerate(nums):
            diff = target - n
            if diff in num_map:
                return i, num_map[diff]
            num_map[n] = i
```
