# <center> 912. Sort an Array </center>


## Problem Description
[Click here](https://leetcode.com/problems/sort-an-array/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Merge sort
1. divide the array into two subarrays
2. sort the subarray
3. merge the sorted arrays


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = size of the array
- check base case
    if n is greater than 1, there are more than 1 elements in the array
    - *divde*
        - find the midpoint 
        - add the elements before the mid index to left subarray and elmenets after the mid index to the right subarray
    - *sort*
        - recursively sort the left and right subarrays
    - *merge*
        - initialize three pointers (i, j, k) for the left, right, and main array
        - traverse both subarrays, find the smallest value, and add it to the main array using pointers
        - check the remaining elements in the left and right subarray and add them to the main array
- else return the array


## Complexity
- Time complexity: O(recursion) → O(total recursion levels * cost at each level) → O(logn * n) → O(nlogn)
    - *recursion levels = logn because in each recursion the array is divided into half*
    - *cost at each level = time for merging = n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack + left array + right array) → O(recursion depth/levels + n1 + n2) → O(logn + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def sortArray(self, nums: List[int]) -> List[int]:
        n = len(nums)
        if n > 1:

            # divide 
            mid = n // 2
            left, right = nums[:mid], nums[mid:]

            # sort
            self.sortArray(left)
            self.sortArray(right)

            # merge
            i, j, k = 0, 0, 0        # pointers for left, right and main array
            while i < len(left) and j < len(right):
                if left[i] < right[j]:
                    nums[k] = left[i]
                    i += 1
                else:
                    nums[k] = right[j]
                    j += 1
                k += 1
            while i < len(left):     # remaining values in left 
                nums[k] = left[i]
                i += 1
                k += 1
            while j < len(right):    # remaining values in right
                nums[k] = right[j]
                j += 1
                k += 1

        return nums
```
