# <center> 217. Contains Duplicate </center>


## Problem Description
[Click here](https://leetcode.com/problems/contains-duplicate/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use a hashset to check for duplicates.


## Approach
<!-- Describe your approach to solving the problem. -->
- set seen = a hashset to check for duplicates
- traverse the input array <br>
for each number in the input
    - if the number is already in the set, it is a duplicate, return true
    - else add the number to the set
- if the loop executes without returning true, there are no duplicates, return false


## Complexity
- Time complexity: O(array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(set) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()
        for n in nums:
            if n in seen:
                return True
            seen.add(n)
        return False
```
