# <center> 2348. Number of Zero-Filled Subarrays </center>


## Problem Description
[Click here](https://leetcode.com/problems/number-of-zero-filled-subarrays/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The number of contiguous 0's represents the number of subarrays.

For example: nums = [1, 3, 0, 0, 2, 0, 0, 0, 4]
- there are 2 contiguous 0's starting at index 2, and 3 contiguous 0's starting at index 5
2 + 3 = 5
    - *the 2 contiguous 0's represent that we can have a subarray of size 1 ([0]) and 2 ([0, 0]) i.e total of 2 subarrays* 
    - *the 3 contiguous 0's represent that we can have a subarray of size 1 ([0]), size 2 ([0, 0]), and size 3 ([0, 0, 0]) i.e total of 3 subarrays*

Traverse the array. Track the number of contiguous zeros and add it to the result.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to count the total 0-filled subarrays 
- set count = 0 to count the number of contiguous 0's 
- traverse the array
    - if the current number is zero, we got another contiguous zero
        - increment count by 1
        - update res by adding the count
    - else the number is not zero, the current 0-filled subarray has ended
        - set count to 0
- return res


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def zeroFilledSubarray(self, nums: List[int]) -> int:
        res, count = 0, 0
        for n in nums:
            if n == 0:
                count += 1
                res += count
            else: 
                count = 0
        return res
```
