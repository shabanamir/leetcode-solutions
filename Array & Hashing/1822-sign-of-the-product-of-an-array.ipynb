## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We only need the sign of the product. So, there's no need to multiply the array values to get the product. 

Use a variable and set it to +1 initially. Traverse the array. Flip the variable sign every time you get a negative number. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 1 to track the product sign
- traverse the numbers
    - if current number is 0
        - return 0 
    - if current number is negative
        - flip the sign by multiplying it with -1
- return res


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def arraySign(self, nums: List[int]) -> int:
        res = 1
        for n in nums:
            if n == 0:
                return 0
            if n < 0:
                res *= -1
        return res
```
