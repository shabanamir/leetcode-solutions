## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
An array is monotonic if it is in ascending or descending order.

Traverse the array to check if it is in ascending or descending order and return the result.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the array size is 1, return true because an array with only one element is monotonic
- set two boolean variables, ascend and descend to true to track if the array is in ascending or descending order 
- traverse the array, *start from the second element*
    - if current element is less than the previous element, the array is not in ascending order
        - set ascend to false
    - if current element is greater than the previous element, the array is not in descending order
        - set descend to false
- check the variables and return the result i.e true if any of them is true else false


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isMonotonic(self, nums: List[int]) -> bool:
        if len(nums) == 1:
            return True
        ascend, descend = True, True
        for i in range(1, len(nums)):
            if nums[i] < nums[i - 1]:
                ascend = False
            if nums[i] > nums[i - 1]:
                descend = False
        return ascend or descend
```
