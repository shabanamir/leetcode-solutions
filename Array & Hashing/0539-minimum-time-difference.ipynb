# <center> 539. Minimum Time Difference </center>


## Problem Description
[Click here](https://leetcode.com/problems/minimum-time-difference/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the minimum difference between array elements.

The difference is minimum between consecutive elements.

Convert the times to minutes, sort the minutes to compare consecutive elements easily, and calculate the difference between the consecutive elements to find the minimum difference. At the end, check for a circular difference, i.e the difference between the last and first time because they are adjacent in a circular manner, e.g 23:59 and 00:00.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the size of input array is more than 1440 (total minutes in 24 hours), there are duplicates
        - return 0 as min difference because difference between duplicate times is 0
- set res = a large number i.e float infinity to track the min difference
- convert the times to minutes
    - set minutes = an array to store the minutes
    - traverse the times array, *for each time string in the array*
        - split the string at ":" to get the hour and minutes and convert them to int
        - find the total minutes and add it to the minutes array
            - *total minutes = hour * 60 + minutes*
- sort the minutes array
- calculate the difference between consecutive elements
    - traverse the minutes array, *start from the second element*
        - find the difference between the current and previous element, take the min of the current difference and the previous min difference, and update res
- check for circular difference
    - take the min of the circular difference and the current min difference, and update res
- return res


## Complexity
- Time complexity: O(time array traversal for conversion to minutes + sorting minutes array + minute array traversal to calculate difference) → O(n + mlogm + m) → O(n + nlogn + n) → O(nlogn)
    - *n = input times array size*
    - *m = minutes array size*
    - *m = n because minutes array size is same as the times array size*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(minutes array + sort) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findMinDifference(self, timePoints: List[str]) -> int:
        if len(timePoints) > 1440:
            return 0
        res = float('inf')
        # convert to minutes
        minutes = []
        for time in timePoints:
            h, m = map(int, time.split(':'))
            minutes.append(h * 60 + m)
        # sort
        minutes.sort()
        # calculate difference 
        for i in range(1, len(minutes)):
            res = min(res, minutes[i] - minutes[i - 1])
        # check circular difference
        res = min(res, (minutes[0] + 1440) - minutes[-1])
        return res
```
