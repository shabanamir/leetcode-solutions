# <center> 242. Valid Anagram </center>


## Problem Description
[Click here](https://leetcode.com/problems/valid-anagram/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Two strings are anagrams if they have the same character and the frequency of characters is also the same.

Use a hashmap to check for a character and its frequency.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the strings are of not equal length, they can't be anagrams, return false
- set count_s = a hashmap to count the characters in string s
    - *key = character*
    - *value = count*
    - *the default dict is used to handle missing keys and avoid the extra check if a key exists*
- traverse string s and update/fill the hashmap
    - for each character in the string s
        - increment the count of the character in the hashmap
- traverse string t and check if all of its characters are present in string s
    - for each character in string t
        - if the character is in the hashmap
            - decrement its count
        - if the count of the character becomes negative, the character is not present in the hashmap
        - return False
- if the loop completes without returning false, the strings are anagrams, return true


## Complexity
- Time complexity: O(string s traversal + string t traversal) → O(m + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(lowercase english letters) → O(26) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        count_s = defaultdict(int)
        for c in s:
            count_s[c] += 1
        for c in t:
            count_s[c] -= 1
            if count_s[c] < 0:
                return False
        return True
```
