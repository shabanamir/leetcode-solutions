# <center> 14. Longest Common Prefix </center>


## Problem Description
[Click here](https://leetcode.com/problems/longest-common-prefix/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
There are two approaches:
1. Sort the strings and find the common prefix between the two most dissimilar strings i.e the first and last string. <br>
Time complexity = O(sorting strings array) → O(n * s * logn)

2. Compare the first string with other strings to find the common prefix. <br>
Time complexity = O(m * n)


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - if there is only one string in the array, return it
  - if there's an empty string in the array, there can't be a common prefix, return an empty string
- traverse the first string <br>
for i in range 0 to the length of the first string
  - traverse the other strings starting from the second string <br>
  for j in range 1 to the length of the strings array
    - if current string is shorter than the first string i.e i is equal to the size of current string or the current character in the first string is not the same as the current character in the current string from the array
      - return the substring of the first string till index i as the longest common prefix
- if the loop completes without returning anything, the first string is the longest common prefix, return it


## Complexity
- Time complexity: O(empty string check + loops to compare strings) → O(n + first string * string array) → O(n + m * n) → O(m * n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(slicing the first string for the LCS) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def longestCommonPrefix(self, strs: List[str]) -> str:
        if len(strs) == 1:
            return strs[0]
        if '' in strs:
            return ''
        for i in range(len(strs[0])):
            for j in range(1, len(strs)):
                if i == len(strs[j]) or strs[0][i] != strs[j][i]:
                    return strs[0][:i]
        return strs[0]
```
