## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
It is the same as [Isomrphic Strings](https://leetcode.com/problems/isomorphic-strings/description/) Problem.

Use two hashmaps to track the 1-1 mapping between the characters in pattern and words in s.


## Approach
<!-- Describe your approach to solving the problem. -->
- split string s to get the words
- check base case
    - if the number of characters in pattern and the number of words in s are not the same, there is no mapping
        - return false
- set char_word = a hashmap to map the characters of pattern to words in s
- set word_char = a hashmap to map the words in s to characters in pattern
- traverse the characters and words in parallel <br>
(either loop in range 0 to the length of pattern because both characters and words are equal in length or use zip())
    - *for each character c and word w in pattern and words*
        - check the mapping i.e if the current char in pattern is not mapped to the current word in s or vice versa
            - return false
        - else map the char to word in char_word hashmap and vice versa
- if the loop completes without returning false, return true


## Complexity
- Time complexity: O(split + traversal) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(words list + hashmap char_word + hashmap word_char) → O(n + 26 + 26) → O(26)
    - *26 because pattern contains lowercase english letters. Hashmap char_word will have at most 26 keys and hashmap word_char will also have 26 because each char will be mapped to one word*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def wordPattern(self, pattern: str, s: str) -> bool:
        words = s.split()
        if len(pattern) != len(words):
            return False
        char_word, word_char = {}, {}
        for c, w in zip(pattern, words):
            if (c in char_word and char_word[c] != w) or (w in word_char and word_char[w] != c):
               return False
            char_word[c] = w
            word_char[w] = c
        return True
```
