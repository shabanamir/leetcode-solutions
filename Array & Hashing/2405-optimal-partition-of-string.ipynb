# <center> 2405. Optimal Partition of String </center>


## Problem Description
[Click here](https://leetcode.com/problems/optimal-partition-of-string/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Traverse the string and make a new partition (substring) whenever you get a duplicate character.

To find duplicates:
1. Use a hash set to store current partition characters and check for duplicates.
2. Use bit manipulation to check for duplicates. Set a bit for every character you visit. If the bit is already set, it means it is a duplicate.


## Approach
<!-- Describe your approach to solving the problem. -->
**Hashset Approach**
- set res = 1 to track the minimum number of partitions
    - *(1 because the string size is at least 1 i.e at least there is one partition)*
- set seen = a hash set to track duplicates i.e to check elements in the current partition
- traverse the string
    - if the current char is in the set i.e current partition, it is a duplicate char, we need a new partition
        - increment res
        - clear the hash set
    - else it is not present in the current partition
        - add it to the current partition i.e hash set
- return res

**Bit Manipulation Approach**
- set res = 1 to track the minimum number of partitions 
- set seen = 0 to check for duplicates using bits 
    - *the string contains lowercase english letters. An integer is represented by at least 32 bits which is more than 26 i.e total lowercase letters. We will set the corresponding bit for the letter. For example, for 'a' bit 0 will be set, for 'b' bit 1, etc*
    - *0 because all bits in 0 are unset*
- traverse the string 
    - find the bit index for the current character and set it using left shift operator
        - *(left shift operator shifts the left operand bits to the left side for the given number of times. in simple terms, it appends 0's at the end of the left operand binary number.
for example, if the character is 'c' 
bit index of c = ascii code of c - ascii code of a = 99 - 97 = 2
1 << 3 = 0001 << 2 = 0100)*
    - if the corresponding bit in seen is already set, the character is a duplicate, we need a new partition
        - *(and operator multiplies corresponding bits, if both are set i.e 1, it returns true else false.
for example if a = 1100 and b = 0100 
a & b will return true because the second bit is set in both)*
        - increment res
        - reset seen to 0
    - else set the corresponding bit in the seen using or operator
        - *(or operator because we want to keep the previous set bits and also set the new bit.
or operator adds the corresponding bits.
for example if a = 1000 and b = 0100
a | b = 1000 | 0100 = 1100)*
- return res


## Complexity
- Time complexity: 
    - Hashset Approach O(string traversal) → O(n)
    - Bit Manipulation Approach O(string traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - Hashset Approach O(hashset) → O(lowercase english letters) → O(26) → O(1)
    - Bit Manipulation Approach O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def partitionString(self, s: str) -> int:

        # hashset appraoch
        res = 1
        seen = set()
        for c in s:
            if c in seen:
                res += 1
                seen.clear()
            seen.add(c)
        return res

        # bit manipulation approach
        res = 1
        seen = 0
        for c in s:
            bit = 1 << (ord(c) - 97)
            if seen & bit: 
                res += 1
                seen = 0
            seen |= bit
        return res
```
