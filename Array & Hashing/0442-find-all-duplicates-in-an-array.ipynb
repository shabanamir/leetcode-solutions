# <center> 442. Find All Duplicates in an Array </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-all-duplicates-in-an-array/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have to find the duplicate numbers in constant space i.e we can't use a hashset.

It is similar to the [First Missing Positive](https://leetcode.com/problems/first-missing-positive/description/) problem.

The numbers in the array are in the range 1 to n and the indices are 0 to n - 1.

Traverse the array and each time mark the index of the current number i.e make the number at that index negative. If you encounter a marked index i.e a negative number, it is the duplicate number.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to store duplicate numbers
- traverse the input list
    - find the index of the current number
    - if the index is already marked i.e the number at that index is negative, the current number is a duplicate
        - add it to the result list *(add its absolute value because we have made it negative for marking)*
    - else mark the index by making the number at index negative
- return the result list


## Complexity
- Time complexity: O(array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result list) → O(n) bit the description says that the result list is not counted → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findDuplicates(self, nums: List[int]) -> List[int]:
        res = []
        for n in nums:
            i = abs(n) - 1
            if nums[i] < 0:
                res.append(abs(n))
            nums[i] *= - 1
        return res
```
