## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
If x + y is greater than y + x, we need to put x first else y.

*(+ means concatenation here)*

For example:
- x = 2, y = 10
x + y = 210, y + x = 102
x + y > y + x
sorted order = x, y
- x = 10, y = 2
x + y = 102, y + x = 210
x + y < y + x
sorted order = y, x

Sort the numbers in the array by comparing the string representation of the numbers and return the string form of the array.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - if all numbers are 0
    - return 0 because the largest number w can form is 0
- convert array elements to string
- define a custom function to compare the string representation of two numbers x and y
    - comparison function returns: 
        - *-1 if the first argument is less*
        - *1 if the first argument is greater*
        - *0 if both are equal*
    - compare(x, y) 
        - if x + y is greater than y + x
            - return -1 because first argument i.e x is less
        - else return 1 because the first argument is greater
- convert the custom comparison function to key function using cmp_to_key() and sort the numbers using the key
- return the string form of the sorted array


## Complexity
- Time complexity: O(check for 0's + conversion to string + sort + join) → O(n + n + nlogn + n) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def largestNumber(self, nums: List[int]) -> str:
        if not any(nums):
            return '0'
        for i in range(len(nums)):
            nums[i] = str(nums[i])
        
        def compare(x: int, y: int) -> int:
            if x + y > y + x:
                return -1
            return 1
        
        nums.sort(key=cmp_to_key(compare))
        return ''.join(nums)
```
