# <center> 392. Is Subsequence </center>


## Problem Description
[Click here](https://leetcode.com/problems/is-subsequence/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use two-pointers approach to track the character and its order in both strings.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if string s is empty, return true because an empty string is always a subseq
    - if string s length is greater than string t, s can't be a subseq of t, return false
- initialize two pointers: i for string s and j for string t
- traverse both strings in parallel, *loop until one of the strings reaches the end*
    - if there's a matching character, increment both pointers      
    - else only increment j
- check if pointer i value and string s length are equal and return the result *(if they are equal string s is a subsequence and vice versa)*


## Complexity
- Time complexity: O(loop) → O(min(s, t)) → O(min(m, n)) → O(n) if m = n
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def isSubsequence(self, s: str, t: str) -> bool:
        if not s:
            return True
        if len(s) > len(t):
            return False
        i, j = 0, 0 
        while i < len(s) and j < len(t):
            if s[i] == t[j]:
                i += 1
                j += 1
            else:
                j += 1
        return i == len(s)
```
