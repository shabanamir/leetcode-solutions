Use a linked list to implement the hashmap.


## Approach
<!-- Describe your approach to solving the problem. -->
*create a separate class for the linked list nodes*

**List Node Class**

**init()**
set the node key, value, and next pointer

**Hash Set Class**

**init()**
- set hashmap size = 10000
*10000 because according to description at most 10$^4$ calls will be made to add() i.e we can have 10$^4$ unique values*
- initialize an array of size 10000 with linked list nodes to represent the hash map

**hash()**

- take the mod of key to get the index and return the result 

**put()**
- find the key index and set current node pointer cur = the first node at the index
- if some nodes are already there due to collision, loop until the end of the list of nodes
    - if key node is already added, update its value and return
- else create a new node for the key and add its value

**get()**
- find index and set cur = the first node at the index
- loop until you find the key
- if the key is found, return its value
- else return -1

**remove()**
- find index and set cur = the first node at the index
- loop until you find the node before the key node
- if you found the previous node
    - remove the key node by updating the next pointer of the previous node


## Complexity
- Time complexity: 
    - init() O(array initialization) → O(10000) → O(1)
    - hash() O(1)
    - put(), get(), and remove() time is O(n) in case of collision otherwise O(1). In the worst case, we can have a linked list of size n if all elements result in the same hash value.
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - init() O(array + linked list) → O(10000 + n) → O(1 + n) → O(n)
    - hash() O(1)
    - put() O(1)
    - get() O(1)
    - remove() (1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class ListNode:
    
    def __init__(self, key= -1, value= -1, next= None):
        self.key = key
        self.val = value
        self.next = next
    

class MyHashMap:
    
    def __init__(self):
        self.size = 10000
        self.map = [ListNode() for _ in range(self.size)]

    def hash(self, key: int) -> int:
        return key % self.size

    def put(self, key: int, value: int) -> None:
        cur = self.map[self.hash(key)]
        while cur.next:
            if cur.next.key == key:
                cur.next.val = value
                return
            cur = cur.next
        cur.next = ListNode(key, value)

    def get(self, key: int) -> int:
        cur = self.map[self.hash(key)]
        while cur and cur.key != key:
            cur = cur.next
        if cur:
            return cur.val
        return -1
        
    def remove(self, key: int) -> None:
        cur = self.map[self.hash(key)]
        while cur.next and cur.next.key != key:
            cur = cur.next
        if cur and cur.next:
            cur.next = cur.next.next
        


# Your MyHashMap object will be instantiated and called as such:
# obj = MyHashMap()
# obj.put(key,value)
# param_2 = obj.get(key)
# obj.remove(key)
```
