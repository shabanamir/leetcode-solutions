## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The array can become non-decreasing if there is no or a single pair of adjacent elements that are not in the sorted order i.e at most a single change in order is required.

Traverse the array. Compare the adjacent elements to track the total number of changes. If more than one change is required, return false.


## Approach
<!-- Describe your approach to solving the problem. -->
- if there are 2 or less than 2 elements, at most one change is required, return true
- set flag = false to track the number of changes
- traverse the array, *start from index 1 to keep the index inbound because we are comparing adjacent elements (current and previous)*
    - if the current element is less than the previous, we need a change in order
        - if flag is true, we have already made a change
            - return false
        - set flag to true because we are making a change
        - if the current element (nums[i]) is less than the second previous element (nums[i-2]), make the current element same as the previous element (i-1) to make the change and fix the violation in order because the previous element is guaranteed to be equal or greater than the second previous
- if the loop completes without returning false, we have made at most 1 change, return true


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def checkPossibility(self, nums: List[int]) -> bool:
        if len(nums) <= 2:
            return True
        flag = False
        for i in range(1, len(nums)):
            if nums[i] < nums[i - 1]:
                if flag:
                    return False
                flag = True
                if i > 1 and nums[i] < nums[i - 2]:
                    nums[i] = nums[i - 1]
        return True
```
