# <center> 205. Isomorphic Strings </center>


## Problem Description
[Click here](https://leetcode.com/problems/isomorphic-strings/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use two hashmaps to track the 1-1 mapping between the characters in strings s and t.


## Approach
<!-- Describe your approach to solving the problem. -->
- set st = a hashmap to map the characters of string s to string t
- set ts = a hashmap to map the characters of string t to string s
- traverse the characters in s and t parallel <br>
(either loop in range 0 to the length of string s because both strings are equal in length or use zip())
    - check the mapping i.e if s character is not mapped to t char or t char is not mapped to s char, return false
    - else map the s char to t char and vice versa
- if the loop executes without returning false, the strings are isomorphic, return true


## Complexity
- Time complexity: O(string traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap st + hashmap ts) → O(128 + 128) → O(1)
    - *128 because s and t consist of valid ascii characters*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isIsomorphic(self, s: str, t: str) -> bool:
        st, ts = {}, {}
        for c1, c2 in zip(s, t):
            if (c1 in st and st[c1] != c2) or (c2 in ts and ts[c2] != c1):
                return False
            st[c1] = c2
            ts[c2] = c1
        return True
```
