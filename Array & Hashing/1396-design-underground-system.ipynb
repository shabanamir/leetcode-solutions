## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use two hashmaps to store check-in and route times information.


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- create a hashmap for check-in info
    - *key = passenger id*
    - *value = (start station, start time)*
- create a hashmap for route info with default value [0, 0] for a key. 
    - *key = (start_station, end_station)*
    - *value = (number of trips, total duration)*

**checkIn()**
- add the check-in info by mapping the id to (start station), start time)

**checkOut()**
- end station and end time is given, 
get the start station and start time by searching the id in the check-in map and remove the check-in entry
- find the route i.e (start station, end station)
-  add/update the number of trips and duration for the route (start station, end station) in the routes map
    - *index 0 represents the number of trips, increment it by 1*
    - *index 1 represents the duration, add the current duration i.e end time - start time*

**getAverageTime()**
- search for the route (start station, end station) in the routes map to get trips and duration info
- divide duration by the number of trips to get the average time and return it


## Complexity
- Time complexity: 
    - init() O(1)
    - checkIn() O(1)
    - checkOut() O(1)
    - getAverageTime O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - init() O(1)
    - checkIn() O(1)
    - checkOut() O(1)
    - getAverageTime O(1)
    - the hashmaps size will be O(n) or more specifically (passengers + stationsÂ²) after n calls to checkIn() and checkOut()
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class UndergroundSystem:

    def __init__(self):
        self.check_in = {}
        self.routes = defaultdict(lambda: [0, 0])

    def checkIn(self, id: int, stationName: str, t: int) -> None:
        self.check_in[id] = (stationName, t)
        
    def checkOut(self, id: int, stationName: str, t: int) -> None:
        start_station, start_time = self.check_in.pop(id)
        route = (start_station, stationName)
        self.routes[route][0] += 1
        self.routes[route][1] += t - start_time
        
    def getAverageTime(self, startStation: str, endStation: str) -> float:
        n_trips, duration = self.routes[(startStation, endStation)]
        return duration / n_trips
        

# Your UndergroundSystem object will be instantiated and called as such:
# obj = UndergroundSystem()
# obj.checkIn(id,stationName,t)
# obj.checkOut(id,stationName,t)
# param_3 = obj.getAverageTime(startStation,endStation)
```
