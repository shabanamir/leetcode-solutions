# <center> 1963. Minimum Number of Swaps to Make the String Balanced </center>


## Problem Description
[Click here](https://leetcode.com/problems/minimum-number-of-swaps-to-make-the-string-balanced/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Traverse the string and track the opening and closing brackets. We need a swap if the closing brackets are more than the opening.


## Approach 
<!-- Describe your approach to solving the problem. -->
**Approach 1**

- set start = 0 to track the starting/opening bracket and end = 0 to track the closing brackets
- set res = 0 to track the min swaps required to make the string balanced
- traverse the string
    - if the current char is an opening bracket, increment start
    - else the char is a closing bracket
        - increment end
        - if there are more closing brackets than opening, we need a        swap
            - increment res, increment start, and decrement end
- return res

**Approach 2**

- set close = 0 to track closing brackets
- set max_close = 0 to track extra closing brackets that need a swap
- traverse the string
    - decrement close if char is an opening bracket
    - increment close if char is a closing bracket
    - update max_close. we need max_close to store the count of extra closing brackets because close will become zero at the end of the loop
- divide max_close by 2 and return the value *(division by 2 is required because each swap takes care of 2 extra closing brackets)*


## Complexity
- Time complexity: 
    - Approach 1 O(string traversal) → O(n)
    - Approach 2 O(string traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - Approach 1 O(1)
    - Approach 2 O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minSwaps(self, s: str) -> int:
        
        # approach 1
        start, end, res = 0, 0, 0
        for c in s:
            if c == '[':
                start += 1
            else:
                end += 1
                if end > start:
                    res += 1
                    start += 1
                    end -= 1
        return res

        # approach 2
        close, max_close = 0, 0
        for c in s:
            if c == '[':
                close -= 1
            else:
                close += 1
            max_close = max(close, max_close)
        return (max_close + 1) // 2
```
