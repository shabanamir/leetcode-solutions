# <center> 523. Continuous Subarray Sum </center>


## Problem Description
[Click here](https://leetcode.com/problems/continuous-subarray-sum/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A good subarray:
- has at least 2 elements
- its sum is multiple of k i.e divisible by k 

An array has a good subarray if there are two subarrays with the same remainder, and the subarray between these two contains at least two elements because such subarray sum is divisible by k.

Examples 
1. arr = [2, 2, 1, 2, 3], k = 6
prefix sum = [0+2, 2+2, 2+2+1, 2+2+1+2, 2+2+1+2+3] = [2, 4, 5, 7, 10]
prefix sum % k = [2/6, 4/6, 5/6, 7/6, 10/6] = [2, **4**, 5, 1, **4**]
The input array has a good subarray because:
    - there are two subarrays with the same remainder (4), one ending at index 1 and the other at index 4
    - the subarray between these two indices *(first index is exclusive, second index is inclusive)* contains 3 elements i.e [1, 2, 3]
    - *(cross-check: subarray sum is divisible by k i.e 1+2+3/6 = 6/6 = 1)*
2. arr = [23,2,4,6,6], k = 7
prefix sum = [23, 25, 29, 35, 41]
prefix sum % k = [2, 4, 1, **0**, 6]
If we see there are no duplicate remainders, but we have 0. Every subarray has an empty subarray whose sum is 0 and 0 % any number is 0. 
The input array has a good subarray because:
    - there are two subarrays with the same remainder (0). One is the empty subarray ending before index 0, let's call this index -1, and the other ending at index 3
    - the subarray between these two indices contains 4 elements i.e [23, 2, 4, 6]
    - *(cross-check: subarray sum is divisible by k i.e 23+2+4+6 = 35/7 = 1)*
3. arr = [23,2,6,4,7], k = 13
prefix sum = [23, 25, 31, 35, 42]
prefix sum % k = [10, 12, 5, 9, 3]
The input array doesn't have a good subarray because there are no duplicate remainders


Use prefix sum to get the subarrays and hashmap to store the subarray reminders and ending indices for fast searching.


## Approach
<!-- Describe your approach to solving the problem. -->
- set remainders = a hashmap to map the remainders and their indices
    - *key = subarray remainder*
    - *value = subarray ending index*
    - *we need indices to check the number of elements in the subarray*
- add empty subarray remainder 0 with index -1 to the hashmap because every array has an empty subarray whose mod is 0
- set pre_sum = 0 to track the prefix sum
- traverse the array
    - add the current number to prefix sum
    - find the remainder of the current subarray (prefix sum)
    - if the remainder is in the hashmap, there is another subarray with same remainder
      - if there are at least two elements i.e difference between current index and subarray ending index >= 2
        - return true because there is a good subarray
      - else continue
    - else the remainder is not in the hashmap, add it with its index
- if the loop completes without returning true, there's no good subarray, return false


## Complexity
- Time complexity: O(array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashmap) → O(subarray remainders) → O(n)
    - *in the worst case, we will add a remainder to the hashmap in each iteration and total iterations are n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def checkSubarraySum(self, nums: List[int], k: int) -> bool:
        remainders = {0: -1}
        pre_sum = 0
        for i, n in enumerate(nums):
            pre_sum += n
            mod = pre_sum % k
            if mod in remainders:
                if i - remainders[mod] >= 2:
                    return True
                continue
            remainders[mod] = i
        return False
```
