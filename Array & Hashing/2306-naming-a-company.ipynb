## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A name is valid if the two words in the name don't exist in the input words array after swapping their first characters. 

For example:
array = ['coffee', 'donuts', 'time', 'toffee']
- name 'coffee donuts' is valid because if we swap the first characters, the name will be 'doffee conuts', and both 'doffee' and 'conuts' don't exist in the array.
- name 'coffee toffee' is invalid because after swapping the first characters, the name is 'toffee coffee', and both 'toffee' and 'coffee' exist in the array.
- name 'time toffee' is invalid because after swapping the first characters, the name is 'time toffee ', and both 'time' and 'toffee' exist in the array.

So, a name is invalid if the words have:
- common prefix 
    - *for example 'time toffee' is invalid because both have a common first letter i.e 't'*
- common suffix  
    - *for example 'coffee toffee' is invalid because both have a common suffix i.e 'offee'*

Group the words based on the prefix (first letter) using a hashmap. Add the prefix as the key and the set of suffixes as the key value. Compare each group (key) with all others and find the number of distinct valid names.

Formula for finding no. of distinct valid names 
= total distinct suffixes in group1 * total distinct suffixes in group2 
= (total suffixes in group1 - common suffixes) * (total suffixes in group2 - common suffixes) 
= (total suffixes in group1 - intersection of group1 and group2) * (total suffixes in group2 - intersection of group1 and group2)


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the total number of distinct valid names
- create a hashmap to group the words
    - *key = prefix i.e first letter*
    - *value = set of suffixes*
- for each word in the input array
    - add the suffix (all word letters the first letter) to the set associated with the prefix (first letter)
- for each group (key) i in the hashmap
    - for each group (key) j in the hashmap
        - if it is the same group, skip
        - else compare the groups
            - find the no. of common suffixes i.e take intersection of the sets associated with each group and get the result set size
                - *(to find intersection of two sets, either use & operator or traverse the set elements)*
            - find total valid names for groups i and j using the formula and add it to res
- return res


## Complexity
- Time complexity: O(filling hashmap + finding valid names) → O(array traversal + hashmap traversal * hashmap traversal * set intersection ) → O(n + 26 * 26 * m) → O(n + m) → O(n)
    - *hashmap keys = lowercase english letters = 26*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashmap + suffix hashset + intersection result set) → O(26 + n + m) → O(n)
    - *if all words in the input array have the same prefix, there will be n suffixes* 
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def distinctNames(self, ideas: List[str]) -> int:
        res = 0
        groups = defaultdict(set)
        for word in ideas:
            groups[word[0]].add(word[1:])
        for i in groups:
            for j in groups:
                if i == j:
                    continue
                common = len(groups[i] & groups[j])
                res += (len(groups[i]) - common) * (len(groups[j]) - common)
        return res
```
