## Approach
<!-- Describe your approach to solving the problem. -->
**concatenate**

- concatenate by using the + operator and return the result

**repeat**

- use array * k to create a kn array
*k is the number of times you want to repeat the array, here k is 2* 

**extend()**

- use extend() and add the input array at the end of the itself
- return the input array

**append()**

- traverse the input array
    - add the current number in the input array at the end of the array
- return the input array


## Complexity
- Time complexity: 
    - concatenate O(n)
    - repeat O(2n) → O(n)
    - extend() O(n)
    - append() O(n)

## Code

```python
class Solution:
    
    def getConcatenation(self, nums: List[int]) -> List[int]:
        
        # return nums + nums

        # return nums * 2

        # nums.extend(nums)
        # return nums

        for i in range(len(nums)):
            nums.append(nums[i])
        return nums
```
