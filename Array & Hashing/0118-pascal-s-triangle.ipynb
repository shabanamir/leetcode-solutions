## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Each number (arr[i][j]) in Pascal's triangle (2d-array) is the sum of the two above numbers i.e previous row previous column element (arr[i-1][j-1]) and previous row same column element (arr[i-1][j]).

arr[i][j] = arr[i-1][j-1] + arr[i-1][j]


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a 2d array with default value 1 to represent the Pascal triangle
    - *rows in range 1 to total rows, 
    (1 because starting with 0 would result in an empty first row)*
    - *cols in range 1 to row number, 
    (row 1 has 1 column, row 2 has 3 columns, row3 has 3 columns etc)*
- loop to update the 2d array
    - for each row *(exclude the first two rows because these contain 1 in Pascal's triangle)*
        - for each column *(exclude the first and last columns because these contain 1 in Pascal's triangle)*
            - update the current cell value using the formula
- return the 2d array i.e res


## Complexity
- Time complexity: O(array creation + array update) → O(row * columns + row * columns) → O(n² + n²) → O(n²)
    - *n² becuause in pascal's traingle the total coulmns in a row = row number for example row1 has 1 column, row2 has 2 columns, and so on. For outer loop iteration 1 the inner loop will run 1 time, for outer loop iteration 2 the inner loop will run two times, and so on.
total iterations = 1 + 2 + 3 + 4 + n = (n² + n)/2 = n²*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(result list) → O(total rows * max column) → O(n * m) → O(n * n) → O(n²) 
    - *m = n because the last row has max columns and we know that total columns in a row = row number for example row1 has 1 column, row2 has 2 columns, and so on*
