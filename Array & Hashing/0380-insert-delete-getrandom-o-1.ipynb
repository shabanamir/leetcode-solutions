## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Besides insertion, deletion, and random, we also need to search in O(1) time.

Insertion, deletion, and search take O(1) time in set but we can't use a set to find a random element. 

Use an array.

- insertion in array takes O(1)
- to delete in O(1), swap the value to be deleted with the last value in the array and then remove the last value in O(1). For swapping, we need the value index. Use hashmap to map the indices and search in O(1)
- for random value, use randint() or choice()


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- create an array to store values
- create a hashmap to map items/values to indices
    - *key = item*
    - *value = index*

**insert()**
- if the value is already inserted/present, return false
- get the index of the value to be inserted and add it to the hashmap
    - *current value index = previous value index + 1 = total values in the list i.e size of the list*
- add value to the array
- return true because insertion was successful

**remove()**
- return false if the current value (value to be deleted) is not in the hashmap
- get the current value index using hashmap and last value in the array
- replace current value with the last value to delete the current value
- update the index of the last value as current index 
- remove the last value from the array because it is now a duplicate
- remove current value from the hashmap
- return true

**getRandom()**
- get a random index in the range 0 to size of the array and return the value at that index
    - *either use randint() or choice()*


## Complexity
- Time complexity: 
    - init() O(1) 
    - insert() O(1)
    - remove() O(1)
    - getRandom() O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - init() O(1) 
    - insert() O(1)
    - remove() O(1)
    - getRandom() O(1)
    - *the list and hashmap size will be n after n calls to insert()*

## Code

```python
class RandomizedSet:

    def __init__(self):
        self.vals = []
        self.index = {}

    def insert(self, val: int) -> bool:
        if val in self.index:
            return False
        self.index[val] = len(self.vals)
        self.vals.append(val)
        return True

    def remove(self, val: int) -> bool:
        if val not in self.index:
            return False
        cur_index, last_value = self.index[val], self.vals[-1]
        self.vals[cur_index] = last_value
        self.index[last_value] = cur_index
        self.vals.pop()
        del self.index[val]
        return True
        
    def getRandom(self) -> int:
        # return self.vals[randint(0, len(self.vals) - 1)]
        return choice(self.vals)
        


# Your RandomizedSet object will be instantiated and called as such:
# obj = RandomizedSet()
# param_1 = obj.insert(val)
# param_2 = obj.remove(val)
# param_3 = obj.getRandom()
```
