# <center> 605. Can Place Flowers </center>


## Problem Description
[Click here](https://leetcode.com/problems/can-place-flowers/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We can plant a flower if:
- if current plot is empty (value is 0)
- previous plot is empty or current plot is the first plot
- next plot is empty or current plot is the last plot


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if n is 0, we don't need to plant any flowers, return true
- traverse the array, *for each plot in the range 0 to n*
    - if all three conditions satisfy
        - plant a flower by updating the plot value to 1 
        - decrement n because we have planted a flower
    - if n <= 0 we have planted all the flowers, return true
- if the loop completes without returning true, we can't plant all flowers, return false


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def canPlaceFlowers(self, flowerbed: List[int], n: int) -> bool:
        if not n:
            return True
        for i in range(len(flowerbed)):
            if flowerbed[i] == 0 and (i == 0 or flowerbed[i - 1] == 0) and (i == len(flowerbed) - 1 or flowerbed[i + 1] == 0):
                flowerbed[i] = 1
                n -= 1
            if n <= 0:
                return True
        return False
```
