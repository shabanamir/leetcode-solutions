# <center> 1189. Maximum Number of Balloons </center>


## Problem Description
[Click here](https://leetcode.com/problems/maximum-number-of-balloons/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
target word instances = min (available/required ratio for each target word character)

Example:
- for 1 instance of the target word i.e ballon, we need:
b = 1, a = 1, l = 2, o = 2 and n = 1
- if we have:
b = 5, a = 5, l = 4, o = 15, n = 3
- b ratio = available b's / required b's = 5/1 = 5
a ratio = 5/1 = 5
l ratio = 4/2 = 2
o ratio = 15/2 = 7
n ratio = 3/1 = 3
- min(all ratios) = min(5, 5, 2, 7, 3) = 2
we can make 2 instances of ballon


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there are less than 7 characters (total characters in the target word i.e 'balloon'), return 0
- set res = a large number to track the maximum instances of the target word we can make
    - *here we have set it to the size of the input text because we can't create more words than that*
- set need = a hashmap and count the freq of the required characters in the target word
- set have = a hashmap to store the freq of the available instances of the required characters
- traverse the input text and fill the *have* hashmap
    - for each character in the input string
        - if the char is in need hashmap, it is a required character
            - increment its count in the hashmap
- traverse the *need* hashmap
    - for each required character
        - find the ratio of available and required instances of the character and update res 
- return res


## Complexity
- Time complexity: O(finding target word freq + input string traversal + need hashmap traversal) → O(5 + n + 5) → O(n)
    - *5 because only the unique characters of the target word are stored in the hashmap and unique characters in the target word 'balloon' are 5*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(need hashmap + have hashmap) → O(5 + 5) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxNumberOfBalloons(self, text: str) -> int:
        if len(text) < 7:
            return 0
        res = len(text)
        need = Counter('balloon')
        have = defaultdict(int)
        for c in text:
            if c in need:
                have[c] += 1
        for c in need:
            res = min(res, have[c] // need[c])
        return res
```
