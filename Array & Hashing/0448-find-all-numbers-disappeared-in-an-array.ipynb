## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The numbers in the array are in the range 1 to n and the indices are 0 to n - 1. 

Mark the indices of the present numbers. The non-marked indices will represent the missing numbers. Add the non-marked indices to the result. 


## Approach
<!-- Describe your approach to solving the problem. -->
- traverse the array
    - find the respective index of the number by subtracting 1 from it because numbers range is 1 to n and indices range is 0 to n-1
    - if the number at that index is not negative i.e greater than 0, the index is not marked
        - mark the index i.e make the value negative at the index to indicate the number is present
    - else no need to do anything because the index is already marked means the current number is a duplicate
- set res =  a list to store the missing numbers
- traverse the array again
    - if the value is positive, the index is not marked and it represents the missing number
        - add the index + 1 to res
- return res


## Complexity
- Time complexity: O(marking indices + checking indices) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(result list) → O(missing numbers) → O(n)
    - *the follow-up says that the output list is not considered extra memory, so space complexity is O(1)*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        for n in nums:
            i = abs(n) - 1
            if nums[i] > 0:
                nums[i] *= -1
        res = []
        for i, n in enumerate(nums):
            if n > 0:
                res.append(i + 1)
        return res
```
