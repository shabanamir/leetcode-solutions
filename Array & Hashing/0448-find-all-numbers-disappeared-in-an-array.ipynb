# <center> 448. Find All Numbers Disappeared in an Array </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-all-numbers-disappeared-in-an-array/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have to find the missing numbers in the range 1 to n in constant space i.e we can't use a hashset, *according to the follow-up question*.

It is similar to the [First Missing Positive](https://leetcode.com/problems/first-missing-positive/description/) problem.

The numbers in the array are in the range 1 to n and the indices are 0 to n - 1. 

Traverse the array and mark the indices of the present number i.e make the number at that index negative. Traverse the array again. If you encounter a non-marked index i.e a positive number, it is the missing number.


## Approach
<!-- Describe your approach to solving the problem. -->
- traverse the array
    - find the index of the current number e.g index of 5 is 4
    - if the index is not marked i.e number at that index is positive
        - mark the index by making the number at the index negative
    - else no need to do anything because the index is already marked means the current number is a duplicate
- set res = a list to store the missing numbers
- traverse the array again
    - if the index is not marked i.e number at that index is positive, it is the missing number
        - add the current index + 1 as missing number to the result list
- return the result list


## Complexity
- Time complexity: O(traversal to mark indices + traversal to check indices) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result list) → O(missing numbers) → O(n) but the follow-up says that the result list is not considered extra memory → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findDisappearedNumbers(self, nums: List[int]) -> List[int]:
        for n in nums:
            i = abs(n) - 1
            if nums[i] > 0:
                nums[i] *= -1
        res = []
        for i, n in enumerate(nums):
            if n > 0:
                res.append(i + 1)
        return res
```
