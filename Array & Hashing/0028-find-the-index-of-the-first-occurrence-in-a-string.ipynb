# <center> 28. Find the Index of the First Occurrence in a String </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-the-index-of-the-first-occurrence-in-a-string/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Traverse the haystack. Each time take a substring of length equal to the needle and compare. 


## Approach
<!-- Describe your approach to solving the problem. -->
- check the base cases
  - if the length of the needle is greater than the haystack, return -1 
  - if both are of equal length, compare them 
    - if they are same, return 0 as the start index
    - else return -1
- set m = size of needle
- set n = size of haystack
- traverse the haystack <br>
loop in range 0 to n - m + 1 to keep index inbound
    - if substring of size m starting with the current index in the haystack is the same as the needle
      - return current index as the start index
- if the loop completes without returning anything, there's no needle in the haystack, return -1


## Complexity
- Time complexity: O(traversal * (string slicing to create substring + substring comparison)) → O((n - m) * (m + m)) → O(n * m)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(substring) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def strStr(self, haystack: str, needle: str) -> int:
        if len(needle) > len(haystack):
            return -1
        if len(needle) == len(haystack):
            if needle == haystack:
                return 0
            return -1
        m, n = len(needle), len(haystack)
        for i in range(n - m + 1):
            if haystack[i: i + m] == needle:
                return i
        return -1
```
