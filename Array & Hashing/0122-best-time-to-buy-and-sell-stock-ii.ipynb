## Intuition
<!-- Describe your first thoughts on how to solve this problem. --> 
It is the same as [Best Time to Buy and Sell I](https://leetcode.com/problems/best-time-to-buy-and-sell-stock/description/), but here we are allowed to buy and sell on the same day and can do multiple transactions i.e we can buy and sell multiple stocks and need to return the total profit we can make.

*In buy and sell stock I, we can't buy and sell on the same day and can only do one transaction i.e we can buy and sell a single stock and we have to find which day to buy and sell to have the max profit.*

Check prices every day. If current day price is greater than the previous day, it is the best day to sell, sell the stock and get a profit. Else, it is the best day to buy.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the total profit
- traverse the prices for each day (start from the second day)
    - if the current day price is greater than the previous day
        - get the profit and update res
- return res as the total profit


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxProfit(self, prices: List[int]) -> int:
        res = 0
        for i in range(1, len(prices)):
            if prices[i] > prices[i-1]:
                res += prices[i] - prices[i-1]
        return res
```
