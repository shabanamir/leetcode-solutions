# <center> 1930. Unique Length-3 Palindromic Subsequences </center>


## Problem Description
[Click here](https://leetcode.com/problems/unique-length-3-palindromic-subsequences/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a length-3 palindrome:
- the first and last characters are the same
- the middle character can be any character


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a hash set to store all unique lenth-3 palindromes
- set left = a hash set to track the characters on the left
- set right = a hashmap to track the characters on the right
- fill the hashmap with the frequency of each character 
    - *key = character*
    - *value = frequency*
    - *either loop to fill the hashmap or use Counter()*
- traverse the characters in the input string
    - *check the possible palindromic subseq with the current character as the middle character of the subseq*
        - remove the current character from the right hashmap by decrementing its count because we want to check the left and right characters for it
        - traverse the left characters
            - if a char same as the left char is on the right side i.e the left character in the right characters hashmap and its count is greater than 0, we found a palindromic subseq
                - add the left character and current character as a pair to the res to represent a palindromic subseq
    - we will move to the next character, the current char will become leftside character, so add it to the left hashset
- return the length of res as the total number of unique length-3 palindromes
 
 
## Complexity
- Time complexity: O(string traversal * hash set traversal) → O(n * 26) → O(n)
  - *26 because input contains lowercase english letters, so at most there can be 26 keys in the set*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(res hashset + left hashset + right hashmap) → O(total possible length-3 palindromes + 26 + 26) → O(n + 26 + 26) → O(n) 
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def countPalindromicSubsequence(self, s: str) -> int:
        res = set()
        left = set()
        right = Counter(s)
        for c in s:
            right[c] -= 1
            for lc in left:
                if lc in right and right[lc] > 0:
                    res.add((lc, c))
            left.add(c)
        return len(res)
```
