# <center> 838. Push Dominoes </center>


## Problem Description
[Click here](https://leetcode.com/problems/push-dominoes/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
- if the dominoes (left and right) are pushed in the same direction
    - if they are pushed to the right side, all dominoes in the middle will fall towards the right 
    'R......R' → 'RRRRRRRR'
    - if they are pushed to the left side, all dominoes in the middle will fall towards the left
'L......L' → 'LLLLLLLL'
- if the dominoes are pushed in the opposite direction 
    - if the left domino is pushed to the left and the right domino is pushed to the right, the middle dominoes will not fall because the left domino will fall towards the left and the right domino will fall towards the right
'L......R' → 'L......R'
    - if the left domino is pushed to the right and the right domino is pushed to the left, the left half of the dominoes will fall towards the right, and the right half will fall towards the left. The middle domino will not fall as it will be balanced by the opposite push
    'R......L' → 'RRRRLLLL' or 'RRRR.LLLL'

Use a two-pointer approach and check which of the above conditions satisfies.


## Approach
<!-- Describe your approach to solving the problem. -->
- add an extra L and R at the beginning and end to prevent crossing the boundary
- set res = an empty list to store the result
- set left pointer l = 0 to track the pushed left domino
- set right pointer r = 1 to track the pushed right domino
- traverse the dominoes  using the right pointer
    - skip if the current (right) domino is not pushed
    - else 
        - find the number of dominoes in the middle 
        - if the left pointer is not 0, the left domino is pushed
            - add the push direction ('L' or 'R') to the result 
        - if the dominoes are pushed in the same direction (cases 1 and 2), the middle dominoes will fall toward the push direction
            - get the result of this push by multiplying the push direction with the number of middle dominoes and add it to res
        - *else dominoes are pushed in opposite directions*
        - if the left domino is pushed to the left and the right domino is pushed to the right (case 4), the middle dominoes will not fall      
            - multiply '.' with the number of middle dominoes and add it to res
        - else the left domino is pushed to the right and the right domino is pushed to the left (case 3), the left half dominoes will fall towards the right, the right half dominoes towards the left, and the middle domino will not fall
            - multiply 'R' with left half dominoes (n_middle_dominoes / 2), '.' with middle domino (n_middle_dominoes % 2), and 'L' with right half dominoes (n_middle_dominoes / 2) to get the result string and add it to res
        - update left pointer
- return string form of the result list


## Complexity
- Time complexity: O(dominoes traversal * substring (push result) creation + join()) → O(n * k + n) → O(n * k) → O(n) because k is very less
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result list) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def pushDominoes(self, dominoes: str) -> str:
        dominoes = 'L' + dominoes + 'R'
        res = []
        l = 0
        for r in range(1, len(dominoes)):
            if dominoes[r] == '.': 
                continue
            middle = r - l - 1
            if l: 
                res.append(dominoes[l])
            if dominoes[l] == dominoes[r]: 
                res.append(dominoes[l] * middle)
            elif dominoes[l] == 'L' and dominoes[r] == 'R':
                res.append('.' * middle)
            else: 
                res.append('R' * (middle // 2) + '.' * (middle % 2) + 'L' * (middle // 2))
            l = r
        return ''.join(res)
```
