# <center> 929. Unique Email Addresses </center>


## Problem Description
[Click here](https://leetcode.com/problems/unique-email-addresses/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Get the valid email by applying the rules and add it to a hashset to avoid duplicates.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a hash set to store (local name, domain name) of unique emails
- traverse the email array
    - get the local name and the domain name by splitting the email at the '@' symbol
    - split the local name at the '+' symbol and get the first part (at index 0) of the result to get characters in the local name part before the '+' symbol 
    - remove the '.' symbols from the local name i.e replace '.' with an empty character ''
    - add the local name and domain name as a pair to the set to represent an email
- return hashset size as the total number of unique emails


## Complexity
- Time complexity: O(array traversal * (split + replace)) → O(n * (m + m)) → O(n * m)
    - *m = size of email string*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashset) → O(unique emails) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def numUniqueEmails(self, emails: List[str]) -> int:
        res = set()
        for e in emails:
            local, domain = e.split('@')
            local = local.split('+')[0].replace('.', '')
            res.add((local, domain))
        return len(res)
```
