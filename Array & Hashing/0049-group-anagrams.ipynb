# <center> 49. Group Anagrams </center>


## Problem Description
[Click here](https://leetcode.com/problems/group-anagrams/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Anagrams have same number of characters with same frequency but different arrangements.

Use a hashmap and group same key anagrams together. The key can be:
1. The sorted string O(n * klogk) <br>
anagrams have same sorted sequence. <br>
For example, the sorted string for eat and ate is aet
2. The frequency string O(n * k) <br>
anagrams have same frequency string. <br>
For example, the frequency string for eat and ate is a1e1t1


## Approach
<!-- Describe your approach to solving the problem. -->

#### Sorted String Approach
- set res = a hashmap to group anagrams
    - *key = frequency string*
    - *value = list of strings with same sorted string i.e anagrams*
    - *the default dict is to handle missing keys and avoid the extra check if a key exists*
- traverse the input list of strings, *for each string s in the array of strings*
    - find the sorted form of the string, convert it to a tuple *(this is because sorted() returns a list and list can't be used as a key)*, use it as a key, and add the string s to its list of values i.e anagrams
- get the hashmap values i.e list of grouped anagrams, convert it to a list *(this is because the expected return type is list of lists and values() returns a view object)*, and return it

#### Frequency String Approach
- set res = a hashmap to group anagrams
    - *key = frequency string*
    - *value = list of strings with same frequency string i.e anagrams*
    - *the default dict is to handle missing keys and avoid the extra check if a key exists*
- traverse the input list of strings, *for each string s in the array of strings*
    - *find the frequency string*
        - initialize a list of size 26 with zeros to get the key 
        *(26 because the characters in anagrams are lowercase English letters)*
        - for each character c in the s
            - find the index of c and increment the value at the index
            *for example index of c  = ascii value of c - ascii value of a  = 99 - 97 = 2*
    - convert the frequency list to tuple *(this is because list can't be used as a key)*, use it as a key, and add the string s to its list of values i.e anagrams
- get the hashmap values i.e list of grouped anagrams, convert it to a list *(this is because the expected return type is list of lists and values() returns a view object)*, and return it


## Complexity
- Time complexity: 
    - Sorted String Approach O(strings list traversal * (sorting string + converting list to tuple)) → O(n * (klogk + k)) → O(n * klogk + n * k) → O(n * klogk)
    - Frequency String Approach O(strings list traversal * (creating frequency list + string traversal + converting list to tuple)) → O(n * (26 + k + 26)) → O(n * 26 + n * k + n * 26) → O(n * k)
    - *n = number of strings in the input*
    - *k = length of a string*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - Sorted String Approach O(hashmap) → O(anagram groups) → O(n)
    - Frequency String Approach O(hashmap + frequency list) → O(anagram groups + 26) → O(n + 26) → O(n)
    - *if there are no anagrams, there can be n groups i.e for each string there will be a key in the hashmap*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    # using sorted string
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        res = defaultdict(list)
        for s in strs:
            res[tuple(sorted(s))].append(s)
        return list(res.values())


    # using frequency string
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        res = defaultdict(list)
        for s in strs:
            freq = [0] * 26
            for c in s:
                freq[ord(c) - 97] += 1    # 97 is ord('a')
            res[tuple(freq)].append(s)
        return list(res.values())
```
