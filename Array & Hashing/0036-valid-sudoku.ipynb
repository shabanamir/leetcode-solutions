# <center> 36. Valid Sudoku </center>


## Problem Description
[Click here](https://leetcode.com/problems/valid-sudoku/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use three hashmaps to check values in columns, rows, and squares.


## Approach
<!-- Describe your approach to solving the problem. -->
- set row = a hashmap to check values in rows
  - *key = row number*
  - *value = set of elements*
- do the same for column and square
- traverse the 2d sudoku array i.e 9x9 grid
    - if the cell is filled
        - if the number (cell value) is present in a column, row, or square
            - return false 
        - else add the number to the column, row, and square set
- if the loop completes without returning false, the sudoku is valid, return true 


## Complexity
- Time complexity: O(grid traversal) → O(rows * columns) → O(9 * 9) → O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(rows hashmap + cols hashmap + squares hashmap) → O(9 + 9 + 9) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isValidSudoku(self, board: List[List[str]]) -> bool:
        row, col, square = defaultdict(set), defaultdict(set), defaultdict(set)
        for r in range(9):
            for c in range(9):
                if board[r][c] != '.':
                    if board[r][c] in row[r] or board[r][c] in col[c] or board[r][c] in square[r//3, c//3]:
                        return False
                    row[r].add(board[r][c])
                    col[c].add(board[r][c])
                    square[r//3, c//3].add(board[r][c])
        return True
```
