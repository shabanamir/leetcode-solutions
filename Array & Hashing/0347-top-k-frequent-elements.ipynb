## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
1. Sorting O(mlogm) *(m = total distinct elements)* 
    - use a map to store element-frequency pair
    - sort the pairs in decreasing order of frequency
    - return the first k pairs
2. Min Heap O(mlogk)
    - use a map to store element-frequency pair
    - create a min-heap with k pairs
    - traverse the remaining pairs, and pop heap if you get a high-freq pair. At the end, there will be k high-freq pairs left in the heap
3. Max Heap O(klogm)
    - use a map to store element-frequency pair
    - create a max-heap with m pairs
    - loop k times and pop heap to get the k max-freq pairs 
4. Bucket/Index sort O(n) 
    - Bucket sort worst time complexity is O(n²), but here we will not do sorting. We will only create n buckets and fill each bucket with elements that have the corresponding frequency i.e frequency equal to the bucket index.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to store the k frequent elements
- set count = a hashmap with a default value of 0 for each key to store the frequency of each element
- set freq = a 2d array of size n for bucket sort
    - *freq[f][n] to store the elements and their frequencies i.e element with frequency 4 will be at index 4 (bucket 4)*
    - *f represents the frequency*
    - *n represents the list of elements that have the frequency f*
- traverse the input array elements and get their frequencies
    - for each number in the input array
        - increment its count in the hashmap
- traverse the count hashmap and fill the frequency array
    - for each element and its count in the hashmap
        - add the element to the list at the index equal to its count in the freq array
- traverse the frequency array in reverse to get the most frequent element first
    - for each index in the freq array
        - for each element in the list at the current index
            - add the element to the result list
        - return when you get k elements 


## Complexity
- Time complexity: O(input array traversal to fill hashmap + hashmap traversal to fill frequency array + frequency array traversal for k elements) → O(n + m + k) → O(n)
    - *m = total distinct elements*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(frequency hashmap + frequency array + result array) → O(m + n + k) → O(n)
  - *frequency array is 2D but at most n elements will be stored*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        res = []
        count = defaultdict(int)
        freq = [[] for _ in range(len(nums) + 1)]
        for n in nums:
            count[n] += 1
        for n, c in count.items():
            freq[c].append(n)
        for i in range(len(freq) - 1, 0, -1):
            for n in freq[i]:
                res.append(n)
                if len(res) == k:
                    return res
```
