# <center> 58. Length of Last Word </center>


## Problem Description
[Click here](https://leetcode.com/problems/length-of-last-word/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Traverse the string from the end and count the non-space characters until you get a space character.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the length of the last word
- traverse the string from the end
    - if you get a space character
        - if it is not a trailing space i.e res > 0 *(this part will execute if there is more than one word in the string)*
            - return res
    - else the character is non-space, increment res
- if the loop completes without returning anything, return res *(this part will execute if string has only one word)*


## Complexity
- Time complexity: O(string traversal) â†’ O(n) 
    - *in the worst case, if the input contains only one word and the rest are trailing spaces, we will end up traversing the whole string*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def lengthOfLastWord(self, s: str) -> int:
        res = 0
        for i in range(len(s) - 1, -1, -1):
            if s[i] == ' ':
                if res: 
                    return res
            else:
                res += 1
        return res
```
