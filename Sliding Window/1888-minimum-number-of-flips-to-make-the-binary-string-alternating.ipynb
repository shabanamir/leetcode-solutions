# <center> 1888. Minimum Number of Flips to Make the Binary String Alternating </center>


## Problem Description
[Click here](https://leetcode.com/problems/minimum-number-of-flips-to-make-the-binary-string-alternating/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We can perform two types of operations to make the binary string alternating: rotate and flip

If the string length is: 
- even, the rotate (type-I) operation will not reduce the number of flips (type-II operation).
- odd, the rotate operation may reduce the flips. So, we need to perform both operations

First, find the number of flips. If the string length is even, return the number of flips. If the string length is odd, rotate it and check if the rotate operation reduces the flips. 

The alternating binary string can start with
- 0 (case I) e.g 0101
    - *in this case, 0's are required at the even indices and 1's at the odd indices*
- 1 (case II) e.g 1010
    - *in this case, 0's are required at the odd indices and 1's at the even indices* 

So, we need to find the minimum flips of the two cases. There are two ways to find the flips:
1. traverse the string separately for each case. If the current character is not the expected character according to alternation, increase the flip count by one  
2. traverse the string once and count even odd indices flips
    - for case I, 0's are required at the even indices and 1's at the odd indices. We need to flip bits that are not at their indices i.e 0's at odd indices and 1's at even indices
    - for case II, 0's are required at the odd indices and 1's at the even indices. We need to flip 0's at even indices and 1's at odd indices


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = size of input string
- set even = a hashmap to count flips required for even index 0's (case ii) and even index 1's (case i)
- set odd = a hashmap to count flips required for odd index 0's (case i) and odd index 1's (case ii)
- *we can also use four variables instead of the two hashmaps but the code is clear this way*
- traverse the string and count the even-odd index flips <br>
for each index i in the range 0 to n
    - if current index is odd
        - increment the count of corresponding character in the odd hashmap
            - *if character is 1, we need to flip it according to case ii because in case ii, 1's are required at even indices*
            - *if character is 0, we need to flip it according to case i because in case i, 0's are required at even indices*
    - else current index is even
        - increment the count of corresponding character in the even hashmap
        - *if character is 1, we need to flip it according to case i because in case i, 1's are required at odd indices*
        - *if character is 0, we need to flip it according to case ii because in case ii, 0's are required at odd indices*
- set res = minimum of case i flips and case ii flips to track the minimum flips
- if n is odd, we need to perform the rotate operation also, Because in even length string, with each rotate operation the odd count is changed to even and vice versa. So, the result remains the same.
In odd length string, odd and even index swap after each rotate operation, and the even count decreases and odd count increases.
    - traverse string s
        - decrement the flip count of current char in even hashmap
        - increment the flip count of current char in odd hashmap
        - swap even odd hashmaps
        - find min flips of the two cases and update res 
- return res


## Complexity
- Time complexity: O(string traversal to find flips + string traversal to rotate in case n is odd) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashmaps) → O(even + odd) → O(2 + 2) → O(1)
    - *2 because each hashmap has two keys i.e 0 and 1*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def minFlips(self, s: str) -> int:
        n = len(s)
        even, odd = {'0': 0, '1': 0}, {'0': 0, '1': 0}
        for i in range(n):
            if i & 1:
                odd[s[i]] += 1
            else: 
                even[s[i]] += 1
        res = min(odd['0'] + even['1'], odd['1'] + even['0'])
        if n & 1: 
            for c in s:
                even[c] -= 1
                odd[c] += 1
                even, odd = odd, even
                res = min(res, odd['0'] + even['1'], odd['1'] + even['0'])
        return res
```
