# <center> 658. Find K Closest Elements </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-k-closest-elements/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the crossover point before which elements are smaller or equal to x and after which elements are greater than x.

The array is sorted, use binary search to find the crossover point


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if x <= first element, the first k elements are closest to x
    - if x >= last element, last k elements are closest to x
- set left pointer l = 0 and right pointer r = array size - k *(to keep the index inbound)*
- do binary search i.e loop until l < r
    - find midpoint 
    - if difference between x and mid value is greater than difference between x and right side value at index mid + k, the right side value is closer to x and k closest elements are on the right side
        - move the left pointer
    - else mid value is closer to x, the k closest elements are on the left side
        - move the right pointer
- return a list of k closest elements starting at index l and ending at index l + k


## Complexity
- Time complexity: O(binary search + slicing to get the result list) → O(log$_2$(n-k) + k) → O(log(n-k) + k)
    - *log$_2$ because in each iteration the array is divided by 2*
    - *n - k because the right pointer starts at index n - k*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(result list) → O(k)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findClosestElements(self, arr: List[int], k: int, x: int) -> List[int]:
        if x <= arr[0]:
            return arr[:k]
        if x >= arr[-1]:
            return arr[len(arr) - k:]
        l, r = 0, len(arr) - k
        while l < r:
            m = (l + r) // 2
            if x - arr[m] > arr[m + k] - x:
                l = m + 1
            else:
                r = m
        return arr[l : l + k]
```
