# <center> 424. Longest Repeating Character Replacement </center>


## Problem Description
[Click here](https://leetcode.com/problems/longest-repeating-character-replacement/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the longest window (substring) with minimum number of replacements required to make all elements the same. 

no. of replacements = window size - window's most frequent character frequency

Use the sliding window approach to check substrings and a hashmap to track the character frequencies. Each time, add an element to the right of window. If window becomes invalid i.e total replacement > k, remove an element from the left.


## Approach
<!-- Describe your approach to solving the problem. -->
- set count = a hashmap to store the character frequencies
- set res = 0 to track the length of the max size valid window
- set max_f = 0 to track the most frequent character frequency
- set left pointer l = 0, and right pointer r = 0 to track the left and right sides of the window
- traverse the string using the right pointer 
    - add current char to the window i.e update the current character count
    - update max_f by taking max of current character freq and previous max
    - if no. of replacements is more than k, the window is not valid, remove a character from the left of window
        - decrement the left char count
        - move the left pointer
    - the window is valid now
        - update res by taking max of current window size and previous max
        - move the right pointer, *(will be auto-updated in for-loop)*
- return res


## Complexity
- Time complexity: O(string traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(uppercase letters) → O(26) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def characterReplacement(self, s: str, k: int) -> int:
        count = defaultdict(int)
        res, max_f = 0, 0
        l = 0
        for r in range(len(s)):
            count[s[r]] += 1
            max_f = max(max_f, count[s[r]])
            if (r - l + 1) - max_f > k:
                count[s[l]] -= 1
                l += 1
            res = max(res, r - l + 1)
        return res
```
