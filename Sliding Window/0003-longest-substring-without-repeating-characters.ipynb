# <center> 3. Longest Substring Without Repeating Characters </center>


## Problem Description
[Click here](https://leetcode.com/problems/longest-substring-without-repeating-characters/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the max-size window (substring) without duplicates. 

Use a sliding window with a hash set. 

Keep track of the max size window. Slide the window whenever there is duplicate i.e char is present in the set. Else extend the window.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if string size is 0 or 1, the longest substring without duplicates is the string itself
        - return string size
- set res = 0 to track the longest substring length 
- set seen = a hash set to check for duplicates 
- set left and right pointers, l = 0 and r = 0, to track the left and right sides of the window
- traverse the string from start to end
    - if current character is not present in the set, it is a duplicate
        - add it to the set
        - increment right pointer to extend the window
        - update res by taking max of current window size and previous max window size
    - else the character is a duplicate
        - remove the left character from the set and move left pinter to slide the window
- return res


## Complexity
- Time complexity: O(loop) → O(input string size) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hash set) → O(unique characters) → O(256) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def lengthOfLongestSubstring(self, s: str) -> int:
        if not s or len(s) == 1:
            return len(s)
        res = 0
        seen = set()
        l, r = 0, 0
        while r < len(s):
            if s[r] not in seen:
                seen.add(s[r])
                r += 1
                res = max(res, r - l)
            else:
                seen.remove(s[l])
                l += 1
        return res
```
