## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to search for a duplicate that is at k distance from its previous occurrence i.e a window of size k where the start and end elements are the same.

Use a hashmap to search for the duplicate and its distance. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set seen = a hashmap to store the seen numbers and their indices
    - *key = number*
    - *value = index*
- traverse the array
    - if the current number is in the hashmap and the distance between its current and previous occurrence index <= k, the current number is a duplicate
        - return true
    - else add the number and its index to the hashmap
- if the loop completes without returning true, there's no duplicate, return false


## Complexity
- Time complexity: O(array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def containsNearbyDuplicate(self, nums: List[int], k: int) -> bool:
        seen = {}
        for i, n in enumerate(nums):
            if n in seen and i - seen[n] <= k:
                return True
            seen[n] = i
        return False
```
