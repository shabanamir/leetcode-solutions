# <center> 1838. Frequency of the Most Frequent Element </center>


## Problem Description
[Click here](https://leetcode.com/problems/frequency-of-the-most-frequent-element/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the longest subarray where the number of operations required to make all elements equal to the maximum value in the subarray is less than k.

Sort the array to easily find the max value of the subarray. Use the sliding window approach to check subarrays. Keep adding elements to the right of the window as long as the window is valid i.e total operations <= k. If the window becomes invalid (total operations > k), remove an element from the left.

operations = desired sum - current sum = subarray size * subarray max - current sum

For example: nums = [1, 9, 2, 6], k = 5
- sort → [1, 2, 6, 9]
- [1] → max = 1, desired sum = 1, current sum = 1, operations = 1 - 1 = 0 < k → valid
- [1, 2] → max = 2, desired sum = 4, current sum = 3, operations = 4 - 3 = 1 < k → valid
- [1, 2, 6]→ max = 6, desired sum = 18, current sum = 9, operations = 18 - 9 = 9 > k → invalid
- [2, 6, 9] → max = 9, desired sum = 27, current sum = 17, operations = 27 - 17 = 10 > k → invalid
- the max size valid window is [1, 2]


## Approach
<!-- Describe your approach to solving the problem. -->
- sort the array
- set res = 0 to track the longest window
- set pre_sum = 0 to track prefix sum (window sum)
- set left pointer l = 0 and right pointer r = 0 to track the left and right sides of the window
- traverse the array using the right pointer
    - add current/right side element to pre_sum 
    - if the window is invalid
        - remove an element from the left of the window by subtracting the value at position l from pre_sum
        - move the left pointer
    - update res by taking max of current window size and previous max window
    - move the right pointer, *(will be auto-updated in for loop)*
- return res


## Complexity
- Time complexity: O(sort + traversal) → O(nlogn + n) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxFrequency(self, nums: List[int], k: int) -> int:
        nums.sort()
        res, pre_sum = 0, 0
        l = 0
        for r in range(len(nums)):
            pre_sum += nums[r]
            if (r - l + 1) * nums[r] - pre_sum > k:
                pre_sum -= nums[l]
                l += 1
            res = max(res, r - l + 1)
        return  res
```
