# <center> 76. Minimum Window Substring </center>


## Problem Description
[Click here](https://leetcode.com/problems/minimum-window-substring/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have to find a minimum size substring (window) in string s such that it contains all the characters of string t.

Use a sliding window with two pointers left and right to move the window. Move the right pointer to find a valid window. When you get a valid window, move the left pointer to find a smaller window.

Use a counter and hashmap to determine if the window is valid i.e contains all the required characters.  


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the length of t is greater than s, return an empty string because there's no way we can find a substring
- set need = a hashmap to count the required characters i.e characters in t
    - *either loop to fill the hashmap or use counter()*
- set have = 0 to track the required characters we currently have
- set start = invalid start index i.e -1, end = invalid end index i.e size of s to track the start and end of the minimum window 
    - these pointers are to avoid slicing/creating substring
    - initially set to invalid indices to handle the case where no substring is found
- set left pointer l = 0, and right pointer r = 0 to track the left and right sides of the current window
- traverse the string s <br>
(you can use the right pointer r or any other pointer for traversal)
    - if current char count is greater than 0, it is a required character
        - increment have because we have found a required character
    - decrement the count of current character in the hashmap 
    - while we have all the required characters, the current window is valid, search for a minimum window by removing characters from the left
        - if current window is smaller than previous
            - update start and end
        - increment the count of left character in the hashmap because we are removing a character from the window
        - if the left character count is greater than 0. it is a required character
            - decrement have because a required character is removed
        - move the left pointer
    - move the right pointer, *(will auto-update in for-loop)*
- if start is still -1, return an empty string because we have not found the substring, else return the minimum length substring starting at index start and ending at index end as the minimum window


## Complexity
- Time complexity: O(finding character frequencies of string t + string s traversal) → O(n + m) → O(m + n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(unique characters in t) → (lowercase and uppercase english letters) → O(52) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        if len(t) > len(s):
            return ""
        need = Counter(t)
        have = 0
        start, end = -1, len(s)
        l, r = 0, 0
        for r in range(len(s)):
            if need[s[r]] > 0:
                have += 1
            need[s[r]] -= 1
            while have == len(t):
                if r - l < end - start:
                    start = l
                    end = r + 1
                need[s[l]] += 1
                if need[s[l]] > 0:
                    have -= 1
                l += 1
        return '' if start == -1 else s[start:end]
```
