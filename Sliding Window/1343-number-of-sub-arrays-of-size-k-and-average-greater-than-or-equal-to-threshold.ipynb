# <center> 1343. Number of Sub-arrays of Size K and Average Greater than or Equal to Threshold </center>


## Problem Description
[Click here](https://leetcode.com/problems/number-of-sub-arrays-of-size-k-and-average-greater-than-or-equal-to-threshold/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find all k-size subarrays with average >= threshold

Use the sliding window approach. Maintain a window of size k. Each time slide the window by removing an element from the left side and adding an element from the right side. If we find a valid window (average >= threshold), increment the subarrays count. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the total valid subarrays 
- set pre_sum = 0 to track the prefix sum (subarray sum) for finding the average
- find the prefix sum of the first k - 1 elements
    - *(this is to start with a window of size k - 1)*
- set left pointer l = 0 to track the left index of the window
- set r = k - 1 to track the right index of the window
- traverse the array using the right pointer 
    - add the right side element to the prefix sum to make the window size equal to k
    - if current window is valid i.e its average is >= threshold
        - increment res
    - slide the window to maintain the window size
        - subtract left element from the pre_sum to remove the left side element
        - move the left and right pointers 
            - *(right pointer will be auto-updated in for loop)*
- return res


## Complexity
- Time complexity: O(finding sum of array first k-1 elements + traversing array remaining elements) → O(k, n - k) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def numOfSubarrays(self, arr: List[int], k: int, threshold: int) -> int:
        res, pre_sum = 0, 0
        for i in range(k - 1):
            pre_sum += arr[i]
        l = 0
        for r in range(k - 1, len(arr)):
            pre_sum += arr[r]
            if pre_sum / k >= threshold:
                res += 1
            pre_sum -= arr[l]
            l += 1
        return res
```
