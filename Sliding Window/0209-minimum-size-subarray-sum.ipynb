# <center> 209. Minimum Size Subarray Sum </center>

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find a minimum size subarray with sum >= target. Use the sliding window approach.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a large number to track the minimum window/subarray size
- set pre_sum = 0 to track the prefix sum (window sum)
- set left pointer l = 0, right pointer r = 0 to track the left and right sides of window
- traverse the array <br>
(you can use the right pointer r or any other pointer for traversal)
    - add an element to the right of window by adding the current element to pre_sum
    - while the current window is valid
        - update res by taking min of current window size and the previous minimum
        - remove an element from the left of the window by subtracting the element at position l from pre_sum
        - move the left pointer
    - move right pointer, *(will auto-update in for-loop)*
- return res if you found a minimum window, else 0


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        res = float('inf')
        pre_sum = 0
        l = 0
        for r in range(len(nums)):
            pre_sum += nums[r]
            while pre_sum >= target:
                res = min(res, r - l + 1)
                pre_sum -= nums[l]
                l += 1
        return res if res != float('inf') else 0
```
