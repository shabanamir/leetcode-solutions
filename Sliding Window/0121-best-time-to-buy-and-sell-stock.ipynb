## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
For max-profit, we need to minimize the buying price and maximize the selling price. 

If current day price is greater than the previous buying price, it is the best day to sell, else it is the best day to buy. 

Traverse the stocks and calculate profit each time the selling price is high, else update the buying price.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the max profit
- set buy = first stock price to track the min buy price
- traverse the stocks to find the max profit
    - if the current stock price is high, it may be the best time to sell
        - check how much profit we can make by selling the stock and update res
    - else it may be the best time to buy
        - update buy
- return res


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxProfit(self, prices: List[int]) -> int:
        res = 0
        buy = prices[0]
        for cur in prices:
            if cur > buy:
                res = max(res, cur - buy) 
            else:
                buy = cur
        return res
```
