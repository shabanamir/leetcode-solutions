## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
It is the same as [Find all anagrams in a string](https://leetcode.com/problems/find-all-anagrams-in-a-string/description/). The difference is here we have to find just one anagram (permutation).

One string is a permutation of another string if both have the same character frequencies. Use a hashmap to store the required character frequencies, and a sliding window to search for a window (substring) of size s1 in string s2.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if s1 size is greater than s2, return false because there's no way we can find a permutation of s1 in s2
- set count = a hashmap to store character frequencies of s1 for tracking the required characters
    - *either loop to fill the hashmap or use Counter()*
- set left pointer l = 0 and right pointer r = 0 to track the left and right sides of the window
- traverse the characters in string s2
    - if current character count is greater than 0, it is a required character
        - decrement its count because we have found a required character and it is a part of the current window
          - *(remember the rule: decrement count of character if it gets inside the window and increment if it gets outside)*
        - if the current window size is the same as s1, we have found the permutation, return True
        - else we have not found all the required characters, move the right pointer to include more characters in the window
    - if left and right pointers are in the same position
      - move both pointers
    - else the current window is not valid because the current character count is 0 i.e it is not part of the permutation. We need to slide the window by removing a character from the left
      - add the left character back to the hashmap because it is a required character and we have removed it
          - *(remember the rule)*
      - move the left pointer   
- if the loop completes without returning true, there is no permutation, return false


## Complexity
- Time complexity: O(counting frequencies in string1 + string2 traversal) → O(m + n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(unique characters in string1) → O(lowercase english letters) → O(26) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def checkInclusion(self, s1: str, s2: str) -> bool:
        if len(s1) > len(s2):
            return False
        count = Counter(s1)
        l, r = 0, 0
        while r < len(s2):
            if count[s2[r]] > 0:
                count[s2[r]] -= 1
                if r - l + 1 == len(s1):
                    return True
                r += 1
            elif l == r:
                l += 1
                r += 1
            else:
                count[s2[l]] += 1
                l += 1
        return False
```
