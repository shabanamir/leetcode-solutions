## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the longest subarray with at most two distinct integers. This is similar to:
[longest substring with at most 2 distinct strings](https://leetcode.com/problems/longest-substring-with-at-most-2-distinct-characters/)
[longest substring with at most k distinct string](https://leetcode.com/problems/longest-substring-with-at-most-k-distinct-characters/)

Use the sliding window to track the subarrays and a hashmap to track the total types of fruits (integers) in a subarray. Keep adding fruits to the right of window as long as the window is valid i.e hashmap has two keys (two types of fruits). If the window becomes invalid, remove fruits from the left side.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the longest window (subarray)
- set count = a hashmap (with a default value of 0 for each key) to count the fruits of each type
- set left pointer l = 0, and right pointer r = 0 to track the left and right sides of the window
- traverse the fruits array using the right pointer
    - increment the count of fruit at position r 
    - while the current window is invalid i.e size of hashmap > 2
        - decrement count of fruit at position l to remove an element from the left 
        - delete the fruit entry in the map if its count is 0
        - move the left pointer
    - else, the window is valid, take max of current window size & previous max window, and update res
    - move the right pointer, *(will auto-update in for loop)*
- return res


## Complexity
- Time complexity: O(array traversal) → (n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashmap) → O(2) → O(1)
    - *2 because at most we are storing two different types of fruits*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def totalFruit(self, fruits: List[int]) -> int:
        res = 0
        count = defaultdict(int)
        l = 0
        for r in range(len(fruits)):
            count[fruits[r]] += 1
            while len(count) > 2:
                count[fruits[l]] -= 1
                if not count[fruits[l]]:
                    del count[fruits[l]]
                l += 1
            res = max(res, r - l + 1)
        return res
```
