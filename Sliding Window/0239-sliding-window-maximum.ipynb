# <center> 239. Sliding Window Maximum </center>


## Problem Description
[Click here](https://leetcode.com/problems/sliding-window-maximum/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the maximum of all subarrays of size k. 

Each time, slide the window and add the max value to the result. To get the max element in a window of size k, we can use:
- linear search O(k)
- max heap O(logk)
- double-ended queue O(1) time

Deque allows insertion and deletion at both ends in O(1) time. We will keep the largest elements of the current window at the front/left end and smallest elements at the rear/right end of the deque.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if k is 1, return the input array 
    - if k is equal to the length of the array, return the max of the input array
- se dq = a deque to store the indices of the current window elements for tracking the max value
- set res = a list to store the max value in each window
- set left pointer l = 0 and right pointer r = 0 to track the left and right sides of the window
- traverses the array <br>
(you can use the right pointer r or any other pointer for traversal)
    - while the current number is greater than the last inserted number in the dq, pop dq to remove all smaller element indices 
    - insert the index of the current number in the dq
    - if current window is valid
        - add the max element (leftmost index element) to the res
        - move left pointer
    - if l is greater than the leftmost index in the dq, remove an element from the left because that element is no longer part of the current window
    - move right pointer, *(will auto update in for-loop)*
- return res


## Complexity
- Time complexity: O(traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(deque) → O(window size) → O(k)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        if k == 1:
            return nums
        if k == len(nums):
            return [max(nums)]
        dq = deque()
        res = []
        l = 0
        for r in range(len(nums)):
            while dq and nums[dq[-1]] < nums[r]:
                dq.pop()
            dq.append(r)
            if r + 1 >= k:
                res.append(nums[dq[0]])
                l += 1
            if l > dq[0]:
                dq.popleft()
        return res
```
