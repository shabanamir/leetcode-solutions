# <center> 35. Search Insert Position </center>


## Problem Description
[Click here](https://leetcode.com/problems/search-insert-position/solutions/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
insert position = position where the previous element is smaller than the target and the next element is greater

The array is sorted, use binary search to search for the target or its possible insertion index.


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = leftmost index and right pointer r = rightmost index
- loop until l <= r
    - compute the midpoint
    - if the value at mid is less than the target
        - move the left pointer to search in the right half
    - if the value at mid is greater
        - move the right pointer to search in the left half
    - else value at mid is equal to the target
        - return the mid as its index
- if the loop completes without returning anything, the target doesn't exist, return l as the possible insertion index


## Complexity
- Time complexity: O(loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def searchInsert(self, nums: List[int], target: int) -> int:
        l = 0
        r = len(nums) - 1
        while l <= r:
            mid = (l + r) // 2
            if nums[mid] < target:
                l = mid + 1
            elif nums[mid] > target:
                r = mid - 1
            else:
                return mid
        return l
```
