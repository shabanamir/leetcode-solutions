# <center> 74. Search a 2D Matrix </center>


## Problem Description
[Click here](https://leetcode.com/problems/search-a-2d-matrix/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The given matrix is row-wise and column-wise sorted. We can get a sorted array if we arrange the elements linearly. Consider the 2D matrix as a 1D array having indices from 0 to (m * n) - 1 and perform a binary search.

Formula to get the index of an element: <br>
2d array converted into 1d: matrix[i][j] → arr[i * m + j] <br>
1d array converted into 2d: arr[i] → matrix[i / m][i % n] <br>
(i = current row, j = current column, m = total rows, n = total columns)


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = total number of rows
- set n = total number of columns
- set left pointer l = leftmost index and right pointer r = rightmost index
- while l <= r
    - get the mid index 
    - get the mid element i.e number at the mid index
    - if the mid element is less than the target, the target is on the right side
        - move the left pointer to search in the right half
    - if the mid element is greater than the target, the target is on the left side
        - move the right pointer to search in the left half
    - else mid element is the target, return true
- if the loop completes without returning true, the target is not in the array, return false


## Complexity
- Time complexity: O(loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$m * n) → O(logmn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        m, n = len(matrix), len(matrix[0])
        l, r = 0, m * n - 1
        while l <= r:
            mid = (l + r) // 2
            x = matrix[mid // n][mid % n]
            if x < target:
                l = mid + 1
            elif x > target:
                r = mid - 1
            else:
                return True
        return False
```
