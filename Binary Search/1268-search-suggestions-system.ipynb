## Problem Description
[Click here](https://leetcode.com/problems/search-suggestions-system/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find three products for each prefix of the search word.

Sort the products to group similar words (products) and use binary search to find the first and last words with the same prefix.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to store the suggested products for each search word character
- sort products
- set left pointer l = leftmost index to track the start of the suggested product range
- set right pointer r = rightmost index to track the end of the suggested products range
- for each character c at index i in the search word
    - find l
        - keep moving l until the product at position l is invalid for suggestion i.e either the product word size is too short or the current char is not equal to the corresponding character in the product
    - find r using the same logic
    - if l <= r, we can choose some products
        - choose 3 or fewer from the array and add them to the result
    - else add an empty list to the result
- return res 


## Complexity
- Time complexity: O(sort + search word traversal * binary search for finding l and r) → O(nlogn + m * logn) → O(nlogn) 
    - *m < n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(sort + result list) → O(n + search word size) → O(n + m) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def suggestedProducts(self, products: List[str], searchWord: str) -> List[List[str]]:
        res = []
        products.sort()
        l, r = 0, len(products) - 1
        for i, c in enumerate(searchWord):
            while l <= r and (i >= len(products[l]) or c != products[l][i]): l += 1
            while l <= r and (i >= len(products[r]) or c != products[r][i]): r -= 1
            if l <= r:
                res.append(products[l: min(l + 3, r + 1)])
            else:
                res.append([])
        return res
```
