## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The range is 1 to n. Consider it a sorted array and use binary search.

If the number we guess is higher, search on the left side for a smaller number, else on the right side for a greater number.


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = 1 and right pointer r = n
- loop til l <= r
    - find the midpoint
    - choose the midpoint number and check if it is correct i.e pass it to guess()
    - if guess() returns 1, the midpoint number is smaller
        -  move the left pointer to search in the right half for a greater number
    - if guess() returns -1, the midpoint number is greater
        - move the right pointer to search in the left part for a smaller number
    - else it is 0, the midpoint number is correct, return it


## Complexity
- Time complexity: O(loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# The guess API is already defined for you.
# @param num, your guess
# @return -1 if num is higher than the picked number
#          1 if num is lower than the picked number
#          otherwise return 0
# def guess(num: int) -> int:


class Solution:

    def guessNumber(self, n: int) -> int:
        l, r = 1, n
        while l <= r:
            mid = (l + r) // 2
            x = guess(mid)
            if x == 1:
                l = mid + 1
            elif x == -1:
                r = mid - 1
            else:
                return mid
```
