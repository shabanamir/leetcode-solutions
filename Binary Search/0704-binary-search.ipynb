# <center> 704. Binary Search </center>


## Problem Description
[Click here](https://leetcode.com/problems/binary-search/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The array is sorted, the left half elements are smaller and the right half are greater, so use binary (half interval) search.
 
Repeatedly divide the array into two parts (left and right) using the midpoint. If the value at mid is greater than the target, search on the left part, else on the right.

Formula for midpoint <br>
mid = (low + high) / 2          *python* <br>
mid = low + (high - low) / 2    *other languages to prevent integer overflow*


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = leftmost index and right pointer r = rightmost index
- loop until l <= r
    - find midpoint using the formula
    - if the value at mid is less than the target, the target lies on the right side
        - move left pointer i.e update l to m+1
    - if the value at mid is greater than the target, the target lies on the left side
        - move the right pointer i.e update r to m-1
    - else value at mid is equal to the target, return mid as index of the target
- if the loop completes without returning anything, the target doesn't exist in the array, return -1


## Complexity
- Time complexity: O(loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def search(self, nums: List[int], target: int) -> int:
        l = 0
        r = len(nums) - 1
        while l <= r:
            m = (l + r) // 2
            if nums[m] < target:
                l = m + 1
            elif nums[m] > target:
                r = m - 1
            else:
                return m
        return -1
```
