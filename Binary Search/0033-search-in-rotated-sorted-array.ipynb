# <center> 33. Search in Rotated Sorted Array </center>


## Problem Description
[Click here](https://leetcode.com/problems/search-in-rotated-sorted-array/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Rotating a sorted array divides the array into two parts, both are sorted except the pivot point. So, we can use binary search.

The rotation/pivot point is where the division occurs. At pivot point
the next element is smaller if the main array (array before rotation) is sorted in ascending order and vice versa.

Examples:
- [0, 1, 2, 4, 5, 6, 7]
sorted in ascending order 
rotate it three times to the left =  [5, 6, **7**, 0, 1, 2, 4], 
pivot = 7 because next to 7 is 0 and 0 < 7
both left and right subarrays are sorted
- [7, 6, 5, 4, 2, 1, 0]
sorted in descending order
rotate it three times to left = [2, 1, **0**, 7, 6, 5, 4] 
pivot = 0 because next to 0 is 7 and 7 > 0
both left and right subarrays are sorted


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = leftmost position and right pointer r = rightmost position
- loop until l <= r
    - find midpoint m
    - if mid value is same as the target, return m
    - if the left element <= mid element, the left half is sorted
        - if the target is in the sorted left half
            - move the right pointer to search in the left half 
        - else target is in the right half
            - move the left pointer to search in the right half
    - else right half is sorted
        - if the target is in the sorted right half
            - move the left pointer to search in the right half
        - else target is in the left half
            - move the right pointer to search in the left half
- if the loop executes completely without returning anything, the target is not in the array, return -1 


## Complexity
- Time complexity: O(loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def search(self, nums: List[int], target: int) -> int:
        l, r = 0, len(nums) - 1
        while l <= r:
            m = (l + r) // 2
            if target == nums[m]:
                return m
            if nums[l] <= nums[m]:
                if nums[l] <= target < nums[m]:
                    r = m - 1
                else:
                    l = m + 1
            else:
                if nums[m] < target <= nums[r]:
                    l = m + 1
                else:
                    r = m - 1
        return -1
```
