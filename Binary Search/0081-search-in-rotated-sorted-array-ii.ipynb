## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
This the same as [Search in Rotated Sorted Array I](https://leetcode.com/problems/search-in-rotated-sorted-array/description/) problem, but here the input contains duplicates.

To handle the duplicates, add a check for duplicates before the [original logic](https://leetcode.com/problems/search-in-rotated-sorted-array/solutions/3839163/o-logn-time-o-1-space-solution-explained/).

In a sorted array, duplicates are adjacent. To search for duplicates in a rotated sorted array, compare mid value with the left and right values. If arr[left] = arr[mid] = arr[right], the mid value is a duplicate.

## Approach 
<!-- Describe your approach to solving the problem. -->
*Same as [Search in rotated and sorted array I solution](https://leetcode.com/problems/search-in-rotated-sorted-array/solutions/3839163/o-logn-time-o-1-space-solution-explained/), with a check for duplicates. Move the left and right pointers whenever you get a duplicate.*

- set left pointer l = leftmost position and right pointer r = rightmost position
- loop until l <= r
    - find midpoint m
    - if mid value is same as the target, return true
    - if mid value is duplicate i.e it is same as the left and right value
        - move both pointers i.e increment l and decrement r
    - if the left element <= mid element, the left half is sorted
        - if the target is in the sorted left half
            - move the right pointer to search in the left half 
        - else the target is in the right half
            - move the left pointer to search in the right half
    - else right half is sorted
        - if the target is in the sorted right half
            - move the left pointer to search in the right half
        - else the target is in the left half
            - move the right pointer to search in the left half
- if the loop executes completely without returning anything, the target is not in the array, return false


## Complexity
- Time complexity: O(loop) → binary search → O(logn) → if all elements are same, we may end up doing linear search → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)

## Code

```python
class Solution:

    def search(self, nums: List[int], target: int) -> bool:
        l, r = 0, len(nums) - 1
        while l <= r:
            m = (l + r) // 2
            if target == nums[m]:
                return True
            if nums[l] == nums[m] == nums[r]:
                l += 1
                r -= 1
            elif nums[l] <= nums[m]:
                if nums[l] <= target < nums[m]:
                    r = m - 1
                else:
                    l = m + 1
            else:
                if nums[m] < target <= nums[r]:
                    l = m + 1
                else:
                    r = m -1
        return False
```
