## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Capacity range is max weight to total weight (weight sum).

Consider it a sorted array and use binary search to find the minimum capacity.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = -1 to track the minimum ship capacity
- set left pointer l = max weight, right pointer r = total weights
- define a helper function to check if a capacity is valid i.e the ship can package all packages in given days with this capacity
    - *if you don't want a helper function, put the function code inside the while loop*
    - can_ship(allowed capacity for a day)
        - set n_days = 1 to track the total days required to ship all packages with the allowed capacity
        - set cur = 0 to track the current sum of weights
        - traverse the weights
            - if adding the current weight to the sum exceeds the allowed capacity, we can't ship the current weight on the same day
                - increment n_days because we need another day for shipping
                - reset cur i.e set cur sum to current weight
            - else we can ship the current weight on the same day
                - add current weight to the current sum
        - check if the days required for shipping with the given capacity <= given days
        and return the result i.e return true if it is <= else false
- loop until l <= r
    - find the mid-capacity
    - if the ship can transport all packages with the mid-capacity
        - update res
        - move the right pointer to search for a lower capacity in the left half
    - else it can't ship
        - move the left pointer to search for a high capacity in the right half
- return res


## Complexity
- Time complexity: O(finding max weight + finding sum + binary search * weights traversal for capacity check) → O(n + n + log$_2$search range * n) → O(n + n + log$_2$m * n) → O(n + n + logm * n) → O(nlogm)
    - *m = binary search range, which is max weight to weights sum*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def shipWithinDays(self, weights: List[int], days: int) -> int:
        res = -1
        l, r = max(weights), sum(weights)

        def can_ship(cap: int) -> bool:
            n_days, cur = 1, 0
            for w in weights:
                if cur + w > cap:
                    n_days += 1
                    cur = w
                else:
                    cur += w
            return n_days <= days
        
        while l <= r:
            m = (l + r) // 2
            if can_ship(m):
                res = m
                r = m - 1
            else:
                l = m + 1
        return res
```
