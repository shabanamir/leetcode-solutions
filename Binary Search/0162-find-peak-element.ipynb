# <center> 162. Find Peak Element </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-peak-element/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
An element (nums[i]) is a peak element if it's greater than its left neighbor (nums[i-1]) and right neighbor (nums[i+1]).

We can use binary search. If the right neighbor is greater than the middle, the peak element is on the right side. If the left neighbor is greater, the peak element is on the left side.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there is only one element, it is the peak element, return 0 as its index
- set n = rightmost index
- set left pointer l = leftmost index, right pointer r = rightmost index
- loop until l <= r
    - find midpoint m
    - if m is inbound and the mid element is less than the left element, the peak element is on the left side
        - move the right pointer to search in the left half
    - if m is inbound and the mid element is less than the right element, the peak element is on the right side
        - move the left pointer to search in the right half
    - else the mid element is the peak element because it is greater than the left and right element, return it


## Complexity
- Time complexity: (loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def findPeakElement(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return 0
        n = len(nums) - 1
        l, r = 0, n
        while l <= r:
            m = (l + r) // 2
            if m > 0 and nums[m] < nums[m - 1]:
                r = m - 1
            elif m < n and nums[m] < nums[m + 1]:
                l = m + 1
            else:
                return m
```
