# <center> 4. Median of Two Sorted Arrays </center>


## Problem Description
[Click here](https://leetcode.com/problems/median-of-two-sorted-arrays/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The median is the point at which the array is divided into two parts.

For the median, we only need the left half of the merged array. Both arrays are sorted, so use binary search to find the elements that contribute to the left half.


## Approach
<!-- Describe your approach to solving the problem. -->
- swap arrays if the first array size is greater than the second
- set m = first array size, n = second array size
- set index_merged_mid = index of the middle element (median) of the merged array 
    - *(it is required to find the left partition size of the second array)*
- set odd = true if the merged array size is odd else false
- set left pointer l = 0, right pointer r = m
- loop until we find the median
    - get left partition index i of the first array
    - get left partition index j of the second array
    - get the leftmost and rightmost elements in both arrays 
    left_a, left_b, right_a, right_b
    - if correct partitioning is done
        - if merged array size is odd, median is the max of left_a and left_b
        - else median is the max of leftmost and min of rightmost elements divide it by 2
    - if left_a > right_b, it means there are more elements in the left part of the first array
        - update right pointer to decrease the size of first array 
    - else there are fewer elements in the left part
        - update left pointer to increase the size 


## Complexity
- Time complexity: O(binary search) → O(log min(nums1, nums2)) → O(log min(m, n)) 
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        if len(nums1) > len(nums2):
            # nums1, nums2 = nums2, nums1
            return self.findMedianSortedArrays(nums2, nums1)
        m, n = len(nums1), len(nums2)
        index_merged_mid = (m + n + 1) // 2
        odd = True if (m + n) % 2 else False
        l, r = 0 , m
        while True:
            i = (l + r) // 2     
            j = index_merged_mid - i  
            left_a = nums1[i - 1] if i > 0 else float('-inf')
            left_b = nums2[j - 1] if j > 0 else float('-inf')
            right_a = nums1[i] if i < m else float('inf')
            right_b = nums2[j] if j < n else float('inf')
            if left_a <= right_b and left_b <= right_a:
                if odd: 
                    return max(left_a, left_b)
                return (max(left_a, left_b) + min(right_a, right_b)) / 2
            elif left_a > right_b:
                r = i - 1
            else:
                l = i + 1
```
