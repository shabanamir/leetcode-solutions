# <center> 875. Koko Eating Bananas </center>


## Problem Description
[Click here](https://leetcode.com/problems/koko-eating-bananas/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The maximum eating speed (k) to eat all bananas within h hours is the maximum bananas of all piles i.e max(piles). 

The minimum eating seed (k) is in the range 1 to max speed. Consider it a sorted array and use binary search.


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = 1, right pointer r = max speed i.e max(piles)
- set res = a max number i.e max speed to track the minimum speed
- while l < r
    - get mid-element 
    - set time = 0 to track the time required to eat all bananas with the mid speed
    - compute the total time required to eat all bananas in piles with mid-element speed
        - for this, iterate all piles, get time for each pile by dividing pile size by mid-element, and update time
        - math.ceil() is to round the result to the nearest greater integer
    - if the time required <= h, we can finish all the bananas with this speed
        - move the right pointer to search in the left half for a min speed
    - else the time required is greater, we can't finish the bananas with this speed
        - move the left pointer to search in the right half
- return res


## Complexity
- Time complexity: O(finding max + binary search * time check) → O(m + logm * n) → O(n * logm)  
    - *m = max value in the piles*
    - *n = piles array size*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        l, r = 1, max(piles)
        res = r
        while l <= r:
            m = (l + r) // 2
            time = 0
            for p in piles:
                time += math.ceil(p / m)
            if time <= h:
                res = m
                r = m - 1
            else:
                l = m + 1
        return res
```
