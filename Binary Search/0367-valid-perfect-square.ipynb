## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A number is a perfect square if its square root is int, not a float. 

The square root of a perfect square lies in the range 1 to perfect square. Consider it a sorted array and use binary search.


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = 1 and right pointer r = square number (num)
- loop until l <= r
    - find midpoint m
    - if the square of m is less than the square number (num)
        - move left pointer to search in the right half
    - if the square of m is greater than num
        - move right pointer to search in the left half
    - else square of m is equal to num, num is a perfect square, so return true
- if the loop completes without returning true, num is not a perfect square, return false


## Complexity
- Time complexity: O(loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isPerfectSquare(self, num: int) -> bool:
        l, r = 1, num
        while l <= r:
            m = (l + r) // 2
            if m * m < num:
                l = m + 1
            elif m * m > num:
                r = m - 1
            else:
                return True
        return False
```
