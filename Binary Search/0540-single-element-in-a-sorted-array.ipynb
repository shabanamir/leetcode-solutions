## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the unique (single) element in O(logn) time.

Example 1 [1,1,**2**,3,3,4,4,8,8]
Example 2 [3,3,7,7,**10**,11,11]

Observations:
- Elements before the single element have the first occurrence at even and the next occurrence at odd index
- Elements after the single elements have the first occurrence at odd and the next occurrence at an even index

The array is sorted, and the duplicates will be adjacent, so use binary search. Find the mid index. If the index is even and the next element is the same or vice versa, search for the single element in the right subarray, else in the left subarray.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there's only one element in the array
        - the first element is the single element, return it
    - if the first element is not equal to the second element
        - the first element is the single element, return it
    - if the last element is not equal to the second last element
        - the last element is the single element, return it
- set left pointer l = leftmost index and right pointer r = rightmost index
- while l < r
    - compute midpoint m
    - if the mid element is not the same as the previous and the next element, it is the single element
        - return mid element
    - if m is even and the mid element is equal to the next element (m + 1) or m is odd and the mid element is equal to the previous element (m - 1), the single element lies in the right half
        - update left pointer 
        - *the xor operator is used to simplify the even-odd and next-previous check*
            - *if x is even, x ^ 1 will return x + 1 as the next higher odd number* 
            - *if x is odd, x ^ 1 will return x - 1 as the lower even number*
    - else the single element lies in the left array
        - update the right pointer


## Complexity
- Time complexity: (loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def singleNonDuplicate(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return nums[0]
        if nums[0] != nums[1]:
            return nums[0]
        if nums[-1] != nums[-2]:
            return nums[-1]
        l, r = 0, len(nums) - 1
        while l <= r:
            m = (l + r) // 2
            if nums[m] != nums[m - 1] and nums[m] != nums[m + 1]:
                return nums[m]
            if nums[m] == nums[m ^ 1]:
                l = m + 1
            else:
                r = m - 1 
```
