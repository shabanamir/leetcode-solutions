# <center> 69. Sqrt(x) </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The square root of the number x is in the range 1 to x. Consider it a sorted array and use binary search to find the square root.


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = 1 and right pointer r = x
- set res = 0 to track the square root in case x is not a perfect square
- while l <= r
    - find the midpoint m
    - if the square of m is less than x, the square root lies in the right half
        - move left pointer and update res
    - if the square of m is greater than x, the square root lies in the left half
        - move the right pointer
    - else the square of m is x (square root of x is m), return m
- if the loop completes without returning anything, x is not a perfect square, return res


## Complexity
- Time complexity: O(loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def mySqrt(self, x: int) -> int:
        l, r = 1, x
        res = 0
        while l <= r:
            m = (l + r) // 2
            if m * m < x:  
                l = m + 1
                res = m
            elif m * m > x:
                r = m - 1
            else:
                return m
        return res
```
