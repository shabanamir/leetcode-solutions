## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Coins required for k row = k*(k+1)/2

- Approach 1 <br>
Easy, we only need to compute (-1 + sqrt(1 + 8*n)) / 2 <br>
because <br>
k*(k+1)/2 = n <br>
k² + k = 2n <br>
k² + k - 2n = 0  -> quadratic equation (ax²+bx+c=0) <br>
put values (a=1, b=1, c=-2n) in quadratic formula (-b±√(b²-4ac))/2a <br>
\-1 + √(1 + 8n) / 2 <br>
(-1 + sqrt(1 + 8*n)) / 2   

- Approach 2 <br>
The range is 1 to n. Consider it a sorted array and use binary search.
Find coins required for midpoint (row). If required coins are less than n, search on the right side to find rows that require more coins. Else search on the left side.

## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = 1 and right pointer r = n
- while l <= r
    - compute the midpoint
    - compute the coins required to build mid-complete rows
    - if required coins are equal to n
        - return mid 
    - if coins < n
        - update l to look for a larger number of rows on the right that use more coins
    - else coins > n
        - update r and look for a smaller number of rows on the left side that use fewer coins
- if the loop completes without returning anything, an exact match is not found, return r as the no. of complete rows


## Complexity
- Time complexity: O(loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def arrangeCoins(self, n: int) -> int:
        l, r = 1, n
        while l <= r:
            mid = (l + r) // 2
            coins = (mid * (mid + 1)) // 2
            if coins == n:
                return mid
            elif coins < n:
                l = mid + 1
            else:
                r = mid - 1
        return r
```
