## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The range to remove characters is from 0 to the size of the removable array. Consider this range a sorted array and use binary search to find k (i.e number of removable characters). If we can remove k characters i.e s remains a subsequence of p after removing the characters, search on the right half to find a max number, else search on the left half.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the max k
- set left pointer l = lower boundary in the range and right pointer r = upper boundary
- loop until l <= r
    - find midpoint m
    - if it is possible to remove m characters 
        - *(use the code for subsequence check here or add it in a separate function can_remove())*
        - set res to m 
        - move the left pointer to search in the right half 
    - else it is not possible 
        - move the right pointer to search in the left half
- return res

**can_remove(k)**
- get k elements (indices) from the removable array and add them to a hashset
    - *(indices are added to a hashset to search in O(1) time)*
- set i = 0 to track index in string s
- set j = 0 to track index in string p
- while indices are inbound
    - if i is not in the hashset, we don't have to remove the char at index i 
        - if current characters in strings s and p are the same, s is currently a subsequence
            - increment j to check further
    - else we have to remove the ith char or move forward in s, so increment i
- check if j = size of string p and return the result
    - *(if it is equal, it is possible to remove k characters, so return true else false)*


## Complexity
- Time complexity: O(binary search * subsequence check) → O(log(removable array size) * min(string s size, string p size)) → O(logn * string p size) → O(logn * m) → O(mlogn)
    - *string s size > string p size*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashset) → O(max number of removable character) → O(size of removable array) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maximumRemovals(self, s: str, p: str, removable: List[int]) -> int:
        res = 0
        l, r = 0, len(removable)

        def can_remove(k: int) -> bool:
            remove = {removable[i] for i in range(k)}
            i, j = 0, 0
            while i < len(s) and j < len(p):
                if i not in remove:
                    if s[i] == p[j]:
                        j += 1
                i += 1
            return j == len(p)

        while l <= r:
            m = (l + r) // 2
            if can_remove(m):
                res = m 
                l = m + 1
            else:
                r = m - 1
        return res
```
