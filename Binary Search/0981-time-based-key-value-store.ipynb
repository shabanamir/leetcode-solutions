## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use a hashmap to store keys and values. 

The timestamp for the next value will always be greater than the previous. So, the values will be stored in sorted order. Use binary search in get().

## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- create a hashmap to store keys 
    - *key = string (a-z)*
    - *value = list of pairs, each pair represents a value (index 0) and timestamp (index 1)*

**set()**
- append the value and timestamp pair to the hashmap

**get()**
- set res = an empty string to track the timestamp value
- if the key is in the hashmap
    - get the list of all values for the keys
    - set left and right pointers to perform a binary search
    - loop until left pointer <= right pointer
        - find midpoint
        - if the midpoint timestamp is the same as the query timestamp
            - return the midpoint element value (at index 0)
        - if midpoint timestamp < query timestamp
            - update res
            - move the left pointer to search in the right half for a greater timestamp
        - else midpoint timestamp is greater
            - move the right pointer to search in the left half 
- return res


## Complexity
- Time complexity:
    - init() O(1)
    - set() O(1)
    - get() O(binary search) â†’ O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - init() O(1)
    - set() O(1)
    - get() O(1)
    - hashmap size will be n after n calls to set()

## Code

```python
class TimeMap:

    def __init__(self):
        self.store = defaultdict(list)
        
    def set(self, key: str, value: str, timestamp: int) -> None:
        self.store[key].append((value, timestamp))
        
    def get(self, key: str, timestamp: int) -> str:
        res = ''
        if self.store[key]:
            values = self.store[key]
            l, r = 0, len(values) - 1
            while l <= r:
                m = (l + r) // 2
                if values[m][1] == timestamp:
                    return values[m][0]
                elif values[m][1] < timestamp:
                    res = values[m][0]
                    l = m + 1
                else:
                    r = m - 1
        return res
        


# Your TimeMap object will be instantiated and called as such:
# obj = TimeMap()
# obj.set(key,value,timestamp)
# param_2 = obj.get(key,timestamp)
```
