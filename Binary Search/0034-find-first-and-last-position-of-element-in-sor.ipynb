## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have to find the first and last index of a duplicate element. 

The array is sorted and in a sorted array the duplicates are adjacent. The first occurrence of a duplicate element lies in the left half and the last occurrence in the right half. Use binary search to find the first and last occurrence of the target duplicate number.

Search for the target number in the array. If the target is in the array, search for its first occurrence in the left half and the last occurrence in the right half.


## Approach
<!-- Describe your approach to solving the problem. -->
- find the first occurrence of the target using binary search
    - *for binary search, either define a helper function or use the library function (bisect_left(), bisect_right())*
- if the first occurrence index is -1, the target is not in the array
    - return -1, -1
- else find the last occurrence using binary search
- return first and last occurrence indices

*Helper function for binary search* <br>
**binary search(target, search first occurrence)**
- set res = -1 to track the target index
- set left pointer l = leftmost index and right pointer r = rightmost index
- loop until l <= r
    - find midpoint
    - if the value at mid is less than the target, the target lies on the right side
        - move the left pointer 
    - if the value at mid is greater than the target, the target lies on the left side
        - move the right pointer 
    - else value at mid is equal to the target, we found the target
        - update res
        - if the search_first parameter is set to true, we need to find the first occurrence
            - move the right pointer to search in the left half        
        - else we need to search for the last occurrence
            - move the left pointer to search in the right half
- return res


## Complexity
- Time complexity: O(binary search for first occurrence + binary search for the last occurrence) → O(logn + logn) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def searchRange(self, nums: List[int], target: int) -> List[int]:
        
        def binary_search(target: int, search_first: bool) -> int:
            res = -1
            l, r = 0, len(nums) - 1           
            while l <= r:
                m = (l + r) // 2
                if nums[m] < target:
                    l = m + 1
                elif nums[m] > target:
                    r = m - 1
                else:
                    res = m
                    if search_first:
                        r = m - 1
                    else:
                        l = m + 1
            return res
        
        first = binary_search(target, search_first=True)
        if first == -1:
            return -1, -1
        last = binary_search(target, search_first=False) 
        return first, last
```
