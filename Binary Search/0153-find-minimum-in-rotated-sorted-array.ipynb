# <center> 153. Find Minimum in Rotated Sorted Array </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-minimum-in-rotated-sorted-array/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a sorted and rotated array. The minimum element is the pivot/rotation point, the elements on its left and right are greater and in sorted order. 

For example: [3, 4, 5, **1**, 2] <br>
1 is the pivot point, all the elements to the left of 1 are greater and in sorted order i.e [3, 4, 5], and all the elements to the right of 1 are greater and in sorted order i.e [2]

Use binary search to find the pivot. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = leftmost index and right pointer r = rightmost index
- loop until l < r
    - find the mid-point
    - if the mid value is less than the previous, it is the pivot, return it
    - if the mid value is greater than the next, the next element is the pivot, return it
    - if the mid value is greater than the right value, the pivot lies in the right half
        - move the left pointer to search in the right half
    - else the pivot lies in the left half
        - move the right pointer to search in the left half
- if the loop completes without returning anything, the left pointer element is the pivot, return it 


## Complexity
- Time complexity: O(loop) → O(each time array is divided by 2) → O(log$_2$array size) → O(log$_2$n) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findMin(self, nums: List[int]) -> int:
        l, r = 0, len(nums) - 1
        while l < r:
            m = (l + r) // 2
            if nums[m] < nums[m - 1]:
                return nums[m]
            elif nums[m] > nums[m + 1]:
                return nums[m + 1]
            elif nums[m] > nums[r]:
                l = m + 1
            else:
                r = m - 1
        return nums[l]
```
