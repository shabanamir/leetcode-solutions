# <center> 977. Squares of a Sorted Array </center>


## Problem Description
[Click here](https://leetcode.com/problems/squares-of-a-sorted-array/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need the output in sorted order. The input array is sorted. Use two-pointer approach to build the output in sorted order. Each time find which number absolute value is greater, the number on the left side or right side and add the greater number square at the right side to the result array. 

This is because if a number's absolute value is greater, its square is also greater.
For example: 2 and -5 
the absolute value of 5 is greater than 2 and therefore square of -5 is greater than the square of 2 i.e 25 > 4

*(we need to compare the absolute values because there are negative numbers in the input and square of negative and positive number is same. For example square of both 2 and -2 is 4)*


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = length of the input array
- check base cases
    - if n is 1
        - square the first element and return it in an array
- set res = an array of size n with zeros for the output
    - *(we can also create an empty array and then add the squares but we will have to reverse the array in the end because we are building the result from right side)*
- set left pointer l = leftmost index and right pointer r = rightmost index
- set k = rightmost index to track the current right index in the result array
    - *(we can also use the difference between the left and right pointer to track the current right index in the result array but the code is clear this way)*
- loop until l <= r
    - if the absolute value of the left number is greater than the absolute value of the right number, the left number square is greater
        - square the left number and add it to the result array at current right index in the result array i.e k 
        - decrement k
        - move the left pointer
    - else the right number square is greater
        - square the right number and add it to the result array at current right index in the result array i.e k 
        - decrement k
        - move the right pointer
- return res


## Complexity
- Time complexity: O(creating result array + loop) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result array) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def sortedSquares(self, nums: List[int]) -> List[int]:
        n = len(nums)
        if n == 1:
            return [nums[0] * nums[0]]
        res = [0] * n
        l, r = 0, n - 1
        k = r   
        while l <= r:
            if abs(nums[l]) > abs(nums[r]):
                res[k] = nums[l] * nums[l]
                k -= 1
                l += 1
            else:
                res[k] = nums[r] * nums[r]
                k -= 1
                r -= 1
        return res
```
