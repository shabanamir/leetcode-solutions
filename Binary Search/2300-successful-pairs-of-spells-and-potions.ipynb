# <center> 2300. Successful Pairs of Spells and Potions </center>


## Problem Description
[Click here](https://leetcode.com/problems/successful-pairs-of-spells-and-potions/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A pair is successful if spell * potion >= success

If a potion forms a successful pair with a spell, all potions greater than it will also form a successful pair. So, for each spell, we need to find a min value potion that forms a successful pair. Sort potions and use binary search to find the minimum value potion.


## Approach
<!-- Describe your approach to solving the problem. -->
- sort potions
- set res = a list to store the number of valid potions for each spell
- set m =  size of potions array
- traverse the spells <br>
for each spell in the spells array
    - set left pointer l = leftmost index and right pointer r = rightmost index in potions array
    - set mini = m to track the index of minimum potion 
    - while l <= r:
        - find midpoint m
        - if the mid potion is valid i.e spell * midpoint potion >= success
            - update mini
            - move right pointer to search in the left half for a minimum value potion
        - else the potion is not valid because the mid potion value is too small
            - move left pointer to search in the right half for a high value potion
    - get the number of valid potions for the spell and add it to res
- return res


## Complexity
- Time complexity: O(potions sort + spells traversal * binary search for min potion) → O(mlogm + n * logm) → O(mlogm + nlogm) 
    - *n = spell array size*
    - *m = potion array size*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(potions sort + result array) → O(m + n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def successfulPairs(self, spells: List[int], potions: List[int], success: int) -> List[int]:
        potions.sort()
        res = []
        m = len(potions)
        for spell in spells:
            l, r = 0,  m - 1
            mini = m
            while l <= r:
                mid = (l + r) // 2
                if spell * potions[mid] >= success:
                    mini = mid
                    r = mid - 1
                else:
                    l = mid + 1
            res.append(m - mini)
        return res  
```
