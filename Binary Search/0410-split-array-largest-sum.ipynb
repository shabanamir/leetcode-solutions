## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have to minimize the maximum sum of any subarray.

The minimized maximum sum can't be less than the max value in the array and can't be greater than the sum of the array. So, the range is array max to array sum. Consider this range a sorted array and use binary search.


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = lower boundary of the range i. e array max, right pointer r = upper boundary of the range i.e array sum
- set res = r to track minimum maximum sum
    - *(r because the minimum maximum sum can't be greater than the upper boundary of the range i.e array sum)*
- loop until l <= r
    - find midpoint m
    - if it is possible to divide the array into k valid subarrays with the sum m
        - *(use the code for sum check here or add it in a separate function can_split())*
        - update res 
        - move the right pointer to search in the left half for a minimum sum
    - else it is not possible
        - move the left pointer to search in the right half

**can_split(given sum)**
- set cur_sum = 0 to track the current subarray sum
- set count = 1 to count subarrays with sum <= given sum
- traverse the array 
    - add the current element to the subarray i.e update cur_sum
    - if current subarray sum > given sum, the current subarray is invalid and we need another subarray        
        - increment count 
        - reset the cur_sum to check the new subarray
- check if the count of subarrays <= k i.e is it possible to divide the array into k valid subarrays and return the result

## Complexity
- Time complexity: O(finding array max + finding array sum + binary search * array traversal to check for valid k subarrays) → O(n + n + log(sum) * n) → O(n * log(sum))
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def splitArray(self, nums: List[int], k: int) -> int:
        l, r = max(nums), sum(nums)
        res = r

        def can_split(sum: int) -> bool:
            cur_sum, count = 0, 1
            for n in nums:
                cur_sum += n
                if cur_sum > sum:
                    count += 1
                    cur_sum = n
            return count <= k

        while l <= r:
            m = (l + r) // 2
            if can_split(m):
                res = m
                r = m - 1
            else:
                l = m + 1
        return res
```
