## Problem Description
[Click here](https://leetcode.com/problems/longest-happy-string/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use the greedy approach and append the most frequent character first to the result list.

To get the most frequent character, use max heap because it returns the max element in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a list for max heap
    - *(Python's heap library doesn't have a max heap class. We can implement the max heap using the min heap class by inverting the values. Flip the signs and then push the numbers to the heap. In this way, the largest number becomes the smallest and moves to the heap top)*
- if the frequency of 'a' is not 0
    - push frequency (flip the sign) and character to heap
- do the same for 'b' and 'c'
- set res = a list list to store the output string
    - *(list because appending to a string is expensive and the list takes O(1) time)*
- loop until all occurrences of a, b, and c are processed
    - pop heap to get the most frequent char and its freq
    - if the last two characters of the result are same as the current most frequent char, we can't add the current most frequent char because a char can't appear consecutively more than two times, so
        - if the heap is not empty
            - pop heap to get the next frequent char and its freq
            - add the next frequent char to the result
            - we have used the next frequent char, so decrement its freq 
                - *(increment in this case because we have added the negative frequency to heap)*
            - if the freq of the next frequent char is not 0, push it to the heap because we still have to process it 
            - push the current most frequent char back to the heap because it was popped but not used
    - else we can add the current most frequent char
        - add the char to result
        - we have used the char, decrement its freq 
            - *(increment in this case because we have added the negative frequency to heap)*
        - if its frequency is not 0, push the char and its freq back to heap because we still have to process it
- convert the result list to a string and return it


## Complexity
- Time complexity: O(loop * heap push pop + converting list to string) → O(a+b+c * log(heap size) + a+b+c) → O(n * log(3) + n) → O(n + n) → O(n)
    - *while loop will run until the heap is empty, and the heap will be empty once all occurrences of a, b, and c are processed*
    - *heap will store at most 3 elements i.e a, b, c*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(heap + result list) → O(3 + a+b+c) → O(3 + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def longestDiverseString(self, a: int, b: int, c: int) -> str:
        max_heap = []
        if a:
            heappush(max_heap, (-a, 'a'))
        if b:
            heappush(max_heap, (-b, 'b'))
        if c:
            heappush(max_heap, (-c, 'c'))
        res = []
        while max_heap:
            freq, char = heappop(max_heap)
            if len(res) >= 2 and res[-1] == res[-2] == char:
                if max_heap:
                    freq2, char2 = heappop(max_heap)
                    res.append(char2)
                    freq2 += 1
                    if freq2:
                        heappush(max_heap, (freq2, char2))
                    heappush(max_heap, (freq, char))
            else:
                res.append(char)
                freq += 1
                if freq:
                    heappush(max_heap, (freq, char))
        return ''.join(res)
```
