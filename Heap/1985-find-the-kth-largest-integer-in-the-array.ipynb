## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To find Kth largest integer, we can use direct sorting or heap but it requires the numbers in integer form. The input is in string form and the string length can be up to 100. So, here the conversion from string to int is expensive.

We can sort the input without converting string to int by defining a custom key function.

If two numbers string form have:
- different lengths, the longer number is the larger number <br>
for example '100' is greater than '99' <br>
- same length, the string comparison gives the correct answer <br>
for example '10' is smaller than '99' 


## Approach
<!-- Describe your approach to solving the problem. -->
- there are some edge cases but handling them is expensive because the numbers are in string form
    - if k == 1, kth largest element is max(Array)
    - if k = size of the array, the kth largest element is min(array) 
- define a custom comparison function
    - A comparison function returns:
        - -1 if the first value is smaller
        - +1 if the first value is greater
        - 0 if both are equal
    - custom function cmp(num1, num2)
        - if num1 and num2 length is different
            - return -1 if num1 length is less else +1
        - else length is same
            - compare the strings, return -1 if num1 is smaller else +1
- convert the comparison function to a key function using cmp_to_key() and sort the number using the key function
- after sorting, kth largest number will be at index n - k, return it


## Complexity
- Time complexity: O(sort) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def kthLargestNumber(self, nums: List[str], k: int) -> str:
        
        def cmp(a: str, b: str) -> int:
            if len(a) != len(b):
                return -1 if len(a) < len(b) else 1
            else:
                return -1 if a < b else 1

        nums.sort(key=cmp_to_key(cmp))
        return nums[len(nums) - k]
```
