## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Approaches:
1. Max heap O(n-k)logn <br>
In the max heap, the largest element is at the top and can be accessed in O(1) time. Create a max heap with all elements and then pop heap k times to get the kth largest element 

2. Min heap O(nlogk) <br>
In the min heap, the smallest element is at the top and can be accessed in O(1) time. Use a Min Heap of size k to store k largest elements. Create a min with k elements. Traverse the remaining elements, if an element is greater than the heap top element, pop heap to remove it and insert the greater element. In the end, the heap will contain k largest elements. Kth largest element will be at the heap top because there are k largest elements in the heap.


## Approach
<!-- Describe your approach to solving the problem. -->
- set min_heap = an empty list
- for each number in the input list
    - if there are less than k elements in the heap, add the number to the heap
    - else there are k elements in the heap. If the number is greater than the heap top element
        - replace the heap top i.e pop heap to remove the top element and push the current number
            - *(heapreplace() is equivalent to pop first and then push)*
- *Note: if you heapify the list first and then pop elements until k elements are left, the time complexity will be O(n + (n-k)logn), and the space complexity will be O(n)*
- return heap top element


## Complexity
- Time complexity: O(array traversal * replacing heap top) → O(n * (heap pop + heap push)) → O(n * (log(heap size) + log(heap size))) → O(n * (logk + logk)) → O(n * logk)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(heap size) → O(k)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def findKthLargest(self, nums: List[int], k: int) -> int:
        min_heap = []
        for num in nums:
            if len(min_heap) < k :
                heapq.heappush(min_heap, num)
            elif num > min_heap[0]:
                heapq.heapreplace(min_heap, num)
        return min_heap[0]
```
