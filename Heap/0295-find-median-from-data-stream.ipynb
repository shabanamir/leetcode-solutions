## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->

1. Insertion Sort O(n) <br>
Keep the list sorted by finding the position to insert the new number in the sorted list.

2. Max and Min heap O(logn) <br>
If the list size is:
    - even, median is the average of two middle numbers i.e last/max number in the left half and first/min number in the right half of the list
    - odd, median is the middle element i.e last/max element in the left half of the list.

Min and max heap returns the minimum and maximum values in O(1) time. Add the left half numbers to the max heap and right half numbers to the min heap.


## Approach
<!-- Describe your approach to solving the problem. -->
**#### Heap Approach**

**init()**
- create empty max and min heap

**addNum()**
- if the number to be added is less than the max heap top, it belongs to the left half
    - add it to the max heap
        - *(Python's heap library doesn't have a max heap class. We can implement the max heap using the min heap class by inverting the values. Flip the signs and then push the numbers to the heap. In this way, the largest number becomes the smallest and moves to the heap top)*
- else, it belongs to the right half, add the number to min heap
- balance the heaps
    - *(heaps are balanced if maxHeap size >= minHeap size and maxHeap - minHeap <= 1)*
    - if max heap size < min heap size
        - pop min heap to remove a smaller value and add it to the max heap
    - if the difference between the max and min heap size > 1
        - pop max heap to remove a larger value and add it to the min heap

**findMedian()**
- if list size is even i.e max and min heap size is equal
    - get the two middle elements i.e top elements of min and max heap and return their average
        - *average = min_heap top element + max_heap top element / 2 = min_heap top element + (-max_heap top element) / 2 = min_heap top - max_heap top / 2*
- else list size is odd, get the middle element i.e max heap top and return it
    - *(flip the sign because we have flipped the signs while adding values to the max heap)*


## Complexity
- Time complexity: 
    - Heap Approach
        - init() O(1)
        - addNum() O(heap push + heap pop) → O(log(heap size) + log(heap size)) → O(logn + logn) → O(logn)
        - findMedian() O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - Heap Approach
        - init() O(1)
        - addNum() O(1)
        - findMedian() O(1)
        - heap size will be n after n calls to addNum()

## Code

```python
class MedianFinder:

    def __init__(self):
        self.max_heap = []
        self.min_heap = []

    def addNum(self, num: int) -> None:
        if not self.max_heap or num <= -self.max_heap[0]:
            heapq.heappush(self.max_heap, -num)
        else:
            heapq.heappush(self.min_heap, num)
        # balance the heaps
        if len(self.max_heap) < len(self.min_heap):
            heapq.heappush(self.max_heap, -heapq.heappop(self.min_heap))
        elif len(self.max_heap) - len(self.min_heap) > 1:
            heapq.heappush(self.min_heap, -heapq.heappop(self.max_heap))

    def findMedian(self) -> float:
        if len(self.max_heap) == len(self.min_heap):
            return (self.min_heap[0] - self.max_heap[0]) / 2.0
        return -self.max_heap[0]
        


# Your MedianFinder object will be instantiated and called as such:
# obj = MedianFinder()
# obj.addNum(num)
# param_2 = obj.findMedian()
```
