# <center> 1675. Minimize Deviation in Array </center>


## Problem Description
[Click here](https://leetcode.com/problems/minimize-deviation-in-array/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to minimize the max deviation (i.e max element - min element).

To minimize the deviation, we can perform the multiplication operation on odd numbers and the division operation on even numbers any number of times.

Handle the multiplication operation on odd numbers first and find the minimum number. Then divide the max element each time if it is even to minimize the deviation. 

We need to track the max element because it will change due to division. Use max heap because it returns the max element in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = max possible number to track the minimum deviation
- set mini = max possible number to track the minimum element in the array
- create a max heap 
    - *(Python's heap library doesn't have a max heap class. We can implement the max heap using the min heap class by inverting the values. Flip the signs and then push the numbers to the heap. In this way, the largest number becomes the smallest and moves to the heap top)*
- traverse the array
    - if the number is odd, multiply it by 2
    - flip the number sign and push it to heap
    - update mini
- loop while the deviation is not 0 or the max element is even
    - pop heap to get the max element and flip the sign because we have added the inverted values to the heap
    - update res by taking minimum of current deviation and previous min deviation
    - if res is 0 or max element is odd, break the loop
    - divide max element by 2
    - update the mini
    - push the new max element to the heap
- return res


## Complexity
- Time complexity: O(for loop * heap push + while loop * heap push pop) → O(n * log(heap size) + n * log(heap size)) → O(n * logn + n * logn) → O(n * logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(heap) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minimumDeviation(self, nums: List[int]) -> int:
        res = math.inf
        mini = math.inf
        max_heap = []
        for num in nums:
            if num & 1:
                num *= 2
            heappush(max_heap, -num)
            mini = min(mini, num)
        while True:
            maxi = -heapq.heappop(max_heap)
            res = min(res, maxi - mini)
            if res == 0 or maxi & 1:
                break
            maxi //= 2
            mini = min(mini, maxi)
            heappush(max_heap, -maxi)
        return res
```
