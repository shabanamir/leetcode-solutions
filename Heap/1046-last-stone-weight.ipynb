# <center> 1046. Last Stone Weight </center>


## Problem Description
[Click here](https://leetcode.com/problems/last-stone-weight/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Each time, we need to find the heaviest stone i.e the max element.

Use max heap because max heap returns max element in O(1) time.

*In max heap, the max element is at the heap top.*


## Approach
<!-- Describe your approach to solving the problem. -->
- create a list of elements with negative stone weights
    - *(Python's heap library doesn't have a max heap class. We can implement the max heap using the min heap class by inverting the values. Flip the signs and then push the numbers to the heap. In this way, the largest number becomes the smallest and moves to the heap top)*
- heapify the list to create a max heap
- loop until there is 1 stone left in the heap
    - pop heap two times to get the two heaviest stones x and y
    - if x and y are not equal, y will become y - x. Push it to heap
        - *(push the opposite of y - x i. e x - y to heap because we have to flip the signs)*
- if the heap is not empty, return heap top element as the last stone weight
else all stones are destroyed, return 0


## Complexity
- Time complexity: O(creating list + heapify + loop * heap push pop) → O(n + n + n * log(heap size)) → O(n + n + n * logn) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(heap) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def lastStoneWeight(self, stones: List[int]) -> int:
        max_heap = [-s for s in stones]
        heapq.heapify(max_heap)
        while len(max_heap) > 1:
            x = heapq.heappop(max_heap)
            y = heapq.heappop(max_heap)
            if x != y:
                heapq.heappush(max_heap, x - y)
        return abs(max_heap[0]) if max_heap else 0   
```
