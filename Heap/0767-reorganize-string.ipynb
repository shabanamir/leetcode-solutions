## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use the greedy approach and append the most frequent character first to the result list.

To get the most frequent character, use max heap because it returns the max element in O(1) time.

One by one take the highest frequency character from the heap, add it to the result, and decrease the character frequency. Add the previous letter back to the heap so that any two adjacent characters are not the same.


## Approach
<!-- Describe your approach to solving the problem. -->
- get the frequency of each character
    - *(either use a library function or create a hashmap and fill it by traversing the string)*
- check base cases
    - if the frequency of the most frequent character > half of the string length, we can't organize the string, return an empty string 
- set res = a list to store the output string
- create a list of characters and their frequencies for the max heap
    - *(Python's heap library doesn't have a max heap class. We can implement the max heap using the min heap class by inverting the values. Flip the signs and then push the numbers to the heap. In this way, the largest number becomes the smallest and moves to the heap top)*
- heapify the list to create a max heap
    - *(if you create a heap by pushing elements instead of heapify(). The time complexity will be O(nlogn))*
- set prev_freq = 0 and prev_char = empty string to track the previous char and its freq
- loop until all characters are processed 
    - pop heap to get the most frequent char and its frequency
    - add char to the result
    - we have used the char, so decrement its frequency
        - *(increment in this case because we have added the negative frequency to heap)*
    - push the previous char and its freq to the heap so that adjacent characters are not the same
    - update prev_freq and prev_char to current frequent char and its freq
- return string form of the result list


## Complexity
- Time complexity: O(counter + finding max + creating max heap list + heapify + loop * heap pop) → O(n + 26 + 26 + 26 + n * log(heap size)) → O(n + 26 + 26 + 26 + n * log26) → O(n)
    - *loop will run until the heap is empty, and the heap will be empty once all characters of the string are processed*
    - *counter hashmap & heap will contain lowercase english letters = 26*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(counter hashmap + heap + result list) → O(26 + 26 + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def reorganizeString(self, s: str) -> str:
        count = Counter(s) 
        if max(count.values()) > (len(s) + 1) // 2:
            return ''
        res = []
        max_heap = [(-freq, c) for c, freq in count.items()]
        heapify(max_heap)
        prev_freq, prev_char = 0, ''
        while max_heap:
            freq, c = heappop(max_heap)
            res.append(c)
            freq += 1
            if prev_freq < 0:
                heappush(max_heap, (prev_freq, prev_char))
            prev_freq, prev_char = freq, c
        return ''.join(res)
```
