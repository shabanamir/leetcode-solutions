# <center> 973. K Closest Points to Origin </center>


## Problem Description
[Click here](https://leetcode.com/problems/k-closest-points-to-origin/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A point is closest to the origin if its distance from the origin is smaller.

Use a max heap of size k to store k closest points i.e points with the smallest distance. 

In max heap, the largest element is at the top and can be accessed in O(1) time. We will add points to the max heap, and if we find a smaller distance point, we will pop the max heap to remove the largest distance and add the smaller distance point.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a list for max heap
- for each x, y pair in the input points array
    - find the distance
    - if there are less than k elements in the heap, push (-distance, x, y) to the heap
        - *(Python's heap library doesn't have a max heap class. We can implement the max heap using the min heap class by inverting the values. Flip the signs and then push the numbers to the heap. In this way, the largest number becomes the smallest and moves to the heap top)*
    - else there are k elements in the heap. If the negative distance is greater than the top distance i.e current distance is smaller than the largest distance in the heap
        - remove the largest distance and push the current distance
            - *(heapreplace() is equivalent to pop first and then push)*
- after the loop, there will be k closest points in the heap, return all x, y pairs from the heap


## Complexity
- Time complexity: O(array traversal * replacing heap top) → O(n * (heap pop + heap push)) → O(n * (log(heap size) + log(heap size))) → O(n * (logk + logk)) → O(n * logk)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(heap) → O(k)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def kClosest(self, points: List[List[int]], k: int) -> List[List[int]]:
        heap = []
        for x, y in points:
            dist = x * x + y * y
            if len(heap) < k:
                heapq.heappush(heap, (-dist, x, y))
            elif -dist > heap[0][0]:
                heapq.heapreplace(heap, (-dist, x, y))
        return [(x, y) for (dist, x, y) in heap]
```
