## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find k projects with minimum capital and maximum profit.

For minimum capital, use min heap because it returns the min element in O(1) time.

For maximum profit, use a max heap because it returns the max element in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the initial capital >= to the max starting capital, we can select all projects and earn max profit
        - return the sum of the initial capital and k largest profits from the profits list
- create a list by pairing the corresponding starting capital and profit of each project
    - *index 0 = capital*
    - *index 1 = profit*
- heapify the list to create min heap
- create an empty max heap
- loop k times to select k project
    - find projects that we can afford i.e projects with starting capital <= available capital using min heap and push their profit to the max heap
        - *(Python's heap library doesn't have a max heap class. We can implement the max heap using the min heap class by inverting the values. Flip the signs and then push the numbers to the heap. In this way, the largest number becomes the smallest and moves to the heap top)*
    - if the max heap is empty, there are no more projects that we can afford, break the loop
    - else pop max heap to get the max profit and add it to the capital w
        - *(flip the sign because we have flipped the signs while adding values to max heap)*
- return w


## Complexity
- Time complexity: O(max() + nlargest() + sum() + min heap list + heapify() + loop * heap push pop) → O(n + nlogk + k + n + n + k * log(heap size)) → O(n + nlogk + k + n + n + klogn) → O(n + nlogk)
    - *the inner while loop will not run n times for every outer iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(min heap + max heap) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findMaximizedCapital(self, k: int, w: int, profits: List[int], capital: List[int]) -> int:
        if w >= max(capital):
            return w + sum(nlargest(k, profits))        
        min_heap = [(capital, profit) for capital, profit in zip(capital, profits)]
        heapq.heapify(min_heap)       
        max_heap = []        
        for _ in range(k):
            while min_heap and min_heap[0][0] <= w:
                c, p = heapq.heappop(min_heap)
                heapq.heappush(max_heap, -p)            
            if not max_heap:
                break
            w += -heapq.heappop(max_heap)
        return w
```
