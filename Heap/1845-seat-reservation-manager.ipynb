## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to reserve the smallest-numbered unreserved seat. Use min heap because it returns the smallest element in O(1) time. 


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- create a list for min heap and add all the numbers (seats) in the range 1 to n
    - *(no need to use heapify() because the list already satisfies the min heap properties i.e the numbers are in ascending sorted order)*

**reserve()**
- pop heap to get the minimum number and return it

**unreserve()**
- push the given number to the heap


## Complexity
- Time complexity:
    - init() O(creating heap) → O(n)
    - reserve() O(heap pop) → O(log(heap size)) → O(logn)
    - unreserve() O(heap push) → O(log(heap size)) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - init() O(heap) → O(n)
    - reserve() O(1)
    - unreserve() O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class SeatManager:

    def __init__(self, n: int):
        self.seats = [i for i in range(1, n + 1)]

    def reserve(self) -> int:
        return heapq.heappop(self.seats)

    def unreserve(self, seatNumber: int) -> None:
        heapq.heappush(self.seats, seatNumber)
        


# Your SeatManager object will be instantiated and called as such:
# obj = SeatManager(n)
# param_1 = obj.reserve()
# obj.unreserve(seatNumber)
```
