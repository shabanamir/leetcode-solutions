## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Same as [Kth Largest Element in an Array](https://leetcode.com/problems/kth-largest-element-in-an-array/description/) but here we will have to handle the incoming new elements also.

Use a Min Heap of size k to store k largest elements of the stream.

The top of the min heap contains the smallest element and can be accessed in O(1) time. In this case, the top would be the Kth largest element because there are k largest elements in the heap.

To process a new element of the stream, compare it with the heap top element. If it is greater than the top, replace the top with it.


## Approach
<!-- Describe your approach to solving the problem. -->
**init()**
- set min_heap = an empty list
- set k
- for each number in the input list
    - if there are less than k elements in the heap, add the number to the heap
    - else there are k elements in the heap. If the number is greater than the heap top
        - replace the heap top i.e pop heap to remove the top element and push the current number 
        (heapreplace() is equivalent to pop first and then push)
- *Note: if you heapify the list first and then pop elements until k elements are left, the time complexity will be O(n + (n-k)logn), and the space complexity will be O(n)*

**add()**
- if there are less than k elements in the heap, add the number to the heap
- else there are k elements in the heap. If the number is greater than the top element, replace the top
- return the heap top


## Complexity
- Time complexity:
    - init() O(array traversal * replacing heap top) → O(n * (heap pop + heap push)) → O(n * (log(heap size) + log(heap size))) → O(n * (logk + logk)) → O(n * logk)
    - add() O(heap pop + heap push) → O(log(heap size) + log(heap size)) → O(logk + logk) → O(logk)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - init() O(heap size) → O(k)
    - add() O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class KthLargest:

    def __init__(self, k: int, nums: List[int]):
        self.min_heap = []
        self.k = k
        for num in nums:
            if len(self.min_heap) < k:
                heapq.heappush(self.min_heap, num)
            elif num > self.min_heap[0]:
                heapq.heapreplace(self.min_heap, num)

    def add(self, val: int) -> int:
        if len(self.min_heap) < self.k:
            heapq.heappush(self.min_heap, val)
        elif val > self.min_heap[0]:
            heapq.heapreplace(self.min_heap, val)
        return self.min_heap[0]
        


# Your KthLargest object will be instantiated and called as such:
# obj = KthLargest(k, nums)
# param_1 = obj.add(val)
```
