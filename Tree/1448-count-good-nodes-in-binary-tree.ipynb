# <center> 1448. Count Good Nodes in Binary Tree </center>


## Problem Description
[Click here](https://leetcode.com/problems/count-good-nodes-in-binary-tree/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to count the good nodes.

A node is a good node if it is the max node of all nodes in the path from the root to that node.

To find path from root node to any other node, do DFS. Traverse the tree and track the max value on the path. If a node value is greater or equal to the max value on the path, the node is a good node.


## Approach
<!-- Describe your approach to solving the problem. -->
- define a helper function for DFS <br>
dfs(current node, max value on the current path)
    - check base cases
        - if current node is null, return 0
    - update the max value by taking the max of the current node value and the previous max value
    - count the good nodes in the left and right subtrees and return the current node count (1 if it is a good node else 0) + good nodes in the left subtree + good nodes in the right subtree
- start DFS to count good nodes in the tree and return the function result i.e good node count


## Complexity
- Time complexity: O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*

<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def goodNodes(self, root: TreeNode) -> int:

        def dfs(node: Optional[TreeNode], max_val: int) -> int:
            if not node:
                return 0
            max_val = max(max_val, node.val)
            return (node.val >= max_val) + dfs(node.left, max_val) + dfs(node.right, max_val)

        return dfs(root, root.val)
```
