## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Do level order traversal (BFS) and swap left and right children. 


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- check base cases
  - if there's no node in the tree, return null
- swap the left and right subtrees
- call invert for left-subtree
- call invert for right-subtree
- the recursion will stop when we reach the leaf node i.e when both the left and right children are null
- return root

**Iterative Approach**
- check base cases
  - if there's no node in the tree, return null
- define a queue and append the root as its first node
- loop until the queue is empty
  - pop a node from queue and store it in a variable cur as the current node
  - swap the left and right children
  - if the current node has left children, append it to queue 
  - if the current node has right children, append it to queue 
- return root


## Complexity
- Time complexity: 
  - Recursive Approach O(tree BFS) → O(n)
  - Iterative Approach O(tree BFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
  - Recursive Approach O(recursion stack) → O(h) → O(n)
    - *tree height and width depend on the tree structure, balanced tree height is logn and skewed tree height is n*
  - Iterative Approach O(queue) → O(max width) → O(w) → O(n)
    - *in a balanced tree, the last level has max nodes*
        - *w = 2$^h$ = 2$^{logn}$ = n* 
    - *in a skewed tree, each level has 1 node*
        - *w = 1*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:

        # recursive
        if not root:
            return None
        root.left, root.right = root.right, root.left
        self.invertTree(root.left)
        self.invertTree(root.right)
        return root

        #iterative
        if not root:
            return None
        q = [root]
        while q:
            cur = q.pop(0)
            cur.left, cur.right = cur.right, cur.left
            if cur.left:
                q.append(cur.left)
            if cur.right:
                q.append(cur.right)
        return root
```
