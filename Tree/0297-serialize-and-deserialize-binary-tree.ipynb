## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
- serialize() <br>
do DFS (preorder traversal) and return the traversal result in string form.
- deserialize() <br>
split the input i.e string to get node values and create a node for each value


## Approach
<!-- Describe your approach to solving the problem. -->

**serialize()**
- set res = an empty array to store serialization result
    - *(array because adding to string is expensive)*
- define a helper function for DFS <br>
dfs(current node)
    - if the node is null
        - add # to res and return
            - *(we can use any special character to represent the null nodes)*
    - add string form of node value to res
    - traverse the left subtree
    - traverse right subtree
- start DFS from the root node
- convert the result array to string and return it

**deserialize()**
- split the string to get node values and create a queue with these values
- define a helper function for DFS <br> 
dfs(current node)
    - pop a node value vode from queue
    - if the node value is #, the node is null
        - return null 
    - convert the node value to int and create a node 
    - recursively create the left and right subtrees
    - return node
- start dfs to create tree and return root node i.e dfs() result


## Complexity
- Time complexity: 
    - serialize() O(tree dfs + join()) → O(n + n) → O(n)
    - deserialize() O(split() + tree dfs) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - serialize() O(result array + recursion stack) → O(n + h) → O(n + n) → O(n)
    - deserialize() O(queue + tree + recursion stack) → O(n + n + h) → O(n + n + n) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Codec:

    def serialize(self, root: Optional[TreeNode]) -> str:
        res = []

        def dfs(node: Optional[TreeNode]) -> str:
            if not node:
                res.append('#')
                return
            res.append(str(node.val))
            dfs(node.left)
            dfs(node.right)
        
        dfs(root)
        return ','.join(res)


    def deserialize(self, data: str) -> TreeNode:
        q = deque(data.split(','))

        def dfs() -> Optional[TreeNode]:
            s = q.popleft()
            if s == '#':
                return None
            node = TreeNode(int(s))
            node.left = dfs()
            node.right = dfs()
            return node

        return dfs()
        


# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# ans = deser.deserialize(ser.serialize(root))
```
