## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
It is a level-order traversal. For even levels, we need to go from left to right, and for odd levels, we need to go reverse i.e. from right to left. 

To do level order traversal, do BFS. 

For the even level, we need to add nodes in the left-right direction to the result and vice versa. Use a double-ended queue (deque) because it allows insertion in both directions in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - return an empty array if the tree is empty
- set res = an array to store the zigzag traversal result
- create a queue to do BFS and add root node to the q
- set even = true to track if the current level is even or odd
    - *(it is initialized with true because the first level (level 0) is even)*
- loop until the end of the tree
    - create a double-ended queue (deque) to store current-level nodes
    - loop in range of size q to get current level nodes
        - pop q to get a current level node
        - if even is true, the current level is even and we need to add nodes in the left-right direction
            - add the current node to the right side of the dq
        - else current level is odd, we need to add nodes in the right-left direction
            - add the current node to the left side of the dq
        - push the current node left and right child to the q to get next level nodes
    - toggle even
        - *(to toggle, either use the not or xor operator. If you don't want to use an operator, keep track of the level number to check for even-odd)*
    - add the current level nodes i.e deque to res
- return res


## Complexity
- Time complexity: O(tree BFS) → O(total nodes) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(queue + deque + result array) → O(max width + max width + total levels * max width) → O(w + w + l * w) → O(n + n + n) → O(n)
    - *the result array is a 2d array, it will have l subarrays of size w i.e one subarray for each level but total n nodes will be stored*
    - *tree height and width depend on the tree structure*
    - *in a balanced tree, the last level has max nodes*
        - *w = 2$^h$ = balanced tree height is logn = 2$^{logn}$ = n* 
    - *in a skewed tree, each level has 1 node*
        - *w = 1*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def zigzagLevelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if not root:
            return []
        res = []
        q = deque([root])
        even = True
        while q:
            dq = deque()
            for _ in range(len(q)):
                node = q.popleft()
                if even:
                    dq.append(node.val)
                else:
                    dq.appendleft(node.val)
                if node.left: 
                    q.append(node.left)
                if node.right: 
                    q.append(node.right)
            even = not even        # even ^= True
            res.append(dq)
        return res
```
