## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to return the root node of duplicate subtrees.

To check if two subtrees are duplicates, compare their encoded traversal result because simple traversal can't uniquely identify a tree due to null nodes. We can do any type of traversal, here pre-order traversal is used.

simple preorder traversal = root → left → right <br>
encoded preorder traversal = root → special char → left → special char → right

Use a hashmap to track duplicate subtrees. Traverse the tree. If current node subtree is already in the hashmap, it is a duplicate subtree, so add the node to the result array.


## Approach
<!-- Describe your approach to solving the problem. -->
- set freq = a hashmap to track the frequency of each subtree 
    - *key = encoded traversal result of a subtree*
    - *value = frequency*
- set res = an empty array to store root nodes of duplicate subtrees
- define a helper function for DFS to do preorder traversal <br>
dfs(current node)
    - check base cases
      - if the current node is null, return an empty string
    - get the encoded traversal result of the subtree starting with the current node
    - increment the frequency of subtree by 1
    - if current subtree frequency is 2, it is a duplicate subtree
      - add current node as the root of the current subtree to the result array
    - return encoded traversal of current subtree
- start DFS from the root node
- return res
  
  
## Complexity
- Time complexity: O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashmap + recursion stack + result array) → O(total unique subtrees + tree height + total nodes) → O(n + h + n) → O(n + n + n) → O(n)
  - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def findDuplicateSubtrees(self, root: Optional[TreeNode]) -> List[Optional[TreeNode]]:
        freq = defaultdict(int)
        res = []

        def dfs(node: Optional[TreeNode]) -> str:
            if not node:
                return ''
            subtree = str(node.val) + ',' + dfs(node.left) + ',' + dfs(node.right)
            freq[subtree] += 1
            if freq[subtree] == 2:
                res.append(node)
            return subtree
            
        dfs(root)
        return res
```
