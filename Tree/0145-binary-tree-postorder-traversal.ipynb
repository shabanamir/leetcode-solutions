## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Post-order traversal = left → right → root

1. traverse the left subtree
2. traverse the right subtree
3. visit root 


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- set res = an array to store postorder traversal output
- define a helper function for postorder traversal (DFS) <br>
dfs(node)
    - if current node is null, return because we reached the end of the tree/subtree
    - recursively traverse the left subtree
        - *(this step will add the left subtree traversal output to the result array)*
    - recursively traverse right subtree
    - add current node value to the result array
- start postorder traversal from root
- return the result array

**Iterative Approach**
- set res = an array to store postorder traversal output
- use a stack for postorder traversal
    - *in postorder, the order is left → right → root  but in stack the order is reversed root → left → right*
- add root node to stack
- loop until stack is empty to traverse all nodes
    - pop stack to get current node
    - if current node is not null
        - add current node value to the result array
        - add the left child to stack
        - add the right child to stack
- reverse the result array and return it


## Complexity
- Time complexity:
    - Recursive Approach O(tree DFS) → O(n)
    - Iterative Approach O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - Recursive Approach O(recursion stack + result array) → O(h + n) → O(n + n) → O(n)
        - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
    - Iterative Approach O(stack + result array) → O(n + n) → O(n)


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def postorderTraversal(self, root: Optional[TreeNode]) -> List[int]:

        # recursive
        res = []

        def dfs(node: Optional[TreeNode]) -> None:
            if not node:
                return
            dfs(node.left)
            dfs(node.right)
            res.append(node.val)
        
        dfs(root)
        return res


        # iterative
        res = []
        stack = [root]
        while stack:
            cur = stack.pop()
            if cur:
                res.append(cur.val)
                if cur.left:
                    stack.append(cur.left)
                if cur.right:
                    stack.append(cur.right)
        return res[::-1]
```
