## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a complete binary tree
  - all levels except the last are completely filled
  - the last level may or may not be complete. If it is not complete, it is left-filled i.e there can't be a non-null node after a null node 

Do level order traversal (BFS) to check each level node. In the traversal, if we get a null node, all following nodes must be null. If they are not, the tree is not complete, and vice versa. 

# Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - if the tree is empty, return true because an empty tree is a complete tree
      - *(no need to check this condition here because according to the constraints, the tree in the test cases is never empty)*
- create a queue for BFS and add the root node
- loop until q is empty i.e until the end of the tree
    - pop q to get current level node
    - if the node is not null
      - add its left and right children to the q
    - else node is null
      - loop to check the remaining nodes in the q
        - if any of the remaining nodes is not null
            - return false because the tree is not complete
- if the loop completes without returning false, the tree is complete, return true 


## Complexity
- Time complexity: O(tree BFS) → O(total nodes) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(queue) → O(max width) → O(w) → O(n)
    - *tree height and width depend on the tree structure*
    - *in a balanced tree, the last level has max nodes*
        - *w = 2$^h$ = balanced tree height is logn = 2$^{logn}$ = n*
    - *in a skewed tree, each level has 1 node*
        - *w = 1*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def isCompleteTree(self, root: Optional[TreeNode]) -> bool:
        # if not root:
        #     return True
        q = deque([root])
        while q:
            cur = q.popleft()
            if cur:
                q.append(cur.left)
                q.append(cur.right)
            else: 
                while q:
                    if q.popleft():
                        return False
        return True
```
