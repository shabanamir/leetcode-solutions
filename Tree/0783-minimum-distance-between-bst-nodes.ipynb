# <center> 783. Minimum Distance Between BST Nodes </center>


## Problem Description
[Click here](https://leetcode.com/problems/minimum-distance-between-bst-nodes/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The minimum difference will be between two consecutive nodes and for this we need the nodes in sorted order. 

In a binary search tree (BST), we can obtain the sorted order by doing in-order traversal. For in-order traversal, do DFS.

In-order traversal = left → root → right


## Approach
<!-- Describe your approach to solving the problem. -->
- set previous node pointer prev = null
- set res = max possible number (i.e float infinity) to track the min difference
- define a helper function for recursive DFS <br>
dfs(current node)
    - return if the current node is null
    - make prev and res nonlocal variables 
        - *(this is to access the outer/main function variable in the nested function)*
    - traverse the left subtree
    - find the difference between the consecutive nodes (i.e current node value and the previous node value) and update res
    - update prev
    - traverse the right subtree
- start dfs from the root node
- return res


## Complexity
- Time complexity: O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def minDiffInBST(self, root: Optional[TreeNode]) -> int:
        prev = None
        res = float('inf')

        def dfs(node: TreeNode) -> None:
            if not node:
                return
            nonlocal prev, res
            dfs(node.left)
            if prev:
                res = min(res, node.val - prev.val)
            prev = node
            dfs(node.right)
        
        dfs(root)
        return res
```
