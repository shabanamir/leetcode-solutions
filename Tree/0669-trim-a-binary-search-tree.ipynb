# <center> 669. Trim a Binary Search Tree </center>


## Problem Description
[Click here](https://leetcode.com/problems/trim-a-binary-search-tree/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have to remove nodes that don't lie in the given range of low to high i.e all nodes whose values are less than the lower boundary or greater than the upper boundary.
 
Do DFS and check each node
- if a node value is less than the lower boundary, check the right subtree for trimming because all the nodes in the left subtree will also be less and therefore will not be included in the result 
    - *(in a BST, all nodes in the left subtree are less than the parent node and all nodes in the right subtree are greater than the parent node)*
- if a node value is greater than the upper boundary, check the left subtree because all nodes in the right subtree will also be greater


## Approach
<!-- Describe your approach to solving the problem. -->
- if the current node is null i.e either the tree is empty or we have reached the end
    - return null
        - *(this step will delete the nodes because setting a node to none is deleting it)*
- if the current node value is less than the lower boundary
    - trim its right subtree and return the result
- if the current node value is greater than the upper boundary
    - trim its left subtree and return the result
- update the left subtree i.e check it for trimming and trim if required 
    - *(this step will either completely delete the left subtree or delete some of its nodes if required)*
- update the right subtree
    - *(this step will either completely delete the right subtree or delete some of its nodes if required)*
- return the root of the trimmed tree


## Complexity
- Time complexity: O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(max height) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def trimBST(self, root: Optional[TreeNode], low: int, high: int) -> Optional[TreeNode]:
        if not root:
            return None
        if root.val < low:
            return self.trimBST(root.right, low, high)
        if root.val > high:
            return self.trimBST(root.left, low, high)
        root.left = self.trimBST(root.left, low, high)
        root.right = self.trimBST(root.right, low, high)
        return root
```
