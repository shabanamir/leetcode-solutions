# <center> 106. Construct Binary Tree from Inorder and Postorder Traversal </center>


## Problem Description
[Click here](https://leetcode.com/problems/construct-binary-tree-from-inorder-and-postorder-traversal/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To create a tree, we need root element, left subtree elements, and right subtree elements.

Postorder traversal = left-right-root <br>
Inoder traversal = left-root-right 

In postorder, the rightmost (last) element is the root. In inorder, the left elements belong to the left subtree and the right elements to the right subtree.

Get the root from postorder. Search for the root in the inorder to get the left and right subtree elements. Use a hashmap to search in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a hashmap and map the elements and indices of the inorder traversal list
    - *key = element*
    - *value = index of element*
- define a helper function with list start (left) and end (right) pointers to avoid slicing and passing lists <br>
helper(left, right)
    - check base case
      - if left > right, return
    - get the root node element (rightmost element in the postorder list) and create a node
    - find the root element index in the inorder list using the hashmap
    - get the elements after the index and recursively create the right subtree
    - get the elements before the index and recursively create the left subtree
    - return root
- call helper() to create the tree and return the root node


## Complexity
- Time complexity: O(filing hashmap + recursion to create tree) → O(inorder list size + total tree nodes) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashmap + recursion stack) → O(n + h) → O(n + n) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def buildTree(self, inorder: List[int], postorder: List[int]) -> Optional[TreeNode]:
        index = {n:i for i, n in enumerate(inorder)}

        def helper(l: int, r: int) -> Optional[TreeNode]:
            if l > r:
                return None
            root = TreeNode(postorder.pop())
            mid = index[root.val]
            root.right = helper(mid + 1, r)
            root.left = helper(l, mid - 1)
            return root
            
        return helper(0, len(inorder) - 1)
```
