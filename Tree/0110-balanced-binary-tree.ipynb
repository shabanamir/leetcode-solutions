## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A binary tree is height-balanced if:
  - the difference between the height of the left and right subtree is not greater than 1
  - both the left and right subtree are also height-balanced.

binary tree height = 1 + max(left subtree height, right subtree height)

Height of a tree is the longest path form root to leaf node. To find the longest path, do DFS.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - return true if the tree is empty because an empty tree is always balanced
- define a helper function to compute the height <br>
dfs(current node)
  - if node is null, return 0
  - compute left subtree height
  - if left subtree height is less than 0, the left subtree is not height- balanced, return -1
      - *(less than 0 height is -1 and -1 indicates that the tree/subtree is not height balanced)*
  - compute right subtree height 
  - if right subtree height is less than 0, the right subtree is not height- balanced, return -1
  - if the difference between the left subtree height and right subtree height is greater than 1, current tree is not height balanced, return -1 
  - return tree height i.e 1 + max of left and right height
- start DFS from root node and check the dfs result. return true if dfs doesn't return -1 else false
    - *(if the dfs returns -1, the tree is not height-balanced else it is)*


## Complexity
- Time complexity: O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def isBalanced(self, root: Optional[TreeNode]) -> bool:
        if not root:
            return True
        
        def height(node:Optional[TreeNode]) -> int:
            if not node:
                return 0
            left = height(node.left)
            if left < 0: 
                return -1
            right = height(node.right)
            if right < 0: 
                return -1
            if abs(left - right) > 1:
                return -1
            return 1 + max(left, right)

        return height(root) != -1
```
