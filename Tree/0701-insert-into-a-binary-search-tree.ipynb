## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a binary search tree (BST), left node values are smaller, and right node values are greater than the root node.

To insert a new value (node) in a binary search tree, find a leaf node and insert the new node as a child of the leaf node.
  - left child if new node value < leaf node value
  - right child if new node value > leaf node value


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- check base cases
    - if current node is null, either the tree is empty or we get a leaf node, create a new node and return it 
- if the new value is less than the current node value, search for a leaf node in the left subtree and insert it
- else value is greater, insert it in the right subtree
- return the root node

**Iterative Approach**
- check base cases
    - if current node is null, either the tree is empty or we get a leaf node, create a new node and return it 
- set current node pointer cur = root
- set leaf node pointer leaf = None
- loop until we find the leaf node for the value to be inserted
    - update the leaf to be the current node
    - if the value is less than the current node, go to the left subtree, else right subtree
- if the value is less than the leaf node, create a left child node
- else create a right child node
- return root


## Complexity
- Time complexity: 
    - Recursive Approach O(recursion) → O(h) → O(n)
    - Iterative Approach O(recursion) → O(h) → O(n)
    - *height (h) depends on tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - Recursive Approach O(recursion stack) → O(h) → O(n)
    - Iterative Approach O(1)

## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def insertIntoBST(self, root: Optional[TreeNode], val: int) -> Optional[TreeNode]:

        # recursive
        if not root:
            return TreeNode(val)
        if val < root.val:
            root.left = self.insertIntoBST(root.left, val)
        else:
            root.right = self.insertIntoBST(root.right, val)
        return root

        # iterative
        if not root:
            return TreeNode(val)
        cur = root
        leaf = None
        while cur:
            leaf = cur
            cur = cur.left if val < cur.val else cur.right
        if val < leaf.val:
            leaf.left = TreeNode(val)
        else:
            leaf.right = TreeNode(val)
        return root
```
