## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
This is a tree problem. Consider the head of the company as the root of a tree and direct subordinates as child nodes. 

The time required to inform all employees is the time to reach from root node to distant leaf node i.e we have to find the longest path from root to leaf node. To find longest path, do DFS.

Here we will do bottom-up DFS. We will traverse each node (employee)  and compute the time to reach the root (head).

Note: The value in manager array is manager id and the index is employee id.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a helper function for DFS <br>
dfs(employee id)
    - if the manager of the current employee != -1 i.e we haven't calculated the time to reach that manager
        - recursively compute the time and add it to the informTime of current employee
        - set the manager to -1 because we have calculated the time
        - return the informtime
    - else, return the previously computed inform time 
- for each employee compute the time to reach the head and return the max time


## Complexity
- Time complexity: O(tree DFS) → O(total nodes) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def numOfMinutes(self, n: int, headID: int, manager: List[int], informTime: List[int]) -> int:
    
        def dfs(i: int) -> int:
            if manager[i] != -1:
                informTime[i] += dfs(manager[i])
                manager[i] = -1
            return informTime[i]
        
        return max(dfs(i) for i in manager)
```
