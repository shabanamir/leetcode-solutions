## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Pre-order traversal = root → left → right

1. visit root 
2. traverse the left subtree
3. traverse the right subtree


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach** 
- set res = an array to store preorder traversal output
- define a helper function for preorder traversal (DFS) <br>
dfs(node)
    - if current node is null, return because we reached the end of the tree/subtree
    - add current node value to the result array
    - recursively traverse the left subtree 
        - *(this step will add the left subtree traversal output to the result array)*
    - recursively traverse right subtree
- start preorder traversal from root
- return the result array

**Iterative Approach**
- set res = an array to store preorder traversal output
- use a stack for preorder traversal
    - *in preorder, the order is root → left → right but in stack the order is reversed right → left → root*
- add root node to stack
- loop until stack is empty to traverse all nodes
    - pop stack to get current node
    - if current node is not null
        - add current node value to the result array
        - add the right child to stack
        - add the left child to stack
- return the result array


## Complexity
- Time complexity:
    - Recursive Approach O(tree DFS) → O(n)
    - Iterative Approach O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - Recursive Approach O(recursion stack + result array) → O(h + n) → O(n + n) → O(n)
        - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
    - Iterative Approach O(stack + result array) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def preorderTraversal(self, root: Optional[TreeNode]) -> List[int]:

        # recursive
        res = []

        def dfs(node: Optional[TreeNode]) -> None:
            if not node:
                return
            res.append(node.val)
            dfs(node.left)
            dfs(node.right)
        
        dfs(root)
        return res


        # iterative
        res = []
        stack = [root]
        while stack:
            cur = stack.pop()
            if cur:
                res.append(cur.val)
                if cur.right:
                    stack.append(cur.right)
                if cur.left:
                    stack.append(cur.left)
        return res
```
