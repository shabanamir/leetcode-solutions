# <center> 538. Convert BST to Greater Tree </center>


## Problem Description
[Click here](https://leetcode.com/problems/convert-bst-to-greater-tree/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Each node value in greater tree = node value + sum of all nodes greater than the node i.e the postfix sum

Do DFS. For each node, find nodes greater than it to track the postfix sum. Replace the node value with the postfix sum.


## Approach
<!-- Describe your approach to solving the problem. -->
- set post_sum = 0 to track the postfix sum
- use a helper function for DFS <br>
dfs(current node)
    - make post_sum variable nonlocal 
        - *(this is to access the outer function variable in nested function)* 
    - if the current node is not null
        - traverse the right subtree to get greater nodes
        - add current node value to postfix sum
        - replace the current node value with the postfix sum
        - traverse the left subtree
- start dfs() from the root node
- return root


## Complexity
- Time complexity: O(DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def convertBST(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        post_sum = 0

        def dfs(node: Optional[TreeNode]) -> None:
            nonlocal post_sum
            if node:
                dfs(node.right)
                post_sum += node.val
                node.val = post_sum
                dfs(node.left)

        dfs(root)
        return root
```
