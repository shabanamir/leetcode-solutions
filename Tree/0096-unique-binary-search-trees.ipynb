## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use the Catalan number formula to find the number of unique binary search trees (BSTs) with n nodes because n is a Catalan number.

For example: <br>
0 nodes : 1 BST <br>
1 nodes : 1 BST <br>
2 nodes : 2 BST <br>
3 nodes : 5 BST <br>
4 nodes : 14 BST <br>
5 nodes : 42 BST <br>

Sequence = 1, 1, 2, 5, 14, 42... = Catalan sequence

Catalan number formula to find the nth Catalan number

${\displaystyle C_{n+1}=\sum _{i=0}^{n}C_{i}\,C_{n-i}\quad {\text{for }}n\geq 0}$

${\displaystyle C_{n}=\sum _{i=1}^{n}C_{i-1}\,C_{n-i}}$

count(n) = summation of count(left) * count(right)) <br>
nodes in left subtree = 1 to i - 1 = i - 1 <br>
nodes in right subtree = i + 1 - n = n - i <br>
i range = 1 to n


## Approach
<!-- Describe your approach to solving the problem. -->
- initialize an array of size n + 1 with zeroes
- set the elements at index 0 and 1 to 1 (see Catalan sequence)
- loop in range 2 to n <br>
for each node i
    - calculate the number of unique BSTs in the left and right subtrees, multiply the left and right subtrees, and add it to the total count at index i
- return value at index n


## Complexity
- Time complexity: O(outer loop * inner loop) → O(n²)
    - *n² becuause for every outer iteration i, the inner loop runs i times. For outer loop iteration 2 the inner loop will run 2 times, for outer loop iteration 3 the inner loop will run 3 times, and so on.
total iterations = 2 + 3 + 4 + n = (n² + n)/2 = n²*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: (dp array) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def numTrees(self, n: int) -> int:
        dp = [0] * (n + 1)
        dp[0] = dp[1] = 1
        for i in range(2, n + 1):
            for j in range(1, i + 1):
                dp[i] += dp[j - 1] * dp[i - j]     # or dp[i] += dp[j] * dp[i - j - 1] 
        return dp[n]
```
