# <center> 662. Maximum Width of Binary Tree </center>


## Problem Description
[Click here](https://leetcode.com/problems/maximum-width-of-binary-tree/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Do level order traversal using BFS, find the width of each level, and return the max width.

Formula to find width of a level <br>
= end nodes position difference + 1 <br>
= rightmost node position - leftmost node position + 1 <br>
= 2 * parentnode position - 2 * parentnode position + 1


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 1 to track the max width
    - *(1 because there's at least one node in the tree)*
- create a queue for BFS 
- push the root node and its position as a pair to q
    - *index 0 = node, index 1 = node position*
- loop until q is empty i.e until the end of the tree
    - loop in range q size to get current level nodes
        - pop q to get the current node and its position
        - append node left child and its position to q
        - append node right child and its position to q
    - if there is more than one node in the queue (current level)
        - compute current level width using the formula (i.e difference between the last node and the first node position + 1) and update the res
- return res


## Complexity
- Time complexity: O(tree BFS) → O(total nodes) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(queue) → O(max width) → O(w) → O(n)
    - *tree height and width depend on the tree structure*
    - *in a balanced tree, the last level has max nodes*
        - *w = 2$^h$ = balanced tree height is logn = 2$^{logn}$ = n*
    - *in a skewed tree, each level has 1 node*
        - *w = 1*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def widthOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        res = 1
        q = deque()
        q.append((root, 1))
        while q:
            for _ in range(len(q)):
                node, pos = q.popleft()
                if node.left:
                    q.append((node.left, 2 * pos))
                if node.right:
                    q.append((node.right, 2 * pos + 1))
            if len(q) > 1:
                res = max(res, q[-1][1] - q[0][1] + 1)
        return res
```
