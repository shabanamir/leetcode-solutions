# <center> 129. Sum Root to Leaf Numbers </center>


## Problem Description
[Click here](https://leetcode.com/problems/sum-root-to-leaf-numbers/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Each root to leaf path represents a number. We need to find all numbers (i.e root to leaf paths) and return the sum.

Use DFS to find root to leaf path, do DFS. 

Do preorder traversal (DFS). Track the value calculated till the current node. For each node, update the value as value * 10 + current node value to get the number.


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- define a helper function for DFS <br>
dfs(node, path value)
    - if the current node is null, return 0
    - update the path value
    - if the node has no children, it is a leaf node, return the value
    - return sum of left and right subtree path values
- start DFS from root node with initial path value as 0

**Iterative Approach**
- set res = 0 to track the result (sum)
- create a stack and push the root node
- loop until the stack is empty
    - pop stack to get the current node
    - if the node has no children, add node value to the res
    - if the node has a left child, update the left child value and push it to stack
    - if the node has a right child, update the right child value and push it to stack
- return res


## Complexity
- Time complexity: 
    - Recursive Approach O(tree DFS) → O(n)
    - Iterative Approach O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - Recursive Approach O(recursion stack) → O(max height) → O(h) → O(n)
        - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
    - Iterative  Approach O(stack) → O(h) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def sumNumbers(self, root: Optional[TreeNode]) -> int:
        
        # recursive
        def dfs(node: Optional[TreeNode], num: int) -> int:
            if not node:
                return 0
            num = num * 10 + node.val
            if not node.left and not node.right:
                return num
            return dfs(node.left, num) + dfs(node.right, num)

        return dfs(root, 0)

        # iterative
        res = 0
        stack = [root]
        while stack:
            node = stack.pop()
            if not node.left and not node.right:
                res += node.val
            if node.left:
                node.left.val += 10 * node.val
                stack.append(node.left)
            if node.right:
                node.right.val += 10 * node.val
                stack.append(node.right)
        return res
```
