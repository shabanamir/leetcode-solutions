# <center> 104. Maximum Depth of Binary Tree </center>


## Problem Description
[Click here](https://leetcode.com/problems/maximum-depth-of-binary-tree/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the max depth (height) of a binary tree i.e longest path from root to leaf. 

longest path from root to leaf <br>
= 1 + max(longest path from left to leaf, longest path from right to leaf)  <br>
= 1 + max(height of left subtree, height of right subtree) 

Do post-order traversal (DFS) and calculate the height of the left and the right subtrees.


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- check base cases
  - if there is no node, return 0
- recursively calculate the height of the left and the right subtrees
- return 1 + max(height of left subtree, height of right subtree)

**Iterative Approach**
- check base cases
  - if there is no node, return 0
- use a stack and push the root node and its depth as a pair
- set res = 0 to track the max depth
- loop until the stack is empty
    - pop the stack top to get current node and its depth
    - update res
    - push the left child and its depth
      - *(depth of child node = depth of parent node + 1)*
    - push the right child and its depth
- return res


## Complexity
- Time complexity: 
  - Recursive Approach O(tree DFS) → O(n)
  - Iterative Approach O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
  - Recursive Approach O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
  - Iterative Approach O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def maxDepth(self, root: Optional[TreeNode]) -> int:

        # recursive
        if not root:
            return 0
        return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))

        # iterative
        if not root:
            return 0
        stack = [(root, 1)]
        res = 0
        while stack:
            node, depth = stack.pop()
            if node:
                res = max(res, depth)
                stack.append((node.left, depth + 1))
                stack.append((node.right, depth + 1))
        return res
```
