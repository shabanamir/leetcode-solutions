## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Do DFS to traverse the tree. At each node, decide whether to rob the node or not and return the max.


## Approach
<!-- Describe your approach to solving the problem. -->
- define a helper function for DFS to decide whether to rob a node or not. The function will return two values:
    - max rob amount if we rob the node
    - max rob amount if we do not rob the node
    - dfs(current node)
        - if the node is null, return (0, 0)
        - compute the rob and not rob values for the left subtree
        - compute the rob and not rob values for the right subtree
        - return (not rob value, rob value) of the current node
            - *not rob value = current node value + not rob left subtree value + not rob right subtree value*
            - *rob value = max of rob and not rob left subtree + max of rob and not rob right subtree*
- start DFS from the root node and return the result i.e max of rob and not rob


## Complexity
- Time complexity: O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(max height) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def rob(self, root: Optional[TreeNode]) -> int:
        
        def dfs(root: Optional[TreeNode]) -> tuple:
            if not root:
                return (0, 0)
            rob_l, not_rob_l = dfs(root.left)
            rob_r, not_rob_r = dfs(root.right)
            return (root.val + not_rob_l + not_rob_r,
                    max(rob_l, not_rob_l) + max(rob_r, not_rob_r))

        return max(dfs(root))
```
