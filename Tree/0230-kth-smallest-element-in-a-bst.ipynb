## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Kth smallest element is the kth number in a sorted list. 

Inorder traversal of a BST traverses the nodes in sorted increasing order. Do inorder traversal (DFS) and track the count of visited nodes. If the count becomes k, the current node is the kth smallest node.

inorder traversal = left → root → right


## Approach
<!-- Describe your approach to solving the problem. -->
- create a stack for DFS
- set cur = root to traverse the tree and track the current node
- loop until we find the kth smallest element
    - push all the left subtree nodes in the stack because the left nodes are smaller
    - pop stack to get the top (smallest) node from the stack 
    - decrement k because we found a smaller node
    - if k becomes 0, current node is the kth smallest node
        - return node value
    - else the kth smallest element is not in the left subtree
        - set the current node to the right child to search for the kth smallest element in the right subtree


## Complexity

- Time complexity: O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(stack) → O(max height) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        stack = []
        cur = root
        while True:   # while stack or cur
            while cur:
                stack.append(cur)
                cur = cur.left
            cur = stack.pop()
            k -= 1
            if not k:
                return cur.val
            cur = cur.right
```
