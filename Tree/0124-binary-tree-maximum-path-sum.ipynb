# <center> 124. Binary Tree Maximum Path Sum </center>


## Problem Description
[Click here](https://leetcode.com/problems/binary-tree-maximum-path-sum/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The max path sum passing through a node is the max of:
1. node value
2. node value + left child max path sum
3. node value + right child max path sum
4. node value + left child max path sum + right child max path sum

Do DFS to find the max path sum in the left and right subtrees.

## Approach
<!-- Describe your approach to solving the problem. -->
- set res = root node value or negative infinity to track the maximum path sum 
- define a helper function for DFS <br>
dfs(current node)
    - if the node is null, return 0
    - make res nonlocal
        - *(this is to access the outer function variable in nested function)*
    - get the max path sum in the left subtree
    - get the max path sum in the right subtree
    - update res i.e maximum path sum with max of the 4 cases mentioned above 
        - *max_path_sum = max(first three cases i.e current max path sum, 4th case)*
    - return the max of the first three cases and 0
        - *max(sum of current node and max of left and right subtree path sum, 0)*
        - *(0 because a node value can be negative)*
- start the dfs
- return res


## Complexity
- Time complexity: O(tree dfs) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        res = root.val
        
        def dfs(node: Optional[TreeNode]) -> int:
            if not node:
                return 0
            nonlocal res
            left_sum = dfs(node.left)
            right_sum = dfs(node.right)
            res = max(res, node.val + left_sum + right_sum)
            return max(node.val + max(left_sum, right_sum), 0)
        
        dfs(root)
        return res
```
