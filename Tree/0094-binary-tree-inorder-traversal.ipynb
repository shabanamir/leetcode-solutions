## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In-order traversal = left → root →  right

1. traverse the left subtree
2. visit root
3. traverse the right subtree


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- set res =  an array to store inorder traversal output
- define a helper function for inorder traversal (DFS) <br>
dfs(node)
  - if current node is null, return because we reached the end of the tree/subtree
  - recursively traverse the left subtree
      - *(this step will add the left subtree traversal output to the result array)*
  - add current node value to the result array
  - recursively traverse the right subtree
- start inorder traversal from root
- return result array

**Iterative Approach**
- set res =  an array to store inorder traversal output
- create an empty stack for inorder traversal
- set current node pointer cur = root
- loop until stack is empty or cur point to a null node
    - push the left subtree nodes to stack and move cur to the leftmost node
    - pop a node from stack
    - add node value to the result array
    - move cur to the right child of the node
- return result array


## Complexity
- Time complexity:
    - Recursive Approach O(tree DFS) → O(n)
    - Iterative Approach O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - Recursive Approach O(recursion stack + result array) → O(h + n) → O(n + n) → O(n)
        - *height (h) depends on tree structure, balanced tree height is logn and skewed tree height is n*
    - Iterative Approach O(stack + result array) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:

        # recursive 
        res = []

        def dfs(node: Optional[TreeNode]) -> None:
            if not node:
                return
            dfs(node.left)
            res.append(node.val)
            dfs(node.right)
        
        dfs(root)
        return res


        # iterative
        res = []
        stack = []
        cur = root
        while cur or stack:
            while cur:
                stack.append(cur)
                cur = cur.left
            node = stack.pop()
            res.append(node.val)
            cur = node.right
        return res
```
