# <center> 513. Find Bottom Left Tree Value </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-bottom-left-tree-value/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Do level order traversal (BFS) to find the leftmost (first) value in the last level.


## Approach
<!-- Describe your approach to solving the problem. -->
- use a queue for BFS
- add root node to q
- loop until q is empty i.e until end of the tree
    - pop q to get current level node
    - add the right and left child to q
        - *(add the right child first, then the left child so that the last node in the queue will be a left node)*
- after the loop, the current node pointer will have the leftmost node of the last level, return its value


## Complexity
- Time complexity: O(tree BFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(queue) → O(max width) → O(w) → O(n)
    - *tree height (h) and width (w) depend on the tree structure*
    - *in a balanced tree, the last level has max nodes*
        - *w = 2$^h$ = balanced tree height is logn = 2$^{logn}$ = n*
    - *in a skewed tree, each level has 1 node*
        - *w = 1*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def findBottomLeftValue(self, root: Optional[TreeNode]) -> int:
        q = deque([root])
        while q:
            cur = q.popleft()
            if cur.right:
                q.append(cur.right)
            if cur.left:
                q.append(cur.left)
        return cur.val
```
