## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The diameter of a binary tree is the number of nodes on the longest path between two end nodes. To find the longest path, use DFS.

The longest path can be of two types:
- split = path that lies in both subtrees i.e it passes through the root
- non-split = path that lies either in the left or right subtree i.e 
it doesn't pass through the root

Formula to find tree diameter <br>
= max(split path, non-split path) <br>
= max(left subtree diameter + right subtree diameter, current tree diameter) <br>
= max(left subtree height + right subtree height, current tree diameter)


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the max diameter
- define a helper function for DFs <br>
  dfs(node)
    - make res nonlocal
        - *(because res is declared in the parent/outer function and we need to access it here in the child/nested function)*
    - if the current node is null i.e we reached the end of the tree, return 0
    - find the left and right subtree height
    - find diameter and update res
    - return current node tree height
- start dfs from root node
- return res


## Complexity
- Time complexity: O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        res = 0

        def dfs(node: Optional[TreeNode]) -> int:
            nonlocal res
            if not node:
                return 0
            height_left = dfs(node.left)
            height_right = dfs(node.right)
            res = max(res, height_left + height_right)
            return 1 + max(height_left, height_right)

        dfs(root)
        return res
```
