# <center> 108. Convert Sorted Array to Binary Search Tree </center>


## Problem Description
[Click here](https://leetcode.com/problems/convert-sorted-array-to-binary-search-tree/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The array is sorted. Find the midpoint of the array and divide it into two halves. Make the middle element the root of the tree. Put the left half elements in the left subtree and the right half elements in the right subtree.


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- define a helper function to divide the array <br>
helper(left pointer (array start), right pointer (array end))
  - if the left pointer is greater than the right pointer, we have reached the end of the array, return
  - find the midpoint
  - create a tree node for the mid element and make it the root of the tree
  - recursively create the left subtree by passing the left subarray start and end pointers to the helper()
  - recursively create the right subtree by passing the right subarray start and end pointers to the helper()
  - return root
- call the helper function with array start and end pointers to create the tree

**Iterative Approach**
- create a queue
- set left pointer l = leftmost position and right pointer r = rightmost position to track the start and end of the array/subarray
- compute midpoint
- make the mid element the root of the tree
- push the root and left subarray start and end pointers to the q
- push the root and right subarray start and end pointers to the q
- loop until q is empty i.e end of the array
    - pop an element from q to get the current node, left, and right pointers
    - if pointers are inbound
        - compute the mid
        - create a new child node for the mid-element
        - if the mid element is less than the current node value, set the child node as the left child of the current node
        - else set it as the right child
        - the child node is now the root of a subtree, we have to create its left and right children/subtrees
            - push the new child node and left subarray start and end pointers to the queue
            - push the new child node and right subarray start and end pointers to the queue
- return root


## Complexity
- Time complexity:
  - Recursive Approach O(n)
  - Iterative Approach O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
  - Recursive Approach O(recursion stack + tree) → O(h + n) → O(n + n) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
  - Iterative Approach O(queue + tree) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:
    
    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
        
        # recursive
        def helper(l, r):
            if l > r:
                return None
            m = (l + r) // 2
            root = TreeNode(nums[m])
            root.left = helper(l, m - 1)
            root.right = helper(m + 1, r)
            return root

        return helper(0, len(nums) - 1)

        # iterative
        q = deque()
        l, r = 0, len(nums) - 1
        m = (l + r) // 2
        root = TreeNode(nums[m])
        q.append((root, l, m - 1))
        q.append((root, m + 1, r))
        while q:
            node, l, r = q.popleft()
            if l <= r:
                m = (l + r) // 2
                child = TreeNode(nums[m])
                if nums[m] < node.val:
                    node.left = child
                else:
                    node.right = child
                q.append((child, l, m - 1))
                q.append((child, m + 1, r))
        return root
```
