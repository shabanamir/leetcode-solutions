## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have to find the lowest common ancestor (LCA) of nodes p and q.

Traverse the binary search tree (BST). If a node value is:
- greater than p and q, then LCA lies on the left side of the node
- smaller than p and q, then LCA lies on the right side 
- else the current node is the LCA 


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- if current node value is greater than p and q, search in the left subtree for LCA
- if current node value is smaller than p and q, search in the right subtree for LCA
- else the current node is the LCA, return it

**Iterative Approach**
- set cur = root to track the current node
- loop until the end of the tree
    - if current node value is greater than p and q, set cur to the left child to search in the left subtree for LCA
    - if the current node value is smaller, set cur to the right child to search in the right subtree
    - else the current node is the LCA, return it


## Complexity
- Time complexity: 
  - Recursive Approach O(h) → O(n)
  - Iterative Approach O(h) → O(n)
  - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
  - Recursive Approach O(recursion stack) → O(h) → O(n)
  - Iterative Approach O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


class Solution:

    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':

        # recursive
        if root.val > p.val and root.val > q.val:
            return self.lowestCommonAncestor(root.left, p, q)
        if root.val < p.val and root.val < q.val:
            return self.lowestCommonAncestor(root.right, p, q)
        return root

        # iterative
        cur = root
        while cur:
            if cur.val > p.val and cur.val > q.val:
                cur = cur.left
            elif cur.val < p.val and cur.val < q.val:
                cur = cur.right
            else:
                return cur
```
