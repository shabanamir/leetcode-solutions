# <center> 105. Construct Binary Tree from Preorder and Inorder Traversal </center>


## Problem Description
[Click here](https://leetcode.com/problems/construct-binary-tree-from-preorder-and-inorder-traversal/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To create a tree, we need root element, left subtree elements, and right subtree elements.

Preorder = root → left → right <br>
Inorder = left → root → right

In preorder, the leftmost element is the root of the tree. In inorder, the left elements belong to the left subtree and the right elements to the right subtree.

Get the root from the preorder. Search for the root in the inorder to get the left and right subtree elements. Use a hashmap to search in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a hashmap and map the elements and indices of the inorder traversal list
    - *key = element*
    - *value = index of element*
- set i = 0 to track the index in preorder traversal list
- define a helper function with only list start (left) and end (right) pointers to avoid slicing and passing lists <br>
helper(left, right)
    - check base case
        - if left > right, return
    - make i nonlocal
        - *(this is to access outer function variables in the nested functions)*
    - get the root node element (current leftmost element in the preorder list) and create a node
    - increment i to pick the next element in the next call
    - find the root element index in the inorder list using the hashmap
    - get the elements before the index and recursively create the left subtree
    - get the elements after the index and recursively create the right subtree
    - return the root
- call helper() to create the tree and return the root node


## Complexity
- Time complexity: O(filing hashmap + recursion to create tree) → O(inorder list size + total tree nodes) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashmap + recursion stack) → O(n + h) → O(n + n) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        index = {n:i for i, n in enumerate(inorder)}
        i = 0

        def helper(l: int, r: int) -> Optional[TreeNode]:
            if l > r:
                return None
            nonlocal i
            root = TreeNode(preorder[i])
            i += 1
            mid = index[root.val]
            root.left = helper(l, mid - 1)
            root.right = helper(mid + 1, r)
            return root
            
        return helper(0, len(inorder) - 1)
```
