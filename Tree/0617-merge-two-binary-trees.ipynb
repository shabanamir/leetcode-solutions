# <center> 617. Merge Two Binary Trees </center>


## Problem Description
[Click here](https://leetcode.com/problems/merge-two-binary-trees/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Do a pre-order traversal (DFS) and merge the left subtrees and right subtrees.


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- check base cases
  - if both trees are null, return null
  - if one of the trees is null, return the other one
- sum the node values and update the first tree node value with the sum 
    - *(here the trees will be merged into the first tree, we can also use a separate tree for merging)*
- merge the left subtrees 
- merge the right subtrees
- return the first tree root
 
**Iterative Approach**
- check base cases
  - if both trees are null, return None
  - if one of the trees is null, return the other one
- create a stack and push the root of both trees as a pair
- loop until the stack is empty 
    - pop the stack top to get nodes of the two trees
    - continue if nodes are null
    - sum the nodes and update the value of the first tree node (node1) with the sum
    - if the left child of the first tree node is null, update it with the left child of the second tree node
    - else push the left child of both trees to the stack
    - if the right child of the first tree node is null, update it with the right child of the second tree node
    - else push the right child of both trees to the stack
- return the first tree root node


## Complexity
- Time complexity:
  - Recursive Approach O(tree DFS) → O(n)
  - Iterative Approach O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
  - Recursive Approach O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
  - Iterative Approach O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def mergeTrees(self, root1: Optional[TreeNode], root2: Optional[TreeNode]) -> Optional[TreeNode]:

        # recursive
        if not root1 and not root2:
            return None
        if not root1:
            return root2
        if not root2:
            return root1
        root1.val += root2.val
        root1.left = self.mergeTrees(root1.left, root2.left)
        root1.right = self.mergeTrees(root1.right, root2.right)
        return root1

        # iterative
        if not root1 and not root2:
            return None
        if not root1:
            return root2
        if not root2:
            return root1
        stack = [(root1, root2)]
        while stack:
            node1, node2 = stack.pop()
            if not node1 or not node2:
                continue
            node1.val += node2.val
            if not node1.left:
                node1.left = node2.left
            else:
                stack.append((node1.left, node2.left))
            if not node1.right:
                node1.right = node2.right
            else:
                stack.append((node1.right, node2.right))
        return root1
```
