# <center> 606. Construct String from Binary Tree </center>


## Problem Description
[Click here](https://leetcode.com/problems/construct-string-from-binary-tree/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Preorder = root → left → right

We need to do preorder traversal (DFS) and put the child nodes in parenthesis i.e root(left child)(right child). 

A null child is represented by an empty parenthesis i.e ()
- The empty parenthesis is unnecessary if both children are null or only the right child is null.
- The empty parenthesis is necessary if only the left child is null because, in preorder traversal, the left child comes first. So, if a node has only right child, we need empty brackets () to indicate that the left child is null otherwise the right child would be considered as the left child.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = an array to store the preorder traversal result. 
    - *(don't use a string because adding to a string is expensive and in array it takes O(1) time)*
- define a helper function for DFS (preorder traversal) <br>
dfs(node, result array) 
    - if the current node is null, return because we have reached the end of the tree/subtree
    - add the string form of the node value to the result array
    - if the node has a left child
        - add an opening bracket to the result array
        - recursively traverse the left child (subtree). This step will add the left subtree traversal output to res
        - add a closing bracket
    - if the node has a right child
        - add empty brackets to the result if the left child is null because brackets for a null left child are necessary
        - add an opening bracket
        - recursively traverse the right child (subtree)
        - add a closing bracket
- start dfs from the root node
- return the string form of the result array


## Complexity
- Time complexity: O(tree DFS + join()) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack + result array) → O(h + n) → O(n + n) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def tree2str(self, root: Optional[TreeNode]) -> str:
        res = []

        def dfs(node: Optional[TreeNode], res: list[str]) -> None:
            if not node:
                return
            res.append(str(node.val))
            if node.left:
                res.append('(')
                dfs(node.left, res)
                res.append(')')
            if node.right:
                if not node.left:
                    res.append('()')
                res.append('(')
                dfs(node.right, res)
                res.append(')')

        dfs(root, res)
        return ''.join(res)
```
