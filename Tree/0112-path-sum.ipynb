# <center> 112. Path Sum </center>


## Problem Description
[Click here](https://leetcode.com/problems/path-sum/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find a root-leaf path, do DFS.

Check the left and right subtree and decrease the sum by the current node value. Return true if you reach a leaf node and the remaining sum is equal to 0.


## Approach
<!-- Describe your approach to solving the problem. -->
**Recursive Approach**
- check base cases
  - if current node is null, we have reached the end of the tree and not found the sum, return false
- subtract current node value from the target sum to get remaining sum
- if remaining sum is 0 and current node is a leaf node, return true
- else check the left and right subtree

**Iterative Approach**
- check base cases
  - if current node is null, we have reached the end of the tree and not found the sum, return false
- create a stack and push the root node and remaining target as a pair
- loop until the stack is empty i.e until the end of the tree
    - pop stack top to get the current node and remaining sum
    - if remaining sum is 0 and current node is a leaf node, return true
    - push the left child and remaining sum to the stack
    - push the right child and remaining sum to the stack
- if the loop completes without returning true, there's no root-leaf path with target sum, return false


## Complexity
- Time complexity: 
  - Recursive Approach O(tree DFS) → O(n)
  - Iterative Approach O(tree DFS) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
  - Recursive Approach O(recursion stack) → O(h) → O(n)
    - *height (h) depends on the tree structure, balanced tree height is logn and skewed tree height is n*
  - Iterative Approach O(stack) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right


class Solution:

    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:

        # recursive
        if not root:
            return False
        targetSum -= root.val
        if targetSum == 0 and not root.left and not root.right:
            return True
        return self.hasPathSum(root.left, targetSum) or self.hasPathSum(root.right, targetSum)
        
        # iterative
        if not root:
            return False
        stack = [(root, targetSum - root.val)]
        while stack:
            node, cur_sum = stack.pop()
            if cur_sum == 0 and not node.left and not node.right:
                return True
            if node.left:
                stack.append((node.left, cur_sum - node.left.val))
            if node.right:
                stack.append((node.right, cur_sum - node.right.val))
        return False
```
