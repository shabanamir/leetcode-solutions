# <center> 435. Non-overlapping Intervals </center>


## Problem Description
[Click here](https://leetcode.com/problems/non-overlapping-intervals/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to return the number of overlapping intervals.

Sort the intervals by end time. Traverse the sorted intervals and check for overlapping. There's overlapping if the current interval start time is less than the previous interval end time and vice versa.


## Approach
<!-- Describe your approach to solving the problem. -->
- sort the intervals by end time
- set res = 0 to count the overlapping intervals
- set prev_end = end time of the first interval to track the previous interval's end time 
- traverse the intervals <br>
start from the second interval
    - if current interval start time >= previous interval end time, there's no overlapping
        - update prev_end to end time of current interval
    - else, there's overlapping
        - increment res
- return res


## Complexity
- Time complexity: O(sort + traversal) → O(nlogn + n) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->

## Code

```python
class Solution:

    def eraseOverlapIntervals(self, intervals: List[List[int]]) -> int:
        intervals.sort(key= lambda x: x[1])
        res = 0
        prev_end = intervals[0][1]
        for i in range(1, len(intervals)):
            if intervals[i][0] >= prev_end:
                prev_end = intervals[i][1]
            else:
                res += 1
        return res
```
