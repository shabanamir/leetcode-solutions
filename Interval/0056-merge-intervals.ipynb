# <center> 56. Merge Intervals </center>


## Problem Description
[Click here](https://leetcode.com/problems/merge-intervals/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Sort the intervals by start time. Traverse the sorted intervals and merge the overlapping ones. There's overlapping if the current interval start time is less than the previous interval end time and vice versa.


## Approach
<!-- Describe your approach to solving the problem. -->
- sort the intervals by start time
- set res = an array to store the merged intervals
- add the first interval to the result array
- traverse the intervals
    - if the current interval start time > end time of the previous interval i.e last interval in the result array, there's no overlapping
        - add the current interval to the array
    - else there's overlapping
        - update the end time of the previous interval by taking max of end time of the current interval and the previous interval
- return the result array

## Complexity
- Time complexity: O(sort + traversal) → O(nlogn + n) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort + result array) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->

## Code

```python
class Solution:

    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        intervals.sort()
        res = [intervals[0]]
        for interval in intervals:
            if interval[0] > res[-1][1]:
                res.append(interval)
            else:
                res[-1][1] = max(res[-1][1], interval[1])
        return res
```
