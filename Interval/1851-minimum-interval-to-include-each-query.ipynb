## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the smallest size interval for each query. Use min heap because it returns the min (smallest) element in O(1) time.

interval size = interval end - interval start + 1


## Approach
<!-- Describe your approach to solving the problem. -->
- sort intervals by start time 
- create an empty min heap
- create a hashmap to store the minimum interval for each query
    - *this is because we need to add the min interval for each query at its input index in the result array. We will be traversing the sorted form of queries and the index order is lost in sorting*
    - *using the hashmap, we will map the query and its min interval and in the end we will traverse the unsorted form of queries again to add the min interval for each query according to its index*
    - *there are also other ways to remember the index but the code is clear this way, and also time and space complexity is better.*
- set i = 0 to track the interval index
- traverse the sorted queries
    - *don't do inplace sort because we need the queries index for the final result*
    - while there are intervals to process and the interval is valid i.e current interval start time <= current query
        - get the start time l and end time r of the current interval
        - push the size and end time of the interval to the heap
            - *index 0 = interval size, index 1 = interval end*
        - increment i
    - while the heap is not empty and the interval is not valid i.e heap top interval end time < current query
       - pop interval
    - if heap is not empty after the loop, there's a min interval for the query
        - add the heap top interval size to the hashmap
    - else
        - add -1 for the query in the hashmap
- create an array and add the min interval for each query at its index and return the array


## Complexity
- Time complexity: O(sorting intervals + sorting queries + queries traversal * heap push pop + creating result array) → O(nlogn + qlogq + q * log(heap size) + q) → O(nlogn + qlogq + qlogn + q) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(sort + heap + hashmap + result array) → O(n + n + n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->

## Code

```python
class Solution:

    def minInterval(self, intervals: List[List[int]], queries: List[int]) -> List[int]:
        intervals.sort()
        min_heap = []
        res = {}
        i = 0
        for q in sorted(queries):
            while i < len(intervals) and intervals[i][0] <= q:
                l, r = intervals[i]
                heapq.heappush(min_heap, (r - l + 1, r))
                i += 1
            while min_heap and min_heap[0][1] < q:
                heapq.heappop(min_heap)
            res[q] = min_heap[0][0] if min_heap else -1
        return [res[q] for q in queries]
```
