# <center> 1288. Remove Covered Intervals </center>


## Problem Description
[Click here](https://leetcode.com/problems/remove-covered-intervals/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to return the number of intervals that can't be covered.

Sort intervals by the start (left bound) and if the start is same, sort by the end (right bound) in reverse order. Traverse the sorted intervals and check for intervals that can't be covered. An interval can't be covered if its end > previous interval end and vice versa.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to count remaining intervals
- sort interval by start. If the start is same, sort by the highest end in reverse order
- set prev_end = 0 to track the previous longest end
- traverse the intervals <br>
for start, end in intervals
    - if current interval end > previous longest end, the interval can't be covered
        - increment res
        - update prev_end because the current end is now the longest
- return res


## Complexity
- Time complexity: O(sort + traversal) → O(nlogn + n) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->

## Code

```python
class Solution:

    def removeCoveredIntervals(self, intervals: List[List[int]]) -> int:
        res = 0
        intervals.sort(key=lambda x: (x[0], -x[1]))
        prev_end = 0
        for _, end in intervals:
            if end > prev_end:
                res += 1
                prev_end = end
        return res
```
