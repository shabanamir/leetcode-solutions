## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use insertion sort

The intervals array is sorted. Find the position to insert the new interval in the sorted array. Merge intervals if there's overlap.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the intervals list is empty, return the new interval in a list
    - if the interval insert position is at corners (start or end).
        - if the last interval end < new interval start, insert the new interval at the end and return the intervals
        - if the first interval start > new interval end, insert the interval at the start and return intervals 
- set res = an empty array to build the result array
- traverse the intervals <br>
for each interval in the intervals
    - if the new interval end < current interval start, we found the insert position, and there's no overlapping
        - add the new interval to the result array
        - return intervals in the result array and remaining intervals from the input array
    - if the new interval start > current interval end, we have not found the insert position
        - add the current interval to the result array
    - else there's overlapping
        - merge the current interval and new interval by updating the start and end of the new interval
        start = min of the start of current and new interval
        end = max of end of current and new interval
- if the loop completes without returning anything, the merged new interval position is at the end, add the updated new interval to the result array
- return the result array


## Complexity
- Time complexity: O(input array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result array) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->

## Code

```python
class Solution:

    def insert(self, intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:
        if not intervals:
            return [newInterval]
        if intervals[-1][1] < newInterval[0]:
            intervals.append(newInterval)
            return intervals
        if intervals[0][0] > newInterval[1]:
            intervals.insert(0, newInterval)
            return intervals
        
        res = []
        for i, interval in enumerate(intervals):
            if newInterval[1] < interval[0]:
                res.append(newInterval)
                return res + intervals[i:]
            elif newInterval[0] > interval[1]:
                res.append(interval)
            else:
                newInterval[0] = min(newInterval[0], interval[0])
                newInterval[1] = max(newInterval[1], interval[1])
        res.append(newInterval)
        return res
```
