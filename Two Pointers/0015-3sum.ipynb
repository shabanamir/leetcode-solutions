## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
For 3 sum, we need three numbers that sum up to the target i.e 0. If we fix one number, we need to find the two others i.e we are left with the two sum problem.

For two sum, we can either use a hashmap or sorting and two-pointers approach. 

Just add an outer loop in the [Two Sum I Solution](https://leetcode.com/problems/two-sum/solutions/3444631/o-n-time-o-n-space-solution-explained/) or [Two Sum II Solution](https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/solutions/3525831/o-n-time-o-1-space-solution-explained/).


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a hashset to avoid the duplicate triplets in the result
- set n = array size
- sort the array in ascending order
- traverse the array to get the first number, *loop until the end - 2 to keep index inbound*
    - if the first num is positive
        - break the loop because we can't make it zero by searching after it
            - *(the array is sorted so the next numbers are greater)*
    - if the first num is a duplicate, skip it
    - use two pointers, left and right, to find the other two numbers
        - *left l = next left index i.e current index + 1*
        - *right r = rightmost index i.e array size - 1*
    - loop until left pointer < right pointer
        - compute the current sum
        - if the sum is less than 0, the left number is too small       
            - move the left pointer
        - if the sum is greater than 0, the right number is too big
            - move the right pointer
        - else the sum is 0
            - add the three numbers to the result set
            - loop to check duplicates on the left side, if there is a duplicate, move the left pointer to skip it
            - do the same for the right side duplicates
            - move both pointers to check other numbers because we need all possible triplets
- return the result set


## Complexity
- Time complexity: O(sort + for loop * while loop) → O(nlogn + n * n) → O(n * n) → O(n²)
    - *the two while loops for the duplicate check will not run n times*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = set()
        n = len(nums)
        nums.sort()
        for i in range(n - 2):
            if nums[i] > 0:
                break
            if i > 0 and nums[i] == nums[i - 1]:
                continue
            l = i + 1
            r = n - 1
            while l < r:
                cur_sum = nums[i] + nums[l] + nums[r]
                if cur_sum < 0:
                    l += 1
                elif cur_sum > 0:
                    r -= 1
                else:
                    res.add((nums[i], nums[l], nums[r]))
                    while l < r and nums[l] == nums[l + 1]:
                        l += 1
                    while l < r and nums[r] == nums[r - 1]:
                        r -= 1
                    l += 1
                    r -= 1
        return res
```
