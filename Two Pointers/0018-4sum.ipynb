## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
1. Simple way to solve 4-sum problem is to reduce it to 3-sum problem which can further be reduced to 2-sum problem. Just add an outer for loop in the [3-sum solution](https://leetcode.com/problems/3sum/solutions/3527300/o-n-time-o-n-space-solution-explained).

2. Design an n-sum solution where n >= 2


## Approach
<!-- Describe your approach to solving the problem. -->


## Complexity
- Time complexity: O(sort + n_sum) → O(nlogn + n$^{n-1}$) → O(nlogn + n$^{4-1}$) → (nlogn + n$^3$) → O(n$^3$)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        return self.n_sum(nums, target, 4, 0, len(nums)-1, [], [])

    def n_sum(self, nums, target, n, l, r, subres, res):
        if r - l + 1 < n or n < 2 or target < nums[l] * n or target > nums[r] * n:  
            return res
        if n == 2: 
            while l < r:
                s = nums[l] + nums[r]
                if s < target:
                    l += 1
                elif s > target:
                    r -= 1
                else:
                    res.append(subres + [nums[l], nums[r]])
                    l += 1
                    while l < r and nums[l] == nums[l-1]:
                        l += 1
        else:
            for i in range(l, len(nums)-n+1):
                if i == l or (i > l and nums[i] != nums[i-1]):
                    self.n_sum(nums, target-nums[i], n-1, i+1, r, subres+[nums[i]], res)
        return res
```
