# <center> 1984. Minimum Difference Between Highest and Lowest of K Scores </center>


## Problem Description
[Click here](https://leetcode.com/problems/minimum-difference-between-highest-and-lowest-of-k-scores/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find a subarray of size k with the minimum difference between its max and min element.

Sort the array because after sorting:
- any contiguous k elements are as close to each other as possible
- the min element is at the start and the max element is at the end. So, we can easily find the difference

Use sliding window to find the subarray of size k. Find the difference between the first and last element of each window (subarray) and return the minimum.


## Approach
<!-- Describe your approach to solving the problem. -->
- sort array
- set res = int max or the last number of the sorted array (array max) to track the minimum difference
- set left pointer l = 0 and right pointer r = k - 1 to track the left and right end of the window i.e subarray of size k
- traverse the sorted array
    - find the difference of the current window and update res by taking min of current window difference and previous min difference
    - slide the window by moving the pointers to check the next window
- return res


## Complexity
- Time complexity: O(sort + loop) → O(nlogn + n-k+1) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def minimumDifference(self, nums: List[int], k: int) -> int:
        nums.sort()
        res = nums[-1]
        l, r = 0, k - 1
        while r < len(nums):
            res = min(res, nums[r] - nums[l])
            l += 1
            r += 1
        return res
```
