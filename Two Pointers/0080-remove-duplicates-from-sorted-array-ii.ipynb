## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
This is same as [Remove Duplicates from Sorted Array I](https://leetcode.com/problems/remove-duplicates-from-sorted-array/description/) problem, but here unique element is the one with count <= 2.

In a sorted array, duplicates are adjacent. 

Use two pointers, one to track the position for placing the unique element (count <= 2) and the other to traverse the array. 

Traverse the array and move unique elements towards the left side using the pointers.


## Approach
<!-- Describe your approach to solving the problem. -->
- set i = 2 to track the index for placing the unique element
    - *2 because the first two elements at index 0 and 1 are unique*
- set j = 2 to traverse the array
- traverse the array starting from the third position
    - if current element is not equal to second previous unique element, it is unique i.e count <=2
        - add the element at index i 
        - move the pointer i
    - else the current element is a duplicate (count > 2), skip it
- return i as total number of unique elements


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def removeDuplicates(self, nums: List[int]) -> int:
        i = 2
        for j in range(2, len(nums)):
            if nums[j] != nums[i - 2]:
                nums[i] = nums[j]
                i += 1
        return i  
```
