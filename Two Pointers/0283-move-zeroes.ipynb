# <center> 283. Move Zeroes </center>


## Problem Description
[Click here](https://leetcode.com/problems/move-zeroes/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use two pointers, one to track current index and other to track zeroes. When you encounter a non-zero element, replace it with the leftmost zero. <br>
*(leftmost zero index = current index - current count of zeroes)*

For example: nums = [0,1,0,3,12]
- i = 0,  nums[0] = 0 → z_count = 1
- i = 1, nums[1] = 1 → swap with leftmost 0 at index cur_index - z_count = 1 - 1 = 0 → [1, 0, 0, 3, 12]
- i = 2, nums[2] = 0 → z_count = 2
- i = 3, nums[3] = 3 → swap with leftmost 0 at index 3 - 2 = 1 → [1,3, 0, 0, 12]
- i = 4, nums[4] = 4 → swap with leftmost 0 at index 4 - 2 = 2 → [1, 3, 12, 0, 0]


## Approach
<!-- Describe your approach to solving the problem. -->
- set z = 0 to count zeroes
- traverse the array
    - if current element is 0
        - increment zero count
    - else the element is not zero, if current count of zeroes is greater than 0
        - swap current element with the leftmost zero 


## Complexity
- Time complexity: O(array traversal) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def moveZeroes(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        z = 0
        for i in range(len(nums)):
            if nums[i] == 0:
                z += 1
            elif z > 0:
                nums[i], nums[i - z] = nums[i - z], nums[i]
```
