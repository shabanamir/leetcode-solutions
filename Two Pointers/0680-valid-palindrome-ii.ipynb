## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a palindrome, the left and right characters are the same.

Use two pointers to check the left and right characters. If they are not equal, check if they can be made equal by skipping a left or right character. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = 0
- set right pointer r = rightmost position i.e length of string - 1
- traverse the string and check the left and right characters
    - if characters at positions l and r are not equal
        - check if removing the right character or right character makes the string palindrome, and return the result
            - *(for this check, we have defined a helper function. You can also use nested loops, slicing, or recursion. Time complexity will remain the same in all the cases. Slicing will increase the space complexity)*
    - else left and right characters are equal
        - move pointers to continue checking the string
- if the loop completes without returning false, the string is a valid palindrome, return true


## Complexity
- Time complexity: O(string traversal * palindrome check) â†’ O(n)
    - *outer loop for string traversal will take n, the inner loop (helper function) for palindrome check will also take at most n, but either the outer or the inner loop will fully execute, so total complexity is n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1) 
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:
    
    def validPalindrome(self, s: str) -> bool:
        
        def is_palindrome(l: int, r: int) -> bool:
            while l < r:
                if s[l] != s[r]:
                    return False
                l += 1
                r -= 1
            return True

        l, r = 0, len(s) - 1
        while l < r:
            if s[l] != s[r]:
                return is_palindrome(l, r - 1) or is_palindrome(l + 1, r)
            l += 1
            r -= 1
        return True
```
