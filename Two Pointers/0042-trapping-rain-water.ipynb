# <center> 42. Trapping Rain Water </center>


## Problem Description
[Click here](https://leetcode.com/problems/trapping-rain-water/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Water can only be trapped if there are high blocks on both the left and right of the current block.<br>
amount of water trapped = current block height - min(max height on the left, max height on the right)

Use two-pointers to find the minimum of the max height on the left and right. If left-max is smaller, find water trapped on the left side, else on the right.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base case
    - if blocks are less than or equal to 2, no water can be trapped
        - return 0 
- set res = 0 to track the total water trapped
- set left pointer l = leftmost index, right pointer r = rightmost index
- set max_l = left block height, max_r = right block height to track the maximum heights on the left and right sides
- loop until left pointer < right pointer to compare height and find the total water trapped
    - if the maximum height on the left is smaller
        - move the left pointer 
        - update max_l
        - compute water trapped on the left side and add it to res
    - else 
        - move the right pointer 
        - update max_r
        - compute water trapped on the right side and add it to res
- return res


## Complexity
- Time complexity: O(loop) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->

## Code

```python
class Solution:

    def trap(self, height: List[int]) -> int:
        if len(height) <= 2:
            return 0
        res = 0 
        l, r = 0, len(height) - 1
        max_l, max_r = height[l], height[r]
        while l < r:
            if max_l < max_r:
                l += 1
                max_l = max(max_l, height[l])
                res += max_l - height[l]
            else:
                r -= 1
                max_r = max(max_r, height[r])
                res += max_r - height[r]
        return res
```
