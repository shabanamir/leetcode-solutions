# <center> 344. Reverse String </center>


## Problem Description
[Click here](https://leetcode.com/problems/reverse-string/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use two pointers, left and right, and swap the characters at left and right positions.


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = leftmost position i.e 0
- set right pointer r = right most position i.e length of string - 1
- loop until left pointer <= right pointer
  - swap the left and right characters
  - move the pointers i.e increment l and decrement r


## Complexity
- Time complexity: O(loop) → O(n/2) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def reverseString(self, s: List[str]) -> None:
        """
        Do not return anything, modify s in-place instead.
        """
        l, r = 0, len(s) - 1
        while l <= r:
            s[l], s[r] = s[r], s[l]
            l += 1
            r -= 1
```
