# <center> 26. Remove Duplicates from Sorted Array </center>

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a sorted array, duplicates are adjacent. 

Use two pointers, one to track the position for placing the unique element (count = 1) and the other to traverse the array. 

Traverse the array and move unique elements towards the left side using the pointers.


## Approach
<!-- Describe your approach to solving the problem. -->
- set i = 1 to track the index for placing the unique element
    - *(1 because the first element at index 0 is always unique)*
- set j = 1 to traverse the array
- traverse the array starting from the second position
    - if the adjacent elements are not equal *(or current element is not same as previous unique element)*, the current element is unique  
        - add the element at index i 
        - move the pointer i
    - else the current element is a duplicate, skip it
- return i as total number of unique elements


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def removeDuplicates(self, nums: List[int]) -> int:
        i = 1
        for j in range(1, len(nums)):
            if nums[j] != nums[j - 1]:  # or nums[j] != nums[i - 1]
                nums[i] = nums[j]
                i += 1
        return i
```
