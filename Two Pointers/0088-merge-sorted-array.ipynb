# <center> 88. Merge Sorted Array </center>


## Problem Description
[Click here](https://leetcode.com/problems/merge-sorted-array/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Both arrays are sorted and we need to use the first array for merging.

Each time, take one element from the end of both arrays using two pointers and compare them. Select the largest element, add it at the end of the first array, and move the pointers accordingly. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set i = -1 to track the right index in the first array 
- set j = -1 to track the right index in the second array
- set k = m + n - 1 to track the right index of the merged array i.e index for placing the largest element in the first array
- loop until we have added all elements of the second array 
    - if the index is inbound and the first array element is the largest
        - add it at position k
        - move the first array pointer i.e decrement i 
    - else second array element is the largest
        - add it at position k
        - move the second array pointer i.e and decrement j
    - move k because in each iteration we are adding/fixing an element in the first array


## Complexity
- Time complexity: O(loop) → O(total elements in both arrays) → O(m + n) 
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        """
        Do not return anything, modify nums1 in-place instead.
        """
        i = m - 1      
        j = n - 1 
        k = m + n - 1 
        while j >= 0:
            if i >= 0 and nums1[i] > nums2[j]:
                nums1[k] = nums1[i]
                i -= 1
            else:
                nums1[k] = nums2[j]
                j -= 1
            k -= 1   
```
