# <center> 1968. Array With Elements Not Equal to Average of Neighbors </center>


## Problem Description
[Click here](https://leetcode.com/problems/array-with-elements-not-equal-to-average-of-neighbors/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
An element nums[i] is the average of its neighbors (nums[i-1], nums[i+1]) if one neighbor is smaller than the number and the other is greater i.e:
- nums[i-1] < nums[i] < nums[i+1]
- nums[i-1] > nums[i] > nums[i+1]

Traverse the array and swap the number (nums[i]) with its neighbor (nums[i+1]) if you encounter any of the above cases.


## Approach
<!-- Describe your approach to solving the problem. -->
- traverse the array, *start traversal from the second position until the last position to keep the index inbound*
    - if there's an increasing sequence (nums[i-1] < nums[i] < nums[i+1]) or a decreasing sequence (nums[i-1] > nums[i] > nums[i+1])
        - swap nums[i] and nums[i+1] to break the sequence
- return the array


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def rearrangeArray(self, nums: List[int]) -> List[int]:
        for i in range(1, len(nums) - 1) :
            if nums[i-1] < nums[i] < nums[i+1] or nums[i-1] > nums[i] > nums[i+1]:
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
        return nums
```
