# <center> 1768. Merge Strings Alternately </center>


## Problem Description
[Click here](https://leetcode.com/problems/merge-strings-alternately/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use two pointers, one pointer for each string. Alternately choose the character from each string and move the pointers.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = an array to store the merged string characters
    - *(use an array because string concatenation is expensive)*
- set i = 0 to track the index in the string1 and j = 0 to track the index in string2
    - *(we can also use one pointer but the code is clear this way)*
- loop until one of the strings ends
    - take a character from both strings one by one and add it to res
    - increment pointers to move to the next character
- if string1 is the longer string
    - add its remaining characters to res
- else add string2 remaining characters
- convert the result array into a string and return it


## Complexity
- Time complexity: O(loop + join) → O(min(string1, string2) + size of merged string) → O(min(m, n) + m + n) → O(m + n)
    - *m = string1 size*
    - *n = string2 size*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(result array) → O(size of the merged string) → O(m + n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def mergeAlternately(self, word1: str, word2: str) -> str:
        res = []
        i, j = 0, 0
        while i < len(word1) and j < len(word2):
            res.append(word1[i])
            res.append(word2[j])
            i += 1
            j += 1
        if len(word1) > len(word2):
            res.append(word1[i:])
        else:
            res.append(word2[j:])
        return ''.join(res)
```
