## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The array is sorted, with smaller numbers on the left side and greater on the right, we can use that to our advantage.

Use two pointers to track the numbers on the left and right. Each time, select a smaller number from the left and a greater one from the right. If their sum is:
- less than the target, it means the left number is too small, we need to move the left pointer
- greater than the target, the right number is too big, move the right pointer
- else we got the target sum


## Approach
<!-- Describe your approach to solving the problem. -->
- set l = leftmost position to track left side elements
- set r = rightmost position to track right side elements
- loop until left pointer < right pointer i.e move pointers towards each other until we get the target sum 
    - compute the current sum by adding elements at positions l and r
    - if the current sum is less than the target
        - move the left pointer i.e increment l
    - if the current sum is greater than the target
        - move the right pointer i.e decrement r
    - else current sum is equal to the target
        - return l+1, r+1 as the positions of the two numbers that sum up to the target


## Complexity
- Time complexity: O(loop) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        l, r = 0, len(numbers) - 1
        while l < r:
            cur_sum = numbers[l] + numbers[r]
            if cur_sum < target:
                l += 1
            elif cur_sum > target:
                r -= 1
            else:
                return l + 1, r + 1
```
