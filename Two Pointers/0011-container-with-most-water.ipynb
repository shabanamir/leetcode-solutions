# <center> 11. Container With Most Water </center>


## Problem Description
[Click here](https://leetcode.com/problems/container-with-most-water/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use two pointers, set one pointer to the left and one to the right of the array. At each step, calculate the area formed by the two pointers. As we want to maximize the area, move the pointer that points to the lower line (height).

area = width * height = distance between the pointers * lower height = (r - l) * lower height


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the max area
- set max_h = max height
- set left pointer l = leftmost index and right pointer r = rightmost index
- traverse the array
    - get current width i.e r - l
    - if the left height is lower
        - compute the area using the formula above and update res by taking max of current area and previous max area
        - move the left pointer i.e increment l
    - else the right height is lower
        - compute the area and update res 
        - move the right pointer i.e decrement r
    - if the current max area is greater than the area formed by current width * max height, there's no need to check further because the next areas are smaller, so break the loop
- return res as the max area


## Complexity
- Time complexity: O(finding max + array traversal) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxArea(self, height: List[int]) -> int:
        res = 0
        max_h = max(height)
        l, r = 0, len(height) - 1
        while l < r:
            width = r - l
            if height[l] < height[r]:
                res = max(res, width * height[l])
                l += 1
            else:
                res = max(res, width * height[r])
                r -= 1
            if res > width * max_h:
                break
        return res
```
