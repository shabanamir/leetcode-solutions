# <center> 1498. Number of Subsequences That Satisfy the Given Sum Condition </center>


## Problem Description
[Click here](https://leetcode.com/problems/number-of-subsequences-that-satisfy-the-given-sum-condition/solutions/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Sort the array to easily check for the minimum and maximum element because the left elements are smaller and the right are greater.

Use two-pointers, left and right. Choose an element from the left side as the minimum and find the maximum number on the right side where nums[l] + nums[r] ≤ target. If we can find such two numbers, we get a valid range. 

Find the valid ranges, and for each valid range find the total number of subsequences.

No. of subseq in a range l to r = 2$^{r - l}$


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the total number of subsequences
- sort the array
- set left pointer l = leftmost index, right pointer r = rightmost index
- loop until left pointer <= right pointer to search for the valid subsequences
    - if the sum of the minimum (left) and maximum (right) of the current range is greater than the target, the range is invalid because the max/right number is too big
        - move the right pointer i.e decrement r
    - else the sum is less or equal to the target, the range is valid
        - compute the number of subsequences in the current range using the formula and add it to res
        - move left pointer to search for other ranges
- mod the result by 10$^9$ + 7 and return it


## Complexity
- Time complexity: O(sort + loop + power) → O(nlogn + n + logn) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def numSubseq(self, nums: List[int], target: int) -> int:
        res = 0
        nums.sort()
        l, r = 0, len(nums) - 1
        while l <= r:
            if nums[l] + nums[r] > target:
                r -= 1
            else:
                res += 1 << (r - l)   # or pow(2, r - l)
                l += 1
        return res % (10**9 + 7)
```
