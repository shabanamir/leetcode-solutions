## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a palindrome, the left and right characters are equal.

Use two-pointers, left and right, and check the characters at positions left and right. Skip characters other than alphanumeric.


## Approach
<!-- Describe your approach to solving the problem. -->
- set left pointer l = leftmost position i.e 0
- set right pointer r = rightmost position i.e length of string - 1
- loop until left pointer is less than right pointer and check the characters at position left and right
  - if left char is not alphanumeric
    - skip it by moving the left pointer i.e increment l
  - if right char is not alphanumeric
    - skip it by moving the right pointer i.e decrement r
  - else both are alphanumeric
  if they are not the same, there's no palindrome
    - return false
  - else both are alphanumeric and same
    - move both pointers
- if the loop completes without returning false, the string is a palindrome, return True


## Complexity
- Time complexity: O(loop) â†’ O(n)

- Space complexity: O(1)

## Code

```python
class Solution:

    def isPalindrome(self, s: str) -> bool:
        l, r = 0, len(s) - 1
        while l < r:
            if not s[l].isalnum():
                l += 1
            elif not s[r].isalnum(): 
                r -= 1
            elif s[l].lower() != s[r].lower():
                return False
            else:
                l += 1
                r -= 1
        return True
```
