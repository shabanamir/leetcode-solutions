## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use reversal algorithm

For example: nums = [1, 12, -3, 4, 5, 6, 2], k = 3 <br>
right rotate = [5, 6, 2, 1, 12, -3, 4] <br>
left rotate = [4, 5, 6, 2, 1, 12, -3] <br>

To right-rotate an array, we need to move the last/right k elements to the start/left
- reverse the whole array, this step will move the last k element to start but in reverse order <br>
[1, 12, -3, 4, 5, 6, 2] → [2, 6, 5, 4, -3, 12, 1]
- reverse the first k elements to correct the order <br>
[**2, 6, 5**, 4, -3, 12, 1] → [**5, 6, 2**, 4, -3, 12, 1]
- reverse the last remaining n-k elements to correct the order <br>
[5, 6, 2, **4, -3, 12, 1**] → [5, 6, 2, **1, 12, -3, 4**]

To left-rotate an array, we need to move the first/left k elements to the end/right
- reverse the whole array to move the first k elements to the end
- reverse the last k elements to correct the order
- reverse the first remaining n-k elements to correct the order

For reversing, either use a library function reverse() or define a function to reverse the array using two-pointers.

## Approach
<!-- Describe your approach to solving the problem. -->
- set n = size of the input array
- check base cases
    - if k is 0 or equal to the length of the array, the rotation will result in the same array so return
- mod k by n to keep k inbound
- define a helper function for reversing 
    - *reverse(array start, array end)*
        - set l = leftmost index i.e start
        - set r = rightmost index i.e array end position - 1
        - loop until left pointer < right pointer
            - swap the elements at index l and r
            - move pointers i.e increment l and decrement r
- reverse all n elements
- reverse first k elements
- reverse remaining last n-k elements


## Complexity
- Time complexity: O(reverse n + reverse k + reverse n - k) → O(n + k + n - k) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def rotate(self, nums: List[int], k: int) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        n = len(nums)
        if k == 0 or k == n:
            return
        k %= n

        def reverse(start: int, end: int) -> None:
            l, r = start, end - 1
            while l < r:
                nums[l], nums[r] = nums[r], nums[l]
                l += 1
                r -= 1
        
        n = len(nums)
        reverse(0, n)
        reverse(0, k)
        reverse(k, n)
```
