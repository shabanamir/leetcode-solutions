# <center> 881. Boats to Save People </center>


## Problem Description
[Click here](https://leetcode.com/problems/boats-to-save-people/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Sort the array so that the lightest person is at the start and the heaviest is at the end. Use two pointers, left and right, to find whether the two persons at position left and right can be in one boat. 


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there's only one person, 1 boat is required
- sort the array in ascending order 
- set res = 0 to track the total boats required
- set left pointer l = leftmost index, right pointer r = rightmost index
- loop until left pointer <= right pointer to compare the weights and find the total number of boats
    - if the weight sum of the two persons at position l and r <= limit, they can be in one boat
        - move both pointers i.e increment l and decrement r
        - increment res because a boat is required
    - else the weight of the person at position r is greater and he has to go in a single boat
        - move the right pointer 
        - increment res because a boat is required
- return res


## Complexity
- Time complexity: O(sort + array traversal) → O(nlogn + n) → O(nlogn) 
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def numRescueBoats(self, people: List[int], limit: int) -> int:
        if len(people) == 1:
            return 1
        people.sort()
        res = 0
        l, r = 0, len(people) - 1
        while l <= r:
            if people[l] + people[r] <= limit:
                l += 1
                r -= 1
                res += 1
            else:
                r -= 1
                res += 1
        return res
```
