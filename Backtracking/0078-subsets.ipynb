## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The total number of subsets for an array of size n is 2$^n$. 

We can make two choices for an element: 
- include the element in the subset and move ahead
- don't include the element ...

We need to explore all possible subsets. Use backtracking algorithm, because it involves considering all possible choices/solutions/paths at each step and choosing the best one.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list to store all subsets
- define a helper function for backtracking, it will consider two paths at each step: including the current index element and excluding it. In this way, it will generate all subsets <br>
backtrack(current index, current path/subset)
    - if the index is out of bounds, it means we have considered all elements
        - add the current subset copy to the result list and return
            - *(add the copy to avoid overwriting this subset during future backtracking)*
    - add the current index element to the current subset and backtrack from next index i.e i + 1 to explore subsets with the current element
    *(an element can appear only once in a subset, so we backtrack from the next index)*
    - remove the current index element from the current subset and backtrack from the next index i.e i + 1 to explore subsets without the current element
- start backtracking from index 0 with an empty subset
- return the result list


## Complexity
- Time complexity: O(copy * generating subsets) → O(subset size * recursion breadth) → O(n * calls at each level ^ total levels) → O(n * 2$^n$)
    - *we will consider the time complexity of copy() here because an array of size has 2$^n$ subsets and we are copying each subset to add to our result*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: (recursion stack + result list) → (recursion depth/levels + 2$^n$ subsets of size 0 to n) → O(n + 2$^n$ * n) → O(n * 2$^n$)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


- *For recursive functions:*
    - *time complexity = recursion breadth = total recursive calls*
    - *space complexity = recursion depth/levels = total return statements*
    - *for this problem, there are n levels and at each level, we make two choices/calls i.e to include or exclude the elements*


## Code

```python
class Solution:

    def subsets(self, nums: List[int]) -> List[List[int]]:
        res = []

        def backtrack(i: int, path: List[int]) -> None:
            if i >= len(nums):
                res.append(path.copy())
                return
            path.append(nums[i])
            backtrack(i + 1, path)
            path.pop()
            backtrack(i + 1, path)

        backtrack(0, [])
        return res
```
