## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Same as [Subset I](https://leetcode.com/problems/subsets/), the difference is that Subset II input contains duplicates.

The total number of subsets for an array of size n is 2$^n$. 

We can make two choices for an element: 
- include the element in the subset and move ahead
- don't include the element ...

Use backtracking to explore all possible subsets. As there are duplicates in the input, so sort the input first to group the duplicates and avoid recursion for them.


## Approach
<!-- Describe your approach to solving the problem. -->
*same as [Subset I solution](https://leetcode.com/problems/subsets/solutions/3914626/o-n-2-n-time-o-n-2-n-space-solution-explained/), just sort the input first to handle the duplicates*
- set res = a list to store the subsets
- sort the input
- define a helper function for backtracking <br>
backtrack(current index, current path/subset)
    - if the index is out of bounds, we have considered all elements
        - add the current subset copy to the result list, and return
            - *(add the copy to avoid overwriting this subset during future backtracking)*
    - add the current index element to the current subset
    - backtrack from next index i.e i + 1 to explore subsets with the current element
        - *(an element can appear only once in a subset, so we backtrack from the next index)*
    - remove the current index element from the current subset
    - increment the index while there are duplicates i.e current and next elements are the same
    - backtrack from next index i.e i + 1 to explore subsets without the current element
- start backtracking from index 0 with an empty subset
- return the result list


## Complexity
- Time complexity: O(sort + copy * generating subsets) → O(nlogn + subset size * recursion breadth) → O(nlogn + n * calls at each level ^ total levels) → O(nlogn + n * 2$^n$) → O(n * 2$^n$)
    - *we will consider the time complexity of copy() here because an array of size has 2$^n$ subsets and we are copying each subset to add to our result*
    - *we will not consider the time complexity of the while loop used to skip duplicates because it will not run n times in each recursion*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: (sort + recursion stack + result list) → (n + recursion depth/levels + 2$^n$ subsets of size 0 to n) → O(n + n + 2$^n$ * n) → O(2$^n$ * n) → O(n * 2$^n$)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def subsetsWithDup(self, nums: List[int]) -> List[List[int]]:
        res = []
        nums.sort()

        def backtrack(i: int, path: List[int]) -> None:
            if i == len(nums):
                res.append(path.copy())
                return
            path.append(nums[i])
            backtrack(i + 1, path)
            path.pop()
            while i + 1 < len(nums) and nums[i] == nums[i + 1]:
                i += 1
            backtrack(i + 1, path)

        backtrack(0, [])
        return res
```
