# <center> 698. Partition to K Equal Sum Subsets </center>


## Problem Description
[Click here](https://leetcode.com/problems/partition-to-k-equal-sum-subsets/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find if it is possible to divide the array into k subsets of equal sum.

The sum of each subset should be equal to  sum(array) / k

For each element, we can make two choices:
- include it in the subset
- don't include...

Use backtracking to explore all possible subsets and check if it is a valid subset i.e sum(subset) = sum(array) / k. To speed up the process, sort the array and use hash set.


## Approach
<!-- Describe your approach to solving the problem. -->
- find the total sum of the array
- check base cases 
    - if the total sum is not divisible by k, it is not possible to have k subsets of equal sum, return false
- sort the array in descending order
- find the target sum for the subset
- create a hashset to check for duplicates
    - *(we will add indices to the hashset, if an index is in the hashset it means we have used that index element)*
- define a helper function for backtracking <br>
backtrack(current index, remaining subsets, current subset sum)
    - if we have found k subsets, return true
    - if current subset sum is equal to the target sum, we have found a subset
        - backtrack to find remaining subsets and return the result if it is possible to find the subsets
    - loop in range i to size of array to check for the remaining elements in the array
        - if current element is a duplicate i.e the index is inbound and we have not used the previous element and previous element is same as the current
            - skip
        - if the current element is already used i.e its index is in the hashset or adding it to the current subset will exceed the target sum
            - skip
        - add current element index to the hashset to include it in the current subset
        - backtrack to find the k subsets, return true if it is possible
        - remove current element index from the set
    - if the loop executes without returning true, it is not possible to divide the array into k subsets, return false
- start backtracking from the first element at index 0


## Complexity
- Time complexity: O(sum + sort + recursion breadth for k subsets) → O(n + nlogn + calls at each level ^ total levels * k) → O(n + nlogn + 2$^n$ * k) → O(k * 2$^n$)
    - *n = total elements in the array*
    - *2 = for each element, we make two choices i.e include or exclude*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(sort + hashset + recursion stack) → O(n + n + recursion depth/levels) → O(n + n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def canPartitionKSubsets(self, nums: List[int], k: int) -> bool:
        total = sum(nums)
        if total % k:
            return False
        nums.sort(reverse = True)
        target = total / k
        seen = set()

        def backtrack(i: int, k: int, cur: int) -> bool:
            if k == 0:
                return True
            if target == cur:
                return backtrack(0, k - 1, 0)
            for j in range(i, len(nums)):
                if j > 0 and (j - 1) not in seen and nums[j] == nums[j - 1]:
                    continue
                if j in seen or cur + nums[j] > target:
                    continue
                seen.add(j)
                if backtrack(j + 1, k, cur + nums[j]):
                    return True
                seen.remove(j)
            return False

        return backtrack(0, k, 0)
```
