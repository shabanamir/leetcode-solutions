## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given some binary strings representing numbers in the range 0 to n. We need to return any one of the missing numbers.

- Approach 1: Backtracking <br>
use backtracking to explore all possible strings of length same as the input string length and check if it is the missing string i.e it is not in the input list. Convert the input list of strings to hash set to search in O(1) time.

- Approach 2 <br>
traverse all decimal numbers in the range 0 to n. Find the binary and check if it is present in the input list. You can also first convert all numbers to binary strings and then check but with that you would be doing some unnecessary conversions. Convert the input to hash set to search in O(1) time.

- Approach 3: Cantor's Diagonalization <br>
according to this, a binary number missing from a binary sequence can be constructed by flipping the 1st bit of s1, flipping the 2nd bit of s2, ... and flipping the nth bit of sn. <br>
For example:  <br>
input strings = ['111', '001', '010'] i.e [7, 1, 2] <br>
missing numbers are 3, 4, 5 <br>
we need only one missing number and the first missing number can be formed as: <br>
get 1st bit of string1 + get 2nd bit of string2 + get 3rd bit of string3 <br>
1 + 0 + 0 <br>
flip the bits <br>
0 + 1 + 1 <br>
concatenate <br>
0 + 1 + 1 = 011 = 3  <br>

**Approach 2**
- convert the input strings list to a set to search in O(1)
- set n = size of the input list
- loop in range 0 to n + 1 to check all numbers in the range 0 to n
    - get the binary string of the current number
        - *(binary string is prefixed with 0b, therefore formatting is required)*
    - if the current number binary string is not in the set, the current number is a missing number
        - return its binary string

**Approach 3**
- set res = a list to store the missing binary number bits
    - *(list because string concatenation is expensive)*
- loop in range 0 to the size of input list to iterate the input list of strings (binary numbers) and get the required bit, *(you can also use enumerate but the code is clear this way)*
    - get the required bit i.e 1st bit of 1st string, 2nd bit of 2nd string etc, flip it, and add it to the result list
- convert the result list to string to get the missing binary number and return it


## Complexity
- Time complexity:
    - Approach 1 O(recursion breadth) → O(calls at each level ^ total levels) → O(2$^n$)
    - Approach 2 O(set conversion + array traversal * decimal to binary conversion) → O(n + n * log(number)) → O(n + n) → O(n)
    - Approach 3 O(array traversal + converting list to string) → O(n + size of a binary string) → O(n + size of binary of 16 because n <= 16) → O(n + 5) → O(n)

## Code

```python
class Solution:

    def findDifferentBinaryString(self, nums: List[str]) -> str:
        
        # appraoch 1

        nums = set(nums)

        def backtrack(i: int, path: List[str]) -> str:
            if i == len(nums):
                res = ''.join(path)
                return res if res not in nums else None
            res = backtrack(i + 1, path)
            if res: 
                return res
            path[i] = '1'
            res = backtrack(i + 1, path)
            if res: 
                return res
            
        return backtrack(0, ['0' for s in nums])
        

        # appraoch 2

        seen = set(nums)
        n = len(nums)
        for num in range(n + 1):
            cur = f'{num:0>{n}b}'
            if cur not in seen:
                return cur


        # approach 3

        # res = ['1' if num[i] == '0' else '0' for i, num in enumerate(nums)]
        res = []
        for i in range(len(nums)):
            bit = '1' if nums[i][i] == '0' else '0'
            res.append(bit)
        return ''.join(res)
```
