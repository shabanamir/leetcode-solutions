# <center> 17. Letter Combinations of a Phone Number </center>


## Problem Description
[Click here](https://leetcode.com/problems/letter-combinations-of-a-phone-number/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given digits, and we need to return all possible letter combinations that the number could represent.

Each digit on a telephone button represents 3-4 letters. <br>
2 → abc <br>
3 → def and so on 

The total combinations of size n that a n digit number could represent is 4$^n$. 4 because in the worst case, a digit could represent 4 letters i.e 7 and 9.

Use a hashmap to map the digits to their corresponding letters and do backtracking to find all possible combinations.

## Approach
<!-- Describe your approach to solving the problem. -->
- check base case
    - if the input is empty, return an empty list
- create a list to store the combinations
- create a hashmap and map the digits to their letters
- define a helper function for backtracking <br>
backtrack(current index to get the current digit, current path/combination)
    - if the current combination and input size are the same, the current combination is valid because there are no more digits to explore
        - add current combination to the result and return
    - for each letter that the current digit represents
        - backtrack i.e explore combinations including the letter
- start backtracking from the first digit at index 0
- return the result list


## Complexity
- Time complexity: O(recursion breadth) → O(calls at each level ^ total levels) → O(4$^n$)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(recursion stack + result list) → O(recursion depth/levels + total combinations) → O(n + 4$^n$) → O(4$^n$) 
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def letterCombinations(self, digits: str) -> List[str]:
        if not digits:
            return []
        res = []
        digit_map = {'2': 'abc',
                     '3': 'def',
                     '4': 'ghi',
                     '5': 'jkl',
                     '6': 'mno',
                     '7': 'qprs',
                     '8': 'tuv',
                     '9': 'wxyz'}

        def backtrack(i: int, path:str):
            if len(path) == len(digits):
                res.append(path)
                return
            for c in digit_map[digits[i]]:
                backtrack(i + 1, path + c)

        backtrack(0, '')
        return res
```
