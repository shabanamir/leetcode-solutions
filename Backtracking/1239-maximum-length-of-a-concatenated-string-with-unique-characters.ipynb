# <center> 1239. Maximum Length of a Concatenated String with Unique Characters </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a list of strings. We need to return the length of the valid subsequence. A subsequence is valid, if it has no duplicate letters.

For each string, we have two choices:
- add it to the subsequence
- don't add it

Use backtracking to explore all possible subsequences and check if the subsequence is valid. To speed the process, remove duplicate strings and strings with duplicate letters, and use the hashset to search in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- preprocess the input i.e remove the duplicate strings and strings with duplicate letters. There are no duplicate strings in any of the test cases, so we will only handle the duplicate letter strings
- if after preprocessing:
    - there are no strings left, return 0
    - only 1 string left, return its length
- use a set to store current subseq letters and avoid duplicates
- define a helper function to check if there's overlap i.e adding current string to current subseq will result in duplicate letters <br>
overlap(current subseq, current string)
    - for each character in the string
        - if the character is in the subseq, there is overlap, return true
    - if the loop completes without returning true, there is no overlap, return false
- define a helper function for backtracking <br>
backtrack(current string index)
    - if the current index is the last, we have considered all strings, return the current subseq length
    - set res = 0 to track the max length valid subseq
    - if adding the current string to the current subseq, will not result in overlap
        - add the current string characters to the current subseq
        - backtrack to explore other subsequence and store the result in res
        - remove the current string characters from the current subseq
    - backtrack and update the res
- start backtracking from the first string at index 0 and return the result
     

## Complexity
- Time complexity: O(list comprehension + recursion breadth * overlap) → O(n + calls at each level ^ total levels * m) → O(n + 2$^n$ * 26) → O(2$^n$ * 26) → O(2$^n$)
    - *2 because for each string, we make two choices: include or exclude*
    - *n = number of strings in the input list*
    - *26 = the string contains lowercase English letters, and the max can be 26, no need to count the duplicates because we remove the strings with duplicate letters in the list comprehension part*

- Space complexity: O(list comprehension + recursion stack * overlap) → O(n + recursion depth/levels * 26) → O(n + n * 26) → O(n * 26) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxLength(self, arr: List[str]) -> int:
        arr = [i for i in arr if len(i) == len(set(i))]
        if not arr:
            return 0
        elif len(arr) == 1:
            return len(arr[0])
        path = set()

        def overlap(path, s):
            for c in s:
                if c in path:
                    return True
            return False

        def backtrack(i):
            if i == len(arr):
                return len(path)
            res = 0
            if not overlap(path, arr[i]):
                for c in arr[i]:
                    path.add(c)
                res = backtrack(i + 1)
                for c in arr[i]:
                    path.remove(c)
            return max(res, backtrack(i + 1)) 

        return backtrack(0)
```
