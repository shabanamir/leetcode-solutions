# <center> 93. Restore IP Addresses </center>


## Problem Description
[Click here](https://leetcode.com/problems/restore-ip-addresses/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given digits and we have to return all valid IP address combinations.

A valid IP address consists of 4 blocks. Each block 
- is separated by a dot
- contains an integer in the range 0 to 255
- contains up to 3 digits

We need to explore all possible combinations and check if it is valid. Use backtracking to explore all possible combinations.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the input string size is more than 12, there can't be a valid combination because an ip address can't have more than 12 digits
        - return an empty list
- create a list to store valid combinations
- define a helper function for backtracking <br>
backtrack(start position i.e current digit index, current path/combination)
    - if there are 4 elements/blocks in the current combination and the current start position is the last position i.e we have used all digits, the current combination is valid
        - add the string form of the current combination to the result and return
    - else if we have four blocks in the current combinations and we have not used all digits or we have used all digits but the current combination doesn't contain 4 blocks
        - return because the current combination is not valid
    - loop to generate the four blocks using backtracking
        - if the current position is out of bound
            - return
        - if there is a leading 0
            - return
        - get the number (substring) representing the current block
        - if the number is greater than 255:
            - return
        - else backtrack to explore and generate the other blocks
- start backtracking from the first digit at index 0


## Complexity
- Time complexity: O(recursion breadth) → O(calls at each level ^ total levels) → O(3^4) → O(81) → O(1)
    - *total levels = total blocks in an ip address = 4*
    - *calls at each level = digits in a single block = 3*
    - *we will not consider the time complexity of join() because join() is not called in each recursion and it complexity is O(4) because it is joining 4 blocks by a dot*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack + result list) → O(recursion depth/levels + total combinations) → O(4 + 19) → O(1)
    - *there is no relation between the input digits and output IP addresses. At most, there can be 19 total combinations in case the input is 11111111*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def restoreIpAddresses(self, s: str) -> List[str]:
        if len(s) > 12:
            return []
        res = []

        def backtrack(start: int, path: List[int]) -> None:
            if len(path) == 4 and start == len(s):
                res.append('.'.join(path))
                return
            if len(path) == 4 or start == len(s):
                return
            for length in range(1, 4):
                if start + length > len(s):
                    return 
                if length > 1 and s[start] == '0':
                    return 
                num = s[start: start + length]
                if int(num) > 255:
                    return
                backtrack(start + length, path + [num])

        backtrack(0, [])
        return res
```
