# <center> 1849. Splitting a String Into Descending Consecutive Values </center>


## Problem Description
[Click here](https://leetcode.com/problems/splitting-a-string-into-descending-consecutive-values/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a string of digits and need to split it into substrings such that the substrings are consecutive numbers in descending order.

Use backtracking to explore all possible splits and check if it is valid. A split is valid if the resulting substring values are in descending order and the difference is 1 i.e substring2 is 1 less than substring1.


## Approach
<!-- Describe your approach to solving the problem. -->
- define a helper function for backtracking <br>
backtrack(current index, current partitions, previous substring value)
    - if the current index is the last, we have considered all digits
        - check if we were able to split the string i.e current partitions >= 2 and return the result
    - loop in range current index i to size of the string s to check for the next digit
        - get the current substring value
        - if the previous substring value is greater or the difference between the previous and current is 1, current split is valid
            - backtrack to explore whether we can make another split and return true if we can
    - if the loop completes without returning true, we can't split the string, return false
- start backtracking from the first digit at index 0 and return the result


## Complexity
- Time complexity: O(recursion breadth) → O(calls at each levele ^ total levels) → O(n$^n$) but the program runs much faster due to pruning 
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(recursion stack) → O(recursion depth/levels) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def splitString(self, s: str) -> bool:
    
        def backtrack(i: int, k: int, prev: int) -> bool:
            if i == len(s):
                return k >= 2
            for j in range(i, len(s)):
                cur = int(s[i:j + 1])
                if prev == math.inf or cur == prev - 1:
                    if backtrack(j + 1, k + 1, cur):
                        return True
            return False
        
        return backtrack(0, 0, math.inf)
```
