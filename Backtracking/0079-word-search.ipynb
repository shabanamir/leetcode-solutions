# <center> 79. Word Search </center>


## Problem Description
[Click here](https://leetcode.com/problems/word-search/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a 2d board and a word. Each board cell represents a letter. The task is to check if the word can be constructed from the letters of adjacent cells.

Use backtracking to explore all possible paths on the board. To avoid revisiting the same cell, mark it as visited before exploring its neighbors and unmark it after. For marking, either use a hash set or temporarily replace the cell letter with a special character '*' (we can use any character except an English letter).

## Approach
<!-- Describe your approach to solving the problem. -->
- set m = no. of rows and n = no. of columns
- define a helper function for backtracking <br>
backtrack(row, column, current character index)
    - if the row or col is out of bounds, return false
    - if the cell has been visited or the cell letter is not the same as the current character in the word, return false
    - if the current index is the last, we've found the word, return true
    - store the current cell letter and mark the current cell as visited
    - backtrack to check if we can find the word by exploring the 4 neighboring paths (bottom, top, right, left) and store the result
    - unmark the cell after backtracking
    - return the backtracking result 
- for each cell, do backtracking and return true if the word is found
- if the loop completes without returning true, the word doesn't exist in the board, return false


## Complexity
- Time complexity: O(board traversal * recursion breadth) → O(rows * columns * calls at each level ^ total levels) → O(m * n * 4$^w$)
    - *w = no of characters in the word*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(recursion stack) → O(recursion depth/levels) → O(w)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def exist(self, board: List[List[str]], word: str) -> bool:
        m, n = len(board), len(board[0])

        def backtrack(r: int, c: int, i: int) -> bool:
            if r < 0 or r == m or c < 0 or c == n:
                return False
            if  board[r][c] == '*' or board[r][c] != word[i]:
                return False
            if i == len(word) - 1:
                return True
            cache = board[r][c]
            board[r][c] = '*'
            res = backtrack(r + 1, c, i + 1) or \
                  backtrack(r - 1, c, i + 1) or \
                  backtrack(r, c + 1, i + 1) or \
                  backtrack(r, c - 1, i + 1)
            board[r][c] = cache
            return res
            
        for r in range(m):
            for c in range(n):
                if backtrack(r, c, 0):
                    return True
        return False
```
