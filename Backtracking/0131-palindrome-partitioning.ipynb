# <center> 131. Palindrome Partitioning </center>


## Problem Description
[Click here](https://leetcode.com/problems/palindrome-partitioning/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The total number of partitions (substring) for n elements (characters) is 2$^n$.

We need only the palindrome substrings. Use backtracking to explore all possible substrings and check if it is a palindrome.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a result list to store palindrome substrings
- define a helper function to check if a string is palindrome or not <br>
is_palindrome(left index, right index)
    - loop until left index < right index
        - if the left character is not same as the right character, the substring can't be a palindrome
            - return false
        - else characters are same, move left and right index pointers
    - if the loop completes without returning false, the substring is a palindrome, return true
- define a helper function for backtracking <br>
backtrack(start position, current path/substring)
    - if current position >= size of the string, current substring is valid
        - add a copy of the current substring to the result list and return
            - *(add the copy to avoid overwriting this subset during future backtracking)*
    - for each character in the range current index (start position) to last index
        - if the substring ending with the current character is a palindrome
            - add it to the current path 
            - backtrack
            - remove the substring from the path
- start backtracking from index 0 as the start position
- return the result list


## Complexity
- Time complexity: O(generating partitions) → O(2$^n$)
    - *we will not consider the time complexity of copy here because copy() is not called in each recursion i.e we are not copying and adding every substring that is generated to our result*
    - *same goes for creating a substring to add to the current path*
    - *palindrome check will also not take n time in each recursion*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion stack + result list) → O(recursion depth/levels + n) → O(n + n) → O(n)
    - *the result list will only contain the palindrome substrings, and the number of palindrome substrings is much less than the number of total substrings (2$^n$)*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def partition(self, s: str) -> List[List[str]]:
        res = []

        def is_palindrome(l: int, r: int) -> bool:
            while l < r:
                if s[l] != s[r]:
                    return False
                l += 1
                r -= 1
            return True

        def backtrack(start: int, path: List[str]) -> None:
            if start >= len(s):
                res.append(path.copy())
                return
            for i in range(start, len(s)):
                if is_palindrome(start, i):
                    path.append(s[start : i + 1])
                    backtrack(i + 1, path)
                    path.pop()

        backtrack(0, [])
        return res
```
