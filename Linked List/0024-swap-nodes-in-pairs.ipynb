## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Two ways to solve this:
1. move two nodes forward at a time and swap adjacent node values if the values are not equal.
2. move two nodes forward at a time and swap adjacent nodes by updating the next pointer

Note: Use approach 2, because it is stated in the description not to change the values.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - if the list is empty or there's a single node in the list, return
- create a dummy node and set its next pointer to head to track the modified list head
- set cur = dummy to traverse the list and track current node
- loop until the second last node
    - take two nodes
    first = 1st node of the pair = cur.next
    second = 2nd node of the pair = cur.next.next
    - connect the first node with the third node by setting the first node next to the second node next
    - make the first node as second node by setting the second node next to the first node
    - make the second node as first node by setting the current node next to the second node
    - move the cur pointer two nodes ahead because we have swapped two nodes
- return dummy.next as the modified list head


## Complexity
- Time complexity:  O(loop) → O(n - 2) → O(n)
    - *we loop until the second last node*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(total new nodes created) → O(1) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next: 
            return head

        # swap nodes 
        dummy = ListNode(None, head)
        cur = dummy
        while cur.next and cur.next.next:
            first = cur.next
            second = cur.next.next
            first.next = second.next
            second.next = first
            cur.next = second
            cur = cur.next.next
        return dummy.next

        # swap values
        # cur = head
        # while cur and cur.next:
        #     if cur.val != cur.next.val:
        #         cur.val, cur.next.val = cur.next.val, cur.val
        #     cur = cur.next.next
        # return head
```
