## Approach

### Singly Linked List Approach

#### List Node Class
*class to create nodes in the linked list*

**init()**
- set the node value and next pointer 

#### LinkedList Class
*class to create a singly linked list*

**init()**
- set head = none to track the head node
- set size = 0 to track linked list size

**get()**
- if the index is out of bound
    - return -1
- set the current node pointer to the head node
- loop index times and move the current node ahead to reach the node at the required index and return the value

**addAtHead()**
- insert a node at the start (index 0) with the specified value

**addAtTail()**
- insert a node at the end (index = linked list length) with the specified value

**addAtIndex()**
- if the index is out of bound
    - return
- set the current node pointer to the head
- if the index is 0
    - create a new node and set its next node as the current node
    - make the new node as the head node
- else:
    - move the current pointer index-1 time ahead to reach the previous node i.e node at index before the required index
    - create a new node and set its next node to the next of the previous node
    - set the previous node next node to the new node 
- increment size because a new node is added

**deleteAtIndex()**
- if the index is out of bound, return
- if the index is 0, we have to delete the head node
    - make the second node the head node
- else:
    - set current node to head
    - move current node pointer index-1 times ahead to reach the previous node i.e node at index before the required index
    - delete the required index node by updating the next pointer of the previous node
- decrement size because a node is removed


## Complexity
- Time complexity:
    - Singly Linked List
        - init() O(1)
        - get() O(index) → O(n)
        - addAtHead() O(1)
        - addAtTail() O(n)
        - addAtIndex() O(index) → O(n)
        - deleteAtIndex() O(index) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity:
    - Singly Linked List
        - init() O(1)
        - get() O(1)
        - addAtHead() O(1)
        - addAtTail() O(1)
        - addAtIndex() O(1)
        - deleteAtIndex() O(1)
        - linked list size will be n after n calls to addAtIndex()

## Code

```python
class ListNode():

    def __init__(self, val, next= None):
        self.val = val
        self.next = next


class MyLinkedList:

    def __init__(self):
        self.head = None
        self.size = 0

    def get(self, index: int) -> int:
        if index >= self.size:
            return -1
        cur = self.head
        for _ in range(index):
            cur = cur.next
        return cur.val

    def addAtHead(self, val: int) -> None:
        self.addAtIndex(0, val)
        
    def addAtTail(self, val: int) -> None:
        self.addAtIndex(self.size, val)

    def addAtIndex(self, index: int, val: int) -> None:
        if index > self.size:
            return
        cur = self.head
        if index == 0:
            new_node = ListNode(val, next= cur)
            self.head = new_node
        else:
            for _ in range(index - 1):
                cur = cur.next
            new_node = ListNode(val, next= cur.next)
            cur.next = new_node
        self.size += 1

    def deleteAtIndex(self, index: int) -> None:
        if index >= self.size:
            return
        if index == 0:
            self.head = self.head.next
        else:
            cur = self.head
            for _ in range(index - 1):
                cur = cur.next
            cur.next = cur.next.next
        self.size -= 1



# Your MyLinkedList object will be instantiated and called as such:
# obj = MyLinkedList()
# param_1 = obj.get(index)
# obj.addAtHead(val)
# obj.addAtTail(val)
# obj.addAtIndex(index,val)
# obj.deleteAtIndex(index)
```
