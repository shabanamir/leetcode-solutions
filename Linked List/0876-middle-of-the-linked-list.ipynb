## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To find the middle, use slow-fast algorithm *(also called Tortoise and hare algorithms or Floyd's algorithm)*.

Use two-pointers (slow and fast) and traverse the list. Move the slow pointer one step and the fast pointer two steps at a time. When the fast pointer reaches the end, the slow pointer reaches the middle.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the list has only one node, the first node is the middle       
        - return the head
- set slow pointer = start node i.e head node and fast pointer = head node
- loop until the last node
    - move the slow pointer one node forward
    - move the fast pointer two nodes forward
- return slow as the middle node


## Complexity
- Time complexity: O(loop) → O(n/2) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def middleNode(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head.next:
            return head
        slow, fast = head, head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        return slow
```
