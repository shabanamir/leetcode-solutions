# <center> 2. Add Two Numbers </center>


## Problem Description
[Click here](https://leetcode.com/problems/add-two-numbers/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use the math rules for addition.
- add the leftmost digits
- store the leftmost digit of the sum 
- pass the rightmost digit of the sum as carry to the next digit


## Approach
<!-- Describe your approach to solving the problem. -->
- create a dummy node to track the head in the result list
- set cur = dummy to track the current node in the result list
- set carry = 0 to track the carry
- loop until one of the lists has not reached the end or there is a carry to process
    - create two variables to store the values of the current node in list1 and list2
      - *(we can use also the carry variable but the code is clear this way)*
    - if list1 is not null
        - get the list1 current node value
        - move to the next node in list1
    - if list2 is not null
        - get the list2 current node value
        - move to the next node in list2
    - get the sum 
    - get the leftmost digits of the sum using %, create a new node for it, and add it as the next node of the current node in the result list
    - get the carry 
    - update cur to move to the next node in the result list
- return dummy.next as the head of the result list


## Complexity
- Time complexity: O(loop) → O(max(list1 size, list2 size)) → O(max(m, n)) → O(n) 
  - *m = n if both lists are equal in size*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(total new nodes created) → O(total digits or nodes in the result) → O(max(list1 size, list2 size)) → O(max(m, n)) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode()
        cur = dummy
        carry = 0
        while l1 or l2 or carry:
            val1, val2 = 0, 0
            if l1:
                val1 = l1.val
                l1 = l1.next
            if l2:
                val2 = l2.val
                l2 = l2.next
            summ = val1 + val2 + carry
            cur.next = ListNode(summ % 10)
            carry = summ // 10
            cur = cur.next
        return dummy.next
```
