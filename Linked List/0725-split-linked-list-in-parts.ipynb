# <center> 725. Split Linked List in Parts </center>


## Problem Description
[Click here](https://leetcode.com/problems/split-linked-list-in-parts/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Find the number of nodes (m) for each part and the extra nodes (r). Distribute the extra nodes in the first r parts.

m = total nodes / k = n / k <br>
r = total nodes % k = n % k


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list of size k to store the splitting result i.e k sublists
- set n = 0 to track linked list size
- set cur = head to track the current node in the linked list
- loop until the last node to find the linked list size
    - increment n
    - move to the next node
- set m = n / k i.e sublist size, r = n % k i.e number of extra nodes
- reset cur to head 
- set prev = null to track the previous/last node
- loop k times to split the linked list into k sublists
    - add the current sublist head i.e cur to the result list
    - loop to get the sublist nodes, *(loop m + 1 times if there are extra nodes left else loop m times)*
        - add current node to the sublist by updating prev
        - move to the next node
    - after the loop, the sublist will have the required number of nodes. Split the linked list i.e disconnect the current subsist by setting next of prev (last node of the sublist) to null
    - decrement r
- return the result list


## Complexity
- Time complexity: O(finding list size + splitting list) → O(n + outer loop * inner loop) → O(n + k * m) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result list) → O(k)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:
    
    def splitListToParts(self, head: Optional[ListNode], k: int) -> List[Optional[ListNode]]:
        res = [None] * k
        size = 0
        cur = head
        while cur:
            size += 1
            cur = cur.next
        m, r = divmod(size, k)
        cur, prev = head, None
        for i in range(k):
            res[i] = cur
            for _ in range(m + (1 if r > 0 else 0)):
                prev = cur
                cur = cur.next
            if prev:
                prev.next = None
            if r > 0:
                r -= 1
        return res
```
