# <center> 61. Rotate List </center>


## Problem Description
[Click here](https://leetcode.com/problems/rotate-list/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To rotate a list k times to the right (clockwise), we need to move the last k elements (nodes) to the start of the list.

To do this, connect the first and last node to form a circular linked list and break the list at the kth node.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - if the list is empty, contains a single node, or k is 0, return because no rotation is required
- set cur = head to traverse the list
- set size = 1 to track list size
- loop until the end to find the last node and list size
  - move the cur pointer and increment size in each iteration
- *after loop, cur will point to the last node*
- if k is equal to the size of the list, return because the rotation will result in the same list
- else
    - connect the first and last nodes by setting the next node of the last node as head 
        - *(cur will point to the first node now)*
    - loop size - k times and move cur to get (k-1)th node
    - set head as kth node (cur.next)
    - break the list at k by setting the next of cur (k-1th node) as null
    - return head


## Complexity
- Time complexity: O(loop to find last node and list size + loop to get k-1th node) → O(n + n - k) → O(n)
  - *n = list size*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if not head or not head.next or k == 0:
            return head
        cur, size = head, 1
        while cur.next:
            cur = cur.next
            size += 1
        if k == size:
            return head
        cur.next = head
        for _ in range(size - k % size):
            cur = cur.next
        head = cur.next
        cur.next = None
        return head
```
