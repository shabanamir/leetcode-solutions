# <center> 23. Merge k Sorted Lists </center>


## Problem Description
[Click here](https://leetcode.com/problems/merge-k-sorted-lists/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
This is same as [Merge Two Sorted Lists](https://leetcode.com/problems/merge-two-sorted-lists/). 

Use the divide and conquer approach.

Each time take two sorted lists and merge them into one list. Repeat this until all the lists are merged.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the input contains no lists i.e [] or contains empty lists i.e [[], []], return null
    - if there's only one list, return it
    - if there are only two lists, merge them 
- set k = number of lists to be merged
- loop until all lists are merged i.e loop until k is zero
    - set left pointer l = 0 and right pointer r = k
    - loop until l < r
        - take two sorted lists (left and right lists), merge them using [Merge Two Sorted Lists Solution](https://leetcode.com/problems/merge-two-sorted-lists/solutions/3604740/o-min-m-n-time-o-1-space-solution-explained/) and store the merged list in the left list place
        - increment l and decrement r to get the next pair
    - divide k by 2 because each time we merge k lists into k/2 lists i.e k/2 lists are left to be merged
- return the first list as it contains the k merged list


## Complexity
- Time complexity: O(outer loop * inner loop * merging two sorted lists) → O(logk * constt * n) → O(logk * n) → O(nlogk)
  - *the outer loop will run logk times because in each iteration, k is divided by 2*
  - *the inner loop iteration is not fixed <br> 
  for example, if k is 8 <br>
the outer loop will run log(k) = log(8) = 3 times <br>
the inner loop will run 8 times for the first outer loop iteration, 4 times for the second iteration, and 2 times for the third iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(total new nodes created) → O(k)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        if not lists or not any(lists):
            return None
        if len(lists) == 1:
            return lists[0]
        if len(lists) == 2:
            return self.merge_sorted_lists(lists[0], lists[1])
        k = len(lists) - 1
        while k:
            l, r = 0, k
            while l < r:
                lists[l] = self.merge_sorted_lists(lists[l], lists[r])
                l += 1
                r -= 1
            k //= 2
        return lists[0]

    def merge_sorted_lists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        if not list1:
            return list2
        if not list2:
            return list1
        head = ListNode()
        cur = head
        while list1 and list2:
            if list1.val < list2.val:
                cur.next = list1
                list1 = list1.next
            else:
                cur.next = list2
                list2 = list2.next
            cur = cur.next
        cur.next = list1 or list2
        return head.next
```
