## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Insertion sort

Iterate the list. In each iteration, take a node, find its position in the sorted list, and insert it there.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the list is empty or contains a single node, no need to sort, return the head
- create a dummy node to track the head of the sorted part of the list
- set cur = head to track the current node in the list
- loop until the last node <br>
in each iteration
    - set prev = sorted_head to traverse and check the previous sorted part of the list
    - get the next node
    - traverse the previously sorted part of the list and find the position to insert the current node i.e keep moving to the next node in the sorted list while current node value is greater
    - insert the node at its position 
    - set cur to next to move to the next node
- return the second node in the sorted list as the head of the sorted list
i.e sorted_head.next
    - *(second node because first node is a placeholder node and the actual list starts at the second node)*


## Complexity
- Time complexity: O(insertion sort) → O(list traversal * sorted list traversal) → O(n * n) → O(n²)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(1) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def insertionSortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        sorted_head = ListNode()
        cur = head
        while cur:
            prev = sorted_head
            next = cur.next
            while prev.next and cur.val >= prev.next.val:
                prev = prev.next
            cur.next, prev.next = prev.next, cur
            cur = next
        return sorted_head.next
```
