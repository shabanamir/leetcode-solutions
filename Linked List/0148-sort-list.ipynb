## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use merge sort
1. split the list into two halves. For this, find the midpoint of the list and split. <br>
[Middle of the Linked List Problem](https://leetcode.com/problems/middle-of-the-linked-list)
2. recursively sort the left and right half <br>
3. merge the two sorted halves <br>
[Merge Two Sorted Lists Problem](https://leetcode.com/problems/merge-two-sorted-lists)


## Approach
<!-- Describe your approach to solving the problem. -->
- if the list is empty or contains a single node, the list is sorted, return head
- find the midpoint of the list
    - [Middle of the Linked List Solution](https://leetcode.com/problems/middle-of-the-linked-list/solutions/3608284/o-n-time-o-1-space-solution-explained/)
    - a little change is to break the list after finding the midpoint i.e set the mid node next to null
- divide the list into two halves and recursively sort the left and right half
    - set left = head to get left half of the list i.e start to mid of list
    - set right = mid node to get the right half of the list i.e mid to end of the list
- merge the sorted left and right half and return the merged list
    - [Merge Two Sorted Lists Solution](https://leetcode.com/problems/merge-two-sorted-lists/solutions/3604740/o-n-time-o-1-space-solution-explained)


## Complexity
- Time complexity: O(merge sort) → O(recursion) → O(recursion depth/levels * cost at each level i.e finding middle, merging, and splitting) → O(logn * n) → O(n * logn)
  - *total levels are logn because in every call we divide the list into two halves*
  - *merging two sorted lists = O(min(list1, list2)) → O(min(m, n) → O(n)*
  - *finding middle to split list = O(n/2) → O(n)* 
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(recursion depth/levels) → O(logn)

## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head or not head.next:
            return head
        mid = self.get_mid(head)
        left, right = self.sortList(head), self.sortList(mid)
        return self.merge_two_sorted(left, right)

    def merge_two_sorted(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        if not list1:
            return list2
        elif not list2:
            return list1
        head = ListNode()
        cur = head
        while list1 and list2:
            if list1.val < list2.val:
                cur.next = list1
                list1 = list1.next
            else:
                cur.next = list2
                list2 = list2.next
            cur = cur.next
        cur.next = list1 or list2
        return head.next

    def get_mid(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head.next:
            return head
        slow, fast = head, head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        mid = slow.next
        slow.next = None
        return mid
```
