# <center> 86. Partition List </center>


## Problem Description
[Click here](https://leetcode.com/problems/partition-list/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Create two dummy lists i.e left and right. Add smaller value nodes to the left list and greater value nodes to the right list. Finally, concatenate the lists.


## Approach
<!-- Describe your approach to solving the problem. -->
- set head and tail pointers for the left list
- set head and tail pointers for the right list
- traverse the list i.e loop until the end of the input/original list 
    - if the current node value is less, add it to the left list
      - add the current node to end of left list i.e update the left list tail pointer
      - move the left list tail pointer
    - else add it to the right list 
      - update the right list tail pointer
      - move the right list tail pointer
    - move the head to the next node
- set the right list tail pointer next to null to terminate the right list
- set the left list tail pointer next to the right list head pointer to concatenate the lists
- return the left list head pointer as the start of the result list


## Complexity
- Time complexity: O(loop) → O(total nodes in the list) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(2) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        head_left = tail_left = ListNode(0)
        head_right = tail_right = ListNode(0)
        while head:
            if head.val < x:
                tail_left.next = head
                tail_left = tail_left.next
            else:
                tail_right.next = head
                tail_right = tail_right.next
            head = head.next
        tail_right.next = None
        tail_left.next = head_right.next
        return head_left.next
```
