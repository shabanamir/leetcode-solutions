# <center> 203. Remove Linked List Elements </center>


## Problem Description
[Click here](https://leetcode.com/problems/remove-linked-list-elements/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A node can be easily deleted by removing its link with the previous node i.e updating the next pointer of the previous node to the next node.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a dummy node to track the new head in case the existing head is the node to be removed
    - *(we need it because in the end we have to return the head of the list because head node is required to get all other nodes in the linked list)*
- set cur = dummy to track current node
    - *(this pointer is used to traverse the list and remove nodes)*
- loop until the last node
    - if the next node is the node to be removed
        - update the next pointer of the current node to remove the node
        - do not break the loop here because there can be more than one node with the same value, and we need to remove all such nodes
    - else move forward to search for the node
- return the dummy next node as the head 
    - *(next because dummy is a placeholder node, the actual list starts at dummy.next)*


## Complexity
- Time complexity: O(loop) → O(total nodes) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(1) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def removeElements(self, head: Optional[ListNode], val: int) -> Optional[ListNode]:
        dummy = ListNode(next=head)
        cur = dummy
        while cur.next:
            if cur.next.val == val:
                cur.next = cur.next.next
            else:
                cur = cur.next
        return dummy.next
```
