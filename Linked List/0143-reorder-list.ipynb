# <center> 143. Reorder List </center>


## Problem Description
[Click here](https://leetcode.com/problems/reorder-list/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to re-order the list by placing the first half (left) and second half (right) nodes alternately. 

first node → last node → second node → second last node and so on

We can use a two-pointer approach to traverse the two halves and place nodes alternately. The problem is that in singly linked list we can only move forward not backward unlike a list (array). To solve this, reverse the second half of the list. 


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there's only one node in the list, return
- find the middle of the list using slow-fast pointer algorithm
    - set two pointers slow and fast at the start of the list
    - move slow one step and fast two steps at a time
    - by the time fast reaches the end, slow reaches the middle of the list
- reverse the second half using pointers
    - set prev = null to track the previous node
    - set cur = start node of the second half i.e slow (we can continue with the slow pointer but the code is clear this way)
    - loop until the last node
        - store the next node in next pointer
        - make the previous node as the next node of the current node
        - make the current node as the previous node and next node as the current node
- place the nodes alternatively using left and right pointers
    - set left = second node of first half i.e head.next
        - *(second node because first left node (head) is correctly place)*
    - set right = start node of reversed second half i.e prev
    - set i = 1 to track the even and odd iteration
        - *(we need to put left nodes in even positions and right nodes in odd)*
    - loop until the left and right nodes meet
        - if iteration is odd
            - put the right node and move right pointer
        - if iteration is even
            - put the left node and move left pointer
        - move head
        - increment i


## Complexity
- Time complexity: O(finding middle + reversing second half + connecting left and right nodes) → O(n/2 + n/2 + n/2) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def reorderList(self, head: Optional[ListNode]) -> None:
        """
        Do not return anything, modify head in-place instead.
        """
        if not head.next:
            return
        
        # find middle 
        slow, fast = head, head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        # reverse second half
        prev, cur = None, slow
        while cur:
            next = cur.next
            cur.next = prev
            prev = cur
            cur = next
        
        # place first and second half nodes alternately 
        l, r = head.next, prev
        i = 1
        while l != r:
            if i & 1:
                head.next = r
                r = r.next
            else:
                head.next = l
                l = l.next
            head = head.next
            i += 1
```
