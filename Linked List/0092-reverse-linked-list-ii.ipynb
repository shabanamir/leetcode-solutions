## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Find the left node and reverse the sublist using pointers.


# Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
  - if left and right node position is the same, return because there's no need to reverse
- create a dummy node to track the head of the result list
    - *(this is to handle the edge case where the head is part of the sublist to be reversed)*
- set previous node pointer prev = dummy
- move prev pointer (left - 1) times to get the node before the left node
- set cur = left node (next node of prev) to track the current node in the sublist
- loop (right - left) times and reverse the sublist by updating the pointers
- return dummy.next as head of the reversed list


## Complexity
- Time complexity: O(loop to find left/start node of the sublist + loop to reverse sublist nodes) → O(left + right - left) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(1) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def reverseBetween(self, head: Optional[ListNode], left: int, right: int) -> Optional[ListNode]:
        if left == right:
            return head
        dummy = ListNode(0, head)
        prev = dummy
        for _ in range(left - 1):
            prev = prev.next 
        cur = prev.next 
        for _ in range(right - left):
            nextt = cur.next
            cur.next = nextt.next
            nextt.next = prev.next
            prev.next = nextt
        return dummy.next
```
