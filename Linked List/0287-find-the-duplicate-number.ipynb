# <center> 287. Find the Duplicate Number </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A duplicate number creates a cycle in a list. To find cycle in a list, use slow-fast pointer algorithm. 

we need to find the duplicate number i.e the cycle entrance point

Use two-pointers. Move the slow pointer one step and the fast pointer two steps at a time. They'll meet in the cycle, but the intersection point is not necessarily the cycle entrance. Now to find the cycle entrance, slow down the fast. Fast will start from the starting position and slow from the intersection point. Move both pointers one step at a time. This time they'll meet at the entrance.

*Note: There are n + 1 numbers in range 1 to n i.e 1, 2, 3, 4, ... <br>
Each new element in the sequence (1-n) is the element at the index equal to the previous element i.e x, nums[x], nums[nums[x]], nums[nums[nums[x]]], ...*


## Approach
<!-- Describe your approach to solving the problem. -->
- set slow and fast at the start of the list i.e first element
- loop until slow pointer meets the fast (intersection point)
    - move slow one step forward
    - move fast two steps forward
    - break when slow = fast
- move the fast pointer back to the starting position 
- loop until the slow pointer meets the fast again (entrance point)
    - move slow one step forward
    - move fast one step forward
- return slow


## Complexity
- Time complexity: O(finding intersection point + finding entrance point) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findDuplicate(self, nums: List[int]) -> int:
        slow, fast = nums[0], nums[0]
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]
            if slow == fast:
                break
        fast = nums[0]
        while slow != fast:
            slow = nums[slow]
            fast = nums[fast]
        return slow
```
