# <center> 19. Remove Nth Node From End of List </center>


## Problem Description
[Click here](https://leetcode.com/problems/remove-nth-node-from-end-of-list/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To remove the nth node from end, we need to update the next pointer of its previous nodes i.e (n+1)th node from the end. In a singly linked list, we can't move in backward direction (end to start) to find the nth node from end.

Approaches:
1. The nth node from end is the same as (total nodes - n)th node from start. But for this, we need to find the list size (total nodes), which will require an extra pass/loop.
2. Use slow-fast pointer approach. Give the fast pointer an earlier start by moving it to (n+1)th node from the start. Then move both pointers at the same pace. By the time, fast pointer reaches the end, the slow pointer will be at (n+1)th node from end.


## Approach
<!-- Describe your approach to solving the problem. -->
- set slow and fast pointers at the start of the list
- loop n times to move the fast pointer (n+1)th node from the start 
    - *(the range is n but due to next pointer, the fast pointer will point to (n+1)th node)*
- if the fast pointer becomes null, it means the first node is the target node
    - delete it by returning the second node as start of the list i.e head.next 
- loop until the fast pointer reaches the end
    - move both pointers one node at a time
- after the loop, the slow pointer will be at (n+1)th node from the end. update its next pointer to remove the next node i.e nth node
- return head


## Complexity
- Time complexity: O(moving fast pointer till (n+1)th node + moving fast pointer til last node) → O(n + total nodes - n) → O(n + m - n) → O(m) 
    - *m >= n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:
    
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        slow = fast = head
        for _ in range(n):
            fast = fast.next
        if not fast:
            return head.next
        while fast.next:
            slow = slow.next
            fast = fast.next
        slow.next = slow.next.next
        return head
```
