# <center> 2130. Maximum Twin Sum of a Linked List </center>


## Problem Description
[Click here](https://leetcode.com/problems/maximum-twin-sum-of-a-linked-list/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The first half nodes are twins of the second half nodes i.e first node is the twin of the last node, second node is the twin of second last node, and so on. 

We need to find twins with max sum. We can use two-pointers to traverse the two halves and compare twins. The problem is, in singly linked list we can only move forward not backward unlike a list (array). To solve this, reverse the second half of the list to easily compare twins.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track max sum
- find the middle of the list using slow-fast pointer algorithm
    - set two pointers slow and fast at the start of the list
    - move slow one step and fast two steps at a time
    - by the time fast reaches the end, slow reaches the middle of the list
- reverse the second half using pointers
    - set prev = null to track the previous node
    - set cur = start node of the second half i.e slow (we can continue with the slow pointer but the code is clear this way)
    - loop until the last node
        - store the next node in next pointer
        - make the previous node as the next node of the current node
        - make the current node as the previous node and next node as the current node
- compare twins using left and right pointers
    - set left = start node of first half i.e head, right = start node of reversed second half i.e prev
    - loop until right reaches the end 
        - update res by taking max of current twins (left and right node) sum and previous max sum
        - move the left and right pointers
- return res


## Complexity
- Time complexity: O(finding middle + reversing second half + comparing twins) → O(n/2 + n/2 + n/2) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def pairSum(self, head: Optional[ListNode]) -> int:
        res = 0
        
        # find middle 
        slow = fast = head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        
        # reverse second half
        prev, cur = None, slow
        while cur:
            next = cur.next
            cur.next = prev
            prev = cur
            cur = next
        
        # compare twins 
        left, right = head, prev
        while right:
            res = max(res, left.val + right.val)
            left = left.next
            right = right.next
        return res
```
