# <center> 21. Merge Two Sorted Lists </center>


## Problem Description
[Click here](https://leetcode.com/problems/merge-two-sorted-lists/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Traverse both lists in parallel and compare their values. If the list1 node is smaller, add the list1 node to the result list and move forward in list1. Else add the list2 node and move forward in list2.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if list1 is empty, the merged list will only contain list2 elements, return list2
    - if list2 is empty, the merged list will only contain list1 elements, return list1
- set head = a dummy node to represent the head of the merged list
- set cur = head to track the current (last inserted) node in the merged list
    - *(this pointer is used to update the merged list)*
- loop until the last element in list1 and list2, *(the loop will end when one of the lists reaches the end)*
    - if list1 node value is smaller
        - add the list1 node to the merged list by updating the cur next pointer
        - move list1 one step forward to get the next node
    - else list2 node value is smaller
        - add the list2 node in the merged list by updating the cur next pointer
        - move list2 one step forward to get the next node
- after the loop, it is possible that one list has some elements left. Add the remaining elements to the merged list by updating the cur next pointer
- return the head next pointer as the head of the merged linked list 
    - *(next because head is a placeholder node, the actual list starts at head.next)*


## Complexity
- Time complexity: O(loop) → O(min(list1 size, list2 size)) → O(min(m, n))
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(1) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        if not list1:
            return list2
        if not list2:
            return list1
        head = ListNode()
        cur = head
        while list1 and list2:
            if list1.val < list2.val:
                cur.next = list1
                list1 = list1.next
            else:
                cur.next = list2
                list2 = list2.next
            cur = cur.next
        cur.next = list1 or list2
        return head.next
```
