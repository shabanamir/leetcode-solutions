## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A list is a palindrome if:
- the reversed list is the same as the original list or
- the reversed right/end half of the list is the same as the original left/first half


## Approach
<!-- Describe your approach to solving the problem. -->
- find the middle of the list using slow-fast pointer algorithm
    - set two pointers slow and fast at the start of the list
    - move slow one step and fast two steps at a time
    - by the time fast reaches the end, slow reaches the middle of the list
- reverse the second half using pointers
    - set prev = null to track the previous node
    - set cur = start node of the second half i.e slow (we can continue with the slow pointer but the code is clear this way)
    - loop until the last node
        - store the next node in next pointer
        - make the previous node as the next node of the current node
        - make the current node as the previous node 
        - make the next node as the current node
- compare the first and second half using left and right pointers
    - set left = start node of first half i.e head, right = start node of reversed second half i.e prev
    - loop until right reaches the end 
        - if the left node value is not same as the the right node value, the list is not a palindrome
            - return false
        - else move the left and right pointers
    - if the loop completes successfully without returning false, the list is a palindrome, return true


## Complexity
- Time complexity: O(finding middle + reversing second half + comparing first and second half) → O(n/2 + n/2 + n/2) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def isPalindrome(self, head: Optional[ListNode]) -> bool:
        
        # find the middle 
        slow, fast = head, head
        while fast and fast.next:
            fast = fast.next.next
            slow = slow.next
            
        # reverse second half
        prev, cur = None, slow
        while cur:
            next = cur.next
            cur.next = prev
            prev = cur
            cur = next
        
        # comapre first and reversed second half
        left, right = head, prev
        while right:
            if left.val != right.val:
                return False
            left = left.next
            right = right.next
        return True
```
