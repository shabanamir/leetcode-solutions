# <center> 83. Remove Duplicates from Sorted List </center>


## Problem Description
[Click here](https://leetcode.com/problems/remove-duplicates-from-sorted-list/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a sorted list, duplicates are adjacent. 

Traverse the list. Compare each node with its next node. If they are same, the next node is a duplicate, delete it by updating the next pointer of the previous node.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there is no node, return null
- set cur = head to track current node
- loop until the last node
    - if the current node value is equal to the next node value         
        - delete the next node 
    - else update the current node
- return head node


## Complexity
- Time complexity: O(loop) → O(total nodes) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:
    
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if not head:
            return None
        cur = head
        while cur and cur.next:
            if cur.val == cur.next.val:
                cur.next = cur.next.next
            else:
                cur = cur.next
        return head
```
