# <center> 141. Linked List Cycle </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Use slow-fast pointer algorithm. 

Traverse linked list using two pointers (slow and fast). Move the slow pointer one step and the fast two steps at a time. There's a cycle if the slow pointer reaches the fast.


## Approach
<!-- Describe your approach to solving the problem. -->
- set both pointers at the start of the list i.e head node
- loop until the fast pointer reaches the end of the list
    - move slow pointer one node forward
    - move fast pointer two nodes forward
    - if slow and fast meet, return true because there's a cycle
- if the loop completes without returning true, there's no cycle, return false


## Complexity
- Time complexity: O(loop) → O(half of the list) → O(n / 2) → O(n)
  - *half of the list because we are moving fast pointer two nodes at a time*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None


class Solution:

    def hasCycle(self, head: Optional[ListNode]) -> bool:
        slow, fast = head, head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False
```
