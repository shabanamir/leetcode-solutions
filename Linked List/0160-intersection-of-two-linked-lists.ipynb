## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
1. O(n) space <br>
Traverse the first list and mark its nodes visited by adding them to a hashset. Traverse the second list, if a node is already present in the set, it is the intersection node.
2. O(1) space <br>
Traverse both nodes in parallel using two-pointers. If list1 pointer reaches the end, redirect it to list2 head and vice-versa. Once both pointers are re-assigned, they will be equidistant from the intersection node.


## Approach
<!-- Describe your approach to solving the problem. -->
- set i = head of list1 to track and traverse nodes in list1
- set j = head of list2 to track and traverse nodes in list2
- loop until the nodes pointed by i and j are not the same
    - if we have not reached list1 end
        - update i to move forward in list1 
    - else reset i to the head of list2
    - if we have not reached list2 end
        - update j to move forward in list2 
    - else reset j to head of list1
- return i or j 
    - *(because both will point to the same node)*


## Complexity
- Time complexity: O(loop) →  O(m + n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None


class Solution:

    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
        i, j = headA, headB
        while i != j:
            i = i.next if i else headB
            j = j.next if j else headA
        return i
```
