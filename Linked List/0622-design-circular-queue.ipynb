Implement the circular queue using a linked list.


## Approach
<!-- Describe your approach to solving the problem. -->

#### ListNode Class
*class for linked list nodes*

**init()**
- set the node value and next node pointer

#### MyCircularQueue Class
*class to implement circular queue using linked list*

**init()**
- set head and tail node pointers to null
- set capacity of the queue to k
- set size of the queue (i.e number of elements currently in the queue) to 0

**enQueue()**
- if the queue is full, insertion is not possible, return false
- create a node for the value to be inserted
- if there are no elements in the queue i.e linked list is empty, set the head and tail of the linked list to the node
- else set the tail and tail next to the node
- increment size
- return true

**deQueue()**
- if the queue is empty, deletion is not possible, return false
- else delete an element from the queue front (linked list start) by updating head to head.next
- decrement size
- return true

**Front()**
- if the queue is empty, return -1, else return the head value (first node value in the linked list)

**Rear()**
- if the queue is empty, return -1, else return the tail value (last node value in the linked list)

**isEmpty()**
- check if size of the queue is 0 and return the result

**isFull()**
- check if size of the queue is equal to the capacity and return the result


## Complexity
- Time complexity: 
    - init() O(1)
    - enQueue() O(1)
    - deQueue() O(1)
    - Front() O(1)
    - Rear() O(1)
    - isEmpty() O(1)
    - isFull() O(1)

- Space complexity:

## Code

```python
class Node:

    def __init__(self, val: int):
        self.val = val
        self.next = None


class MyCircularQueue:

    def __init__(self, k: int):
        self.head = self.tail = None
        self.capacity = k
        self.size = 0

    def enQueue(self, value: int) -> bool:
        if self.isFull():
            return False
        node = Node(value)
        if self.size == 0:
            self.head = self.tail = node
        else:
            self.tail.next = self.tail = node
        self.size += 1
        return True

    def deQueue(self) -> bool:
        if self.isEmpty():
            return False
        self.head = self.head.next
        self.size -= 1
        return True

    def Front(self) -> int:
        return -1 if self.isEmpty() else self.head.val

    def Rear(self) -> int:
        return -1 if self.isEmpty() else self.tail.val

    def isEmpty(self) -> bool:
        return self.size == 0

    def isFull(self) -> bool:
        return self.capacity == self.size
    


# Your MyCircularQueue object will be instantiated and called as such:
# obj = MyCircularQueue(k)
# param_1 = obj.enQueue(value)
# param_2 = obj.deQueue()
# param_3 = obj.Front()
# param_4 = obj.Rear()
# param_5 = obj.isEmpty()
# param_6 = obj.isFull()
```
