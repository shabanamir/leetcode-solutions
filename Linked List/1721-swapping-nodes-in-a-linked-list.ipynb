# <center> 1721. Swapping Nodes in a Linked List </center>


## Problem Description
[Click here](https://leetcode.com/problems/swapping-nodes-in-a-linked-list/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the kth node from the start and kth node from the end and swap the values.

Use slow-fast pointer approach. Give the fast pointer an earlier start by moving it to the kth node from the start. Then move both pointers at the same pace. By the time, fast pointer reaches the end, the slow pointer will reach the kth node from the end.


## Approach
<!-- Describe your approach to solving the problem. -->
- set slow and fast pointers to start of the list
- loop k times to move the fast pointer kth node from the start and store the kth node
- loop until the fast pointer reaches the end
    - move both pointers one node at a time
- after the loop, the slow pointer will be at the kth node from end 
- swap the k_start and k_end node values
- return head


## Complexity
- Time complexity: O(moving fast pointer til kth node + moving fast pointer till last node) → O(k + n - k) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(total new nodes created) → O(0) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next


class Solution:

    def swapNodes(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        slow = fast = head
        for _ in range(k - 1):
            fast = fast.next 
        k_start = fast 
        while fast.next:
            slow = slow.next
            fast = fast.next
        k_start.val, slow.val = slow.val, k_start.val 
        return head
```
