## Intuition

1. Doubly Linked List O(n)

2. Stack O(n) <br>
use two stacks: one for back history, and one for forward history

3. Array O(1) <br>
use an array to store and access urls by index in constant time


## Approach
<!-- Describe your approach to solving the problem. -->
#### Array Approach

**init()**
- initialize an array with a homepage to store visited website urls
- set current = 0 to track the index of current url
- set end = 0 to track the last visited url index

**visit()**
- increment cur because we are visiting a new url
- if we are at the end of the array, add url at the end 
- else add it next to the currently visited url i.e update the value at cur index
- update end to clear the forward urls

**back()**
- move k steps backward to get the url index and set cur to the new index because it is now the current url index 
    - *(the index can go out of bounds, so take the max to find the url we can go back to)*
- return the url at the current index

**forward()**
- move k steps forward to get the url index and set cur to the new index because it is now the current url index 
    - *(the index can go out of bounds, to handle this take min)*
- return the url at the current index


## Complexity
- Time complexity: 
    - Array Approach
        - init() O(1)
        - visit() O(1)
        - back() O(1)
        - forward() O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - Array Approach
        - init() O(1)
        - visit() O(1)
        - back() O(1)
        - forward() O(1)
        - array size will be O(n) after n calls to visit()

## Code

```python
class BrowserHistory:

    def __init__(self, homepage: str):
        self.history = [homepage]
        self.cur = 0
        self.end = 0

    def visit(self, url: str) -> None:
        self.cur += 1
        if self.cur == len(self.history):
            self.history.append(url)
        else:
            self.history[self.cur] = url
        self.end = self.cur

    def back(self, steps: int) -> str:
        self.cur = max(0, self.cur - steps)
        return self.history[self.cur]

    def forward(self, steps: int) -> str:
        self.cur = min(self.end, self.cur + steps)
        return self.history[self.cur]
        


# Your BrowserHistory object will be instantiated and called as such:
# obj = BrowserHistory(homepage)
# obj.visit(url)
# param_2 = obj.back(steps)
# param_3 = obj.forward(steps)
```
