## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
With JFK as the source (initial airport), we need to return the destinations (airports) in lexical (ascending) order. Solve it using a directed graph.

Create a graph with airports as nodes and tickets as edges and do DFS.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = a list for itinerary
- create a hashmap  to represent the graph (adjacency list)
    - *key = node*
    - *value = list of neighbors*
- sort the tickets in reverse order so that the smallest destination will be at the end
- traverse the tickets and add edges to the graph
    - for each source node, and destination node in the sorted edges (tickets)
        - add an edge from the source node to the target node i.e add the destination node to the neighbors' list of the source node
            - *(the tickets are sorted in reverse, so the neighbors will be added in descending order)*
- define a helper function for DFS <br>
dfs(node)
    - loop until you find a node with no outgoing edges (neighbors)
        - remove the last neighbor (smallest destination node) from the neighbors' list of the node and do DFS
            - *(in this way, each time we will pop the smallest node, find the smallest destination node for it, and add it to the result list)*
        - the node has no outgoing edges, add it to the result list
- start DFS from the JFK
- reverse the result list because we have constructed the itinerary in reverse and return it


## Complexity
- Time complexity: O(sorting tickets (edges) + graph (adjacency list) creation + dfs + reversing result list) → O(ElogE + E + V + E + E) → O(ElogE) → O(nlogn)
    - *V = total vertices or nodes = total airports*
    - *E = total edges = total tickets = n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(sort + graph  + dfs stack + result list) → O(E + V + E + V + E) → O(V + E) → O(E) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findItinerary(self, tickets: List[List[str]]) -> List[str]:
        res = []
        graph = defaultdict(list)
        for src, dst in sorted(tickets, reverse=True):
            graph[src].append(dst)
        
        def dfs(node: str) -> None:
            while graph[node]:
                dfs(graph[node].pop())
            res.append(node)
        
        dfs('JFK')
        return res[::-1]
```
