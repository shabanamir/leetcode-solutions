# <center> 1523. Count Odd Numbers in an Interval Range </center>


## Problem Description
[Click here](https://leetcode.com/problems/count-odd-numbers-in-an-interval-range/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
- If the range n is even, count of even and odd numbers will be the same i.e half numbers (n/2) will be even and half will be odd
- If the range is odd, count of odd numbers will be n/ 2 + 1

Range = high - low <br>
Range is odd if either high or low is odd


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = high - low to get the total numbers in the range
- if high is odd or low is odd, the range is odd
    - *you can also use % operator to check for even-odd*
    - return n/2 + 1
- else range is even, return n/2


## Complexity
- Time complexity: O(1)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def countOdds(self, low: int, high: int) -> int:
        n = high - low 
        if high & 1 or low & 1:
            return n // 2 + 1
        return n // 2
```
