# <center> 9. Palindrome Number </center>


## Problem Description
[Click here](https://leetcode.com/problems/palindrome-number/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A number x is a palindrome if:
- it is the same as its reverse form or
- the left half is equal to the reversed right half

Approach 1 <br>
Reverse the full number and compare the original and reversed number.

Approach 2 <br>
Only reverse the right half of the number and compare the left half and right half. This approach will also handle the integer overflow.

For an explanation of reversing a number, check [Reverse Integer Solution](https://leetcode.com/problems/reverse-integer/solutions/3845859/o-logx-time-o-1-space-solution-explained/)
 
 
## Approach
<!-- Describe your approach to solving the problem. -->
**Approach 1**

- check base cases
    - if the integer x is negative or ends with a zero i.e x is not 0 but a number that ends with 0, it can't be a palindrome, return false
- set y = x and rev = 0 to track the digits in the original and reversed part
- loop until y is not 0 i.e there are digits left to be processed
    - get the rightmost digit of y using %, multiply rev by 10, and add the digit 
    - divide y by 10 to remove the rightmost digit 
- compare x and reverse of x i.e rev and return the result

**Approach 2**

- check base cases
    - if the integer x is negative or ends with a zero i.e x is not 0 but a number that ends with 0, it can't be a palindrome, return false
- set left half = x and right half = 0 to track the left and reversed right half
- loop until the left half is greater than the right half
    - get the rightmost digit using %. to add the digit to right half, multiply the right half by 10 and add the digit 
    - divide x by 10 to remove the digit from the left half
- after the loop, left will contain the left half, and right will contain the reverse of the right half
- check if the left half is equal to the right half or left half is equal to the right half with one more digit removed *(this can be the case if no. of digits in x is odd, in that case, we remove the middle digit)*


## Complexity
- Time complexity:
    - Approach 1 O(log$_{10}$x) → O(logx)
    - Approach 2 O(log$_{10}$x) → O(logx)
    - *log$_{10}$x because in each iteration x is divided by 10*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - Approach 1 O(1)
    - Approach 2 O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isPalindrome(self, x: int) -> bool:
        
        # approach 1
        if x < 0 or x != 0 and x % 10 == 0:
            return False
        y, rev = x, 0
        while y:
            rev = rev * 10 + y % 10
            y //= 10
        return x == rev

        # approach 2
        if x < 0 or x != 0 and x % 10 == 0:
            return False
        left, right = x, 0
        while left > right:
            right = right * 10 + left % 10
            left //= 10
        return left == right or left == right // 10
```
