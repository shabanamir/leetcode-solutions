# <center> 149. Max Points on a Line </center>


## Problem Description
[Click here](https://leetcode.com/problems/max-points-on-a-line/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Two points are on the same line if they share the same slope.

For each point, compute the slope w.r.t to every other point. Use a hashmap to count the points with the same slope and get the max count.

P1 = (x1, y1) <br>
P2 = (x2, y2) <br>
slope = change in y / change in x = y2 – y1 / x2 – x1


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 1 to track the slope with max points count
    - *1 because there's at least one point*
- set n = length of the input points array
- create a hashmap to count the points with the same slope
- loop in range n
    - get the first point p1
    - clear the hashmap
    - loop in range i + 1 to n to compute slope with every other point
        - get the secind point p2
        - if points are the same, set the slope to a very large number i.e float infinity
        - else, compute the slope
        index 0 is x, index 1 is y
        - increment the slope count by 1
        - update res by taking the max of previous max and current slope count
- return res


## Complexity
- Time complexity: O(outer loop * inner loop) → O(n * n) → O(n²)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def maxPoints(self, points: List[List[int]]) -> int:
        res = 1
        n = len(points)
        count = defaultdict(int)
        for i in range(n):
            p1 = points[i]
            count.clear()
            for j in range(i + 1, n):
                p2 = points[j]
                if p2[0] == p1[0]:
                    slope = float('inf')
                else:
                    slope = (p2[1] - p1[1]) / (p2[0] - p1[0])
                count[slope] += 1
                res = max(res, count[slope] + 1)
        return res
```
