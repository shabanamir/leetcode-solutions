## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a string and rows. We have put the string letters in given rows in a zigzag pattern and return the string formed by getting the letters in each row one by one.

For example: 
input string = "PAYPALISHIRING", rows = 3

zigzag pattern
```
P   A   H   N
A P L S I I G
Y   I   R
```

output string = row1 letters + row2 letters + row3 letters = PAHN + APLSIIG + YIR = 'PAHNAPLSIIGYIR'

zigzag pattern observations:
- we move from top to bottom and bottom to top and put letters in rows
- if we are in the first row, we move down 
- if we are in the last row, we move up 

Traverse the string and put letters in their appropriate rows by tracking the row number and moving direction. 


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if total rows are 1 or equal to the length of string s, return the string
- set res = an array of strings of size equal to total rows to track each row letters
- set row = 0 to track the index of the current row
- set down = 1 to track the direction
    - *1 means to move down, and -1 to move up*
- traverse the string
    - add the current character to the current row
    - if the current row is the first row, move down
    - if the current row is the last row, move up
    - update the current row i.e increment the row if we are moving down else decrement the row
        - *if the direction is down (+1), it will increment the row i.e row + 1*
        - if the direction is up (-1), it will decrement the row i.e row - 1*
- convert the return result array to a string and return it


## Complexity
- Time complexity: O(string traversal + converting result array to string) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result array) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

````python
class Solution:

    def convert(self, s: str, numRows: int) -> str:
        if numRows == 1 or numRows == len(s):
            return s
        res = [''] * numRows
        row = 0
        down = 1
        for c in s:
            res[row] += c
            if row == 0:
                down = 1
            elif row == numRows - 1:
                down = -1
            row += down
        return ''.join(res)
````
