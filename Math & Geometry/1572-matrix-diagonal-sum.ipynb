## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Primary diagonal indices <br>
- 00, 11, 22 *(3 x 3 matrix)* 
- row-col condition is row = column
- matrix[row][col] = matrix[i][i]

Secondary diagonal indices
- 02, 11, 21 *(3 x 3 matrix)* 
- row-col condition is row = total number of rows - col - 1
- matrix[row][col] = matrix[i][n-i-1] or matrix[i][~i]


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = number of rows in the matrix
- set res = 0 to track the sum
- traverse the rows
    - get the sum of primary and secondary diagonal elements for the current row and update the res
- if n is odd, there is a common element between the primary and secondary diagonals
    - subtract the middle element (matrix[n/2][n/2]) form res
- return res


## Complexity
- Time complexity: O(rows traversal) â†’ O(n)
    - *n = number of rows*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->



## Code

```python
class Solution:

    def diagonalSum(self, mat: List[List[int]]) -> int:
        n = len(mat)
        res = 0
        for i in range(n):
            res += mat[i][i] + mat[i][n - i - 1] 
        if n & 1:
            res -= mat[n // 2][n // 2]
        return res
```
