# <center> 2028. Find Missing Observations </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-missing-observations/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Find the sum of missing n rolls
*missing sum = target sum - sum of m rolls = mean * (m + n) - sum of m rolls*

Check if the missing sum can be distributed to n rolls. If yes, find the average *(missing sum / n)* and distribute it. Distribute the remainder also if any.


## Approach
<!-- Describe your approach to solving the problem. -->
- compute the target sum i.e mean * (m + n)
- compute the missing sum i.e target sum - sum of m rolls
- if the missing sum is less than n or more than n * 6, return an empty array because it can't be distributed
- find the average n_avg and remainder n_mod
- create a result array of size n and initialize it with n_avg to give each roll distribution of the average
- distribute the remainder
    - loop in range n_mod
        - increment the value in the result array by 1
- return the result array


## Complexity
- Time complexity: O(finding sum of m rolls + creating result array to distribute the average + distributing remainder) → O(m + n + n) → O(m + n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result array) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def missingRolls(self, rolls: List[int], mean: int, n: int) -> List[int]:
        target_sum = (len(rolls) + n) * mean
        missing_sum = target_sum - sum(rolls)
        if missing_sum < n or missing_sum > n * 6:
            return []
        n_avg, n_mod = missing_sum // n, missing_sum % n
        res = [n_avg] * n
        for i in range(n_mod):
            res[i] += 1
        return res
```
