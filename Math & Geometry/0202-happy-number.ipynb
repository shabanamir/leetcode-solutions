## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A number is happy if it enters a cycle and becomes 1 by replacing it with the sum of the square of its digits after some iterations.

Use Floyd’s Cycle detection algorithm. It is based on two-pointers: fast and slow. The slow pointer moves ahead one step at a time, and the fast pointer moves ahead two steps at a time. The pointers meet if there is a cycle.


## Approach
<!-- Describe your approach to solving the problem. -->
- define a helper function to get the sum of squares of digits <br>
sum_square_digits(number)
    - set res = 0 to track the sum
    - loop until the number becomes 0
        - get the rightmost digit of the number using mod operator
        - get the square of the digits and add it to res
        - remove the digit from the number
    - return res
- set slow = n and fast = sum of squares of digits of n
- while the pointers are not equal, *the loop will break when the pointers enter a cycle i.e their values become equal*
    - move slow pointer 1 time
        - replace slow with the sum of squares of digits of slow
    - move fast pointer 2 times
        - replace fast with the sum of squares of digits of fast
        - replace fast with the sum of squares of digits of fast
- return true if fast value is 1, else false


## Complexity
- Time complexity: O(loop for cycle detection * finding sum of square of digits) → O(constt * logn) → O(logn)
    - *the outer while loop will run a few times*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isHappy(self, n: int) -> bool:
        
        def sum_square_digits(n: int) -> int:
            res = 0
            while n:
                digit = n % 10
                res += digit * digit
                n //= 10
            return res

        slow, fast = n, sum_square_digits(n)
        while slow != fast:
            slow = sum_square_digits(slow)
            fast = sum_square_digits(fast)
            fast = sum_square_digits(fast)
        return fast == 1
```
