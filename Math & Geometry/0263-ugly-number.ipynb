# <center> 263. Ugly Number </center>


## Problem Description
[Click here](https://leetcode.com/problems/ugly-number/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Divide the number n by 2, 3, and 5. If n becomes 1, it is an ugly number.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if n <= 0, return false because the ugly number is a positive integer
- loop until n is greater than 1
    - if it is divisible by 2, 3, or 5, divide it
    - else it is not divisible, return false
- if the loop completes without returning false, n has become 1, return true


## Complexity
- Time complexity: O(log$_5$n) â†’ O(logn)
    - *logn because in each iteration n is divided by 5, 3, or 2*
    - *log$_5$n is greater than log$_3$n and log$_2$n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isUgly(self, n: int) -> bool:
        if n <= 0:
            return False
        while n > 1:
            if n % 2 == 0:
                n //= 2
            elif n % 3 == 0:
                n //= 3
            elif n % 5 == 0:
                n //= 5
            else:
                return False
        return True
```
