# <center> 43. Multiply Strings </center>


## Problem Description
[Click here](https://leetcode.com/problems/multiply-strings/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Multiply digit by digit using multiplication rules. To get the digits in int form, use ASCII values.

Multiplication rules
- start from the right side
- each time take a digit of num1 and multiply it with all digits of num2
- if the product is a two-digit number, keep the rightmost digit and pass (add) the leftmost digits as a carry to the next digit


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if num1 is 0 or num2 is 0, return 0 because multiplication with 0 results in 0
- set m = size of string1, n = size of string2
- set res = a list of size m + n initialized with zeroes to store the digit multiplication result 
    - *m + n size list because the no. of digits in the multiplication result is equal to the sum of no. of digits of the two numbers*
- traverse num1 digits from the right side i.e in reverse
    - traverse num2 digits from the right side
        - multiply num1 digit at position i with num2 digit at position j
            - *get the digit by subtracting the ascii value of the current char from ascii value of 0 i.e 48*
        - add the product with the carry at its respective index in the result list i.e at index i + j + 1
        - store the rightmost digit of the sum
            - *rightmost digit = sum % 10*
        - pass the leftmost digit as carry to the next digit at position i + j
            - *carry = leftmost digit = sum // 10* 
- convert the values in the result list to string, remove the leading zeroes, and return the list in string form 


## Complexity
- Time complexity: O(string1 traversal * string2 traversal + join) → O(m * n + m + n) → O(m * n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result list) → O(m + n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def multiply(self, num1: str, num2: str) -> str:
        if num1 == '0' or num2 == '0':
            return '0'
        m, n = len(num1), len(num2)
        res = [0] * (m + n)
        for i in range(m - 1, -1, -1):
            for j in range(n - 1, -1, -1):
                prod = (ord(num1[i]) - 48) * (ord(num2[j]) - 48)
                summ = prod + res[i + j + 1]
                res[i + j + 1] = summ % 10
                res[i + j] += summ // 10
        return ''.join(map(str, res)).lstrip('0')
```
