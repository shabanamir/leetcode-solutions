# <center> 50. Pow(x, n) </center>


## Problem Description
[Click here](https://leetcode.com/problems/powx-n/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
even power = x$^2$ = x * x        
odd power = x$^3$ = x * x * x     

Use a loop. If the power is odd, multiply the result with the base first. Take the square of the base and divide power by 2.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - return 0 if x is 0 
        - *(result is 0 if base is 0)*
    - return 1 if n is 0 
        - *(result is 1 if power is 0)*
- set res = 1 to track the result
- if n (power) is negative
    - make n positive
    - take the reciprocal of x
        - *x$^{-2}$ = 1/x$^2$*
- loop until n becomes 0
    - if n is odd
        - multiply res with x first
    - take square of x
    - divide n by 2 
- return res


## Complexity
- Time complexity: O(log$_2$n) â†’ O(logn)
    - *logn because each time n is divided by 2*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def myPow(self, x: float, n: int) -> float:
        if x == 0:
            return 0
        if n == 0:
            return 1
        if n < 0:
            n = abs(n)
            x = 1 / x
        res = 1
        while n:
            if n & 1:      
                res *= x
            x *= x
            n //= 2       
        return res
```
