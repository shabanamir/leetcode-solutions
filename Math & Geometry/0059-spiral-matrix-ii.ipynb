## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
An n x n matrix has n/2 square/cycles. The first square is formed by the first row, last column, last row, and first column. The second square is formed by the second row, second-last column, second-last row, and second column, etc.

Create a n x n matrix. Divide the matrix into squares/cycles and fill elements in each square in clockwise direction i.e top → right → bottom → left


## Approach
- check base cases
    - if n is 0, return an empty list
- initialize nxn matrix with zeroes
- initialize four variables to track the indices for the corner points
    - *top = 0*
    - *left = 0*
    - *right = rightmost index i.e total columns - 1*
    - *bottom = bottom index i.e total rows - 1*
- set val = 1 to track the value to fill in
- run a loop until we get all squares i.e while left <= right and top <= bottom
    - *move in the square in clockwise direction i.e top row → right column → bottom row → left column and fill the cells*
    - fill the top row of the square by moving left to right using a loop
        - increment val after filling each cell
    - increment top after the loop to point to the top of the next square
    - fill the right col by moving top to bottom using a loop
        - increment val after filling each cell
    - decrement right after the loop to point to the right col of the next square
    - if the pointers are not out of bounds, break the loop
        - *this is helpful in case we are left with one row or column in the square*
    - fill the bottom row in reverse by moving right to left using a loop
    - decrement bottom
    - fill the left col in reverse by moving bottom to top using a loop
    - increment left
- return the matrix


## Complexity
- Time complexity: O(matrix creation + filling) → O(n * n + n * n) → O(n²)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result matrix) → O(n * n) → O(n²)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def generateMatrix(self, n: int) -> List[List[int]]:
        if not n:
            return []
        matrix = [[0] * n for _ in range(n)]
        left, right = 0, n - 1
        top, bottom = 0, n - 1
        val = 1
        while left <= right and top <= bottom:
            # move left to right and fill top row
            for c in range(left, right+1):
                matrix[top][c] = val
                val += 1
            top += 1
            # move top to bottom and fill right col
            for r in range(top, bottom+1):
                matrix[r][right] = val
                val += 1
            right -= 1
            if not (left <= right and top <= bottom):
                break
            # move right to left and fill bottom row
            for c in range(right, left-1, -1):
                matrix[bottom][c] = val
                val += 1
            bottom -= 1
            # move bottom to top and fill left col
            for r in range(bottom, top-1, -1):
                matrix[r][left] = val
                val += 1
            left += 1
        return matrix
```
