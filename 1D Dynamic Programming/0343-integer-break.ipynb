## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
- Approach 1: O(n/3) <br>
n = 10, max product = 3 x 3 x 4 <br>
n = 11, max prduct = 3 x 3 x 3 x 2 <br>
If we observe the pattern, the max product is obtained by repeatedly cutting parts of size 3 while the size is greater than 4, keeping the last part as 2, 3, or 4.
- Approach 2: O(log(n/3)) <br>
According to the maxima-minima concept, an integer n should be broken down into equal parts x to maximize the product (x$^{n/x}$). The value of x should be e (natural log base). As 2 < e < 3, x should be 2 or 3. <br>
So we will keep breaking the integer n in power of 3 only, until n becomes 4 or 2. <br>
Max product = 3$^{n/3}$ x last part  (last part can be 4, 2, or null)


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if n <= 3, the max prdouct is n - 1 
        - *n = 2 break into 1 + 1, max product = 1 x 1 = 1*
        - *n = 3 break into 2 + 1, max product = 2 x 1 = 2*
- find the last part and get the product of the power of 3 and the last part
    - if n is divisible by 3, there's no last part, return 3$^{n/3}$
        - *n = 6 break it into 3 + 3* 
    - if n mode 3 is 1, the last part is 4, return 3$^{n/3 - 1}$ * 4
        - *n = 10 break into 3 + 3 + 4* 
    - else n mode 3 is 2, last part is 2, return 3$^{n/3}$ * 2
        - *n = 11 break into 3 + 3 + 3 + 2*


## Complexity
- Time complexity: O(power) → O(log(exponent)) → O(log(n/3))
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def integerBreak(self, n: int) -> int:
        if n <= 3:
            return n - 1
        if n % 3 == 0:
            return 3 ** (n // 3)
        if n % 3 == 1:
            return 3 ** ((n // 3) - 1) * 4
        return 3 ** (n // 3) * 2
```
