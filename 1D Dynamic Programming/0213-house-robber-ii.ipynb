## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
This is the same as [House Robber I](https://leetcode.com/problems/house-robber/) but in House Robber II, the houses are arranged in a circle, which means the first and last houses are adjacent and cannot be robbed together. 

We have two cases:
1. include the first house in the rob list and exclude the last house i.e all elements except the last
2. vice versa 

Find the max amount of money you can rob in each case using the logic of [House robber I solution](https://leetcode.com/problems/house-robber/solutions/3578630/o-n-time-o-1-space-solution-explained/). Compare the robbing of both cases and return the max.


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = total number of houses i.e size of the input list
- check base cases:
    - if there are one or two elements (houses), return the max of the array as the max money you can rob
- define a helper function to find the max amount of money you can rob in each case, using the logic of [House robber I solution](https://leetcode.com/problems/house-robber/solutions/3578630/o-n-time-o-1-space-solution-explained/) <br>
helper(array start, array end)
- return the max of case i and case ii


## Complexity
- Time complexity: O(case i + case ii) → O(n + n) → O(2n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def rob(self, nums: List[int]) -> int:
        n = len(nums)
        if n == 1 or n == 2:
            return max(nums)
        
        def helper(l: int, r: int) -> int:
            rob1, rob2 = 0, 0
            for i in range(l, r):
                res = max(nums[i] + rob1, rob2)
                rob1 = rob2
                rob2 = res
            return res
            
        return max(helper(0, n - 1), helper(1, n))
```
