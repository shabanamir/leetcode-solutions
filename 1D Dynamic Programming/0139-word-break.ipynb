# <center> 139. Word Break </center>


## Problem Description
[Click here](https://leetcode.com/problems/word-break/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have to find out if the string s can be segmented into the dictionary words i.e it is formed by using the words in the input array.

*Note: we have to use the complete word, we can't use a part of it.*

A string of size n can be segmented if its substring of size n - 1 can be segmented.  So, to solve the main problem (string segmentation), we have to solve the sub-problems (substring segmentation), use dynamic programming.


## Approach
<!-- Describe your approach to solving the problem. -->
- set dp = a boolean array of size len(s) + 1, where dp[i] represents whether the substring up to index i can be segmented 
    - *index = substring size*
    - *value = is it possible to segment*
- set value at index 0 to True, because an empty string can always be segmented
- for every substring in the string s, check if the substring can be segmented
    - for each index (substring size) i 
        - for each word w in the dictionary
            - if current index (substring size) >= current word size and substring before the current can be segmented and the current substring ends with the current word
                - the current substring can be segmented, set dp[i] to true
                - no need to check the remaining words, break the inner loop
- return the last index value from the dp array i.e dp[-1] that represents whether the entire string s can be segmented or not


## Complexity
- Time complexity: O(dp array creation + outer loop * inner loop * endswith() for substring check) → O(input string size + input string size * words array size * max substring size) → O(m + m * n * 20) → O(m * n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(dp array) → O(string s size) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def wordBreak(self, s: str, wordDict: List[str]) -> bool:
        dp = [False] * (len(s) + 1)
        dp[0] = True
        for i in range(1, len(s) + 1):
            for w in wordDict:
                if i >= len(w) and dp[i - len(w)] and s[:i].endswith(w):
                    dp[i] = True
                    break
        return dp[-1]
```
