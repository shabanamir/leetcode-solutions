# <center> 152. Maximum Product Subarray </center>


## Problem Description
[Click here](https://leetcode.com/problems/maximum-product-subarray/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find a subarray with max product. It is similar to the [Maximum Sum Subarray](https://leetcode.com/problems/maximum-subarray/description/) problem. 

Use Kadane's algorithm.

## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there is only one element in the array, it is the max product, return it
- set res = the first element to track the global max product
- set cur_min and cur_max to track the current minimum and maximum product
    - *(we need to track both max and min product because there are negative values in the input array)*
- traverse the nums array <br>
for each number n in the array
    - add the number to the current subarray and get the maximum and minimum product of the current subarray i.e multiply the number with cur_max and cur_min
    - update cur_max by taking the max of current subarray max, current subarray min, and current element
        - *(in this step, we decide whether to add the current number to the subarray (if maxi or mini is greater) or start a new subarray (if n is greater))*
    - update cur_min by taking the min of current subarray max, current subarray min, and current element
    - update res by taking max of current max and previous max product
- return res


## Complexity
- Time complexity: O(input array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxProduct(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return nums[0]
        res = nums[0]
        cur_min = cur_max = 1
        for n in nums:
            maxi = n * cur_max
            mini = n * cur_min
            cur_max = max(maxi, mini, n)
            cur_min = min(maxi, mini, n)
            res = max(res, cur_max)
        return res
```
