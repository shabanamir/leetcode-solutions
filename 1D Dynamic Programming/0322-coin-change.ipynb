## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a list of coins and can use each coin any number of times. We need to find the minimum number of coins required to make the target amount.

If we choose a coin from the coins list, <br>
min coins required to make target amount = 1 + min coins required to make the remaining amount <br>
*(1 because we have chosen a coin)*

For example: <br>
coins = [1, 3, 4], target amount = 7 
- if we choose coin 1 <br>
remaining amount = target amount - coin value = 7 - 1 = 6 <br>
min coins required to make amount 7 = 1 + min coins required to amount 6
- if we choose coin 3 <br>
remaining amount = 7 - 3 = 4 <br>
min coins to make amount 7 = 1 + min coins required to make amount 4
- if we choose 4 <br>
remaining amount = 7 - 4 = 3 <br>
min coins to make amount 7 = 1 + min coins required to make amount 3

So, each time we also need to find the minimum coins to make the remaining amount i.e to solve the main problem (target amount), we need to solve the sub-problem (remaining amount). Use dynammic programming. 

For each amount in the range 0 to target amount, traverse all coins and find the min coins that can make that amount. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = a value greater than the target amount i.e amount + 1
- set dp = a list of size m where dp[i] represents the minimum number of coins required to make amount i
    - *index = amount*
    - *value = min required coins*
 - *set value at index 0 to 0 because 0 coins are required for amount 0*
 - *set the value at other indices to m (value greater than the given amount)*
- for each amount in the range 1 to m
    - for each coin
        - if the remaining amount is not negative
            - find the min coins required to make that amount using the formula
                - *min coins to make current amount = min(current value, 1 + min coins to make remaining amount)*
- if the value at index equal to target amount in the dp array is still m, we couldn't make the target amount using the coins, return -1. Else return dp[target amount] as the minimum coins required


## Complexity
- Time complexity: O(amount * coins) → O(m * n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(dp list) → O(amount) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def coinChange(self, coins: List[int], amount: int) -> int:
        m = amount + 1
        dp = [m] * (m)
        dp[0] = 0
        for a in range(1, m):
            for c in coins:
                if a - c >= 0:
                    dp[a] = min(dp[a], 1 + dp[a - c])
        return -1 if dp[amount] == m else dp[amount]
```
