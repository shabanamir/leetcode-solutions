## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The total combinations that sum up to a target value x depend on the combinations that sum up to value x - i, where i is a number from the input numbers list that is <= x.  

C(x) = sum of all C(x-i) where i <= x <br>

For example: <br>
target = 4, nums list = [1, 2, 5, 3] <br>
C(4) =  C(4-1) + C(4-2) + C(4-3) <br>
C(4) = C(3) + C(2) + C(1)

To solve the main problem C(x), we need to solve the subproblems C(x-i), use dynamic programming.


## Approach
<!-- Describe your approach to solving the problem. -->
- set dp = an array of size target + 1 with zeroes where dp[i] represents the number of combinations that sum up to i.
- set value at index 0 to 1 because there is one way to reach the sum of 0 i.e by selecting no numbers
- traverse all sums in range 1 to the target sum
    - traverse all numbers in the input arrsy
        - if current number <= current sum, we can select the current number and can reach the current sum 
            - update the value at index equal to current sum in the dp array i.e dp[i] using the formula
- return the last index value from the dp array that represents the number of combinations that sum up to the target 


## Complexity
- Time complexity: O(outer loop * inner loop) → O(target * input array) → O(t * n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(dp array) → O(target) → O(t)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def combinationSum4(self, nums: List[int], target: int) -> int:
        dp = [0] * (target + 1)
        dp[0] = 1
        for i in range(1, target + 1):
            for n in nums:
                if n <= i:
                    dp[i] += dp[i - n]
        return dp[-1]  
```
