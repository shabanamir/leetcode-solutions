## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We can't rob adjacent houses.

We have two options. 
1. rob current house i <br>
(the robber can't rob the previous house, he can rob the current house i and the house before the previous i-2)

2. don't rob the current house i <br>
(the robber can't rob the current house, he can rob the previous house i-1 and any next)

Traverse the houses, and choose the option with the max value. <br>
max(option 1, option 2) <br>
max(rob(i-2) + rob(i), rob(i-1))


## Approach
<!-- Describe your approach to solving the problem. -->
- initialize two variables rob1 and rob2 to represent the two options
- traverse the houses
    - find the option with max value
    - update rob1 and and rob2
- return res or rob2


## Complexity
- Time complexity: O(loop) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def rob(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return nums[0]
        rob1, rob2 = 0, 0
        for n in nums:
            res = max(n + rob1, rob2)
            rob1 = rob2
            rob2 = res
        return res
```
