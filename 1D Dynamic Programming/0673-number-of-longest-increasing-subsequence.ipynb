# <center> 673. Number of Longest Increasing Subsequence </center>


## Problem Description
[Click here](https://leetcode.com/problems/number-of-longest-increasing-subsequence/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
To find the total number of longest increasing subsequence (LIS) in the given array i.e LIS ending at index n, we need to find the number of LIS ending at each index.

So, to solve the main problem (LIS at index n), we need to solve the sub-problem (LIS at each index), use dynamic programming.


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = input array size
- set size = a list of size n where size[i] represents length of the LIS ending at index i
- set freq = a list of size n where freq[i] represents total number of LIS ending at index i
- set maxi =  1 to track the length of the LIS
- set count = 1 to track the total number of LIS
- for i in range 1 to n
    - for j in range 0 to i
        - if current subseq is an increasing subseq i.e index i element is greater than index j element
            - if size of the LIS ending at index j is greater than the current LIS ending at i, we found a LIS ending at i
                - update size[i] and freq[i]
            - if size of the LIS ending at j = current LIS ending at i, we found another LIS with the same length as the current LIS ending at i
                - only update the freq[i]
    - if length of the current LIS ending at i is greater than maxi, update maxi and count
    - if length of the current LIS ending at i is equal to maxi, only update the count
- return count as the total number of longest increasing subsequences


## Complexity
- Time complexity: O(creating size list + creating freq list + loops) → O(n + n + n * constt) → (n * constt) → (n)
    - *the inner loop doesn't run n times for each outer iteration*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(size list + freq list) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findNumberOfLIS(self, nums: List[int]) -> int:
        n = len(nums)
        size, freq = [1] * n, [1] * n
        count, maxi = 1, 1
        for i in range(1, n):
            for j in range(i):
                if nums[i] > nums[j]:
                    if size[j] + 1 > size[i]:
                        size[i] = size[j] + 1
                        freq[i] = freq[j]
                    elif size[j] + 1 == size[i]:
                        freq[i] += freq[j]
            if size[i] > maxi:
                maxi = size[i]
                count = freq[i]
            elif size[i] == maxi:
                count += freq[i]
        return count
```
