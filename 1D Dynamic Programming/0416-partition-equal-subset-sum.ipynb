## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to divide the input array into two subarrays of equal sum (if possible).

subarray sum = input array sum  / 2

Constraints:
- we can put an element as a whole either in subarray1 or subarray2
- sum of each subarray should be half of the input array sum and a whole number because the input array elements are integers and we have to include them as a whole in a subarray i.e we can't take a fraction/part of it. So if the input array sum is odd, it cannot be divided into two subarrays

We can do the partition, if the input array sum is even and we can find elements that contribute to the subarray sum i.e half of the input array sum.

For example: 
- arr = [2, 4, 3, 3] <br>
 array sum = 12 → even <br>
 subarray sum = 6 <br>
 partition is possible because 2 and 4 sum to 6
- arr = [4, 3, 3] <br>
 array sum = 10 → even <br>
 subarray sum = 5 <br>
 partition is not possible because no elements in the input array sum to 5
- arr = [4, 3, 3, 5] <br>
 array sum = 15 → odd <br>
 partition is not possible because array sum is odd

This problem is similar to the 0/1 knapsack problem because the constraints are similar. In 0/1 knapsack, we have n items with weights and profits and a bag capacity. We need to put items in the bag so that the profit is maximized and the bag weight does not exceed the capacity. For each item, we have two choices: to include it as a whole in the bag or exclude it. Similarly, in this problem, we are given n items and need to put them into two bags (subarrays) such that each bag weight is equal to the target sum. We have two choices for each item: include it in bag1 or exclude it (include it in bag2).


## Approach
<!-- Describe your approach to solving the problem. -->
- find the sum of the input array
- check base cases
    - if array sum is odd, we can not do the partition
        - return false
- find the target sum i.e half of the total sum
- set dp = a boolean list of size equal to the target sum + 1 with a default value of false
    - *dp[i] represents whether it is possible to get sum i using the input array elements we’ve seen so far*
- set dp[0] to true because we can always get sum 0 by selecting no elements 
- traverse the input array <br>
for each number in the input array
    - traverse the dp list in reverse order <br>
    for each sum i in range target sum  to current number
        - check if we can get the current sum i by including the current number or excluding it and update the value at index i in the dp array 
            - *if dp[i] is true it means we can get the current sum i by including the current element*
            - *if dp[i - current num] is true it means we can get the current sum i by excluding the current element*
    - the last index represents the target sum. If after the inner loop, the last index value is true, we can get the target sum i.e it is possible to do the partition
        - return true
- return the last index value from the dp array i.e dp[-1] that represents whether it is possible to get the target sum or not 


## Complexity
- Time complexity: O(outer loop * inner loop) → O(input array * target sum) → O(n * k)
    - *k = target sum = sum(input array sum) / 2*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(dp array) → O(target sum) → O(k)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def canPartition(self, nums: List[int]) -> bool:
        summ = sum(nums)
        if summ & 1:
            return False
        target = summ // 2
        dp = [False] * (target + 1)
        dp[0] = True
        for num in nums:
            for i in range(target, num - 1, -1):
                dp[i] = dp[i] or dp[i - num]
            if dp[-1]:
                return True
        return dp[-1]
```
