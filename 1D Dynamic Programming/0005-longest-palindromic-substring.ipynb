# <center> 5. Longest Palindromic Substring </center>


## Problem Description
[Click here](https://leetcode.com/problems/longest-palindromic-substring/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
In a plaindrome, the left and right characters are the same. 
- in an odd-length palindrome, there is a central character for example 'aba'
- in an even-length palindrome there is no central character for example 'abba'.

For each char in the input string, expand to the left and right to check even and odd length palindromes using a two-pointer approach.
Keep track of the palindrome sizes and return the max length palindrome.


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = length of the string
- check base cases
    - if the string size is 1, it is a palindrome
        - return the input string as the longest palindrome
- set two variables start = 0, end = 0 to track the start and end indices of the longest palindromic substring
- traverse the string
    - break the loop if the current substring length (n - i) is less than the max length substring we have found
    - check odd-length palindromes
        - set two variables l and r to track the left and right of odd-length palindromes with the current char as the left and right char of the palindrome
        - loop until l and r are inbound and the left and right characters are the same 
            - if the length of the current palindrome (r - l) is greater than the current max length (end - start)
                - update the start and end
            - update l and r to expand to the left and right by one character
    - check even-length palindromes
        - set two variables l and r to track the left and right of an even-length palindrome with the current char as the starting left char and the next char as the starting right char of the palindrome
        - loop until l and r are inbound and the left and right characters are the same 
            - if the length of the current palindrome (r - l) is greater than the current max length (end - start)
                - update the start and end
        - update l and r to expand to the left and right by one character
- return the max length palindrome using the start and end indices


## Complexity
- Time complexity: O(string traversal * (odd-length palindrome check + even-length palindrome check) + slicing) → O(n * (n + n) + n) → O(n * n + n * n + n) → O(n² + n² + n) → O(n²)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def longestPalindrome(self, s: str) -> str:
        n = len(s)
        if n == 1:
            return s
        start, end = 0, 0
        for i in range(n):
            # possibility of max length substring
            if n - i < (end - start) / 2:
                break
            # odd length
            l, r = i, i
            while l >= 0 and r < n and s[l] == s[r]:
                if r - l > end - start:
                    start = l
                    end = r
                l -= 1
                r += 1
            # even length
            l, r = i, i + 1
            while l >= 0 and r < n and s[l] == s[r]:
                if r - l > end - start:
                    start = l
                    end = r
                l -= 1
                r += 1
        return s[start:end+1]
```
