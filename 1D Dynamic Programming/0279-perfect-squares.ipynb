## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the minimum number of perfect squares that sum to the given number.

Approaches:
1. Dynamic programming <br>
The number of perfect squares that sum to a number x is in range 0 to sqrt(x). We can reach a sum i using a number j as a perfect square if we can reach sum i - j$^2$. So, to solve the main problem we have to solve the sub-problems. Use dynammic programming .

2. Lagrange's four-square theorem <br>
according to the theorem, any positive number can be represented by at most 4 square number sum. The theorem steps are:
    - keep dividing the number by 4 as long as it is divisible
    - if after the division by 4, dividing it by 8 results in a remainder of 7, the answer is 4 i.e the number is a sum of four square numbers
    - else if any two numbers can square sum to n, the answer is 1 or 2
    - else answer is 3


## Approach
<!-- Describe your approach to solving the problem. -->
**Dynamic Programming Approach**
- check base cases
    - if n is a perfect square for example 4, 9, 16 etc, return 1
- set dp = an array of size n with a default max value (use n or max integer) where dp[i] represents the number of perfect squares that sum to i
- set dp[0] to 0 and dp[1] to 1 because the total square numbers that sum to 0 and 1 are 0 and 1 respectively
- for each sum i in range 2 to n
    - for each perfect square j in range 1 to sqrt(i)
        - update dp[i] using the formula
            - *(we can reach sum i using the current number j as a perfect square if we can reach sum i - j$^2$)* 
- return number of perfect squares that sum to n i.e dp[n]
    
**Math Approach**
- check base cases
    - if n is a perfect square for example 4, 9, 16 etc, return 1
- apply the theorem steps
    - loop to keep dividing the number by 4 as long as it is divisible
    - if after the division by 4, dividing it by 8 results in a remainder of 7
        - return 4
    - else if any two numbers can square sum to n, the answer is 1 or 2. We have already handled the 1 in the base case, check for 2 using a loop
    - else return 3


## Complexity
- Time complexity:
    - Dynamic Programming  O(base case + outer loop * inner loop) → O(logn + n * sqrt(n)) → O(n * sqrt(n))
        - *for each outer iteration i the inner loop runs sqrt(i) times*
    - Math Approach O(base case + while loop + for loop) → O(log$_2$(n) + log$_4$(n) + sqrt(n)) → O(sqrt(n))

- Space complexity:
    - Dynamic Programming Approach O(dp array) →  O(n)
    - Math Approach O(1)

## Code

```python
class Solution:

    def numSquares(self, n: int) -> int:
        # dynammic  approach
        if sqrt(n).is_integer():
            return 1 
        dp = [n] * (n + 1)
        dp[0], dp[1] = 0, 1
        for i in range(2, n + 1):
            for j in range(1, int(i ** 0.5) + 1):
                dp[i] = min(dp[i], dp[i - j * j] + 1)
        return dp[n]

        # math approach
        if sqrt(n).is_integer():
            return 1 
        while n % 4 == 0: 
            n /= 4
        if n % 8 == 7: 
            return 4
        for i in range(int(sqrt(n)) + 1):
            if int(sqrt(n - i * i))**2 == n - i * i: 
                return 2
        return 3
```
