## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
This is similar to [Longest Palindromic Substring](https://leetcode.com/problems/longest-palindromic-substring/description/) problem but here we have to count the palindromes.

In a plaindrome, the left and right characters are the same.
- in an odd-length palindrome, there is a central character for example 'aba'
- in an even-length palindrome there is no central character for example 'abba'.

For each char in the input string, expand to the left and right to check even and odd length palindromes using a two-pointer approach. Keep track of the palindromes count and return it.


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = input string length
- if string length is 1, it is a palindrome
    - return 1 as total number of palindromes
- set res = 0 to count the palindromes
- traverse the string <br>
for every char in the string
    - check odd-length palindromes
        - set two variables l and r to track the left and right of odd-length palindromes with the current char as the left and right char of the palindrome
        - loop until l and r are inbound and the left and right characters are the same 
            - add 1 to res because we found a palindrome
            - update l and r to expand to the left and right by one character
    - check even-length palindromes
        - set two variables l and r to track the left and right of an even-length palindrome with the current char as the starting left char and the next char as the starting right char of the palindrome
        - loop until l and r are inbound and the left and right characters are the same 
            - add 1 to res because we found a palindrome
            - update l and r to expand to the left and right by one character
- return res


## Complexity
- Time complexity: O(string traversal * (odd-length palindrome check + even-length palindrome check)) → O(n * (n + n)) → O(n * n + n * n) → O(n² + n²) → O(n²)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def countSubstrings(self, s: str) -> int:
        n = len(s)
        if n == 1:
            return 1
        res = 0
        start, end = 0, 0
        for i in range(n):
            # odd length
            l, r = i, i
            while l >= 0 and r < n and s[l] == s[r]:
                res += 1
                l -= 1
                r += 1
            # even length
            l, r = i, i + 1
            while l >= 0 and r < n and s[l] == s[r]:
                res += 1
                l -= 1
                r += 1
        return res        
```
