## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a triangle array and need to find a path whose sum from top to bottom is minimum by moving to lower-left or lower-right elements in the path.

The minimum path sum to reach a row depends on the minimum path sum to reach the previous row. So, to solve the main problem (min path sum to reach the target row), we need to solve the sub-problems (min path sum to reach the previous row). Use dynammic programming.

Calculate the path sum from backward (i.e bottom to top) using bottom-up dynamic programming.


## Approach
<!-- Describe your approach to solving the problem. -->
- set dp = last row where dp[i] represents the min path sum to reach row i from the bottom (last row)
    - *index = row number*
    - *value = min path sum to reach the row from the last row*
- traverse the rows in reverse <br>
for each row starting from the second last row
    - traverse the columns <br>
    for each col in the row
        - find the cost of moving lower left
        - find the cost of moving lower right
        - update the cost of the current row in the dp array by taking the minimum of the two costs
- return the first value in the dp array that represents the min path sum to reach the top row


## Complexity
- Time complexity: O(outer loop * inner loop) → O(rows * columns) → O(n²)
    - the input array is a triangle array i.e each row contains a different number of columns (elements) and the number of cols in a row = row number <br>
for example, there is 1 element in row 1, 2 in row 2, 3 in row 3 etc <br>
so, the pattern of execution is 1 + 2 + 3 + 4 + ... + n times = n(n+1)/2 (n² + n)/2 = O(n²)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(dp array) → O(last row size) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minimumTotal(self, triangle: List[List[int]]) -> int:
        dp = triangle[-1]
        for i in range(len(triangle) - 2, -1, -1):
            for j in range(i + 1):
                # dp[j] = triangle[i][j] + min(dp[j], dp[j + 1])
                lower_left = triangle[i][j] + dp[j]
                lower_right = triangle[i][j] + dp[j + 1]
                dp[j] = min(lower_left, lower_right)
        return dp[0]
```
