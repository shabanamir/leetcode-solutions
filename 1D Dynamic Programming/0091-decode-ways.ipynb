# <center> 91. Decode Ways </center>


## Problem Description
[Click here](https://leetcode.com/problems/decode-ways/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The English letters are 26 and represented by numbers 1 to 26 inclusive. At most, we can group two digits if they are inbound. 

The number of ways to decode a string of size n is the sum of:
- no. of ways to decode string of size n-1 if n-1 is not 0
- no. of ways to decode string of size n-2 if the number represented by n-1 and n-2 is inbound (i.e 10-26)

So, to solve the main problem (n), we need to solve the sub-problems i.e (n-1 and n-2). Use dynamic programming.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if there is a leading zero, we can't decode the string, return 0
- set n = string size
- set dp = a list of size n where dp[i] represents number of ways to decode string of size i
    - *index = string size*
    - *value = ways to decode*
- set the value at index 0 and 1 to 1 because there is 1 way to decode an empty or 1-letter string
- traverse the string starting from the third index (2) to keep index inbound
    - check for the above two conditions and fill the dp array
        - if the last digit is not 0
            - update dp[i]
        - if the last two digits are inbound (10-26) i.e the second last digit is either 1 or 2 and the last digit is less than 7
            - update dp[i]
- return the value at index n (last index) as the number of ways to decode the string of size n

## Complexity
- Time complexity: O(loop) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(dp list) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def numDecodings(self, s: str) -> int:
        if s[0] == '0': 
            return 0
        n = len(s)
        dp = [0] * (n + 1)
        dp[0] = dp[1] = 1          
        for i in range(2, n + 1):
            if s[i - 1] != '0':
                dp[i] += dp[i - 1]
            if s[i - 2] == '1' or s[i - 2] == '2' and s[i - 1] < '7':
                dp[i] += dp[i - 2]
        return dp[n]
```
