# <center> 740. Delete and Earn </center>


## Problem Description
[Click here](https://leetcode.com/problems/delete-and-earn/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
If we earn an element nums[i]:
- we earn all its duplicates
- we can not earn its first neighbors (immediate left and right values) i.e nums[i] - 1 and nums[i] + 1

Use dynamic programming. For each element, find the points we can earn i.e sum of its duplicates. Traverse the points/sums array and check if we can take the points. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set dp = a list of size equal to the max of the input array and set all values to zeroes where dp[i] represents the sum of duplicates of i. 
    - for example, input array = [2,2,3,3,3,4], dp = [0, 0, 4, 9, 4]
- for each number n in the input array
    - update dp[n] by adding n to the value at index n 
- set prev = 0 to track previous max sum
- set cur = 0 to track current max sum
- for each value in the dp list
    - check if taking the current value can maximize the current sum and update cur and prev
- return cur


## Complexity
- Time complexity: O(dp array creation + input array traversal + dp array traversal) → O(max(n) + n + max(n)) → O(n)
    - *max(n) < n*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(dp array)  → O(max(n))
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def deleteAndEarn(self, nums: List[int]) -> int:
        dp = [0] * (max(nums) + 1)  
        for n in nums:
            dp[n] += n
        prev = cur = 0
        for n in dp:
            cur, prev = max(cur, prev + n), cur
        return cur
```
