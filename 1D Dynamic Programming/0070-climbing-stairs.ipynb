## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The number of ways to reach the top using n steps is equal to the n+1 Fibonacci number. For example:
#steps #ways to the reach top <br>
n  fib(n) <br>
0  0 <br>
1  1 <br>
2  1 <br>
3  2 <br>
4  3 <br>
5  5 <br>
6  8 

*Note: Each number in the Fibonacci series is the sum of two previous numbers.*


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if n is less than or equal to 3 return n because:
      - *n = 1, fib(n+1) = fib(2) = 1 way to reach the top*
      - *n = 2, fib(2+1) = fib(3) = 2 ways to reach the top*
      - *n = 3, fib(3+1) = fib(4) = 3 ways to reach the top*
- initialize two variables n1 and n2 to track the previous two numbers
    - *n1 = 2 and n2 = 3 (because we start from fib4)*
- start from fib 4 and loop until n + 1
    - compute the next fibinoacci number i.e res by taking sum of n1 and n2. 
    - update n1, n2
        - *(res is for clarity, we can skip res and directly update n1 and n2 i.e n1, n2 = n2, n1 + n2)*
- return res *(or n2 if we directly update the numbers)*


## Complexity
- Time complexity: O(loop) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def climbStairs(self, n: int) -> int:
        if n <= 3:
            return n
        n1 = 2
        n2 = 3
        for _ in range(4, n + 1):
            res = n1 + n2
            n1 = n2
            n2 = res
        return res
```
