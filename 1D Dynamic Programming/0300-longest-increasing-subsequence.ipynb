## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Find the longest increasing subsequences (LIS) ending at each index and return the max.

LIS at index i <br>
= max(length of increasing subsequences til index i) <br>
= 1 (if index i element is greater) + length of LIS until index i - 1

So, to solve the main problem (LIS at index i), we need to solve the sub-problems (LIS at index i-1). We can use dynamic programming.

1. Dynamic programming
2. Dynamic programming with binary search


## Approach
<!-- Describe your approach to solving the problem. -->
**Dynamic Programming**
- set dp = an array of size n with default value of ones where dp[i] represents the length of an increasing subsequence up to index i
    - *initialize the array with ones because the length of a LIS ending at each index is at least 1*
- for i in range 1 to n i.e 2nd element to last <br>
start from the second index because the length of a subsequence ending at index 0 is 1
    - for j in range 0 to i i.e first element to ith element
        - if the ith element is greater and length of current LIS of i <= LIS of j
            - update dp[i] using the formula above
- return the max of all LIS lengths i.e max(dp)

**Dynamic Programming With Binary Search**


## Complexity
- Time complexity:
    - Approach 1 O(nums * nums) → O(n * n) → O(n$^2$)
    - Approach 2 O(nums * binary search) → O(n * logn) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: 
    - Approach 1 O(dp array) → O(n)
    - Approach 1 O(dp array) → O(n)

## Code

```python
class Solution:

    def lengthOfLIS(self, nums: List[int]) -> int:
        
        # dynamic programming O(n^2)
        dp = [1] * len(nums)
        for i in range(1, len(nums)):
            for j in range(i):
                if nums[i] > nums[j] and dp[i] <= dp[j]:
                    dp[i] = dp[j] + 1
        return max(dp)
        
        # dynamic programming with binary search O(nlogn)
        dp = []
        for num in nums:
            if not dp or num > dp[-1]:
                dp.append(num)
            else:
                idx = bisect_left(dp, num)
                dp[idx] = num
        return len(dp)
```
