## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We have two options: 

1. take 1 steps
2. take 2 steps

Traverse the cost array, each time compare the cost of the two options, and choose the option with minimum cost.


## Approach
<!-- Describe your approach to solving the problem. -->
- initialize two variables n1 and n2 to represent the two options
- start from the 3rd step in the stair and loop until n + 1
    - find the current minimum cost i.e  min(n1 + cost[i - 2], n2 + cost[i - 1])
    - update n1 and n2
        - *n1 = previous min cost i.e n2*
        - *n2 = current min cost i.e res*
        - *res is for clarity, we can also directly update n1 and n2*
- return res or n2


## Complexity
- Time complexity: O(loop) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minCostClimbingStairs(self, cost: List[int]) -> int:
        n1 = n2 = 0
        for i in range(2, len(cost) + 1):
            res = min(n1 + cost[i - 2], n2 + cost[i - 1])
            n1 = n2
            n2 = res
        return res
```
