# <center> 1254. Number of Closed Islands </center>


## Problem Description
[Click here](https://leetcode.com/problems/number-of-closed-islands/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
An island is a closed island if it is surrounded by water i.e it is not on the boundary.

Do DFS to find islands (connected components) in the grid (graph) and check if it is closed or not.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the number of closed islands
- set m = total rows, n = total columns
- define a helper function for DFS <br>
dfs(row, column)
    - if the indices are out of bounds, the cell doesn't belong to a closed island, return 0
    - if the cell value is not 0, it is either a water cell or a visited land cell, we got a closed island, return 1
    - else cell value is 0 i.e we got an unvisited land cell
        - mark it visited
            - *here, we are updating the cell value to mark it as visited, we can also use a hashset for marking*
            - *0 = land cell, 1 = water cell, 2 = visited land cell*
        - explore all 4 neighbors (top, bottom, left, right) using DFS and return 1 if all the neighbors belong to a closed island, else return 0. In other words, return the minimum of the neighbors' dfs result
- traverse all cells <br>
for each cell on the grid
    - if the cell is an unvisited land cell (value = 0)
        - do DFS to find if this belongs to a closed island and update the res
- return res


## Complexity
- Time complexity: O(grid/graph traversal) → O(adjacency matrix DFS) → O(V * V) → O(rows * columns) → O(m * n) → O(m * n)
    - *V = total vertices or nodes or rows or sublists or keys*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(DFS recursion stack) → O(V) → O(rows) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def closedIsland(self, grid: List[List[int]]) -> int:
        res = 0
        m, n = len(grid), len(grid[0])

        def dfs(r: int, c: int) -> int:
            if r < 0 or r == m or c < 0 or c == n:
                return 0
            if grid[r][c] != 0:
                return 1
            grid[r][c] = 2
            return min(dfs(r - 1, c), dfs(r + 1, c), dfs(r, c - 1), dfs(r, c + 1))

        for row in range(m):
            for col in range(n):
                if grid[row][col] == 0:
                    res += dfs(row, col)
        return res
```
