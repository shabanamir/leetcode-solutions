## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given an undirected connected graph represented using an adjacency list and we need to clone it.

Traverse the graph using DFS and clone each node. Use a hashmap to track the visited/cloned nodes.


## Approach
<!-- Describe your approach to solving the problem. -->
- if there are no nodes, return None
- create a hashmap
- define a helper function for DFS <br>
dfs(current node)
    - if current node is already visited and cloned, return its copy node from the hashmap
    - else copy the node and add it to the hashmap
    - for each neighbor of the current node
        - recursively copy the neighbor node and add it to the neighbor node list of the current node copy
    - return the current node copy
- start DFS from the head node


## Complexity
- Time complexity: O(graph traversal) → O(adjacency list DFS) → O(V + E)
    - *V = total vertices or nodes or rows or sublists or keys*
    - *E = edges, for n vertices there are n(n-1)/2 edges in a complete graph*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(DFS recursion stack) → O(V)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
"""
# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
"""


class Solution:

    def cloneGraph(self, node: 'Node') -> 'Node':
        if not node:
            return None
        clone = {}

        def dfs(node: 'Node') -> 'Node':
            if node in clone:
                return clone[node]
            copy = Node(node.val)
            clone[node] = copy
            for n in node.neighbors:
                copy.neighbors.append(dfs(n))
            return copy

        return dfs(node)
```
