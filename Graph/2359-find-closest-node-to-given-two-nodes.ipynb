# <center> 2359. Find Closest Node to Given Two Nodes </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-closest-node-to-given-two-nodes/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the closest (shortest distance) node to node1 and node2.

- use BFS to find the shortest distance from node1 to all nodes and store them in a hashmap
- do the same for node2
- finally, iterate over all nodes, and find the node that is closest to node1 and node2 using the hashmaps 


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if node1 and node2 are the same, return any of the two nodes as both are closest to each other
- set dist_node1 = a hashmap to store the shortest distance of each node from node 1
    - *key = node*
    - *value = shortest distance from node 1*
- set dist_node2 = a hashmap to store the shortest distance of each node from node 2
    - *key = node*
    - *value = shortest distance from node 2*
- set res = -1 to track the closest node from both node 1 and node 2
- set min_dist to a very large number to track the closest node distance
- define a helper function for BFS <br>
bfs(source node, distance hashmap)
    - create a queue for BFS and add the source node
    - add the distance to reach the source node in the distance hashmap
    - traverse all levels i.e loop until the queue is empty
        - pop queue to get the current node
        - get the neighbor of the current node
        - if there is an edge to the neighbor node and the neighbor node is not visited
            - add the neighbor to the queue
            - add the distance to reach the neighbor node from the source node in the hashmap
                - *neighbor distance = current node distance + 1*
- do BFS from node1 to find the shortest distance of each node from node1
- do BFS for node2
- traverse all node, *the nodes are in range 0 to n i.e size of the input edges list*
    - if the node is in both hashmaps i.e we can reach the node from node1 and node2
        - find its distance
        - if the distance is less than the current minimum distance, this node is closest to node1 and node2
            - update the min_dist and res
- return res


## Complexity
- Time complexity: O(node1 BFS + node2 BFS + nodes traversal) → O(V + V + V) → O(V) → O(n)
    - *V = total vertices or nodes*
    - *1D array directed graph BFS = O(V)*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(node1 BFS queue + node2 BFS queue + node1 hashmap + node2 hashmap) → O(V + V + V + V) → O(V) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def closestMeetingNode(self, edges: List[int], node1: int, node2: int) -> int:
        if node1 == node2:
            return node1
        dist_node1 = {}
        dist_node2 = {}
        res = -1
        min_dist = float('inf')

        def bfs(src, dist_map) -> None:
            q = deque([src])
            dist_map[src] = 0
            while q:
                node = q.popleft()
                neighbor = edges[node]
                if neighbor != -1 and neighbor not in dist_map:
                    q.append(neighbor)
                    dist_map[neighbor] = dist_map[node] + 1
        
        bfs(node1, dist_node1)
        bfs(node2, dist_node2)
        for node in range(len(edges)):
            if node in dist_node1 and node in dist_node2:
                cur_dist = max(dist_node1[node], dist_node2[node])
                if cur_dist < min_dist:
                    min_dist = cur_dist
                    res = node
        return res
```
