# <center> 127. Word Ladder </center>


## Problem Description
[Click here](https://leetcode.com/problems/word-ladder/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the shortest path from start word to end word. Use BFS.


## Approach
<!-- Describe your approach to solving the problem. -->
- check bases
    - if start and words are the same, return 0
        - *(no need to check for it here because no test cases are added for it)*
    - if end word is not in the word list, there is no path, return 0
- create a hashmap to represent the graph (adjacency list)
    - *key = node*
    - *value = list of neighbor nodes*
- create a queue for BFS
- create a hashset to mark visited nodes
- add the start word to the word list
- create/fill the graph. The adjacent words in the path from the start word to the end word should differ by one character. We will find the intermediate words of all the words and treat them as graph nodes. For example for the word 'POON', the intermediate words are '*OON', 'P*ON', 'PO*N', and 'POO*'.
    - for each word
        - for each character index in the word
            - find the intermediate word
            intermediate word = prefix + * + postfix
            - add an edge from word to intermediate word i.e add word to the neighbors' list of the intermediate word
- add (start word, distance) to the queue
- mark the start word as visited
- Do BFS
    - loop until q is empty
        - pop q to get a node/word and distance to reach that node
        - if the node is the same as the end word, we get the shortest path, return the distance
        - else find the intermediate words for the node and add unvisited nodes to the queue with updated distance and mark them visited
- if BFS completes without returning anything, there is no path, return 0

## Complexity
- Time complexity: O(wordList check for endWord + graph(adjacency list) creation + BFS) → O(total words + E + V + E) → O(n + V + E) → O(V + E)
    - *V = total vertices or nodes*
    - *E = total edges*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(graph (adjacency list) + BFS queue + seen hashset) → O(V + E + deepest level nodes + total nodes) → O(V + E + V + V) →  O(V + E)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def ladderLength(self, beginWord: str, endWord: str, wordList: List[str]) -> int:
        if endWord not in wordList:
            return 0
        graph = defaultdict(list)
        q, seen = deque(), set()
        wordList.append(beginWord)
        for word in wordList:
            for i in range(len(word)):
                pattern = word[:i] + '*' + word[i + 1:]
                graph[pattern].append(word)
        q.append((beginWord, 1))
        seen.add(beginWord)
        while q:
            word, dist = q.popleft()
            if word == endWord:
                return dist
            for i in range(len(word)):
                node = word[:i] + "*" + word[i + 1:]
                for neighbor in graph[node]:
                    if neighbor not in seen:
                        seen.add(neighbor)
                        q.append((neighbor, dist + 1))
        return 0  
```
