# <center> 2101. Detonate the Maximum Bombs </center>


## Problem Description
[Click here](https://leetcode.com/problems/detonate-the-maximum-bombs/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the maximum bombs that can be detonated using a single bomb.

A bomb x can detonate another bomb y if the distance between x and y is <= blast range of bomb x.

Consider it a graph with bombs as nodes and their ranges to detonate other bombs as directed edges between the nodes. 

The number of bombs a given bomb can detonate is the number of connected nodes i.e length of the longest path. Run a DFS from every node (bomb) to find the longest path (all/max connected nodes).


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the max bombs
- create a hashmap to represent the adjacency list (graph)
    - *key = nodes*
    - *values = list of neighbor nodes*
- find and add edges to the graph. For each bomb, find the bombs that it can detonate directly. Here, the bombs are represented by indices and the value at each index in the bombs list represents the coordinates and the bomb blast range i.e (x, y, range)
    - for each index i, x1, y1 and range1 in the bomb list
        - for each index j, x2, y2, and range2
            - if i and j are the same, this is the same bomb, skip to avoid adding a self-edge
            - find the distance between bombs i and j 
            - if the distance <= bomb i range, bomb i can detonate bomb j, add an edge
            - if distance <= bomb j range, bomb j can detonate bomb i, add an edge
- define a helper function for DFS
    - dfs(source node, visited nodes set)
        - if the node is already visited, return 0
        - mark the node as visited
        - visit neighbors of the node and do DFS to find other connected nodes
        - the visited nodes will represent the total connected nodes to the source node, return its length
- for each node/bomb
    - do DFS to find the total connected nodes and update res
- return res


## Complexity
- Time complexity: O(graph creation + graph traversal) → O(adjacency list creation + adjacency list DFS for every node) → O(E + (V + E) * V) → O(V * (V + E)) → O(V * V + V * E) → O(n * n + n * n²) → O(n * n²)
→ O(n$^3$) 
    - *V = total vertices or nodes = total bombs = n*
    - *E = total edges = n²*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(graph + dfs stack + seen hashset) → O(adjacency list + V + total nodes) → O(V + E + V + V) → O(V + E) → O(n + n²) → O(n + n²) → O(n²)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maximumDetonation(self, bombs: List[List[int]]) -> int:
        res = 0
        graph = defaultdict(list)
        for i, (xi, yi, ri) in enumerate(bombs):
            for j, (xj, yj, rj) in enumerate(bombs):
                if i == j: continue
                dist = sqrt((xi - xj)**2 + (yi - yj)**2)
                if dist <= ri:
                    graph[i].append(j)
                if dist <= rj:
                    graph[j].append(i)

        def dfs(node: int, seen: Set[int]) -> int:
            if node in seen:
                return 0
            seen.add(node)
            for neighbor in graph[node]:
                dfs(neighbor, seen)
            return len(seen)

        for i in range(len(bombs)):
            res = max(res, dfs(i, set()))
        return res
```
