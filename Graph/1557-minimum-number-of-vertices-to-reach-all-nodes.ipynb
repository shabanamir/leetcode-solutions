# <center> 1557. Minimum Number of Vertices to Reach All Nodes </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The node with no incoming edges can reach all other nodes.

Find the nodes that have no incoming edges.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a list of size n (total nodes) to find the incoming edges for each node
    - *index = node*
    - *value = number of edges*
- traverse the edges list <br>
for each edge (source node, destination node)
    - increment the number of incoming edges for the destination node
- traverse the incoming edges list and add nodes to the result list for which there is no incoming edge
- return the result list


## Complexity
- Time complexity: O(edges traversal to get in_degree edges + in_degree traversal to get result) → O(E + V) → O(E) → O(n)
    - *E = total edges*
    - *V = total vertices or nodes*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(indegree list + result list) → O(V + V) → O(V) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def findSmallestSetOfVertices(self, n: int, edges: List[List[int]]) -> List[int]:
        in_degree = [0] * n
        for src, dst in edges:
            in_degree[dst] += 1
        return [node for node, degree in enumerate(in_degree) if not degree]
```
