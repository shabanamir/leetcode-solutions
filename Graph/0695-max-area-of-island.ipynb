## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given m x n binary grid. 1 represents land cell and 0 represents water cell. An island (land block) is formed by connecting land cells horizontally or vertically. The area of an island is the number of cells. We need to find the max cell island.

The grid is an adjacency matrix (graph) and we need to find the subgraph (connected component) with max vertices.

Traverse the graph using DFS and search for subgraphs (islands). When you get a subgraph (an unvisited land cell represents the start vertex of a subgraph), find its total vertices (other connected land cells) and mark them visited.

For marking a vertex/cell as visited, use a hash set or replace its value with 2 (you can use any number or character except 1).


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = total rows and n = total column
- set res = 0 to track the max area
- define a helper function for DFS <br>
dfs(row, column)
    - if row or column index is out of bound or current cell value is not 1 i.e either it is a visited land cell (value 2) or a water cell (value 0), return 0
    - else cell is an unvisited land cell
        - mark it as a visited
        - visit its 4 neighbors (left, right, top, bottom) using DFS to find other connected land cells. For each connected cell, add 1 to the area
- for each cell
    - if it is an unvisited land cell i.e value is 1, we found an island (subgraph). Find the area and update res by taking max of current island area and previous max area
- return res


## Complexity
- Time complexity: O(graph traversal) → O(adjacency matrix DFS) → O(V * V) → O(rows * columns) → O(m * n)
    - *V = total vertices or nodes or rows or sublists or keys*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(DFS recursion stack) → O(V) → O(rows) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxAreaOfIsland(self, grid: List[List[int]]) -> int:
        m, n = len(grid), len(grid[0])
        res = 0

        def dfs(r: int, c: int) -> int:
            if r < 0 or r == m or c < 0 or c == n or grid[r][c] != 1:
                return 0
            grid[r][c] = 2
            return 1 + dfs(r + 1, c) + dfs(r - 1, c) + dfs(r, c + 1) + dfs(r, c - 1)

        for row in range(m):
            for col in range(n):
                res = max(res, dfs(row, col))
        return res
```
