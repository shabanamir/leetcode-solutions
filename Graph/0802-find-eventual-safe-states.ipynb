# <center> 802. Find Eventual Safe States </center>


## Problem Description
[Click here](https://leetcode.com/problems/find-eventual-safe-states/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
A node is safe if it is not part of a cycle.

Do DFS and check for cycles. Add all cycle-free nodes to the result.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a hashmap to track the status of a node
    - *key = node*
    - *value = status (0 = unvisited, -1 = visiting i.e on the current path, 1 = visited)*
- define a helper function for DFS <br>
dfs(cuurent node/course)
    - if status is -1, we encounter the current node again in the path, there is a cycle
        - return false
    - if status is 1, current node is visited and is cycle-free
        - return True
    - else it is an unvisited node
        - add it to the current path by updating the status to -1
        - visit all its neighbors
            - if a neighbor is part of a cycle
                - return false
        - mark the current node as visited and cycle-free by updating the status to 1
        - return true
- for each node
    - do DFS to check if the node is safe, if it is safe
        - add it to the result list
- return the result list


## Complexity
- Time complexity: O(graph traversal) → O(adjacency list DFS) → O(V + E)
    - *V = total vertices or nodes or rows or sublists or keys*
    - *E = total edges*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(DFS stack + result list) → O(V + V) → O(V)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def eventualSafeNodes(self, graph: List[List[int]]) -> List[int]:
        state = defaultdict(int)

        def dfs(node: int) -> bool:
            if state[node] == -1:
                return False
            if state[node] == 1:
                return True
            state[node] = -1
            for neighbor in graph[node]:
                if not dfs(neighbor):
                    return False
            state[node] = 1
            return True

        return [node for node in range(len(graph)) if dfs(node)]
```
