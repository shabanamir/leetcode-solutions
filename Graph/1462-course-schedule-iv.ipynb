## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
For a query (a, b), we need to find if course a is a prerequisite of b.

Find all prerequisites (direct and indirect) for each course and store them in a hashmap or an array. Use the hashmap/array to answer the queries. 

Create a graph with courses as vertices/nodes and prerequisites as edges. Do topological sorting using DFS to find the prerequisites.


## Approach
<!-- Describe your approach to solving the problem. -->
- create a list to store query result
- create a hashmap for prerequisites
    - *key = nodes (courses)*
    - *value = set of all neighbor nodes (prerequisites)*
- create a hashmap for graph
    - *key = nodes (courses)*
    - *value = list of direct neighbor nodes (prerequisites)*
- create a graph
    - for each edge (course, prerequisite) in the prerequisites list
        - add the prerequisite as a neighbor to the course node
- define a helper function for DFS <br>
dfs(cuurent node/course)
    - if the node is not in the prerequisites hashmap, we need to find all its prerequisites
        - for each direct neighbor of the node
            - do dfs on the direct neighbor to find the indirect neighbors and add them to the neighbors set of the current node i.e update the current node neighbors by taking the union of direct & indirect neighbor sets
    - else we have its prerequisites set, return it
- for each course
    - do DFS to find all direct and indirect prerequisites
- for each query (a, b)
    - if a is in the prerequisite set of node b i.e a is a prerequisite of b, add True to the result list, else false
- return the result list


## Complexity
- Time complexity: O(graph creation + graph traversal + queries traversal) → O(adjacency list creation + topological sort for each node + Q) → O(E + adjacency list DFS * total nodes + Q) → O(E + (V + E) * V + Q) → O(V * (V + E) + Q)
    - *V = total vertices or nodes or rows or sublists or keys = courses*
    - *E = total edges = prerequisites*
    - *Q = total queries*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(graph + DFS recursion stack + prerequisite hashmap + result list) → O(adjacency list + V + V + Q) → O(V + E + V + V + Q) → O(V + E + Q)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def checkIfPrerequisite(self, numCourses: int, prerequisites: List[List[int]], queries: List[List[int]]) -> List[bool]:
        res = []
        pre_req = defaultdict(set) 
        graph = defaultdict(list)
        for a, b in prerequisites:
            graph[b].append(a)
        
        def dfs(node: int):
            if node not in pre_req:
                for neighbor in graph[node]:
                    pre_req[node] |= dfs(neighbor)
            pre_req[node].add(node)
            return pre_req[node]

        for course in range(numCourses):
            dfs(course)
        for u, v in queries:
            res.append(u in pre_req[v])
        return res
```
