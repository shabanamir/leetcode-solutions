# <center> 1905. Count Sub Islands </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given two m x n binary grids and need to return the number of sub-islands in grid2.

The grid is an adjacency matrix (graph). We need to find the subgraphs (islands) and check if it is valid i.e a sub-island.

Do parallel DFS. Traverse the second graph (grid2) and search for subgraphs (islands). When you get a subgraph (an unvisited land cell represents the start vertex of a subgraph), check if the corresponding cell in the first graph (grid1) is also a land cell. If it is, we get a sub-island.

For marking a vertex/cell as visited, use a hash set or replace its value with 2 (you can use any number or character except 1).


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = total rows and n = total columns
- define a helper function for DFS <br>
dfs(row, column)
    - if the indices are out of bounds or current cell value in grid2 is not 1 i.e either it is a visited land cell (value 2) or a water cell (value 0), we get a sub-island
        - return 1
    - else cell is an unvisited land cell (start cell of an island (subgraph)) 
        - mark it as a visited
        - visit its 4 neighbors (left, right, top, bottom) using DFS to find other cells on the island and check the corresponding cell in grid1 to know if the island is a sub-island. Return the result (1 if it is a sub-island, 0 if it is not)
- set res = 0 to count the sub-islands
- for each cell in grid2
    - if it is an unvisited land cell i.e value is 1, we found an island (subgraph)
        - do dfs to find other cells on the island and check if it is a sub-island and add the result to res
- return res


## Complexity
- Time complexity: O(graph traversal) → O(adjacency matrix DFS) → O(V * V) → O(rows * columns) → O(m * n)
    - *V = total vertices or nodes or rows or sublists or keys*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(DFS recursion stack) → O(V) → O(rows) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def countSubIslands(self, grid1: List[List[int]], grid2: List[List[int]]) -> int:
        m, n = len(grid2), len(grid2[0])

        def dfs(r: int, c: int) -> int:
            if r < 0 or c < 0 or r == m or c == n or grid2[r][c] != 1:
                return 1
            grid2[r][c] = 2
            return dfs(r + 1, c) & dfs(r - 1, c) & dfs(r, c - 1) & dfs(r, c + 1) & grid1[r][c]

        res = 0
        for r in range(m):
            for c in range(n):
                if grid2[r][c] == 1:
                    res += dfs(r, c)
        return res
```
