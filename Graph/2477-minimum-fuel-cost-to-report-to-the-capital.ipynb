## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Total fuel = sum of cars required at each child node to reach the parent node

Total cars required at a node = ceil (passengers arriving at the node / total seats in the car)

ceil is used to round up the number <br>
For example:  <br>
passenger = 1, seats = 2 <br>
cars required = passengers/seats = 1/2 = 0.5 = ceil(0.5) = 1 <br>

Do DFS to traverse all nodes, find the child nodes and the cars required.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the total fuel
- create a hashmap for the adjacency list representation of the graph
    - *key = node*
    - *value = list of neighbor nodes*
- create the graph (adjacency list)
    - for each edge (source node, destination node)
        - add a bidirectional edge from the source to the destination i.e add destination node to the neighbors list of the source node and vice versa
- define a helper function for DFS <br>
dfs(current node, parent node)
  - mark res as nonlocal
      - *(this is to access outer function variables in nested functions)*
  - set total = 1 to track the total passengers arriving at the current node
      - *(1 because initially there is one passenger at each node)*
  - for each neighbor (child node) of the current node
    - if the neighbor is not the parent i.e it is a child node, *(this is to make sure we are not moving backward)*
      - do DFS to find the passengers arriving at the child node
      - add the child node passengers to the total passengers
      - find the total cars required for the child node passengers and update res
  - return total
- start DFS from the root node
- return res


## Complexity
- Time complexity:  O(graph creation + graph traversal) → O(adjacency list creation + adjacency list DFS) → O(E + V + E) → O(V + E) → O(n + n) → O(n + n)
    - *V = total vertices or nodes or rows or sublists or keys = cities*
    - *E = total edges = roads*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(graph + DFS recursion stack) → O(adjacency list + V) → O(V + E + V) → O(V + E) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:
    
    def minimumFuelCost(self, roads: List[List[int]], seats: int) -> int:
        res = 0
        graph = defaultdict(list)
        for u, v in roads:
            graph[u].append(v)
            graph[v].append(u)
        
        def dfs(node: int, parent: int) -> None:
            nonlocal res
            total = 1
            for neighbor in graph[node]:
                if neighbor != parent:
                    cur = dfs(neighbor, node)
                    total += cur
                    res += ceil(cur / seats)
            return total

        dfs(0, None)
        return res
```
