# <center> 130. Surrounded Regions </center>


## Problem Description
[Click here](https://leetcode.com/problems/surrounded-regions/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to convert all surrounded O's to X's.

An 'O' is surrounded if it is not on the border (1st row, 1st columns, last row, last column) or not connected to a border 'O'. So, instead of finding all surrounded O's, find unsurrounded O's and change all O's except the unsurrounded O's.

The grid is an adjacency matrix (graph), using DFS traverse and find the unsurrounded O's ('O' cells on the border or 'O' cells that are connected to border 'O' cells)


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = total rows, n = total columns
- define a helper function for DFS <br>
dfs(row, column)
    - if the indices are out of bounds or current cell value is not 'O' i.e either it is a visited unsurrounded 'O' cell or an 'X' cell
        - return
    - else cell is an unvisited unsurrounded 'O' cell 
        - mark it as an unsurrounded cell by temporarily replacing its value with '.' 
            - *(we can use any character except O or X)*
        - visit its 4 neighbors (left, right, top, bottom) using DFS to find other unsurrounded cells 
- for each cell on the grid
    - if it is an 'O' cell and lies on the border
        - do DFS to find other unsurrounded cells and mark them
- for each cell on the grid
    - if it is an unsurrounded 'O' cell i.e its value is '.'
        - change it back to 'O'
    - if it is a surrounded 'O' i.e value is 'O':
        - change it to 'X'


## Complexity
- Time complexity: O(graph traversal for finding border O's + travesal to flip O's) → O(adjacency matrix DFS + rows * columns) → O(V * V + m * n) → O(rows * columns + m * n) → O(m * n + m * n) → O(m * n)
    - *V = total vertices or nodes or rows or sublists or keys*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(DFS recursion stack) → O(V) → O(rows) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def solve(self, board: List[List[str]]) -> None:
        """
        Do not return anything, modify board in-place instead.
        """
        m, n = len(board), len(board[0])

        def dfs(i: int, j: int) -> None:
            if i < 0 or i == m or j < 0 or j == n or board[i][j] != 'O':
                return
            board[i][j] = '.'
            dfs(i + 1, j)
            dfs(i - 1, j)
            dfs(i, j + 1)
            dfs(i, j - 1)

        for r in range(m):
            for c in range(n):
                if board[r][c] == 'O' and r * c == 0 or r == m - 1 or c == n - 1:
                    dfs(r, c)
        for r in range(m):
            for c in range(n):
                if board[r][c] == '.':
                    board[r][c] = 'O' 
                elif board[r][c] == 'O':
                    board[r][c] ='X'
```
