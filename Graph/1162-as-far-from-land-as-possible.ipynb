## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find a water cell with max distance to the nearest land cell.
- Approach 1 <br>
Do BFS for each water cell to search for the nearest land cells and find the max distance land cell. <br>
Time complexity = O(no. of water cells * grid BFS) → O(m * V²) → O(m * n²) 
- Approach 2 <br>
The distance from a water cell to max distance nearest land cell is same as the distance from a land cell to max distance nearest water cell. Do multisource BFS i.e start BFS from all land cells at the same time, search for the nearest water cells, and track the max distance from the land cells. We'll go level by level. All land cells will be at level 0 and the last level where we find a 0 will be the answer i.e total levels = max distance <br>
Time complexity = O(grid BFS) → O(V²) → O(n²)


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = -1 to track the max distance
- set n = total rows
- create a list and add all 4 directions (right, left, bottom, top)
- create a queue for BFS
- traverse the grid
    - if you get a land cell (cell value 1) 
        - add its coordinates (row, column) to the queue
- if q is empty (all cells in the grid are water cells) or q size is equal to the total rid cells (all are land cells)
    - return -1
- loop until q is empty
    - for each level
        - pop queue to get coordinates of current level cell
        - explore all 4 directions to find unvisited water cells <br>
        for each direction
            - get the neighbor cell coordinates/indices
            - if the indices are inbound and the neighbor is an unvisited water cell
                - add it to the queue
                - mark the neighbor as visited by updating the value to -1
                    - *(we can use any value or a hashset for marking visited cells)*
    - after each level, increment res
- return res


## Complexity
- Time complexity: O(grid traversal for land cells + grid (adjacency matrix) BFS) → O(rows * columns + V * V) → O(n * n + n * n) →  O(n * n) → O(n²)
    - *V = total vertices or nodes or rows or sublists*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(BFS queue) → O(width) → O(V) → O(rows) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxDistance(self, grid: List[List[int]]) -> int:
        res = -1
        n = len(grid)
        directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        q = deque()
        for r in range(n):
            for c in range(n):
                if grid[r][c] == 1:
                    q.append((r, c))
        if not q or len(q) == n * n:
            return res
        while q:
            for _ in range(len(q)):
                row, col = q.popleft()
                for x, y in directions:
                    r, c = x + row, y + col
                    if 0 <= r < n and 0 <= c < n and grid[r][c] == 0:
                        q.append((r, c))
                        grid[r][c] = -1
            res += 1
        return res
```
