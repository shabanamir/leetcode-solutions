## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Total lock combinations = n$^k$ = slots$^{wheels}$ = 10$^4$ = 10000

There are 10,000 combinations (nodes) and we need to find the minimum turns required to find the target combination. It is the same as finding the shortest path in a graph, use BFS. Start BFS from the root node (0000) to check the nodes level by level.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if the target is the start node (combination), we unlocked the lock without turning the wheel, return 0
    - convert the deadends to a set to search in O(1) time
    - if the start node is in the deadends, we can't unlock it, return -1
- set seen = a set to store the visited nodes and add the start node
- create a queue for BFS and add the start node
- set res = 0 to track the total turns
- traverse the graph level by level i.e loop until the queue is empty
    - for current level
        - pop queue to get current node
        - if node is the target node, return res
        - create the child nodes <br>
        for each of the 4 wheels
            - get current wheel digit i.e char at index i in the current node
            - turn it up and get the child node i.e increment the digit at index i to get a combination
            - if the child node is not in visited and not in deadends, add it to the queue and set
            - turn it down and get the child node i.e decrement the digit at index i to get a combination
            - if the child node is not in visited and not in deadends, add it to the queue and set
    - increment res after checking a level
- if the BFS completes without returning anything, there's no shortest path i.e we can't unlock it, return -1


## Complexity
- Time complexity: O(graph BFS) → O(V) → O(total nodes/combinations) → O(10$^4$)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(BFS queue + hashset) → O(x + total nodes) → O(x + 10$^4$)
    - *at each level, at most 8 (2$^4$) child nodes are created for a node*
    - *the nodes at last level = 8$^{last level}$ = some constant less than 10$^4$ = x*
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def openLock(self, deadends: List[str], target: str) -> int:
        if target == '0000':
            return 0
        deadends = set(deadends)
        if '0000' in deadends:
            return -1
        seen = set('0000')
        q = deque(['0000'])
        res = 0
        while q:
            for _ in range(len(q)):
                cur = q.popleft()
                if cur == target:
                    return res
                for i in range(4):
                    digit = int(cur[i])
                    up = cur[:i] + str((digit + 1) % 10) + cur[i+1:]
                    if up not in seen and up not in deadends:
                        q.append(up)
                        seen.add(up)
                    down = cur[:i] + str((digit - 1) % 10) + cur[i+1:]
                    if down not in seen and down not in deadends:
                        q.append(down)
                        seen.add(down)
            res += 1
        return -1
```
