## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given n x n grid (adjacency matrix) and we need to find the minimum moves to reach the last position i.e n². It is same as finding the shortest path in a graph, use BFS.


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = total rows
- set target = last position value
- there are n² cells and n² position but the positions are arranged in reversed rows and alternating columns fashion i.e 1 (start position) is in the last row and n² (last position) is in the first row. So, reverse the board to get the rows' order correct
- create a hash set to store the visited positions
- create a queue for BFS. Add (root node / start position, moves required to reach root node) to queue
- do BFS i.e loop until the queue is empty
    - pop queue to get current position and moves required to reach it
    - roll the dice, (a dice has 6 values, loop 6 times to get all possible next positions) <br>
    for each dice value
        - get the next position
        - get the coordinates (row, column) of the next position
            - *row = next position - 1 / total rows*
            - *column = next position - 1 % total rows*
        - if the row is odd, the column index = total columns - 1 - column
            - *(in columns, the positions alternate i.e even rows have positions in ascending order and odd in descending order)*
        - if the value at the next position is not -1 (a ladder or a snake)
            - update the next position
        - if the next position is the target position
            - increment moves and return it
        - if the next position is not visited
            - add the position to the set
            - add the position and moves to the queue
- if the BFS completes without returning anything, we can't reach the last position, return -1


## Complexity
- Time complexity: O(reverse + graph traversal) → O(board size + adjacency matrix BFS) → O(rows * columns + V * V) → O(n * n + rows * columns) → O(n * n + n * n) → O(n * n) → O(n²)
    - *V = total vertices or nodes or rows or sublists*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashset + BFS queue) → O(total positions + V) → O(rows * columns + rows) → O(n * n + n) → O(n * n) → O(n²)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def snakesAndLadders(self, board: List[List[int]]) -> int:
        n = len(board)
        target = n * n
        board.reverse()
        seen = set()
        q = deque([(1, 0)])
        while q:
            square, moves = q.popleft()
            for i in range(1, 7):
                next_sq = square + i
                r, c = (next_sq - 1) // n, (next_sq - 1) % n
                if r % 2:
                    c = n - 1 - c
                if board[r][c] != -1:
                    next_sq = board[r][c]
                if next_sq == target:
                    return moves + 1
                if next_sq not in seen:
                    seen.add(next_sq)
                    q.append((next_sq, moves + 1))
        return -1
```
