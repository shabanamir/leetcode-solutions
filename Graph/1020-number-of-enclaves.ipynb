# <center> 1020. Number of Enclaves </center>


## Problem Description
[Click here](https://leetcode.com/problems/number-of-enclaves/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the number of land cells in grid for which we cannot walk off the boundary of the grid.

The land cells for which we can walk off the boundary are part of the border islands and vice versa. 

Instead of finding land cells that are not part of the border islands, find the land cells that are part of the border islands using DFS. Return the difference between total land cells and border land cells.


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = total rows, n = total columns
- create a list and add the neighbor cells directions (top, bottom, left, right)
- define a helper function for DFS <br>
dfs(row, column)
    - if the indices are out of bounds or the cell is not an unvisited land cell (value=1) i.e either it is a water cell (value=0) or a visited land cell (value=2)
        - return 0   
    - else the cell is an unvisited land cell
        - mark it visited
            - *here, we are updating the cell value to mark it as visited, we can also use a hashset for marking*
            - *0 = water cell, 1 = unvisited land cell, 2 = visited land cell*
        - set res = 1 to count the cells in the island
        - visit all 4 neighbors to find other cells on the island and update res
        - return res
- set total = 0 to track the total land cells
- set border = 0  to track the total border land cells
- traverse all cells <br>
for each cell on the grid
    - if the cell is an unvisited land cell (value = 1)
        - increment total
        - if the cell is on the border (first row, first column, last row, or last column)
            - do DFS to find other border cells and update the border cells count
    - if the cell is a visited land cell (value=2)
        - increment the total
- return the difference between total and border cells count


## Complexity
- Time complexity: O(grid/graph traversal) → O(adjacency matrix DFS) → O(V * V) → O(rows * columns) → O(m * n) → O(m * n)
    - *V = total vertices or nodes or rows or sublists or keys*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(DFS recursion stack) → O(V) → O(rows) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def numEnclaves(self, grid: List[List[int]]) -> int:
        m, n = len(grid), len(grid[0])
        directions = [[-1, 0], [1, 0], [0, -1], [0, 1]]

        def dfs(r: int, c: int) -> int:
            if r < 0 or r == m or c < 0 or c == n or grid[r][c] != 1:
                return 0
            grid[r][c] = 2
            res = 1
            for x, y in directions:
                res += dfs(r + x, c + y)
            return res

        total , border = 0, 0
        for r in range(m):
            for c in range(n):
                if grid[r][c] == 1:
                    total += 1
                    if r == 0 or r == m - 1 or c == 0 or c == n - 1:
                        border += dfs(r, c) 
                elif grid[r][c] == 2:
                    total += 1
        return total - border
```
