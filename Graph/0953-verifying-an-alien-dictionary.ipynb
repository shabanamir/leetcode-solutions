## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given a list of words and we need to check if the words are in sorted order according to the given key.

The words are sorted if:
- next word is not a prefix of the previous 
- order/index of the first different letter in the next word is greater than the previous

For example <br>
words = ['hello', 'hi'] <br>
key/order = 'hlabcdefgijkmnopqrstuvwxyz' <br>
The answer is true because: 
- hi is not a prefix for hello 
- the first differing letters of the two words are e and i, and the index of i in the key is 9 which is greater than e (6)

Use a hashmap to map the letters to their order/index. Traverse the words and check the above conditions. Return false if any of it doesn't hold.

## Approach
<!-- Describe your approach to solving the problem. -->
- map the letters of the key to the indices 
- loop in range 0 to n - 1 to traverse and compare words
    - get current and next words
    - loop in range 0 to the length of the current word to traverse and compare the letters
        - if current iteration is equal to the length of the next word, the next word is a prefix of the current word
            - return false
        - if current letter of current word is not the same as the next word, we get the first different letter
            - if order of next word letter is less, the words are not sorted
                - return false
            - else the words are sorted, we don't need to check the rest of the letters, break the loop
- if the loop completes without returning false, all words are sorted, return true


## Complexity
- Time complexity: O(hashmap creation + words traversal) → O(order key size + total words * max letter word) → O(26 + n * 20) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(hashmap) → O(26) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isAlienSorted(self, words: List[str], order: str) -> bool:
        order_index = {c:i for i, c in enumerate(order)}
        for i in range(len(words) - 1):
            w1, w2 = words[i], words[i+1]
            for j in range(len(w1)):
                if j == len(w2):
                    return False
                if w1[j] != w2[j]:
                    if order_index[w2[j]] < order_index[w1[j]]:
                        return False
                    break
        return True
```
