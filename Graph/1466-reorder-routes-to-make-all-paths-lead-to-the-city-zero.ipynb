## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given n cities (vertices) and n one-way roads (edges). We need to reorient roads that don't lead to capital (city 0) .

Visit all cities from city 0 and count the roads that need to be oriented. 

To do this, create an undirected graph and do DFS. Start from the root vertex (city 0), if we get a forward edge (one-way road from city 0 to another city), we need to reverse (reorient) it. Count the total forward edges.


## Approach
<!-- Describe your approach to solving the problem. -->
- use a hashmap to create an undirected graph (adjacency list)
    - *key = vertex or node i.e city*
    - *value = list of neighbors*
    - *we can also use a nested list*
- create a stack for DFS and add the root node 
- create a set to store visited nodes
- set res = 0 to count the forward edges (roads that need to be reoriented)
- for each edge (start city, end city) in the connection list
    - add bidirectional edge to the graph i.e add the end city as a neighbor to the neighbors list of the start city and vice versa
    - also add an indicator with each neighbor to indicate that the edge is forward or backward
    - in the input list the edges are forward (from start city to end city), so use 1 for forward and 0 for backward
- do DFS i.e loop until the stack is empty
    - pop stack to get current node
    - add the current node to the visited set
    - for each neighbor and edge indicator in the neighbor's list of the current node
        - if we haven't visited the node
            - if the edge is a forward edge i.e value is 1
                - increment res
            - add the node to the stack
- return res


## Complexity
- Time complexity: O(graph creation + graph traversal) → O(adjacency list creation + adjacency list DFS) → O(E + V + E) → O(V + E) → O(n + n) → O(n)
    - *V = total vertices or nodes or rows or sublists or keys = roads*
    - *E = total edges = connections*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(graph + DFS stack) → O(adjacency list + V) → O(V + E + V) → O(V + E) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minReorder(self, n: int, connections: List[List[int]]) -> int:
        graph = defaultdict(list)
        stack = [0]
        seen = set()
        res = 0
        for a, b in connections:
            graph[a].append((b, 1))
            graph[b].append((a, 0))
        while stack:
            node = stack.pop()
            seen.add(node)
            for neighbor, edge_type in graph[node]:
                if neighbor not in seen:
                    stack.append(neighbor)
                    if edge_type == 1: 
                        res += 1
        return res
```
