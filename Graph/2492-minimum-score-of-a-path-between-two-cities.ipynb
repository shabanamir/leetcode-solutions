## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given n cities and the distance between cities. We need to find the min distance road on the path from the first city (1) to the last city (n) i.e we need to find the min weight edge on the path.

score of a path = minimum weight edge on the path

Use DFS to find the path from node 1 to n and keep track of the minimum weight edge on the path.

# Approach
<!-- Describe your approach to solving the problem. -->
- set res = a very large number to track minimum weight edge
- create a hashset to track the visited nodes
- create a hashmap for the adjacency list representation of the graph
    - *key = node, value = list of neighbor nodes*
    - *each neighbor node value will be a pair i.e (neighbor node, edge weight)*
- create the graph (adjacency list)
    - for each edge (source node, destination node, distance or edge weight)
        - add a bidirectional weighted edge i.e add (destination node, edge weight) to the neighbor nodes list of the source node and vice versa
- define a helper function for DFS <br>
dfs(current node)
  - mark res as nonlocal 
    - *this is to access outer function variables in nested functions*
    - *res is a float variable, not an object like the seen hashset, and to update (reassign) a variable defined outside the function, we need to declare it as nonlocal*
  - if the node is in the hashset, it is already visited
    - return
  - add the node to the set to mark it visited
  - traverse the neighbors of the current node <br>
  for each neighbor node, edge weight of the current node
    - update res by taking min of the current edge weight and previous min 
    - do dfs to explore the neighbor node
- start DFS from the root node i.e city 1
- return res


## Complexity
- Time complexity: O(graph creation + graph traversal) → O(adjacency list creation + adjacency list DFS) → O(E + V + E) → O(V + E) → O(n + n) → O(n + n)
    - *V = total vertices or nodes or rows or sublists or keys = cities*
    - *E = total edges = roads*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(graph + DFS recursion stack + hashset) → O(adjacency list + V + V) → O(V + E + V + V) → O(V + E) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minScore(self, n: int, roads: List[List[int]]) -> int:
        res = float('inf')
        seen = set()
        graph = defaultdict(list)
        for u, v, distance in roads:
            graph[u].append((v, distance))
            graph[v].append((u, distance))
        
        def dfs(node: int) -> None:
            nonlocal res
            if node in seen:
                return
            seen.add(node)
            for neighbor, distance in graph[node]:
                res = min(res, distance)
                dfs(neighbor)

        dfs(1)
        return res
```
