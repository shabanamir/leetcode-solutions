# <center> 1553. Minimum Number of Days to Eat N Oranges </center>


## Problem Description
[Click here](https://leetcode.com/problems/minimum-number-of-days-to-eat-n-oranges/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The max days can be n if we eat one orange a day. 

We will start with n oranges and eat oranges every day to reduce n to 0 or 1. For each day, we have two options/paths: eat n/2 oranges a day or n/3 oranges

Do DFS to explore all paths and choose the best. 


## Approach
<!-- Describe your approach to solving the problem. -->
- create a hashmap for memoization i.e to store the min days required to eat x oranges
    - *key = oranges left*
    - *value = min days required to eat the remaining oranges*
- add base cases to the hashmap i.e days required to eat 0 and 1 oranges
- define a helper function for DFS <br>
dfs(oranges left)
    - if x is in the hashmap, we have already found the min days required to eat x oranges, return it
    - recursively find the days required to eat x oranges if we eat x/2 oranges a day or x/3 oranges and update the hashmap with the minimum
    - return the min days from the hashmap
- call dfs to find min days required to eat n oranges 


## Complexity
- Time complexity: O(dfs recursion) → O(calls at each level ^ total levels) → O(2$^{logn}$) → (memoization using hashmap) → O(logn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(dfs recursion stack + hashmap) → O(total levels + n) → O(logn + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minDays(self, n: int) -> int:
        dp = {0: 0, 1: 1}

        def dfs(x: int) -> int:
            if x in dp:
                return dp[x]
            dp[x] = 1 + min(x % 2 + dfs(x // 2), x % 3 + dfs(x // 3))
            return dp[x]

        return dfs(n)
```
