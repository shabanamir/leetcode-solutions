# <center> 1129. Shortest Path with Alternating Colors </center>


## Problem Description
[Click here](https://leetcode.com/problems/shortest-path-with-alternating-colors/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the shortest path from root node to every other with alternating edge color.

Do BFS with tracking of the previous edge color to find the shortest path with alternating colors.

# Approach
<!-- Describe your approach to solving the problem. -->
- initialize a result list of size n (total nodes) with -1 to track the distance of each node from the root node
- create a hashmap for the graph
    - *key = node*
    - *value = list of pairs of neighbor nodes, each pair will represent an edge of color x from the key node to the neighbor node i.e (neighbor node, edge color)*
    - *0 is for red color edge, 1 for blue, and -1 for none *
- create a queue for BFS and add (root node, color of the previous edge i.e edge to reach root node)
- traverse the given red edges <br>
for each edge (source node, destination node)
    - add the destination node and the edge color to the neighbor nodes list of the source node
- do the same for the blue edges
- set step = 0 to track the levels
- traverse all levels i.e loop until q is empty
    - for each level
        - pop q to get the current node and previous edge color
        - if the current node distance is not added to the result list i.e value is -1
            - update the node distance in the result list 
                - *(current node distance from root node = current level)*
        - traverse the neighbor nodes <br>
        for each neighbor node index, neighbor node, current edge color
            - if the node is visited (value = -1) or the edge color is not alternating (same as the previous edge)
                - don't consider this path and continue
            - else the node is not visited and the edge is an alternating color edge
                - take this path by adding the neighbor node and its edge color to the q
                - mark the neighbor node as visited
                    - *(here we mark the visited node by replacing the node value with -1 using the index of the neighbor node in the neighbor nodes list, we can also use a hashset for marking visited nodes but that will require extra memory)*
    - increment level
- return the result list


## Complexity
- Time complexity: O(graph creation + graph traversal) → O(adjacency list creation + adjacency list BFS) → O(E + V + E) → O(V + E) → O(n + E)
    - *V = total vertices or nodes or rows or sublists or keys = n*
    - *E = total edges =  red + blue edges*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(graph + BFS queue + result list) → O(adjacency list + V + V) → O(V + E + V + V) → O(V + E) → O(n + E)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def shortestAlternatingPaths(self, n: int, redEdges: List[List[int]], blueEdges: List[List[int]]) -> List[int]:
        res = [-1] * n
        graph = defaultdict(list)
        q = deque([(0, -1)])
        for u, v in redEdges:
            graph[u].append((v, 0))
        for u, v in blueEdges:
            graph[u].append((v, 1))
        step = 0
        while q:
            for _ in range(len(q)):
                node, prev_color = q.popleft()
                if res[node] == -1:
                    res[node] = step
                for i, (neighbor, cur_color) in enumerate(graph[node]):
                    if neighbor == -1 or cur_color == prev_color:
                        continue
                    q.append((neighbor, cur_color))
                    graph[node][i] = (-1, -1) 
            step += 1
        return res
```
