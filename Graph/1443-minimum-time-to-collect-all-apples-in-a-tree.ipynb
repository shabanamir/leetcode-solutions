# <center> 1443. Minimum Time to Collect All Apples in a Tree </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
It is a graph problem because there are n nodes and an edges list. We need to find a path starting from the root node that includes all apple nodes and has minimum edges.

The time here means no. of edges because passing 1 edge requires 1 second. 

Do DFS to visit every node and check if it has an apple. If a node has an apple, count the number of edges (nodes) to reach that node from the root node. Double the number of edges to include the comeback time.

1 second to pass an edge + 1 second to come back


## Approach
<!-- Describe your approach to solving the problem. -->
- create a hashmap to represent the graph (adjacency list)
    - *key = node*
    - *value = list of neighbor nodes*
    - *(we can also use a 2D array for the adjacency list)*
- traverse the edges list and fill the graph
  - for each edge between nodes a and b
    - add forward edge i.e add node b to the neighbor list of node a
    - add backward edge i.e node a to the neighbor list of node b
- define a helper function to do DFS <br>
dfs(current node, parent node)
    - set time = 0 to track the time required to visit a node with an apple and come back to the current node
    - traverse the neighbor nodes of the current node
      - if the neighbor is a child node
        - *(we only count the child node time because we have already visited the parent node)*
        - recursively check for apples and update time
    - if the current node has an apple and is not the root node (i.e 0)
        - add 2 to time 
            - *(2 = 1 second to pass an edge + 1 second to come back)*
    - return time
- start dfs from the root node and return the result


## Complexity
- Time complexity: O(graph creation + graph DFS) → O(adjacency list creation + adjacency list DFS) → O(E + V + E) → O(V + E) 
  - *V = total vertices or nodes or hashmap keys = n*
  - *E = total edges*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(graph/hashmap + recursion stack) → O(V + V) → O(V)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def minTime(self, n: int, edges: List[List[int]], hasApple: List[bool]) -> int:
        adj = defaultdict(list)
        for a, b in edges:
            adj[a].append(b)
            adj[b].append(a)

        def dfs(node, parent):
            time = 0
            for neighbor in adj[node]:
                if neighbor != parent:
                    time += dfs(neighbor, node)
            if (time or hasApple[node]) and node != 0:
                time += 2
            return time
            
        return dfs(0, None)
```
