## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given m x n binary matrix. 1 represents land cell and 0 represents water cell. Each cell is a square and the land cells form the island. We need to find the island perimeter.

Perimeter of a cell = total sides = 4 <br>
but here we will only count those sides that are adjacent to water <br>
Perimeter of island = total sides adjacent to water

Traverse each cell, if you get a land cell:
- Approach 1 <br>
    Check all 4 sides/neighbors (top, bottom, left, right). If the neighbor is a water cell, increment the perimeter count.
- Approach 2 (fast) <br>
    Increment the perimeter count by 4 and check only two sides/neighbors i.e. either top & left or bottom & right. If any of the neighbors is a land cell, decrement the count by 2 .
  

## Approach
<!-- Describe your approach to solving the problem. -->
- set m = total rows
- set n = total columns
- set res = 0 to track the perimeter
- for each cell
    - if it is a land cell
        - increment res by 4
        - if row index is inbound and the bottom cell is a land cell, decrement res by 2
        - if column index is inbound and the right cell is a land cell, decrement res by 2
- return res


## Complexity
- Time complexity: O(grid traversal) → O(rows * columns) → O(m * n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def islandPerimeter(self, grid: List[List[int]]) -> int:

        m, n = len(grid), len(grid[0])
        res = 0
        for r in range(m):
            for c in range(n):
                if grid[r][c] == 1:
                    res += 4
                    if r + 1 < m and grid[r + 1][c] == 1:
                        res -= 2
                    if c + 1 < n and grid[r][c + 1] == 1:
                        res -= 2
        return res
```
