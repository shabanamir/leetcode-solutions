## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The graph is bipartite if we can divide the nodes into two sets (A, B) such that no nodes in the same sets are neighbors (have an edge connecting them). For example, if node 0 has neighbors 2 and 3 then node 0 should be in a separate set and nodes 2, 3 in a separate one.

Do BFS to check if the neighbors can be placed in separate sets.

Instead of using two sets to put the nodes, mark/color the nodes. <br>
color white (0) → not placed in a set <br>
color blue (1) → set A (nodes set) <br>
color red (-1) → set B (neighbors set)


# Approach
<!-- Describe your approach to solving the problem. -->
- set n = total number of nodes
- initialize an array of size n with zeroes to color the nodes
    - *index = node*
    - *value = color*
- define a helper function for BFS <br>
bfs(node)
    - if the node has been colored (value = 1 or -1), it has been visited and placed in a set, return True
    - create a queue and add the node
    - color the node
    - while q is not empty
        - pop a node from the q to get current level node
        - visit the neighbors of the currrent node
            - if the neighbor is colored and the color is same as the current node, the neighbor node will be placed in the same set, return false
            - if the neighbor is not colored
                - add it to the queue
                - color the neighbor with the opposite color of the current node
- for each node in the graph
    - if the node is not colored (not placed in a set), do BFS to check if we can place the node and its neighbors in separate sets. If we can not, return False
- if the loop completes without returning false, the graph is bipartite, return true


## Complexity
- Time complexity: O(graph BFS) → O(adjacency list BFS) → O(V + E) → O(n + n) → O(n)
    - *V = total vertices or nodes*
    - *E = total edges*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(colors array + BFS queue) → O(total nodes + deepest level nodes) → O(V + V) → O(V) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isBipartite(self, graph: List[List[int]]) -> bool:
        n = len(graph)
        color = [0] * n

        def bfs(node: int) -> bool:
            if color[node] != 0:
                return True
            q = deque([node])
            color[node] = 1
            while q:
                cur = q.popleft()
                for neighbor in graph[cur]:
                    if color[neighbor] == color[cur]:
                        return False
                    if color[neighbor] == 0:
                        q.append(neighbor)
                        color[neighbor] = -color[cur]
            return True
        
        for node in range(n):
            if color[node] == 0 and not bfs(node):
                return False
        return True
```
