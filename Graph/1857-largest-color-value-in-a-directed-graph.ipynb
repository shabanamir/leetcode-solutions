# <center> 1857. Largest Color Value in a Directed Graph </center>


## Problem Description
[Click here](https://leetcode.com/problems/largest-color-value-in-a-directed-graph/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Do topological sorting to detect cycles and compute the color value of each node's path. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = 0 to track the max color value 
- set seen = 0 to track the total visited nodes
- set n = total number of nodes
- create a hashmap to represent the graph (adjacency list)
    - *key = node*
    - *value = list of neighbor nodes*
- create a queue for BFS for topological sorting
- create a list of size n to store the in-degree (total number of incoming edges) of each node
    - *index = node*
    - *value = in-degree*
- create a list to store the frequency of each color for each node's path
    - *index = node*
    - *value = hashmap of color values*
        - *hashmap key = color*
        - *value = frequency of the color*
- add edges to the graph
    - for each source node, destination node in the edges
        - add an edge from destination node to source node
        - increment the in-degree of the destination node
- add nodes with 0-indegree to the queue
- do topological sort
    - loop until q is empty
        - pop q to get a node to process
        - get the color of the node and increment the count of that color
        - update res
        - we have visited a new node, increment seen
        - visit the neighbors of the current node, update the colors, and reduce the neighbor in-degree by 1. Add neighbor to q if in-degree becomes 0
- if we have not visited all n nodes, there is a cycle, return -1
- else return res


## Complexity
- Time complexity: O(graph (adjacency list) creation + nodes traversal to add nodes with 0 in-degree to queue + BFS) → O(E + V + V + E) → O(V + E) → O(n + m)
    - *V = total vertices or nodes*
    - *E = total edges*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(graph (adjacency list) + in-degree list + count list + BFS queue) → O(V + E + V + V + V) → O(V + E) → O(n + m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def largestPathValue(self, colors: str, edges: List[List[int]]) -> int:
        res, seen, n = 0, 0, len(colors)
        graph, q = defaultdict(list), deque()
        in_degree = [0] * n
        count = [defaultdict(int) for _ in range(n)]
        for u, v in edges:
            graph[v].append(u)
            in_degree[u] += 1
        for node in range(n):
            if not in_degree[node]:
                q.append(node)
        while q:
            cur = q.popleft()
            count[cur][colors[cur]] += 1
            res = max(res, count[cur][colors[cur]])
            seen += 1
            for v in graph[cur]:
                for c in count[cur]:
                    count[v][c] = max(count[v][c], count[cur][c])
                in_degree[v] -= 1
                if not in_degree[v]:
                    q.append(v)
        if seen < n:
            return -1
        return res
```
