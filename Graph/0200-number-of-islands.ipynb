# <center> 200. Number of Islands </center>


## Problem Description
[Click here](https://leetcode.com/problems/number-of-islands/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We are given m x n binary grid. 1 represents land cell and 0 represents water cell. An island (land block) is formed by connecting land cells horizontally or vertically. We need to return the total number of islands.

The grid is an adjacency matrix (graph), and we need to find the subgraphs (connected components). 

Traverse the graph using DFS and search for subgraphs. If you get an unvisited land cell, explore its neighbors to find other connected land cells and mark them visited. Count the number of times you visit an unvisited land cell because it represents the start vertex of a subgraph (island).

For marking a vertex/cell as visited, use a hash set or replace its value with 2 (you can use any number or character except 0 and 1).


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = total rows
- set n = total column
- set res = 0 to track the total number of subgraphs (islands)
- define a helper function for DFS <br>
dfs(row, column)
    - if row or column index is out of bound or current cell value is not 1 i.e either it is a visited land cell (value 2) or a water cell (value 0)
        - return
    - else cell value is 1 i.e it is an unvisited land cell
        - mark it as a visited land cell by replacing the value with 2
        - visit its 4 neighbors (left, right, top, bottom) to find other connected land cells
- for each cell in the grid
    - if it is an unvisited land cell i.e value is 1, we found a subgraph (island)
        - increment res by 1 
        - do DFS to find and mark the connected land cells
- return res


## Complexity
- Time complexity: O(graph traversal) → O(adjacency matrix DFS) → O(V * V) → O(rows * columns) → O(m * n)
    - *V = total vertices or nodes or rows or sublists*
    - *DFS time complexity = O(V X). V because we traverse each vertex only once. X is the additional complexity that comes from how you traverse the outgoing edges i.e how the graph is implemented.*
        - *X → * V if the graph is implemented using adjacency matrix*
            - *DFS = V * V*
        - *X → + E if the graph is implemented using adjacency list*
            - *DFS = V + E*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(DFS recursion stack) → O(V) → O(rows) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def numIslands(self, grid: List[List[str]]) -> int:
        m, n = len(grid), len(grid[0])
        res = 0

        def dfs(r: int, c: int) -> None:
            if r < 0 or r == m or c < 0 or c == n or grid[r][c] != '1':
                return
            grid[r][c] = '2' 
            dfs(r, c - 1)
            dfs(r, c + 1)
            dfs(r + 1, c)
            dfs(r - 1, c)

        for row in range(m):
            for col in range(n):
                if grid[row][col] == '1':
                    res += 1
                    dfs(row, col)
        return res
```
