# <center> 934. Shortest Bridge </center>


## Problem Description
[Click here](https://leetcode.com/problems/shortest-bridge/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Do DFS to find the first island. Do BFS to find the shortest path from island1 to island2.


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = total rows
- create a queue for BFS
- create a list and add all 4 directions (bottom, top, right, left)
- define a helper function for DFS <br>
dfs(row, column)
    - if the indices are inbound and the current cell is an unvisited land cell (value 1)
        - add it to the queue
        - mark the cell as visited by updating the value to -1
            - *(you can use any value except 0 and 1)*
        - explore its 4 neighbors to find other cells of the island1
- define a helper function for BFS <br>
bfs()
    - set res = 0 to track the steps of the shortest path i.e levels to reach the island2
    - loop until the queue is empty i.e til last level
        - loop to process current-level cells
            - pop queue to get current cell coordinates
            - explore all 4 neighbors of the cell
            for each direction (x, y)
                - get the neighbor cell coordinates
                - if the coordinates are out of bounds or the neighbor cell is already visited
                    - continue to check another neighbor
                - if the neighbor is an unvisited land cell (value 1), we reached island2
                    - return res
                - else add the cell coordinates to the queue 
                - mark the cell as visited
        - after each level, increment res
- traverse the matrix
    - when you get the first land cell (value = 1)
        - do DFS to find other cells of the island1
        - do BFS to find the shortest path to island2 and return the path size i.e number of steps/levels


## Complexity
- Time complexity: O(finding island1 + shortest distance to island2) → O(adjacency matrix DFS + adjacency matrix BFS) → O(V * V + V * V) → O(rows * columns + rows * columns) → O(n * n + n * n) → O(n * n) → O(n²)
    - *V = total vertices or nodes or rows or sublists*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(DFS recursion stack + BFS queue) → O(height/depth + width) → O(V + V) → O(rows + rows) → O(n + n) → O(n) 
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def shortestBridge(self, grid: List[List[int]]) -> int:
        n = len(grid)
        q = deque()
        directions = [[1, 0], [-1, 0], [0, 1], [0, -1]]
        
        def dfs(r: int, c: int) -> None:
            if 0 <= r < n and 0 <= c < n and grid[r][c] == 1:
                q.append((r, c))
                grid[r][c] = -1
                for x, y in directions:
                    dfs(r + x, c + y)

        def bfs() -> int:
            res = 0
            while q:
                for _ in range(len(q)):
                    row, col = q.popleft()
                    for x, y in directions:
                        r = x + row
                        c = y + col
                        if r < 0 or c < 0 or r == n or c == n or grid[r][c] == -1:
                            continue
                        if grid[r][c] == 1:
                            return res
                        q.append((r, c))
                        grid[r][c] = -1
                res += 1

        for r in range(n):
            for c in range(n):
                if grid[r][c] == 1:
                    dfs(r, c)
                    return bfs()
```
