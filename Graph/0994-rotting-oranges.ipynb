## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to return the minimum time required to rot all oranges or -1 if it is impossible to rot all oranges.

The gird is an adjacency matrix (graph) and a rotten orange makes its neighbor (top, bottom, left, right) oranges rotten by level.

Do BFS to traverse the graph and rot neighbor fresh oranges level by level. For each level, increment the minute counter. In the end, if all fresh oranges are rotten, return the total minutes, else return -1.


## Approach
<!-- Describe your approach to solving the problem. -->
- set m = total rows, n = total columns
- create a queue to store same-level nodes for BFS (level-order traversal)
- set fresh = 0 to track fresh oranges
- set res = 0 to track the time (minutes) required to rot all oranges
- for each cell in the grid
    - if it is a fresh orange (value 1)
        - increment fresh
    - if it is a rotten orange (value 2)
        - add its coordinates to the queue
- traverse the graph level by level i.e loop until q is not empty and we have fresh oranges left
    - at each level
        - pop queue to get the rotten orange in the current level
        - visit its all 4 neighbors (bottom, top, left, right)
            - if the indices are inbound and neighbor is a fresh orange (value 1)
                - rot it by changing the value to 2
                - add its coordinates to the queue
                - decrement fresh
    - after each level traversal, a minute is passed, increment res
- return res if no fresh oranges are left else return -1


## Complexity
- Time complexity: O(traversal to get fresh and rotten oranges + graph traversal to rot fresh oranges) → O(rows * columns + adjacency matrix BFS) → O(m * n + V * V) → O(m * n + rows * columns) → O(m * n + m * n) → O(m * n)
    - *V = total vertices or nodes or rows or sublists or keys*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(BFS queue) → O(V) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def orangesRotting(self, grid: List[List[int]]) -> int:
        m, n = len(grid), len(grid[0])
        q = collections.deque()
        fresh, res = 0, 0
        for r in range(m):
            for c in range(n):
                if grid[r][c] == 1:
                    fresh += 1
                elif grid[r][c] == 2:
                    q.append((r, c))
        while fresh > 0 and q:
            for i in range(len(q)):
                r, c = q.popleft()
                for i, j in [(r + 1, c), (r - 1, c), (r, c - 1), (r, c + 1)]:
                    if 0 <= i < m and 0 <= j < n and grid[i][j] == 1:
                        grid[i][j] = 2
                        q.append((i, j))
                        fresh -= 1
            res += 1
        return res if fresh == 0 else -1
```
