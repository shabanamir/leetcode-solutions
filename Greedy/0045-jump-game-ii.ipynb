## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Greedy approach

For each index, find the next max index we can reach in order to get closer to the last index.


## Approach
<!-- Describe your approach to solving the problem. -->
- set cur = -1 to track the current max index reached
- set next = 0 to track the next max reachable index
- set res = 0 to count the number of jumps
- set i = 0 to track the current index
- traverse the array i.e loop until we reach the last index
  - if current index > max index reached
    - jump i.e increment res
    - update cur 
  - find the next max reachable index and update next 
  - increment i 
- return res


## Complexity
- Time complexity: O(loop) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def jump(self, nums: List[int]) -> int:
        cur, next = -1, 0
        res = i = 0
        while next < len(nums) - 1:
            if i > cur:
                res += 1
                cur = next
            next = max(next, nums[i] + i)
            i += 1
        return res
```
