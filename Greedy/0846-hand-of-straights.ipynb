# <center> 846. Hand of Straights </center>


## Problem Description
[Click here](https://leetcode.com/problems/hand-of-straights/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to divide the array into groups of k consecutive numbers.

Find the minimum available number (card) and search for the next numbers in the group of size k. If all next numbers are available, return true, else false

For example, if the minimum number is 2 and k (group size) is 4, then the next numbers in the group are 3, 4, 5. 

Use a hashmap to check the availability of a number in O(1) time.
Use min heap to find the min number in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if total number of cards is not divisible by k (group size), it is not possible to arrange the cards in k groups, return false
- use a hashmap and count the cards
    - *either use a library function or create an empty hashmap and fill it* 
- get the hashmap keys and convert them to a list 
- heapify the list to create a min heap
- loop until the heap is empty i.e there are no cards left to process
    - get the min number from the min heap as start of the current sequence of consecutive number
        - *heap top (index 0) contains the min element*
    - loop to find the next numbers in the group
        - if the next number is not available
            - return false
        - else the next number is available
            - use it i.e decrement its count
            - if the count of the next number becomes 0, we have to remove it from heap
                - if the number is not the current minimum number
                    - return false because removing it will create a hole
                    - *(because this number is the next consecutive number of the current minimum number. If we remove this number, the minimum number will be still in the heap. However, without the next consecutive number, we will not be able to create a group. For example, if the current minimum number is 2 and its next number is 3, but the count of 3 becomes 0. If we remove 3, 2 will remain in the heap. In the next iteration, 2 will be the start of a sequence, but without 3, we will not be able to form a group)*
                - else the number is the minimum number, pop heap to remove it
- if the loop completes without returning false, it is possible to rearrange the numbers (cards), return true


## Complexity
- Time complexity: O(filing hashmap + creating list of hashmap keys + heapify + while loop * for loop * heap pop) → O(n + m + m + n * k * log(heap size)) → O(n + m + m + n * k * logm) → O(n * k * logm)
    - *n = total numbers in the input array*
    - *m = total unique numbers*
    - *k = group size*
<!-- Add your time complexity here, e.g. $$O(n)$$ -->


- Space complexity: O(hashmap + heap) → O(m + m) → O(m)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def isNStraightHand(self, hand: List[int], groupSize: int) -> bool:
        if len(hand) % groupSize:
            return False
        count_card = Counter(hand)
        min_heap = list(count_card.keys())
        heapq.heapify(min_heap)
        while min_heap:
            start = min_heap[0]
            for i in range(start, start + groupSize):
                if i not in count_card:
                    return False
                count_card[i] -= 1
                if count_card[i] == 0:
                    if i != min_heap[0]:
                        return False
                    heapq.heappop(min_heap)
        return True
```
