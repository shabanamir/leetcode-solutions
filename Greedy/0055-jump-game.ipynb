# <center> 55. Jump Game </center>

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Greedy approach.

At each index, check the maximum index we can reach from that index and jump to that index. In the end, check if the maximum index reached is the last index or not.

Note: For an index to be reachable, each of the previous indices has to be reachable.


## Approach
<!-- Describe your approach to solving the problem. -->
- set i = 0 to track the index
- set reach = 0 to track the max index reached
- traverse the array, *loop til the index is inbound i.e i is less than the last index and current index i is reachable*
    - find the max reachable index and jump i.e update reach
    - increment i
- check if reach = size of the array i.e we reached the last index and return the result


## Complexity
- Time complexity: O(loop) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def canJump(self, nums: List[int]) -> bool:
        i = 0
        reach = 0
        while i < len(nums) and i <= reach:
            reach = max(reach, i + nums[i])
            i += 1
        return i == len(nums)
```
