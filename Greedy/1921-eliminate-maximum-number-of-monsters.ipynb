# <center> 1921. Eliminate Maximum Number of Monsters </center>


## Problem Description
[Click here](https://leetcode.com/problems/eliminate-maximum-number-of-monsters/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Sort the monster by their arrival time. Traverse the sorted array. In every iteration, eliminate a monster and recharge the weapon. If the recharge time (iteration number) takes more than the monster's arrival, you lose else you kill all monsters.

Note: recharge time is 1 minute and no recharge is required at the start.


## Approach
<!-- Describe your approach to solving the problem. -->
- set n = total number of monsters
- calculate the arrival time (distance - 1 / speed) of each monster and sort the arrival times
- set recharge = 0 to track the recharge time and number of monsters you killed
    - *(recharge time also shows the number of monsters killed because for killing each monster the weapon is recharged)*
- traverse the sorted array <br> 
for each index (recharge time), arrival time in the sorted array
    - if recharge time > arrival time, you lose
        - return recharge time 
- if the loop completes, you killed all monsters, return n


## Complexity
- Time complexity: O(sort + traversal) → O(nlogn + n) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort + arrival time array) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def eliminateMaximum(self, dist: List[int], speed: List[int]) -> int:
        n = len(dist)
        monster_arrival = sorted([(dist[i] - 1) // speed[i] for i in range(n)])
        for recharge, arrival in enumerate(monster_arrival):
            if recharge > arrival:
                return recharge
        return n
```
