# <center> 978. Longest Turbulent Subarray </center>


## Problem Description
[Click here](https://leetcode.com/problems/longest-turbulent-subarray/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the longest turbulent subarray.

A subarray is turbulent if the comparison sign flips between adjacent elements.
nums[0] < nums[1] > nums[2] < nums[3] > nums[4] ...  or
nums[0] > nums[1] < nums[2] > nums[3] < nums[4] ...

An element nums[i] is part of the turbulant subarray if:
- nums[i-2] < nums[i-1] > nums[i] or 
- nums[i-2] > nums[i-1] < nums[i]

Traverse the array and track the longest turbulent subarray size. If current element is part of the turbulent subarray, increment the size else reset the size 


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases:
    - if array max and min are same, it means all elements are same, return 1
- set res = 1 to track the max turbulent subarray size
- set cur = 1 to track the current turbulent subarray size
- traverse the array starting from the second element
    - if current element is part of the current turbulent subarray i.e turbulent subarray conditions satisfy
        - increment cur
    - *else the current turbulent subarray has ended*
    if the adjacent elements i.e previous and current elements are not same, the make a new turbulent subarray that contains two elements so reset the size to 2 
    - update res by taking max of current turbulent subarray size i.e cur and previous max size turbulent subarray i.e res
- return res


## Complexity
- Time complexity: O(finding min + finding max + array traversal) → O(n + n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxTurbulenceSize(self, arr: List[int]) -> int:
        if min(arr) == max(arr):
            return 1
        res, cur = 1, 1
        for i in range(1, len(arr)):
            if arr[i-2] > arr[i-1] < arr[i] or arr[i-2] < arr[i-1] > arr[i]:
                cur += 1
            elif arr[i-1] != arr[i]:
                cur = 2
            res = max(res, cur)
        return res
```
