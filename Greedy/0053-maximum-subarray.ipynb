# <center> 53. Maximum Subarray </center>


## Problem Description
[Click here](https://leetcode.com/problems/maximum-subarray/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to find the contiguous subarray with the max sum. There are also negative numbers in the array.


Use Kadane's algorithm

Traverse the array and track the prefix sum. If the prefix sum is less than 0 i.e current subarray sum is less, discard the subarray by resetting the prefix sum. Else the current subarray sum is greater. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = first element in the array to track the max sum
- set pre_sum = 0 to track the prefix sum
- traverse the array
    - add the current number to pre_sum to include it in the current subarray
    - update res by taking the max of current prefix sum and previous max sum
    - if prefix sum is less than 0, the current subarray sum is less    - reset pre_sum to 0
- return res


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxSubArray(self, nums: List[int]) -> int:
        res = nums[0]
        pre_sum = 0
        for n in nums:
            pre_sum += n
            res = max(res, pre_sum)
            if pre_sum < 0:
                pre_sum = 0
        return res
```
