## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to send half people to city A and half to B with minimum cost.

Use the greedy approach. Sort the costs based on the cost difference between the cities. The lower the cost difference, the higher the priority to be sent to city A. Send the first half of the sorted array to city A and the second half to city b.


## Approach
<!-- Describe your approach to solving the problem. -->
- sort the costs based on cost difference to get people with low costs for city A at the start (left half) and with high costs at the end (right half)
- set res = 0 to track the minimum cost
- set n = mid of array
- traverse the left half array
    - add city A cost of left element (i) and city b cost of right element (i + n) to res
    *index 0 = city A cost, index 1 = city B cost*
- return res

## Complexity
- Time complexity: O(sort + loop) → O(nlogn + n/2) → O(nlogn)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sort) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def twoCitySchedCost(self, costs: List[List[int]]) -> int:
        costs.sort(key=lambda x: x[0] - x[1])
        res = 0
        n = len(costs) // 2
        for i in range(n):
            res += costs[i][0] + costs[i + n][1]
        return res 
```
