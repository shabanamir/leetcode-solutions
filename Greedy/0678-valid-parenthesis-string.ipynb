# <center> 678. Valid Parenthesis String </center>


## Problem Description
[Click here](https://leetcode.com/problems/valid-parenthesis-string/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
The string is invalid if there are too many unpaired opening or closing brackets.

An asterisk can be considered as an opening bracket, closing bracket, or an empty string. Treat asterisk as a closing bracket only when there are more opening brackets.

Track the lower and upper boundary of the opening bracket. 
- lower boundary is the minimum number of '(' that must be paired
- high boundary is the maximum '(' that could be paired
- if high < 0, there are too many closing brackets, and if low > 0, there are too many opening brackets.


## Approach
<!-- Describe your approach to solving the problem. -->
- set low and high to 0 to track the upper and lower boundary of the opening brackets
- traverse the string, *for each character c in the string s*
    - if c is an opening bracket
        - increment high and low because we got a new unpaired opening bracket 
    - if c is a closing bracket
        - decrement high because we got a closing bracket to pair with the opening bracket 
        - decrement low if there are unpaired opening brackets left 
    - else c is an asterik
        - for high, treat it as an opening bracket and increment high because we got a new opening bracket
        - for low, treat it as a closing bracket if there are unpaired opening brackets left, else treat it as an empty string
    - if high becomes negative, the string is invalid because there are too many closing brackets, return false
- after processing all characters check if low is 0 i.e all necessary opening brackets are paired, and return the result


## Complexity
- Time complexity: O(string traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def checkValidString(self, s: str) -> bool:
        low, high = 0, 0
        for c in s:
            if c == "(":
                high += 1
                low += 1
            elif c == ")":
                high -= 1
                if low > 0:
                    low -= 1
            else:
                high += 1
                if low > 0:
                    low -= 1
            if high < 0:
                return False
        return low == 0
```
