# <center> 1423. Maximum Points You Can Obtain from Cards </center>


## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We need to select k cards with maximum points i.e a subarray of size k with maximum sum.

Use the sliding window approach and choose the greater value card from left or right.  We can pick a card either from left or right at any point. So, start by taking k cards from one side and slide the window to maximize the points. There are two ways:
1. start by taking all cards from the left side. Each time add a card from the right and remove a card from the left 
2. start by taking all cards from the right side. Each time add a card from the left and remove a card from the right 


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if k = total number of cards, we have to choose all cards
        - return the sum of all cards
- set left pointer l at the start
- set right pointer r at (n - k) position
- set total = sum of last k cards to track total current points
- set res = total to track the max points i.e max sum
- start from position r and loop until the last position
    - get current points by adding a card from the left and removing a card from the right
        - total = add left card to total and remove right card <br>
        total = total + left card - right card <br>
        total += left card - right card*
    - update res by taking max of current subarray sum and previous max sum
        - *(this step will check which card was greater, the removed right card or the added left card, and will add it to the result)*
    - increment l and r to move
- return res


## Complexity
- Time complexity: O(sum of last k cards + loop) → O(k + n - k) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(sliced list) → O(k)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxScore(self, cardPoints: List[int], k: int) -> int:
        if k == len(cardPoints):
            return sum(cardPoints)
        l, r = 0, len(cardPoints) - k
        total = sum(cardPoints[r:])
        res = total
        while r < len(cardPoints):
            total += cardPoints[l] - cardPoints[r]
            res = max(res, total)
            l += 1
            r += 1
        return res 
```
