## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
1. Binary Search <br> 
We have to minimize the max value of the array. So the result will be in the range of min(array) to max(array). Consider this range a sorted array and use binary search. Find the midpoint, if it is valid (present in the array), move left to find a smaller value, else move right to find a valid value. 
The complexity will be n * log(max-low). log for binary search and n to check if the number is valid.

2. Math/Greedy Approach <br> 
The minimized max value of an array is the max of all subarray ceil averages. Compute the ceil average of each subarray and return the max.
    - ceil average because to minimize max of an array, we need to decrease a number and increase the other i.e to evenly distribute them. <br>
    For example [4, 7] <br>
    array max is 7 <br>
    find the average i.e 4 + 7 / 2 = 11/ 2 = 5.5 <br>
    take ceil of the average i.e 5.5 = 5 <br>
    5 is the minimized max of array <br>
    - formula for ceil of subarray average <br>
= ceil (subarray average) <br>
= ceil (subarray sum / total elements in the subarray) <br>
= ceil (prefix sum  till index i / index i + 1) *(i is subarray ending index)* <br>
= ceil (prefix sum / i + 1) <br>
= prefix sum + i + 1 - 1 / i + 1  *because ceil(a/b) = a + b - 1 / b* <br>
= prefix sum + i / i + 1

## Approach
<!-- Describe your approach to solving the problem. -->
**Math Approach**
- set res = first element to track the minimized max
    - *(we are using the first element to skip a step. The first subarray contains only the first element. So, the sum and ceil of the first subarray is the first number.)*  
- set pre_sum = first element to track the sum of each subarray
- traverse the array starting from the second number because we have already done the computation for the first subarray 
    - add the current number to pre_sum to include the number in the subarray
    - find the subarray ceil avg and update the res
- return res

## Code

```python
class Solution:
    
    def minimizeArrayValue(self, nums: List[int]) -> int:
        res, pre_sum = nums[0], nums[0]
        for i in range(1, len(nums)):
            pre_sum += nums[i]
            res = max(res, (pre_sum + i) // (i + 1))
        return res
```
