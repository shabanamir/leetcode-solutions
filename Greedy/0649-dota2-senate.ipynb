## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
- senators on the left have an earlier turn to vote
- senators on the left compete against their counterparts on the right
- the winner eliminates (removes from the left) the opposing senator and proceeds to the next round (moves to the right).

Use a queue to insert (from right) and delete (from left) in O(1) time.


## Approach
<!-- Describe your approach to solving the problem. -->
- use a separate queue for each party
- for each senator (character c) in the senators list
    - if c is 'D'
        - add the position to the dire queue
    - else add the position to the radiant que
- loop until one of the queues is empty i.e senators will compete until there is one winner
    - pop both queues to get the senators' position for fight/voting
    - if the dire senator position is less, he will win because he has an early turn to vote
        - add the dire senator position to the right of the queue to move him to the next round
    - else move the radiant senator to the next round
- in the end, the non-empty queue will have the winner. check the queues and return the result


## Complexity
- Time complexity: O(array traversal + queue loop) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(dire queue + radiant queue) → O(n + n) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def predictPartyVictory(self, senate: str) -> str:
        dire, radiant = deque(), deque()
        for i, c in enumerate(senate):
            if c == 'D':
                dire.append(i)
            else:
                radiant.append(i)
        while radiant and dire:
            d_turn = dire.popleft()
            r_turn = radiant.popleft()
            if d_turn < r_turn:
                dire.append(d_turn + len(senate))
            else:
                radiant.append(r_turn + len(senate))
        return 'Dire' if dire else 'Radiant'
```
