# <center> 1899. Merge Triplets to Form Target Triplet </center>


## Problem Description
[Click here](https://leetcode.com/problems/merge-triplets-to-form-target-triplet/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
triplets array = (a1, b1, c1), (a2, b2, c2), ... , (an, bn, cn) <br>
target triplet = (A, B, C)

Find a subset (ai, bi, ci) in the triplets array such that <br>
max(ai) = A, max(bi) = B, max(ci) = C


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = an array of size 3 initialized with zeroes to track the result triplet
- traverse the triplets array <br> 
for each triplet subset t in the triplets array
    - if first number in the current triplet <= first number in the target triplet, second <= target triplet second target number and third <= target triplet third number
        - update the first, second, and third numbers in the result array
- check if the result triplet is equal to the target triplet and return the result


## Complexity
- Time complexity: O(array traversal + target and result triplet comparison) → O(n + 3) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result array) → O(3) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def mergeTriplets(self, triplets: List[List[int]], target: List[int]) -> bool:
        res = [0, 0, 0]
        for t in triplets:
            if t[0] <= target[0] and t[1] <= target[1] and t[2] <= target[2]:
                res[0], res[1], res[2] = max(res[0], t[0]), max(res[1], t[1]), max(res[2], t[2])
        return res == target
```
