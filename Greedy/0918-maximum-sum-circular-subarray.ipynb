## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
max sum circular subarray = max (maximum subarray sum, total sum - minimum subarray sum)

Use Kadane's algorithm to find the minimum subarray sum and the maximum subarray sum. 


## Approach
<!-- Describe your approach to solving the problem. -->
- set total = 0 to track the total sum of the array
- set cur_max = 0 to track the sum of the current maximum subarray
- set cur_min = 0 to track the sum of the current minimum subarray
- set global_max = a very small number to track the maximum subarray sum
- set global_min = a very large number to track the minimum subarray sum
- traverse the array <br> 
for each number in the array
    - add the number to total
    - update cur_max, cur_min, global_max, global_min
- if maximum subarray sum (global_max) is less than 0, it means all numbers in the array are negative, return global_max
- else return the max sum circular subarray using the formula


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def maxSubarraySumCircular(self, nums: List[int]) -> int:
        total = 0
        cur_max = 0
        cur_min= 0
        global_max = -math.inf
        global_min = math.inf
        for num in nums:
            total += num
            cur_max = max(cur_max + num, num)
            cur_min= min(cur_min + num, num)
            global_max = max(global_max, cur_max)
            global_min = min(global_min, cur_min)
        return global_max if global_max < 0 else max(global_max, total - global_min)
```
