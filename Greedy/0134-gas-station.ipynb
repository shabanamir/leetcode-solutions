## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
There are n stations, represented by the array indices. We have to find a station, starting from which we can travel to all other stations. We can travel to a station if we have enough gas in the tank.

gas required to travel from station i to i+1 = gas available at station i - gas required to go to station i+1 = gas[i] - cost[i]

Visit all gas stations and track the current gas in the tank. If the current gas in the tank becomes negative, reset it and update the start station. 


## Approach
<!-- Describe your approach to solving the problem. -->
- check base cases
    - if sum(gas array) < sum(cost array), we can't travel all stations, return -1 because we can only travel all the stations if sum(gas) >= sum(cost)
        - *(here we will not check the base case because it requires the sum of both arrays so time will be O(n + n + n) i.e O(finding gas sum + finding cost sum + finding start station), instead we will check net gas in the end. If we use this condition then there is no need to calculate or check the net gas)*
- set start = 0 to track the start station
- set net = 0 to track the net gas available
- set cur = 0 to track the current gas in the tank
- traverse all gas stations <br> 
at each station
    - fill the tank and update cur i.e get the difference of gas available at the current station and gas required to reach the next
    - update net
    - if gas in the tank becomes negative i.e cur < 0
        - reset cur to 0 
        - update the start station to the current station
- if net gas < 0, we can't travel around the circuit
    - return -1
- else return the start station


## Complexity
- Time complexity: O(array traversal) â†’ O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
        # if sum(gas) < sum(cost):
        #     return -1
        start, net, cur = 0, 0, 0                  
        for i in range(len(gas)):  
            cur += gas[i] - cost[i] 
            net += gas[i] - cost[i]          
            if cur < 0: 
                cur = 0 
                start = i + 1
        return -1 if net < 0 else start  
```
