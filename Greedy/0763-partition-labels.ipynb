# <center> 763. Partition Labels </center>


## Problem Description
[Click here](https://leetcode.com/problems/partition-labels/description/)


## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
Each character can appear only in one partition. 

The repeating character with the max last position decides the partition. 
For example abcafadcbghtkhg
- partition 1 = abcafadcb because b last index is maximum
- partition 2 = ghtkhg, g last index is max

We need the last index of a character to decide the partition. Use a hashmap and map the last index of each character.


## Approach
<!-- Describe your approach to solving the problem. -->
- set res = an array to store each partition size
- create a hashmap and map the last index of each character
    - *key = character*
    - *value = last occurring index*
- set left pointer l = 0 to track the start index of a partition
- set right pointer r = 0 to track the last index of a partition
- traverse the string, *for each character c at index i*
    - find r by taking max of the current character's last index and previous max last index
    - if current index is the max last index i.e i == r, you get a partition
        - add the partition size to res
            - *partition size = partiton end - partiton start + 1*
        - update l to the index after the current partition end i.e r + 1
- return res

## Complexity
- Time complexity: O(hashmap creation + string traversal) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(result array + hashmap) → O(possible partitions + hashmap keys) → (no. of lowercase english letters + no. of lowercase english letters) → O(26 + 26) → O(1)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def partitionLabels(self, s: str) -> List[int]:
        res = []
        last_index = {c: i for i, c in enumerate(s)}
        l, r = 0, 0
        for i, c in enumerate(s):
            r = max(r, last_index[c])
            if i == r:
                res.append(r - l + 1)
                l = r + 1
        return res
```
