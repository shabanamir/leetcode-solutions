## Problem Description

## Intuition
<!-- Describe your first thoughts on how to solve this problem. -->
We can move forward if:
- current index + minJump <= next index <= min(current index + maxJump, last index)
- value at next index is '0'

Next index status (reachable or not reachable) depends on the previous. So, to solve the main problem, we need to solve the sub-problems. Use dynamic programming.


## Approach
<!-- Describe your approach to solving the problem. -->
- check base case
    - if the character at the last index is not 0, return false
- set reached = 0 to track the max index reached
- initialize an array of size equal to string size with false to track whether an index is reachable from the start or not
- set index 0 value to true because it satisfies both conditions
- start from the minJump index and loop till the last index
    - update reached
    - if the difference between current index and maxJump is greater than 0, we can't reach the current index from the start (index 0) by max jump
        - decrement reached
    - update the status of the current index in the dp array by checking if the index is reachable and the character at the index is '0'
- return status of the last index i.e last value from the dp array
    - *(if true we can reach the last index and vice versa)*


## Complexity
- Time complexity: O(dp array creation + loop) → O(n + n) → O(n)
<!-- Add your time complexity here, e.g. $$O(n)$$ -->

- Space complexity: O(dp array) → O(n)
<!-- Add your space complexity here, e.g. $$O(n)$$ -->


## Code

```python
class Solution:

    def canReach(self, s: str, minJump: int, maxJump: int) -> bool:
        if s[-1] != '0':
            return False
        reached = 0
        dp = [False] * len(s)
        dp[0] = True
        for i in range(minJump, len(s)):
            reached += dp[i - minJump]
            if i - maxJump > 0:
                reached -= dp[i - maxJump - 1]
            dp[i] = reached and s[i] == '0'
        return dp[-1] 
```
